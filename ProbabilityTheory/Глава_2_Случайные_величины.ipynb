#Глава 2.  Случайные величины

## Введение

Случайные величины являются ключевым понятием в теории вероятностей и статистике. Они позволяют моделировать неопределенность и варьировать результаты в различных ситуациях. В данной лекции мы рассмотрим основные аспекты случайных величин, включая их типы, функции распределения, распространенные распределения, а также характеристики, такие как математическое ожидание, дисперсия, медиана, мода и квантиль.

## 1. Случайные величины

### 1.1. Дискретные и непрерывные случайные величины

**Случайная величина** — это числовая функция, которая сопоставляет каждому исходу некоторого случайного эксперимента определённое число.

- **Дискретная случайная величина** — это случайная величина, которая может принимать только счётное число значений. Примеры: количество выпавших орлов при подбрасывании монеты, число автомобилей, проходящих через определённый пункт за час.

- **Непрерывная случайная величина** — это случайная величина, которая может принимать любое значение из некоторого диапазона. Примеры: рост людей, температура воздуха.

### 1.2. Функции распределения и плотности распределения

**Функция распределения (CDF)** — это функция, которая показывает вероятность того, что случайная величина $X$ примет значение меньше или равно $x$:

$$
F_X(x) = P(X \leq x)
$$

**Плотность распределения (PDF)** — это функция, которая описывает вероятность того, что случайная величина $X$ примет значение в определённом интервале. Для непрерывной случайной величины плотность вероятности $f_X(x)$ связана с функцией распределения следующим образом:

$$
F_X(x) = \int_{-\infty}^{x} f_X(t) \, dt
$$

Для дискретной случайной величины плотность вероятности выражается через функцию вероятности:

$$
P(X = x_i) = p_X(x_i)
$$

### 1.3. Распределения

1. **Распределение Бернулли**: Моделирует эксперимент с двумя возможными исходами (успех и неуспех). Вероятность успеха обозначается $p$, неуспеха — $1-p$.

   $$
   P(X=1) = p, \quad P(X=0) = 1 - p
   $$

2. **Биномиальное распределение**: Обобщает распределение Бернулли на $n$ независимых экспериментов.

   $$
   P(X = k) = C(n, k) p^k (1-p)^{n-k}, \quad k = 0, 1, \ldots, n
   $$

   где $C(n, k) = \frac{n!}{k!(n-k)!}$ — биномиальный коэффициент.

3. **Гауссово (нормальное) распределение**: Непрерывное распределение, которое описывается двумя параметрами — математическим ожиданием $\mu$ и дисперсией $\sigma^2$.

   $$
   f_X(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
   $$

4. **Распределение Пуассона**: Моделирует количество событий, происходящих в фиксированном интервале времени или пространства при известной средней интенсивности $\lambda$.

   $$
   P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad k = 0, 1, 2, \ldots
   $$

5. **Экспоненциальное распределение**: Непрерывное распределение, часто используемое для описания времени между событиями в процессе Пуассона.

   $$
   f_X(x) = \lambda e^{-\lambda x}, \quad x \geq 0
   $$

### 1.4. Математическое ожидание и дисперсия

**Математическое ожидание** $E(X)$ случайной величины является средним значением, которое она принимает.

- Для дискретной случайной величины:

$$
E(X) = \sum_{i=1}^{n} x_i P(X = x_i)
$$

- Для непрерывной случайной величины:

$$
E(X) = \int_{-\infty}^{\infty} x f_X(x) \, dx
$$

**Дисперсия** $Var(X)$ показывает, насколько значения случайной величины разбросаны относительно её математического ожидания.

- Для дискретной случайной величины:

$$
Var(X) = E(X^2) - (E(X))^2
$$

- Для непрерывной случайной величины:

$$
Var(X) = \int_{-\infty}^{\infty} (x - E(X))^2 f_X(x) \, dx
$$

## 2. Характеристики случайных величин

### 2.1. Медиана, мода и квантиль

- **Медиана** — это значение, которое делит распределение на две равные части. Для дискретной случайной величины медиана — это среднее двух центральных значений, если количество значений чётное.

- **Мода** — это значение, которое встречается наиболее часто в распределении.

- **Квантиль** — это значение $q$, такое что:

$$
F_X(q) = P(X \leq q) = p
$$

где $p$ — вероятность, обычно $0 < p < 1$.

## 3. Примеры

### Пример 1: Дискретная случайная величина

Рассмотрим случайную величину $X$, которая описывает количество орлов, выпавших при подбрасывании двух монет. Возможные значения: $0, 1, 2$.

- $P(X=0) = \frac{1}{4}$
- $P(X=1) = \frac{1}{2}$
- $P(X=2) = \frac{1}{4}$

**Математическое ожидание:**

$$
E(X) = 0 \cdot \frac{1}{4} + 1 \cdot \frac{1}{2} + 2 \cdot \frac{1}{4} = 0 + \frac{1}{2} + \frac{1}{2} = 1
$$

**Дисперсия:**

Сначала вычислим $E(X^2)$:

$$
E(X^2) = 0^2 \cdot \frac{1}{4} + 1^2 \cdot \frac{1}{2} + 2^2 \cdot \frac{1}{4} = 0 + \frac{1}{2} + 1 = \frac{3}{2}
$$

Теперь находим дисперсию:

$$
Var(X) = E(X^2) - (E(X))^2 = \frac{3}{2} - 1^2 = \frac{3}{2} - 1 = \frac{1}{2}
$$

### Пример 2: Непрерывная случайная величина

Рассмотрим случайную величину $Y$, которая имеет нормальное распределение с параметрами $\mu = 0$ и $\sigma^2 = 1$.

**Математическое ожидание:**

$$
E(Y) = \mu = 0
$$

**Дисперсия:**

$$
Var(Y) = \sigma^2 = 1
$$

**Плотность вероятности:**

$$
f_Y(y) = \frac{1}{\sqrt{2\pi}} e^{-\frac{y^2}{2}}
$$

### Пример 3: Биномиальное распределение

Пусть мы подбрасываем монету $n = 10$ раз, и вероятность успеха $p = 0.5$. Рассмотрим случайную величину $Z$, которая описывает количество орлов.

**Вероятность получить 5 орлов:**

$$
P(Z = 5) = C(10, 5) (0.5)^5 (0.5)^5 = \frac{10!}{5!5!} (0.5)^{10} = 252 \cdot \frac{1}{1024} \approx 0.246
$$

**Математическое ожидание и дисперсия:**

$$
E(Z) = n \cdot p = 10 \cdot 0.5 = 5
$$
$$
Var(Z) = n \cdot p \cdot (1 - p) = 10 \cdot 0.5 \cdot 0.5 = 2.5
$$

### Пример 4: Экспоненциальное распределение

Рассмотрим случайную величину $W$, которая

 имеет экспоненциальное распределение с параметром $\lambda = 2$.

**Математическое ожидание:**

$$
E(W) = \frac{1}{\lambda} = \frac{1}{2} = 0.5
$$

**Дисперсия:**

$$
Var(W) = \frac{1}{\lambda^2} = \frac{1}{4} = 0.25
$$

### Пример 1: Дискретная случайная величина (Распределение Бернулли)

Предположим, мы подбрасываем монету, и случайная величина $X$ описывает результат подбрасывания, где:
- $X = 1$ (орел) с вероятностью $p = 0.6$
- $X = 0$ (решка) с вероятностью $1 - p = 0.4$

**Вероятности:**
- $P(X = 1) = 0.6$
- $P(X = 0) = 0.4$

**Математическое ожидание:**

$$
E(X) = 1 \cdot P(X = 1) + 0 \cdot P(X = 0) = 1 \cdot 0.6 + 0 \cdot 0.4 = 0.6
$$

**Дисперсия:**

Сначала найдем $E(X^2)$:

$$
E(X^2) = 1^2 \cdot P(X = 1) + 0^2 \cdot P(X = 0) = 1 \cdot 0.6 + 0 \cdot 0.4 = 0.6
$$

Теперь находим дисперсию:

$$
Var(X) = E(X^2) - (E(X))^2 = 0.6 - (0.6)^2 = 0.6 - 0.36 = 0.24
$$

### Пример 2: Непрерывная случайная величина (Нормальное распределение)

Пусть случайная величина $Y$ имеет нормальное распределение с математическим ожиданием $\mu = 10$ и стандартным отклонением $\sigma = 2$.

**Математическое ожидание:**

$$
E(Y) = \mu = 10
$$

**Дисперсия:**

$$
Var(Y) = \sigma^2 = 2^2 = 4
$$

**Плотность вероятности:**

Для нахождения вероятности, например, что $Y$ меньше 12, используем стандартную нормальную таблицу. Сначала преобразуем $Y$ в стандартную нормальную величину $Z$:

$$
Z = \frac{Y - \mu}{\sigma} = \frac{12 - 10}{2} = 1
$$

Теперь находим $P(Y < 12) = P(Z < 1)$. Из таблицы стандартного нормального распределения:

$$
P(Z < 1) \approx 0.8413
$$

### Пример 3: Биномиальное распределение

Рассмотрим случайную величину $Z$, которая представляет количество успехов при 5 подбрасываниях монеты с вероятностью успеха $p = 0.5$ (орел).

**Вероятности:**

Используем формулу биномиального распределения:

$$
P(Z = k) = C(n, k) p^k (1-p)^{n-k}
$$

где $n = 5$.

**Расчеты для $k = 2$ (два орла):**

$$
P(Z = 2) = C(5, 2) (0.5)^2 (0.5)^{5-2} = \frac{5!}{2!3!} (0.5)^5 = 10 \cdot \frac{1}{32} = \frac{10}{32} = 0.3125
$$

**Математическое ожидание:**

$$
E(Z) = n \cdot p = 5 \cdot 0.5 = 2.5
$$

**Дисперсия:**

$$
Var(Z) = n \cdot p \cdot (1-p) = 5 \cdot 0.5 \cdot 0.5 = 1.25
$$

### Пример 4: Экспоненциальное распределение

Пусть случайная величина $W$ описывает время между событиями с параметром $\lambda = 3$.

**Математическое ожидание:**

$$
E(W) = \frac{1}{\lambda} = \frac{1}{3} \approx 0.333
$$

**Дисперсия:**

$$
Var(W) = \frac{1}{\lambda^2} = \frac{1}{3^2} = \frac{1}{9} \approx 0.111
$$

**Плотность вероятности:**

Для $x \geq 0$:

$$
f_W(x) = \lambda e^{-\lambda x} = 3 e^{-3x}
$$

**Вероятность того, что $W < 0.5$:**

$$
P(W < 0.5) = 1 - e^{-\lambda x} = 1 - e^{-3 \cdot 0.5} \approx 1 - e^{-1.5} \approx 1 - 0.2231 \approx 0.7769
$$

### Пример 5: Распределение Пуассона

Предположим, в некотором магазине в среднем наблюдается 4 покупателя в час. Случайная величина $N$ описывает количество покупателей за час.

**Вероятности:**

Используем формулу распределения Пуассона:

$$
P(N = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

где $\lambda = 4$.

**Вероятность, что в магазине будет 2 покупателя:**

$$
P(N = 2) = \frac{4^2 e^{-4}}{2!} = \frac{16 e^{-4}}{2} = 8 e^{-4} \approx 8 \cdot 0.0183 \approx 0.1464
$$

**Математическое ожидание:**

$$
E(N) = \lambda = 4
$$

**Дисперсия:**

$$
Var(N) = \lambda = 4
$$
