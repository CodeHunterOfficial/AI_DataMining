##Глава 8. Процесс Маркова и марковские цепи

### Введение

Процесс Маркова — это стохастический процесс, обладающий свойством «безопасности памяти». Это означает, что будущее состояния процесса зависит только от его текущего состояния и не зависит от того, как это состояние было достигнуто. В данной лекции мы рассмотрим основные понятия, связанные с процессами Маркова, марковскими цепями и их приложениями, в частности в области обработки естественного языка (NLP).

### 1. Марковские процессы

**Определение**: Марковский процесс — это последовательность случайных величин $X_1, X_2, \ldots$, где вероятность следующего состояния зависит только от текущего состояния:

$$
P(X_{n+1} = x | X_n = x_n, X_{n-1} = x_{n-1}, \ldots, X_1 = x_1) = P(X_{n+1} = x | X_n = x_n)
$$

**Свойства**:

1. **Марковское свойство**: Для марковских процессов характерно, что состояние процесса в момент времени $n$ зависит только от состояния в момент времени $n-1$.
2. **Трансляция вероятностей**: Если $S$ — это множество состояний процесса, то вероятность перехода из состояния $i$ в состояние $j$ обозначается как $P_{ij}$.

### 2. Марковские цепи и их свойства

**Марковская цепь** — это специальный случай марковского процесса, где состояния дискретны, и процесс принимает значения в конечном или счётном множестве.

#### 2.1. Определение и матрица переходов

Марковская цепь можно описать с помощью матрицы переходов $P$, где элемент $P_{ij}$ представляет вероятность перехода из состояния $i$ в состояние $j$:

$$
P =
\begin{pmatrix}
P_{11} & P_{12} & \ldots & P_{1n} \\
P_{21} & P_{22} & \ldots & P_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
P_{n1} & P_{n2} & \ldots & P_{nn}
\end{pmatrix}
$$

Свойства матрицы переходов:
- Все элементы матрицы неотрицательны: $P_{ij} \geq 0$
- Сумма вероятностей для каждого состояния равна 1:

$$
\sum_{j} P_{ij} = 1, \quad \forall i
$$

#### 2.2. Достижимость и рекурсия

Состояние $j$ считается достижимым из состояния $i$, если существует такая последовательность переходов, что можно попасть из $i$ в $j$.

**Рекуррентные и транзиторные состояния**:
- **Рекуррентное состояние**: Если, начиная из состояния $i$, вероятность возвращения в это состояние равна 1.
- **Транзиторное состояние**: Если вероятность возвращения меньше 1.

#### 2.3. Эразмус и периодичность

- Состояние называется периодическим, если возвращение в это состояние происходит с периодом $d$, где $d$ — это наибольшее общее кратное всех периодов возвращения.

### 3. Стационарные распределения

**Стационарное распределение** $\pi$ — это распределение вероятностей, которое остаётся неизменным при переходах по марковской цепи. Оно удовлетворяет следующему уравнению:

$$
\pi P = \pi
$$

Где $P$ — матрица переходов. Вектор $\pi$ должен удовлетворять условиям:

$$
\sum_{i} \pi_i = 1, \quad \pi_i \geq 0 \quad \forall i
$$

Стационарные распределения помогают понять долгосрочные свойства системы.

### 4. Приложения в NLP

В NLP марковские цепи и скрытые марковские модели (HMM) находят широкое применение. Скрытые марковские модели представляют собой статистическую модель, в которой система описывается марковской цепью с скрытыми состояниями.

**Применения**:
- **Анализ текста**: Например, в задаче тегирования частей речи (POS tagging), где каждое слово может быть связано с определённой частью речи.
- **Распознавание речи**: Использование HMM для преобразования звуковых сигналов в текст.

#### Пример: Тегирование частей речи

Предположим, у нас есть предложения, и мы хотим определить части речи для каждого слова. Мы можем использовать HMM следующим образом:

1. Состояния модели: части речи (существительное, глагол, прилагательное и т.д.).
2. Наблюдаемые события: слова в предложении.
3. Вероятности переходов между частями речи: $P(\text{тег}_{t} | \text{тег}_{t-1})$.
4. Вероятности наблюдений: $P(\text{слово}_{t} | \text{тег}_{t})$.

### 5. Примеры

#### Пример 1: Простая марковская цепь

Рассмотрим марковскую цепь с состояниями $S = \{A, B\}$ и следующими вероятностями переходов:

$$
P =
\begin{pmatrix}
0.7 & 0.3 \\
0.4 & 0.6
\end{pmatrix}
$$

- $P_{AA} = 0.7$, $P_{AB} = 0.3$
- $P_{BA} = 0.4$, $P_{BB} = 0.6$

Сначала состояние $A$:

$$
\text{Вероятности после одного перехода} = \begin{pmatrix} 1 \\ 0 \end{pmatrix} P = \begin{pmatrix} 0.7 \\ 0.3 \end{pmatrix}
$$

После второго перехода:

$$
\begin{pmatrix} 0.7 \\ 0.3 \end{pmatrix} P = \begin{pmatrix} 0.7 \cdot 0.7 + 0.3 \cdot 0.4 \\ 0.7 \cdot 0.3 + 0.3 \cdot 0.6 \end{pmatrix} = \begin{pmatrix} 0.61 \\ 0.39 \end{pmatrix}
$$

#### Пример 2: Стационарное распределение

Для предыдущей марковской цепи найдем стационарное распределение $\pi$:

$$
\pi P = \pi \implies \begin{pmatrix} \pi_A & \pi_B \end{pmatrix} \begin{pmatrix} 0.7 & 0.3 \\ 0.4 & 0.6 \end{pmatrix} = \begin{pmatrix} \pi_A & \pi_B \end{pmatrix}
$$

Записываем систему уравнений:

1. $\pi_A = 0.7\pi_A + 0.4\pi_B$
2. $\pi_B = 0.3\pi_A + 0.6\pi_B$

Сложив уравнения, получаем:

$$
\pi_A + \pi_B = 1
$$

Решая систему уравнений, находим:

- Из первого уравнения: $0.3\pi_A - 0.6\pi_B = 0$ или $\pi_A = 2\pi_B$.
- Подставляем в условие: $2\pi_B + \pi_B = 1 \Rightarrow 3\pi_B = 1 \Rightarrow \pi_B = \frac{1}{3}, \pi_A = \frac{2}{3}$.

Таким образом, стационарное распределение:

$$
\pi = \begin{pmatrix} \frac{2}{3} \\ \frac{1}{3} \end{pmatrix}
$$

## Задача 1: Стационарное распределение

**Условие**: Дана марковская цепь с состояниями $S = \{1, 2\}$ и матрицей переходов:

$$
P =
\begin{pmatrix}
0.3 & 0.7 \\
0.4 & 0.6
\end{pmatrix}
$$

Найдите стационарное распределение $\pi = (\pi_1, \pi_2)$.

### Решение:

1. **Система уравнений**:
   Для нахождения стационарного распределения необходимо решить уравнения:

   $$
   \pi P = \pi
   $$

   Это означает:

   $$
   \begin{pmatrix}
   \pi_1 & \pi_2
   \end{pmatrix}
   \begin{pmatrix}
   0.3 & 0.7 \\
   0.4 & 0.6
   \end{pmatrix}
   =
   \begin{pmatrix}
   \pi_1 & \pi_2
   \end{pmatrix}
   $$

   Записываем уравнения для каждой компоненты:

   - Уравнение 1:
     $$
     \pi_1 = 0.3\pi_1 + 0.4\pi_2
     $$

   - Уравнение 2:
     $$
     \pi_2 = 0.7\pi_1 + 0.6\pi_2
     $$

2. **Упрощение уравнений**:
   Перепишем уравнения:

   $$
   \pi_1 - 0.3\pi_1 - 0.4\pi_2 = 0 \implies 0.7\pi_1 - 0.4\pi_2 = 0 \tag{1}
   $$

   $$
   \pi_2 - 0.7\pi_1 - 0.6\pi_2 = 0 \implies -0.7\pi_1 + 0.4\pi_2 = 0 \tag{2}
   $$

3. **Запишем систему**:
   Теперь имеем систему уравнений:

   - $0.7\pi_1 - 0.4\pi_2 = 0$
   - $0.7\pi_1 - 0.4\pi_2 = 0$

   Упростим первое уравнение:

   $$
   \pi_2 = \frac{0.7}{0.4}\pi_1 = \frac{7}{4}\pi_1 \tag{3}
   $$

4. **Условие нормировки**:
   Используем условие нормировки:

   $$
   \pi_1 + \pi_2 = 1
   $$

   Подставляем из уравнения (3):

   $$
   \pi_1 + \frac{7}{4}\pi_1 = 1 \implies \frac{11}{4}\pi_1 = 1 \implies \pi_1 = \frac{4}{11}
   $$

   Теперь найдём $\pi_2$:

   $$
   \pi_2 = \frac{7}{4} \cdot \frac{4}{11} = \frac{7}{11}
   $$

5. **Ответ**:
   Стационарное распределение:

   $$
   \pi = \left( \frac{4}{11}, \frac{7}{11} \right)
   $$



## Задача 2: Анализ переходов в марковской цепи

**Условие**: В марковской цепи с состояниями $S = \{A, B, C\}$ матрица переходов задается следующим образом:

$$
P =
\begin{pmatrix}
0.2 & 0.5 & 0.3 \\
0.1 & 0.4 & 0.5 \\
0.6 & 0.2 & 0.2
\end{pmatrix}
$$

Если начальное состояние $A$, найдите вероятности состояния цепи после двух шагов.

### Решение:

1. **Начальное состояние**:
   Начальное состояние можно записать как вектор:

   $$
   \mathbf{v_0} =
   \begin{pmatrix}
   1 \\
   0 \\
   0
   \end{pmatrix}
   $$

2. **Первый переход**:
   Вычисляем вектор вероятностей после первого шага:

   $$
   \mathbf{v_1} = \mathbf{v_0} P =
   \begin{pmatrix}
   1 \\
   0 \\
   0
   \end{pmatrix}
   \begin{pmatrix}
   0.2 & 0.5 & 0.3 \\
   0.1 & 0.4 & 0.5 \\
   0.6 & 0.2 & 0.2
   \end{pmatrix}
   =
   \begin{pmatrix}
   0.2 \\
   0.5 \\
   0.3
   \end{pmatrix}
   $$

3. **Второй переход**:
   Теперь вычисляем вектор вероятностей после второго шага:

   $$
   \mathbf{v_2} = \mathbf{v_1} P =
   \begin{pmatrix}
   0.2 \\
   0.5 \\
   0.3
   \end{pmatrix}
   \begin{pmatrix}
   0.2 & 0.5 & 0.3 \\
   0.1 & 0.4 & 0.5 \\
   0.6 & 0.2 & 0.2
   \end{pmatrix}
   $$

   Расчитаем каждую компоненту:

   - Для состояния $A$:
     $$
     v_{2A} = 0.2 \cdot 0.2 + 0.5 \cdot 0.1 + 0.3 \cdot 0.6 = 0.04 + 0.05 + 0.18 = 0.27
     $$

   - Для состояния $B$:
     $$
     v_{2B} = 0.2 \cdot 0.5 + 0.5 \cdot 0.4 + 0.3 \cdot 0.2 = 0.1 + 0.2 + 0.06 = 0.36
     $$

   - Для состояния $C$:
     $$
     v_{2C} = 0.2 \cdot 0.3 + 0.5 \cdot 0.5 + 0.3 \cdot 0.2 = 0.06 + 0.25 + 0.06 = 0.37
     $$

4. **Ответ**:
   Вероятности состояния цепи после двух шагов:

   $$
   \mathbf{v_2} = \begin{pmatrix} 0.27 \\ 0.36 \\ 0.37 \end{pmatrix}
   $$



## Задача 3: Достижимость состояний

**Условие**: Рассмотрите марковскую цепь с состояниями $S = \{1, 2, 3, 4\}$ и следующей матрицей переходов:

$$
P =
\begin{pmatrix}
0.1 & 0.2 & 0.7 & 0.0 \\
0.0 & 0.5 & 0.5 & 0.0 \\
0.4 & 0.0 & 0.0 & 0.6 \\
0.0 & 0.0 & 1.0 & 0.0
\end{pmatrix}
$$

Определите, достижимо ли состояние 3 из состояния 1.

### Решение:

1. **Проверка достижимости**:
   Состояние $j$ достижимо из состояния $i$, если существует последовательность переходов, ведущая от $i$ к $j$. Для этого мы можем использовать матрицу переходов.

2. **Переходы**:
   Рассмотрим первый переход из состояния 1:

   - Из состояния 1 можно перейти в состояния 1, 2 и 3. Вероятность перехода в состояние 3 составляет 0.7.

3. **Состояние 3**:
   Из состояния 1 в состояние 3 можно перейти сразу за один шаг. Это означает, что состояние 3 достижимо из состояния 1.

4. **Проверка через два шага**:
   Проверим, возможно ли достижение состояния 3 через другие состояния:

   - Из состояния 1 можно попасть в состояние 2 (вероятность 0.2) или оставаться

 в состоянии 1 (вероятность 0.1).

   Из состояния 2:
   - Из состояния 2 можно перейти в состояния 2 (вероятность 0.5) или в состояние 3 (вероятность 0.5).

   Таким образом, состояние 3 достижимо через состояние 2.

5. **Ответ**:
   Состояние 3 достижимо из состояния 1.

Эти три задачи и их подробные решения помогают лучше понять основные концепции и методы работы с марковскими процессами и марковскими цепями.





Давайте реализуем три задачи, которые мы обсудили ранее, на Python с использованием библиотек `NumPy` и `Matplotlib` для визуализации.

### Задача 1: Стационарное распределение

Мы найдем стационарное распределение для марковской цепи с заданной матрицей переходов.

### Задача 2: Анализ переходов в марковской цепи

В этой задаче мы вычислим вероятности состояний цепи после двух шагов.

### Задача 3: Достижимость состояний

Мы проверим, достижимо ли состояние 3 из состояния 1.

Для реализации этих задач, сначала необходимо установить библиотеки. Вы можете сделать это с помощью следующих команд:

Теперь давайте перейдем к коду:


````python
import numpy as np
import matplotlib.pyplot as plt

# Задача 1: Стационарное распределение

# Определим матрицу переходов
P = np.array([[0.3, 0.7],
              [0.4, 0.6]])

# Определение стационарного распределения
def stationary_distribution(P):
    # Система уравнений
    n = P.shape[0]
    A = np.vstack((P.T - np.eye(n), np.ones(n)))
    b = np.zeros(n + 1)
    b[-1] = 1
    return np.linalg.lstsq(A, b, rcond=None)[0]

pi = stationary_distribution(P)
print("Стационарное распределение:", pi)

# Визуализация
states = ['Состояние 1', 'Состояние 2']
plt.bar(states, pi, color=['blue', 'orange'])
plt.ylabel('Вероятность')
plt.title('Стационарное распределение')
plt.show()

# Задача 2: Анализ переходов в марковской цепи

# Определим новую матрицу переходов
P2 = np.array([[0.2, 0.5, 0.3],
               [0.1, 0.4, 0.5],
               [0.6, 0.2, 0.2]])

# Начальное состояние
v0 = np.array([1, 0, 0])  # Начинаем из состояния A

# Расчет вероятностей после двух шагов
v1 = np.dot(v0, P2)  # После 1 шага
v2 = np.dot(v1, P2)  # После 2 шагов

print("Вероятности после двух шагов:", v2)

# Визуализация
states2 = ['Состояние 1', 'Состояние 2', 'Состояние 3']
plt.bar(states2, v2, color=['green', 'yellow', 'red'])
plt.ylabel('Вероятность')
plt.title('Вероятности состояний после двух шагов')
plt.show()

# Задача 3: Достижимость состояний

# Определим матрицу переходов для третьей задачи
P3 = np.array([[0.1, 0.2, 0.7, 0.0],
                [0.0, 0.5, 0.5, 0.0],
                [0.4, 0.0, 0.0, 0.6],
                [0.0, 0.0, 1.0, 0.0]])

# Проверка достижимости
def check_reachability(P, start_state, target_state):
    n = P.shape[0]
    reachable = np.zeros(n, dtype=bool)
    reachable[start_state] = True

    # Итерируемся по возможным переходам
    for _ in range(n):
        new_reachable = reachable.copy()
        for i in range(n):
            if reachable[i]:
                new_reachable |= (P[i] > 0)
        reachable = new_reachable

    return reachable[target_state]

start = 0  # Состояние 1 (индекс 0)
target = 2  # Состояние 3 (индекс 2)
is_reachable = check_reachability(P3, start, target)

if is_reachable:
    print(f"Состояние {target + 1} достижимо из состояния {start + 1}.")
else:
    print(f"Состояние {target + 1} не достижимо из состояния {start + 1}.")
````

1. **Задача 1**: Определите, является ли данная последовательность случайных величин марковским процессом: $X_1, X_2, X_3$ с вероятностями $P(X_3 = x | X_2 = x_2, X_1 = x_1) = P(X_3 = x | X_2 = x_2)$.
  
2. **Задача 2**: Опишите марковский процесс с двумя состояниями и вероятностями перехода:
   - $P_{00} = 0.8, P_{01} = 0.2, P_{10} = 0.4, P_{11} = 0.6$.

3. **Задача 3**: Для матрицы переходов $P = \begin{pmatrix} 0.5 & 0.5 \\ 0.2 & 0.8 \end{pmatrix}$ найдите вероятности перехода после двух шагов, начиная из состояния 1.

4. **Задача 4**: Дайте определение стационарного распределения. Какое значение имеет это распределение для марковской цепи с бесконечным состоянием?

5. **Задача 5**: Проверьте, является ли вектор $\pi = \begin{pmatrix} 0.4 & 0.6 \end{pmatrix}$ стационарным распределением для матрицы переходов $P = \begin{pmatrix} 0.3 & 0.7 \\ 0.6 & 0.4 \end{pmatrix}$.

### Задачи на расчёт

   \end{pmatrix}

   найдите вероятности, если начальное состояние равно 2.

   вычислите распределение после 3-х шагов, начиная с состояния 1.

9. **Задача 9**: Определите, являются ли состояния $A$ и $B$ рекуррентными или транзиторными в следующей марковской цепи:

   \end{pmatrix}
   $$
