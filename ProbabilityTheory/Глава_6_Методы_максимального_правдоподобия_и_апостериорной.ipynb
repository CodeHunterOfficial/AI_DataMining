## Глава 6. Методы максимального правдоподобия и апостериорной вероятности

### Введение

Методы статистической оценки играют важную роль в теории вероятностей и статистике. Среди них два наиболее распространенных подхода — это метод максимального правдоподобия (MLE) и байесовский подход с использованием апостериорной вероятности. В этой лекции мы рассмотрим каждую из этих методик, их математические основы и применение на практике.

### 1. Оценка максимального правдоподобия (MLE)

**Определение**

Метод максимального правдоподобия (MLE) является одним из самых популярных методов статистической оценки. Он используется для нахождения значений параметров модели, которые максимизируют функцию правдоподобия.

**Функция правдоподобия**

Пусть у нас есть выборка данных $x_1, x_2, \ldots, x_n$ из некоторого распределения с параметрами $\theta$. Функция правдоподобия $L(\theta)$ определяется как:

$$
L(\theta) = P(X = x_1, X = x_2, \ldots, X = x_n \mid \theta)
$$

Если данные независимы и идентично распределены (iid), то:

$$
L(\theta) = \prod_{i=1}^{n} f(x_i \mid \theta)
$$

где $f(x \mid \theta)$ — функция плотности вероятности.

**Логарифм правдоподобия**

Для удобства вычислений, часто используют логарифм функции правдоподобия:

$$
\ell(\theta) = \log L(\theta) = \sum_{i=1}^{n} \log f(x_i \mid \theta)
$$

**Оценка MLE**

Чтобы найти оценку максимального правдоподобия, необходимо максимизировать логарифм правдоподобия. Это достигается путем нахождения производной и приравнивания её к нулю:

$$
\frac{d\ell(\theta)}{d\theta} = 0
$$

Решение этого уравнения даст значение $\hat{\theta}$, которое и будет оценкой максимального правдоподобия.

### 2. Метод моментов

**Определение**

Метод моментов — это ещё один способ оценки параметров, основанный на равенстве выборочных и теоретических моментов.

**Моменты**

- **Первый момент** (математическое ожидание):
  $$
  \mu = E[X]
  $$
- **Второй момент**:
  $$
  E[X^2]
  $$
- **Дисперсия**:
  $$
  \sigma^2 = E[X^2] - (E[X])^2
  $$

**Применение метода моментов**

1. Вычислите выборочные моменты:
   - Выборочный первый момент:
   $$
   \hat{\mu} = \frac{1}{n} \sum_{i=1}^{n} x_i
   $$

2. Установите равенство между выборочными и теоретическими моментами для определения параметров.

### Пример

Рассмотрим случай, когда $X$ — нормально распределенная случайная величина $N(\mu, \sigma^2)$.

- Для первого момента:
  $$
  \hat{\mu} = \bar{x}
  $$

- Для второго момента:
  $$
  E[X^2] = \sigma^2 + \mu^2
  $$

Таким образом, выбираем $\hat{\sigma}^2$ следующим образом:
$$
\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} x_i^2 - \hat{\mu}^2
$$

### 3. Байесовский подход и апостериорное распределение

**Определение**

Байесовский подход использует теорему Байеса для оценки параметров на основе апостериорной вероятности.

**Теорема Байеса**

$$
P(\theta \mid x) = \frac{P(x \mid \theta) P(\theta)}{P(x)}
$$

где:
- $P(\theta \mid x)$ — апостериорная вероятность;
- $P(x \mid \theta)$ — правдоподобие;
- $P(\theta)$ — априорная вероятность;
- $P(x)$ — маргинальная вероятность данных.

**Апостериорное распределение**

Рассмотрим апостериорное распределение для параметра $\theta$:

$$
P(\theta \mid x) \propto P(x \mid \theta) P(\theta)
$$

Таким образом, апостериорное распределение пропорционально произведению функции правдоподобия и априорного распределения.

### Пример

Допустим, у нас есть данные, которые следуют нормальному распределению $N(\mu, \sigma^2)$, и мы выбираем априорное распределение для $\mu$ как нормальное $N(\mu_0, \sigma_0^2)$.

**Априорное распределение**:
$$
P(\mu) = \frac{1}{\sqrt{2 \pi \sigma_0^2}} \exp\left(-\frac{(\mu - \mu_0)^2}{2 \sigma_0^2}\right)
$$

**Правдоподобие**:
$$
P(x \mid \mu) = \prod_{i=1}^{n} \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{(x_i - \mu)^2}{2 \sigma^2}\right)
$$

### Итоговое апостериорное распределение

Скомбинировав, получаем:

$$
P(\mu \mid x) \propto P(x \mid \mu) P(\mu)
$$

### Конкретные числовые примеры

#### Пример 1: Метод максимального правдоподобия

Допустим, у нас есть выборка: $x = [2.1, 2.5, 1.8, 2.2, 2.0]$, и мы предполагаем, что $X \sim N(\mu, \sigma^2)$.

1. **Оценка $\hat{\mu}$**:
   $$
   \hat{\mu} = \frac{2.1 + 2.5 + 1.8 + 2.2 + 2.0}{5} = \frac{10.0}{5} = 2.0
   $$

2. **Оценка $\hat{\sigma}^2$**:
   $$
   E[X^2] = \frac{(2.1^2 + 2.5^2 + 1.8^2 + 2.2^2 + 2.0^2)}{5} = \frac{(4.41 + 6.25 + 3.24 + 4.84 + 4)}{5} = \frac{22.74}{5} = 4.548
   $$

   $$
   \hat{\sigma}^2 = 4.548 - (2.0^2) = 4.548 - 4 = 0.548
   $$

#### Пример 2: Байесовский подход

Предположим, что мы имеем те же данные, но теперь мы используем априорное распределение для $\mu$: $N(2, 0.5^2)$.

1. **Априорное распределение**:
   $$
   P(\mu) \sim N(2, 0.25)
   $$

2. **Правдоподобие**:
   Используя формулы правдоподобия, мы получаем:

   $$
   P(x \mid \mu) \propto \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
   $$

3. **Апостериорное распределение**:
   Скомбинировав результаты, получаем:

   $$
   P(\mu \mid x) \propto P(x \mid \mu) P(\mu)
   $$

   В результате апостериорное распределение $\mu$ будет также нормальным, с новыми параметрами, которые можно вычислить на основе данных.

### Заключение

Методы максимального правдоподобия и байесовский подход представляют собой мощные инструменты для оценки параметров статистических моделей. Они имеют свои сильные и слабые стороны, и выбор метода часто зависит от конкретной задачи и доступной информации о параметрах.


# Задачи
### Задача 1: Оценка параметров нормально распределенной выборки

**Условие:** Пусть у нас есть выборка из нормально распределенной случайной величины:  
$x = [4.8, 5.1, 5.3, 4.9, 5.0]$. Найдите оценки параметров $\mu$ и $\sigma^2$ методом максимального правдоподобия.

#### Решение:

1. **Определение функции правдоподобия:**

   Функция правдоподобия для нормально распределенной случайной величины $X \sim N(\mu, \sigma^2)$ имеет вид:

   $$
   L(\mu, \sigma^2) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right)
   $$

   Для данной выборки:

   $$
   L(\mu, \sigma^2) = \frac{1}{(2\pi\sigma^2)^{n/2}} \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^{n} (x_i - \mu)^2\right)
   $$

   где $n = 5$.

2. **Логарифм функции правдоподобия:**

   Для удобства вычислений используем логарифм функции правдоподобия:

   $$
   \ell(\mu, \sigma^2) = \log L(\mu, \sigma^2) = -\frac{n}{2} \log(2\pi) - \frac{n}{2} \log(\sigma^2) - \frac{1}{2\sigma^2} \sum_{i=1}^{n} (x_i - \mu)^2
   $$

3. **Нахождение производных:**

   Теперь найдем частные производные логарифма функции правдоподобия по $\mu$ и $\sigma^2$ и приравняем их к нулю.

   - **Частная производная по $\mu$**:

   $$
   \frac{\partial \ell}{\partial \mu} = \frac{1}{\sigma^2} \sum_{i=1}^{n} (x_i - \mu) = 0
   $$

   Отсюда:

   $$
   \sum_{i=1}^{n} x_i = n\mu \implies \hat{\mu} = \frac{1}{n} \sum_{i=1}^{n} x_i
   $$

   Подставим значения:

   $$
   \hat{\mu} = \frac{4.8 + 5.1 + 5.3 + 4.9 + 5.0}{5} = \frac{25.1}{5} = 5.02
   $$

   - **Частная производная по $\sigma^2$**:

   $$
   \frac{\partial \ell}{\partial \sigma^2} = -\frac{n}{2\sigma^2} + \frac{1}{2\sigma^4} \sum_{i=1}^{n} (x_i - \mu)^2 = 0
   $$

   Умножив на $2\sigma^4$:

   $$
   -n\sigma^2 + \sum_{i=1}^{n} (x_i - \mu)^2 = 0
   $$

   Таким образом:

   $$
   \hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \hat{\mu})^2
   $$

   Рассчитаем $\hat{\sigma}^2$:

   $$
   \hat{\sigma}^2 = \frac{(4.8 - 5.02)^2 + (5.1 - 5.02)^2 + (5.3 - 5.02)^2 + (4.9 - 5.02)^2 + (5.0 - 5.02)^2}{5}
   $$

   $$
   = \frac{(0.22)^2 + (0.08)^2 + (0.28)^2 + (0.12)^2 + (0.02)^2}{5}
   $$

   $$
   = \frac{0.0484 + 0.0064 + 0.0784 + 0.0144 + 0.0004}{5} = \frac{0.148}{5} = 0.0296
   $$

4. **Ответ:**

   Оценки параметров:
   - $\hat{\mu} = 5.02$
   - $\hat{\sigma}^2 = 0.0296$



### Задача 2: Оценка параметров экспоненциального распределения

**Условие:** Пусть выборка из экспоненциального распределения:  
$x = [1, 2, 3, 4, 2]$. Найдите оценку параметра $\lambda$ методом максимального правдоподобия.

#### Решение:

1. **Определение функции правдоподобия:**

   Для экспоненциального распределения $X \sim \text{Exp}(\lambda)$:

   $$
   L(\lambda) = \prod_{i=1}^{n} \lambda e^{-\lambda x_i} = \lambda^n e^{-\lambda \sum_{i=1}^{n} x_i}
   $$

   где $n = 5$.

2. **Логарифм функции правдоподобия:**

   $$
   \ell(\lambda) = \log L(\lambda) = n \log \lambda - \lambda \sum_{i=1}^{n} x_i
   $$

3. **Нахождение производной:**

   Найдем производную по $\lambda$ и приравняем её к нулю:

   $$
   \frac{\partial \ell}{\partial \lambda} = \frac{n}{\lambda} - \sum_{i=1}^{n} x_i = 0
   $$

   Отсюда:

   $$
   \hat{\lambda} = \frac{n}{\sum_{i=1}^{n} x_i}
   $$

   Рассчитаем $\sum_{i=1}^{n} x_i$:

   $$
   \sum_{i=1}^{5} x_i = 1 + 2 + 3 + 4 + 2 = 12
   $$

   Подставляем:

   $$
   \hat{\lambda} = \frac{5}{12} \approx 0.4167
   $$

4. **Ответ:**

   Оценка параметра $\lambda$:
   - $\hat{\lambda} \approx 0.4167$


### Задача 3: Байесовская оценка параметров нормального распределения

**Условие:** Пусть выборка из нормально распределенной случайной величины:  
$x = [3.5, 3.7, 3.9, 4.0]$. Предположим, что априорное распределение для $\mu$ имеет вид $N(3.5, 0.2^2)$. Найдите апостериорное распределение для параметра $\mu$.

#### Решение:

1. **Определение априорного распределения:**

   Априорное распределение для $\mu$:

   $$
   P(\mu) \sim N(3.5, 0.04) \quad (\sigma^2 = 0.04)
   $$

2. **Определение функции правдоподобия:**

   Функция правдоподобия для выборки $x$:

   $$
   P(x \mid \mu) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right)
   $$

   Используем $\sigma^2 = 0.2^2 = 0.04$.

   $$
   P(x \mid \mu) \propto \exp\left(-\frac{1}{2 \cdot 0.04} \sum_{i=1}^{n} (x_i - \mu)^2\right)
   $$

3. **Нахождение параметров:**

   - Найдем $\hat{\mu}$ и $\hat{\sigma}^2$:

   $$
   \hat{\mu} = \frac{3.5 + 3.7 + 3.9 + 4.0}{4} = \frac{15.1}{4} = 3.775
   $$

  

 $$
   \hat{\sigma}^2 = \frac{(3.5 - 3.775)^2 + (3.7 - 3.775)^2 + (3.9 - 3.775)^2 + (4.0 - 3.775)^2}{4}
   $$

   $$
   = \frac{(0.275)^2 + (0.075)^2 + (0.125)^2 + (0.225)^2}{4}
   $$

   $$
   = \frac{0.075625 + 0.005625 + 0.015625 + 0.050625}{4} = \frac{0.1475}{4} = 0.036875
   $$

4. **Вывод апостериорного распределения:**

   Апостериорное распределение для $\mu$ будет нормально распределенным:

   $$
   \mu \mid x \sim N\left( \frac{\frac{3.5}{0.04} + \frac{3.775}{0.036875}}{\frac{1}{0.04} + \frac{1}{0.036875}}, \frac{1}{\frac{1}{0.04} + \frac{1}{0.036875}} \right)
   $$

   - Рассчитаем значения:

   $$
   \text{Обозначим } a = \frac{3.5}{0.04}, \quad b = \frac{3.775}{0.036875}
   $$
   
   $$
   \text{И } c = \frac{1}{0.04} + \frac{1}{0.036875}
   $$

   $$
   a = 87.5, \quad b \approx 102.85, \quad c \approx 54.09
   $$

   Подставляя в формулу:

   $$
   \mu \mid x \sim N\left( \frac{87.5 + 102.85}{54.09}, \frac{1}{54.09} \right)
   $$

   Рассчитаем:

   $$
   \hat{\mu}_{\text{апост.}} \approx \frac{190.35}{54.09} \approx 3.52
   $$

   Дисперсия:

   $$
   \hat{\sigma}^2_{\text{апост.}} = \frac{1}{54.09} \approx 0.0185
   $$

5. **Ответ:**

   Апостериорное распределение:

   $$
   \mu \mid x \sim N(3.52, 0.0185)
   $$


Эти задачи и решения показывают, как применять методы максимального правдоподобия и байесовские методы для оценки параметров распределений.




Давайте реализуем задачи на Python с использованием библиотек `numpy`, `scipy`, и `matplotlib` для вычислений и визуализации.

### Задача 1: Оценка параметров нормально распределенной выборки


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Данные
data = np.array([4.8, 5.1, 5.3, 4.9, 5.0])

# Оценка параметров методом максимального правдоподобия
mu_hat = np.mean(data)
sigma_hat = np.var(data)

print(f"Оценка параметра mu: {mu_hat:.2f}")
print(f"Оценка параметра sigma^2: {sigma_hat:.4f}")

# Визуализация
x = np.linspace(4.5, 5.5, 100)
pdf = (1 / (np.sqrt(2 * np.pi * sigma_hat))) * np.exp(-0.5 * ((x - mu_hat) ** 2) / sigma_hat)

plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, stat="density", label='Гистограмма выборки', color='lightgray')
plt.plot(x, pdf, label='PDF N(μ, σ²)', color='red')
plt.axvline(mu_hat, color='blue', linestyle='--', label='μ_hat (Оценка)')
plt.title('Оценка параметров нормально распределенной выборки')
plt.xlabel('x')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.show()
```



### Задача 2: Оценка параметров экспоненциального распределения


```python
# Данные
data_exp = np.array([1, 2, 3, 4, 2])

# Оценка параметра lambda методом максимального правдоподобия
lambda_hat = len(data_exp) / np.sum(data_exp)

print(f"Оценка параметра lambda: {lambda_hat:.4f}")

# Визуализация
x_exp = np.linspace(0, 6, 100)
pdf_exp = lambda_hat * np.exp(-lambda_hat * x_exp)

plt.figure(figsize=(10, 6))
sns.histplot(data_exp, kde=False, stat="density", label='Гистограмма выборки', color='lightgray')
plt.plot(x_exp, pdf_exp, label='PDF Exp(λ)', color='green')
plt.axvline(1/lambda_hat, color='blue', linestyle='--', label='λ_hat (Оценка)')
plt.title('Оценка параметра экспоненциального распределения')
plt.xlabel('x')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.show()
```



### Задача 3: Байесовская оценка параметров нормального распределения


```python
import scipy.stats as stats

# Данные
data_bayes = np.array([3.5, 3.7, 3.9, 4.0])
n = len(data_bayes)

# Априорные параметры
mu_prior = 3.5
sigma_prior = 0.2

# Параметры оценок
mu_hat_bayes = np.mean(data_bayes)
sigma_hat_bayes = np.var(data_bayes)

# Вычисление апостериорного распределения
n_prior = 1 / sigma_prior**2
n_likelihood = n / sigma_hat_bayes

mu_post = (mu_prior * n_prior + mu_hat_bayes * n_likelihood) / (n_prior + n_likelihood)
sigma_post = 1 / (n_prior + n_likelihood)

print(f"Априорное распределение: mu = {mu_prior}, sigma^2 = {sigma_prior**2}")
print(f"Постериорное распределение: mu = {mu_post:.4f}, sigma^2 = {sigma_post:.4f}")

# Визуализация
x_bayes = np.linspace(3, 4.5, 100)
pdf_prior = stats.norm.pdf(x_bayes, mu_prior, sigma_prior)
pdf_likelihood = stats.norm.pdf(x_bayes, mu_hat_bayes, np.sqrt(sigma_hat_bayes))
pdf_post = stats.norm.pdf(x_bayes, mu_post, np.sqrt(sigma_post))

plt.figure(figsize=(10, 6))
plt.plot(x_bayes, pdf_prior, label='Априорное распределение', color='orange')
plt.plot(x_bayes, pdf_likelihood, label='Функция правдоподобия', color='purple')
plt.plot(x_bayes, pdf_post, label='Постериорное распределение', color='brown')
plt.axvline(mu_post, color='blue', linestyle='--', label='μ_post (Оценка)')
plt.title('Байесовская оценка параметров нормального распределения')
plt.xlabel('x')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.show()
```
