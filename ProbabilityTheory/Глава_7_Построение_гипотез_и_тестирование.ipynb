# Глава 7. Построение гипотез и тестирование

## Введение
В статистике процесс тестирования гипотез является важным этапом, позволяющим исследователям делать выводы о популяциях на основе выборочных данных. В этой лекции мы рассмотрим ключевые аспекты построения гипотез и тестирования, включая формулирование нулевой и альтернативной гипотез, ошибки первого и второго рода, использование p-value и доверительных интервалов, а также критерии согласия.

## 1. Формулирование нулевой и альтернативной гипотез

### Нулевая гипотеза $(H_0$)
Нулевая гипотеза представляет собой утверждение о том, что не существует эффекта или различия между группами. Обычно она формулируется как равенство, например:

$$
H_0: \mu_1 = \mu_2
$$

где \$mu_1$ и \$mu_2$ — средние значения двух групп.

### Альтернативная гипотеза $(H_a$)
Альтернативная гипотеза — это утверждение, которое противоречит нулевой гипотезе. Она может быть односторонней или двусторонней:

- **Двусторонняя**: \(H_a: \mu_1 \neq \mu_2$
- **Односторонняя**: \(H_a: \mu_1 > \mu_2$ или \(H_a: \mu_1 < \mu_2$

### Пример
Предположим, мы хотим исследовать, влияет ли новый препарат на уровень холестерина. Нулевая гипотеза может звучать как:

$$
H_0: \text{средний уровень холестерина в группе с препаратом} = \text{средний уровень холестерина в контрольной группе}
$$

А альтернативная гипотеза:

$$
H_a: \text{средний уровень холестерина в группе с препаратом} < \text{средний уровень холестерина в контрольной группе}
$$

## 2. Ошибки первого и второго рода

### Ошибка первого рода $$alpha$)
Ошибка первого рода возникает, когда мы отвергаем нулевую гипотезу, когда она на самом деле верна. Уровень значимости \$alpha$ — это вероятность совершения ошибки первого рода. Обычно \$alpha$ устанавливается на уровне 0,05.

### Ошибка второго рода $$beta$)
Ошибка второго рода происходит, когда мы не отвергаем нулевую гипотезу, когда она на самом деле ложна. Вероятность ошибки второго рода обозначается как \$beta$, а мощность теста (вероятность правильного отвергнуть нулевую гипотезу, когда она ложна) равна \(1 - \beta$.

### Пример
Предположим, что нулевая гипотеза \(H_0$ верна. Если мы отказываемся от \(H_0$ (ошибка первого рода), то это значит, что мы утверждаем, что препарат эффективен, хотя это не так. В случае ошибки второго рода мы можем не отвергать \(H_0$ и неверно утверждать, что препарат не эффективен, когда на самом деле он эффективен.

## 3. p-value и доверительные интервалы

### p-value
p-value — это вероятность получения наблюдаемых данных (или более экстремальных) при условии, что нулевая гипотеза верна. Низкое значение p-value (обычно меньше 0,05) говорит о том, что наблюдаемые данные несовместимы с нулевой гипотезой.

### Формула p-value
Для оценки p-value в зависимости от типа теста используется различная статистика:

- Для теста Стьюдента:
$$
t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}
$$
где:
- \$bar{x}$ — выборочное среднее,
- \$mu$ — гипотетическое среднее,
- \(s$ — стандартное отклонение выборки,
- \(n$ — размер выборки.

- Для других тестов может использоваться Z-статистика, аналогичная формуле для t, но с известной дисперсией.

### Доверительные интервалы
Доверительный интервал (ДИ) — это диапазон значений, который с определенной вероятностью содержит истинное значение параметра.

### Формула доверительного интервала для среднего
Для нормального распределения доверительный интервал для среднего значения может быть вычислен по формуле:

$$
\text{ДИ} = \bar{x} \pm z_{\frac{\alpha}{2}} \cdot \frac{s}{\sqrt{n}}
$$

где \(z_{\frac{\alpha}{2}}$ — критическое значение Z, соответствующее уровню значимости.

### Пример
Предположим, что среднее значение уровня холестерина в выборке составляет \$bar{x} = 200$, стандартное отклонение \(s = 30$, размер выборки \(n = 50$, и мы используем уровень значимости \$alpha = 0.05$. Критическое значение \(z_{0.025} \approx 1.96$.

Доверительный интервал будет:

$$
\text{ДИ} = 200 \pm 1.96 \cdot \frac{30}{\sqrt{50}} \approx 200 \pm 8.3
$$

То есть ДИ: (191.7, 208.3).

## 4. Критерии согласия: критерий χ² и тесты Стьюдента

### Критерий согласия χ²
Критерий \$chi^2$ используется для оценки того, насколько хорошо наблюдаемые данные соответствуют ожидаемым. Он применяется в основном для категориальных данных.

### Формула критерия χ²
$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

где:
- \(O_i$ — наблюдаемые частоты,
- \(E_i$ — ожидаемые частоты.

### Пример
Предположим, что у нас есть данные о предпочтениях 100 человек по трем видам напитков (A, B, C):

- Ожидаемое распределение: A - 30, B - 40, C - 30.
- Наблюдаемое распределение: A - 20, B - 50, C - 30.

Рассчитаем \$chi^2$:

$$
\chi^2 = \frac{(20 - 30)^2}{30} + \frac{(50 - 40)^2}{40} + \frac{(30 - 30)^2}{30} = \frac{100}{30} + \frac{100}{40} + 0 = 3.33 + 2.5 + 0 = 5.83
$$

### Тесты Стьюдента
Тест Стьюдента используется для сравнения средних двух групп.

### Формула для t-статистики
$$
t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$

где:
- \$bar{x}_1, \bar{x}_2$ — выборочные средние,
- \(s_1, s_2$ — стандартные отклонения,
- \(n_1, n_2$ — размеры выборок.

### Пример
Допустим, мы имеем две группы с следующими данными:
- Группа 1: \$bar{x}_1 = 100$, \(s_1 = 15$, \(n_1 = 30$
- Группа 2: \$bar{x}_2 = 110$, \(s_2 = 20$, \(n_2 = 30$

Тогда t-статистика будет:

$$
t = \frac{100 - 110}{\sqrt{\frac{15^2}{30} + \frac{20^2}{30}}} = \frac{-10}{\sqrt{\frac{225}{30} + \frac{400}{30}}} = \frac{-10}{\sqrt{7.5 + \frac{400}{30}}} = \frac{-10}{\sqrt{7.5 + 13.33}} = \frac{-10}{\sqrt{20.83}} \approx \frac{-10}{4.57} \approx -2.19
$$

## Заключение
В данной лекции мы рассмотрели основные аспекты построения гипотез и тестирования, включая формулирование гипотез, ошибки первого и второго рода, использование p-value и доверительных интервалов, а также критерии согласия. Тестирование гипотез является мощным инструментом в статистике, позволяющим делать обоснованные выводы на основе данных.

## Конкретные числовые примеры

1. **Пример с гипотезами**:  
   - Нулевая гипотеза: \(H_0: \mu = 200\

)  
   - Альтернативная гипотеза: \(H_a: \mu < 200$  
   - Выборка: \$bar{x} = 195$, \(s = 10$, \(n = 25$  
   - \(t = \frac{195 - 200}{10/\sqrt{25}} = \frac{-5}{2} = -2.5$  
   - p-value (по таблице критических значений) соответствует значению 0.01. Следовательно, если \$alpha = 0.05$, отвергаем \(H_0$.

2. **Пример с p-value и доверительными интервалами**:  
   - Выборка: \$bar{x} = 150$, \(s = 20$, \(n = 36$  
   - Доверительный интервал:  
   \$text{ДИ} = 150 \pm 1.96 \cdot \frac{20}{\sqrt{36}} = 150 \pm 6.53$  
   - ДИ: (143.47, 156.53).

3. **Пример с критерием χ²**:  
   - Наблюдаемые: A - 40, B - 30, C - 30; Ожидаемые: A - 30, B - 40, C - 30.  
   - \$chi^2 = \frac{(40-30)^2}{30} + \frac{(30-40)^2}{40} + \frac{(30-30)^2}{30} = \frac{100}{30} + \frac{100}{40} = 3.33 + 2.5 = 5.83$.

4. **Пример с тестом Стьюдента**:  
   - Группа 1: \$bar{x}_1 = 120$, \(s_1 = 10$, \(n_1 = 25$  
   - Группа 2: \$bar{x}_2 = 115$, \(s_2 = 15$, \(n_2 = 25$  
   - \(t = \frac{120 - 115}{\sqrt{\frac{10^2}{25} + \frac{15^2}{25}}} = \frac{5}{\sqrt{4 + 9}} = \frac{5}{\sqrt{13}} \approx 1.39$.

Эти примеры иллюстрируют применение методов тестирования гипотез в реальных сценариях и дают понимание о том, как использовать статистику для принятия обоснованных решений.

## Задача 1: Тестирование средней оценки

### Условия задачи
У нас есть выборка из 30 студентов, которые прошли тест по математике. Средняя оценка в выборке составила 78, стандартное отклонение — 10. Проверим, является ли средняя оценка студентов, прошедших тест, статистически значимо выше 75.

### Шаг 1: Формулировка гипотез
- Нулевая гипотеза $(H_0$): \$mu = 75$ (средняя оценка равна 75).
- Альтернативная гипотеза $(H_a$): \$mu > 75$ (средняя оценка больше 75).

### Шаг 2: Определение уровня значимости
Установим уровень значимости \$alpha = 0.05$.

### Шаг 3: Расчет t-статистики
Используем формулу для t-статистики:

$$
t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}
$$

где:
- \$bar{x} = 78$ (средняя оценка),
- \$mu = 75$ (гипотетическое среднее),
- \(s = 10$ (стандартное отклонение),
- \(n = 30$ (размер выборки).

Подставляем значения:

$$
t = \frac{78 - 75}{\frac{10}{\sqrt{30}}} = \frac{3}{\frac{10}{5.477}} = \frac{3}{1.8257} \approx 1.64
$$

### Шаг 4: Определение критического значения
Для одностороннего теста с \$alpha = 0.05$ и \(n - 1 = 29$ степенями свободы находим критическое значение t из таблицы:

$$
t_{крит} \approx 1.699
$$

### Шаг 5: Сравнение t-статистики с критическим значением
Сравниваем \(t$ со значением \(t_{крит}$:

- \(t \approx 1.64 < t_{крит} \approx 1.699$

### Шаг 6: Вывод
Не отклоняем нулевую гипотезу. Нет достаточных оснований утверждать, что средняя оценка студентов выше 75.



## Задача 2: Тестирование разности средних

### Условия задачи
У нас есть две группы пациентов, которые принимают разные лекарства. Группа 1 (15 пациентов) имеет среднюю оценку улучшения состояния 80 (стандартное отклонение 10), а группа 2 (20 пациентов) имеет среднюю оценку 75 (стандартное отклонение 12). Проверим, есть ли статистически значимая разница между группами.

### Шаг 1: Формулировка гипотез
- Нулевая гипотеза $(H_0$): \$mu_1 = \mu_2$ (разницы между средними нет).
- Альтернативная гипотеза $(H_a$): \$mu_1 \neq \mu_2$ (разница между средними есть).

### Шаг 2: Определение уровня значимости
Установим уровень значимости \$alpha = 0.05$.

### Шаг 3: Расчет t-статистики
Используем формулу для t-статистики для независимых выборок:

$$
t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$

где:
- \$bar{x}_1 = 80$,
- \$bar{x}_2 = 75$,
- \(s_1 = 10$,
- \(s_2 = 12$,
- \(n_1 = 15$,
- \(n_2 = 20$.

Подставляем значения:

$$
t = \frac{80 - 75}{\sqrt{\frac{10^2}{15} + \frac{12^2}{20}}} = \frac{5}{\sqrt{\frac{100}{15} + \frac{144}{20}}}
$$

Расчитаем:

$$
t = \frac{5}{\sqrt{6.67 + 7.2}} = \frac{5}{\sqrt{13.87}} \approx \frac{5}{3.72} \approx 1.34
$$

### Шаг 4: Определение критического значения
Находим критические значения t для двухстороннего теста с \$alpha = 0.05$ и \(df = n_1 + n_2 - 2 = 33$.

Находим по таблице:

$$
t_{крит} \approx 2.034
$$

### Шаг 5: Сравнение t-статистики с критическим значением
Сравниваем:

- \(t \approx 1.34 < t_{крит} \approx 2.034$

### Шаг 6: Вывод
Не отклоняем нулевую гипотезу. Нет статистически значимой разницы между группами.



## Задача 3: Тестирование на согласие

### Условия задачи
Имеются наблюдаемые частоты по трем категориям: A (30), B (50), C (20). Ожидаемые частоты: A (40), B (40), C (20). Проверим, согласуются ли наблюдаемые данные с ожидаемыми.

### Шаг 1: Формулировка гипотез
- Нулевая гипотеза $(H_0$): наблюдаемые частоты согласуются с ожидаемыми.
- Альтернативная гипотеза $(H_a$): наблюдаемые частоты не согласуются с ожидаемыми.

### Шаг 2: Расчет χ²
Используем формулу для критерия согласия:

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

где \(O_i$ — наблюдаемые частоты, \(E_i$ — ожидаемые частоты.

Подставляем значения:

$$
\chi^2 = \frac{(30 - 40)^2}{40} + \frac{(50 - 40)^2}{40} + \frac{(20 - 20)^2}{20}
$$

Расчитаем:

$$
\chi^2 = \frac{100}{40} + \frac{100}{40} + 0 = 2.5 + 2.5 + 0 = 5.0
$$

### Шаг 3: Определение критического значения
Определим критическое значение χ² для \$alpha = 0.05$ и \(k - 1 = 3 - 1 = 2$ степенями свободы:

$$
\chi^2_{крит} \approx 5.991
$$

### Шаг 4: Сравнение χ² с критическим значением
Сравниваем:

- \$chi^2 = 5.0 < \chi^2_{крит} = 5.991$

### Шаг 5: Вывод
Не отклоняем нулевую гипотезу. Наблюдаемые частоты согласуются с ожидаемыми.



Эти задачи показывают, как пошагово решать задачи, используя методы тестирования гипотез, и помогают закрепить материал.





Для реализации задач на Python с визуализацией мы будем использовать библиотеки `scipy` для статистических расчетов, `matplotlib` для построения графиков и `numpy` для работы с массивами данных. В каждой задаче мы покажем результаты в виде текстового вывода и графиков.



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Задача 1: Тестирование средней оценки
def task1():
    # Данные
    n1 = 30
    x_bar1 = 78
    mu1 = 75
    s1 = 10
    alpha = 0.05

    # Расчет t-статистики
    t_stat1 = (x_bar1 - mu1) / (s1 / np.sqrt(n1))

    # Критическое значение t
    t_critical1 = stats.t.ppf(1 - alpha, n1 - 1)

    # Вывод результатов
    print(f"Задача 1: t-статистика = {t_stat1:.2f}, Критическое значение = {t_critical1:.2f}")

    # Визуализация
    plt.figure(figsize=(10, 6))
    x = np.linspace(-4, 4, 100)
    plt.plot(x, stats.t.pdf(x, n1 - 1), label='t-распределение', color='blue')
    plt.axvline(t_stat1, color='red', linestyle='--', label='t-статистика')
    plt.axvline(t_critical1, color='green', linestyle='--', label='Критическое значение')
    plt.title('Задача 1: Тестирование средней оценки')
    plt.xlabel('t-значения')
    plt.ylabel('Плотность вероятности')
    plt.legend()
    plt.grid()
    plt.show()

# Задача 2: Тестирование разности средних
def task2():
    # Данные
    n1, n2 = 15, 20
    x_bar1, x_bar2 = 80, 75
    s1, s2 = 10, 12
    alpha = 0.05

    # Расчет t-статистики
    t_stat2 = (x_bar1 - x_bar2) / np.sqrt((s1**2 / n1) + (s2**2 / n2))

    # Критическое значение t
    df = n1 + n2 - 2
    t_critical2 = stats.t.ppf(1 - alpha / 2, df)

    # Вывод результатов
    print(f"Задача 2: t-статистика = {t_stat2:.2f}, Критическое значение = ±{t_critical2:.2f}")

    # Визуализация
    plt.figure(figsize=(10, 6))
    x = np.linspace(-4, 4, 100)
    plt.plot(x, stats.t.pdf(x, df), label='t-распределение', color='blue')
    plt.axvline(t_stat2, color='red', linestyle='--', label='t-статистика')
    plt.axvline(t_critical2, color='green', linestyle='--', label='Критические значения')
    plt.axvline(-t_critical2, color='green', linestyle='--')
    plt.title('Задача 2: Тестирование разности средних')
    plt.xlabel('t-значения')
    plt.ylabel('Плотность вероятности')
    plt.legend()
    plt.grid()
    plt.show()

# Задача 3: Тестирование на согласие
def task3():
    # Наблюдаемые и ожидаемые частоты
    observed = np.array([30, 50, 20])
    expected = np.array([40, 40, 20])

    # Расчет χ²
    chi2_stat = np.sum((observed - expected) ** 2 / expected)

    # Степени свободы
    df = len(observed) - 1

    # Критическое значение χ²
    chi2_critical = stats.chi2.ppf(0.95, df)

    # Вывод результатов
    print(f"Задача 3: χ²-статистика = {chi2_stat:.2f}, Критическое значение = {chi2_critical:.2f}")

    # Визуализация
    x = np.linspace(0, 15, 100)
    plt.figure(figsize=(10, 6))
    plt.plot(x, stats.chi2.pdf(x, df), label='χ²-распределение', color='blue')
    plt.axvline(chi2_stat, color='red', linestyle='--', label='χ²-статистика')
    plt.axvline(chi2_critical, color='green', linestyle='--', label='Критическое значение')
    plt.title('Задача 3: Тестирование на согласие')
    plt.xlabel('χ²-значения')
    plt.ylabel('Плотность вероятности')
    plt.legend()
    plt.grid()
    plt.show()

# Выполнение задач
task1()
task2()
task3()
```
