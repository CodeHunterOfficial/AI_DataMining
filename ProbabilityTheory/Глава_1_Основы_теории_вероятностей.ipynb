### Пространство элементарных событий

Пространство элементарных событий (или пространство исходов) $\Omega$ представляет собой множество всех возможных исходов некоторого случайного эксперимента. Эксперимент может иметь один или более исходов, и каждый из них является элементарным событием. Примерами могут служить бросание монеты, бросание кубика или измерение случайной величины.

### События и их вероятность

Событие $A$ — это подмножество пространства элементарных событий $\Omega$. Вероятность события $A$, обозначаемая $P(A)$, измеряет вероятность того, что событие $A$ произойдет при выполнении случайного эксперимента.

#### Основные аксиомы теории вероятностей

1. **Неотрицательность вероятности:**
   $$ P(A) \geq 0 $$
   Вероятность любого события не может быть отрицательной.

2. **Нормировка:**
   $$ P(\Omega) = 1 $$
   Вероятность всего пространства элементарных событий равна 1.

3. **Аддитивность:**
   Если события $A$ и $B$ несовместны (их пересечение пусто), то вероятность их объединения равна сумме вероятностей:
   $$ P(A \cup B) = P(A) + P(B) $$

#### Вычисление вероятности

Вероятность события $A$ может быть вычислена через отношение количества благоприятных исходов к общему числу исходов:

$$ P(A) = \frac{\text{Число благоприятных исходов для } A}{\text{Общее число исходов}} $$

### Пример

Представим, что у нас есть справедливая монета, которую мы бросаем. Пространство элементарных событий $\Omega = \{Орел, Решка\}$. Пусть событие $A$ состоит в том, чтобы выпал орел.

В этом случае, вероятность события $A$ будет:
$$ P(A) = \frac{\text{Число благоприятных исходов для } A}{\text{Общее число исходов}} = \frac{1}{2} $$

Таким образом, вероятность выпадения орла при однократном броске справедливой монеты составляет $\frac{1}{2}$, что эквивалентно 50%.

Это основные концепции теории вероятностей, которые позволяют нам математически определять вероятности различных событий в случайных экспериментах.

# Аксиомы вероятности

## Введение в теорию вероятностей

Теория вероятностей — это математическая дисциплина, изучающая случайные явления и закономерности, связанные с ними. Основные концепции вероятности формализуются через аксиомы, которые служат основой для дальнейших исследований и приложений.

### Определение вероятности

Вероятность события $A$ (обозначается как $P(A)$) — это числовая мера, которая показывает, насколько вероятно это событие. Вероятность принимает значения в диапазоне от 0 до 1:

$$
0 \leq P(A) \leq 1
$$

Если вероятность равна 0, событие невозможно, если равна 1 — событие обязательно произойдет.

## Аксиомы вероятности

В математической теории вероятностей аксиомы формулируются следующим образом, согласно работам Андрея Колмогорова:

### Аксиома 1: Неположительность вероятности

Вероятность любого события $A$ не может быть отрицательной:

$$
P(A) \geq 0
$$

### Аксиома 2: Полная вероятность

Вероятность, что произойдет хотя бы одно из возможных событий, равна 1. Если $S$ — это пространство элементарных событий (все возможные исходы), то:

$$
P(S) = 1
$$

### Аксиома 3: Аддитивность вероятностей

Если $A_1, A_2, \ldots, A_n$ — это несовместные события (то есть события, которые не могут произойти одновременно), то вероятность их объединения равна сумме вероятностей этих событий:

$$
P(A_1 \cup A_2 \cup \ldots \cup A_n) = P(A_1) + P(A_2) + \ldots + P(A_n)
$$

## Связанные понятия

### Элементарное событие

Элементарное событие — это одно конкретное исходное состояние в пространстве $S$. Например, при броске монеты элементарные события: "орел" и "решка".

### Событие

Событие — это любое подмножество пространства элементарных событий. Например, событие $A$ может быть "выпадение орла при броске монеты".

### Примеры

1. **Пример с игральной костью**:
    - Пространство элементарных событий: $S = \{1, 2, 3, 4, 5, 6\}$.
    - Событие $A$: "выпадает четное число". Тогда $A = \{2, 4, 6\}$.
    - Вероятность события $A$:
    
    $$
    P(A) = \frac{|A|}{|S|} = \frac{3}{6} = \frac{1}{2}
    $$

2. **Пример с выбором карты из колоды**:
    - Пространство элементарных событий: $S$ — все 52 карты.
    - Событие $B$: "вытащить туза". Тогда $B = \{ \text{туз червей, туз бубен, туз треф, туз пик} \}$, и $|B| = 4$.
    - Вероятность события $B$:
    
    $$
    P(B) = \frac{|B|}{|S|} = \frac{4}{52} = \frac{1}{13}
    $$

### Формулы для комбинирования событий

- **Вероятность объединения двух событий**:

Если события $A$ и $B$ не являются несовместными, то:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

- **Вероятность пересечения двух событий**:

Если события $A$ и $B$ независимы (то есть вероятность наступления одного события не влияет на вероятность наступления другого), то:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

## Числовой пример

Рассмотрим задачу о броске двух игральных костей.

1. Определим пространство элементарных событий. Каждая кость имеет 6 граней, поэтому общее число исходов:
$$
|S| = 6 \times 6 = 36
$$

2. Определим событие $C$: "сумма чисел на костях равна 7". Возможные сочетания:
   - (1, 6)
   - (2, 5)
   - (3, 4)
   - (4, 3)
   - (5, 2)
   - (6, 1)

Итого 6 благоприятных исходов.

3. Вероятность события $C$:
$$
P(C) = \frac{|C|}{|S|} = \frac{6}{36} = \frac{1}{6}
$$

Таким образом, вероятность того, что сумма чисел на двух брошенных костях равна 7, составляет $\frac{1}{6}$.

## Пример 5: Определение вероятности совместных событий

### Условие
Бросаем одну игральную кость и вытаскиваем карту из колоды. Найдем вероятность того, что на кости выпадает 2, а на карте — туз.

### Решение

1. **Определим события**:
   - Событие $E$: "выпадает 2 на кости".
   - Событие $F$: "вытаскивается туз".

2. **Вычислим вероятности отдельных событий**:
   - Вероятность события $E$:
   $$
   P(E) = \frac{1}{6} \quad (\text{т.к. 1 из 6 возможных исходов})
   $$

   - Вероятность события $F$:
   $$
   P(F) = \frac{4}{52} = \frac{1}{13} \quad (\text{т.к. 4 туза из 52 карт})

# Условная вероятность и формула Байеса

## Введение в условную вероятность

**Условная вероятность** — это вероятность события, при условии, что произошло другое событие. Если события $A$ и $B$ являются двумя событиями, то условная вероятность события $A$ при условии, что произошло событие $B$, обозначается как $P(A|B)$ и вычисляется по следующей формуле:

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}, \quad P(B) > 0
$$

где:
- $P(A|B)$ — условная вероятность события $A$ при условии, что событие $B$ произошло.
- $P(A \cap B)$ — вероятность того, что оба события $A$ и $B$ произошли.
- $P(B)$ — вероятность события $B$.

### Пример 1: Условная вероятность

Рассмотрим следующий пример:

**Задача:** В классе 30 студентов, из которых 18 — мальчики и 12 — девочки. Если известно, что выбранный студент — мальчик, какова вероятность того, что он играет на музыкальном инструменте, если известно, что 10 мальчиков играют на музыкальном инструменте?

Для решения задачи используем определение условной вероятности:

- Обозначим события:
  - $A$: Студент играет на музыкальном инструменте.
  - $B$: Студент — мальчик.

Сначала найдем $P(A \cap B)$ и $P(B)$:

- Общее количество студентов, играющих на инструменте среди мальчиков $P(A \cap B) = \frac{10}{30} = \frac{1}{3}$
- Вероятность того, что выбранный студент — мальчик: $P(B) = \frac{18}{30} = \frac{3}{5}$

Теперь можем вычислить условную вероятность:

$$
P(A|B) = \frac{P(A \cap B)}{P(B)} = \frac{\frac{1}{3}}{\frac{3}{5}} = \frac{1}{3} \cdot \frac{5}{3} = \frac{5}{9}
$$

Таким образом, вероятность того, что мальчик играет на музыкальном инструменте, составляет $\frac{5}{9}$.

## Формула Байеса

**Формула Байеса** — это один из основополагающих результатов в теории вероятностей, который позволяет вычислять обратные условные вероятности. Она выражается следующим образом:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

где:
- $P(A|B)$ — условная вероятность события $A$ при условии события $B$.
- $P(B|A)$ — условная вероятность события $B$ при условии события $A$.
- $P(A)$ — априорная вероятность события $A$.
- $P(B)$ — априорная вероятность события $B$.

### Пример 2: Применение формулы Байеса

**Задача:** Пусть в популяции 1% людей страдают определенной болезнью. У нас есть тест, который дает положительный результат с вероятностью 90% для больных и 5% для здоровых. Какова вероятность того, что человек действительно болен, если тест показал положительный результат?

Обозначим события:
- $A$: Человек болен.
- $B$: Тест положительный.

Из условия мы знаем:
- $P(A) = 0.01$
- $P(B|A) = 0.90$
- $P(B|\neg A) = 0.05$

Сначала найдем $P(B)$:

$$
P(B) = P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A)
$$

Здесь $P(\neg A) = 1 - P(A) = 0.99$.

Теперь можем подставить значения:

$$
P(B) = (0.90 \cdot 0.01) + (0.05 \cdot 0.99) = 0.009 + 0.0495 = 0.0585
$$

Теперь используем формулу Байеса для нахождения $P(A|B)$:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} = \frac{0.90 \cdot 0.01}{0.0585} = \frac{0.009}{0.0585} \approx 0.1538
$$

Таким образом, вероятность того, что человек действительно болен, если тест показал положительный результат, составляет примерно 15.38%.

## Заключение

Условная вероятность и формула Байеса — это мощные инструменты для анализа вероятностных событий. Они широко применяются в различных областях, таких как медицина, финансы, информатика и многие другие.

### Числовой пример без кода

Рассмотрим следующую задачу:

В компании 80 сотрудников, из которых 60 — мужчины и 20 — женщины. Из мужчин 30 работают в IT, а из женщин 5 работают в IT. Какова вероятность того, что случайно выбранный сотрудник из IT — мужчина?

   - $P(A) = \frac{60}{80} = 0.75$

   - $P(B|\neg A) = \frac{5}{20} = 0.25$

3. Теперь находим $P(B)$:
$$
P(B) = P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A) = 0.5 \cdot 0.75 + 0.25 \cdot 0.25 = 0.375 + 0.0625 = 0.4375

4. Применим формулу Байеса для нахождения $P(A|B)$:
$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} = \frac{0.5 \cdot 0.75}{0.4375} = \frac{0.375}{0.4375} \approx 0.8571

Итак, вероятность того, что случайно выбранный сотрудник из IT — мужчина, составляет примерно 85.71%.

# Независимость событий

## Введение

В теории вероятностей события могут быть независимыми или зависимыми. Понимание независимости событий является ключевым для правильного применения вероятностных методов в различных областях, таких как статистика, экономика, наука и инженерия. В данной лекции мы рассмотрим:

1. Определение независимости событий.
2. Математическую формулировку независимости.
3. Примеры независимых и зависимых событий.
4. Свойства независимых событий.
5. Применение в практических задачах.

## 1. Определение независимости событий

События A и B называются независимыми, если вероятность наступления события A не зависит от того, произошло ли событие B, и наоборот. Это означает, что знание о том, что одно событие произошло, не изменяет вероятность другого события.

## 2. Математическая формулировка независимости

Для событий A и B независимость формулируется следующим образом:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

где:
- $P(A \cap B)$ — вероятность одновременного наступления событий A и B.
- $P(A)$ — вероятность события A.
- $P(B)$ — вероятность события B.

Если вышеуказанное равенство выполняется, то события A и B считаются независимыми.

### Примечание

Если $P(A \cap B) \neq P(A) \cdot P(B)$, то события A и B являются зависимыми.

## 3. Примеры независимых и зависимых событий

### Пример 1: Независимые события

Рассмотрим два события:
- Событие A: Выпадение орла при броске монеты.
- Событие B: Выпадение четного числа при броске стандартного кубика.

Вероятности:
- $P(A) = \frac{1}{2}$ (орел или решка)
- $P(B) = \frac{3}{6} = \frac{1}{2}$ (четные числа: 2, 4, 6)

Теперь найдем $P(A \cap B)$:

Событие A и событие B являются независимыми, если:

$$
P(A \cap B) = P(A) \cdot P(B) = \frac{1}{2} \cdot \frac{1}{2} = \frac{1}{4}
$$

### Пример 2: Зависимые события

Рассмотрим два события:
- Событие C: Выбор красной карты из колоды (52 карты, 26 красных).
- Событие D: Выбор карты, которая является дамой.

Вероятности:
- $P(C) = \frac{26}{52} = \frac{1}{2}$
- $P(D) = \frac{4}{52} = \frac{1}{13}$

Однако, если мы выбираем карту из колоды, и она оказывается красной, вероятность того, что эта карта является дамой, изменится. В этом случае события C и D являются зависимыми.

### Проверка на независимость:

Для проверки независимости вычислим $P(C \cap D)$:

- Из 26 красных карт в колоде 2 являются дамами. Значит, $P(C \cap D) = \frac{2}{52} = \frac{1}{26}$.

Теперь проверим, равняется ли:

$$
P(C \cap D) = P(C) \cdot P(D)
$$

Сравним:

$$
P(C) \cdot P(D) = \frac{1}{2} \cdot \frac{1}{13} = \frac{1}{26}
$$

Хотя равенство выполняется, события являются зависимыми, так как вероятность выбора дамы зависит от того, была ли выбрана красная карта.

## 4. Свойства независимых событий

1. **Независимость более чем двух событий**:
   Для трех событий A, B и C независимость будет следующей:

   $$
   P(A \cap B \cap C) = P(A) \cdot P(B) \cdot P(C)
   $$

   Если же события независимы, то это условие будет выполняться.

2. **Независимость и условная вероятность**:
   Если события A и B независимы, то:

   $$
   P(A | B) = P(A)
   $$

   и

   $$
   P(B | A) = P(B)
   $$

3. **Объединение независимых событий**:
   Если события A и B независимы, то вероятность их объединения:

   $$
   P(A \cup B) = P(A) + P(B) - P(A \cap B)
   $$

## 5. Применение в практических задачах

Рассмотрим практический пример, в котором нужно вычислить вероятность нескольких независимых событий. Пусть:

- Вероятность того, что студент сдаст экзамен по математике, равна $P(A) = 0.8$.
- Вероятность того, что студент сдаст экзамен по физике, равна $P(B) = 0.9$.
- Вероятность того, что студент сдаст экзамен по химии, равна $P(C) = 0.7$.

### Задача

Какова вероятность того, что студент сдаст все три экзамена?

### Решение

Так как события A, B и C независимы, то:

$$
P(A \cap B \cap C) = P(A) \cdot P(B) \cdot P(C)
$$

Подставляем известные значения:

$$
P(A \cap B \cap C) = 0.8 \cdot 0.9 \cdot 0.7
$$

### Вычисление

$$
P(A \cap B \cap C) = 0.8 \cdot 0.9 = 0.72
$$
$$
0.72 \cdot 0.7 = 0.504
$$

Таким образом, вероятность того, что студент сдаст все три экзамена, составляет **0.504** или **50.4%**.

## Заключение

Понимание независимости событий позволяет делать более точные прогнозы и принимать более обоснованные решения на основе вероятностных моделей. Важно уметь различать независимые и зависимые события, так как это влияет на применение различных формул и методов анализа.


 Давайте рассмотрим несколько конкретных примеров, которые иллюстрируют концепцию независимости событий, с вычислениями и объяснениями.

### Пример 1: Бросок монеты и бросок кубика

**События:**
- **A:** Выпадение орла при броске монеты.
- **B:** Выпадение четного числа при броске стандартного кубика.

**Вероятности:**
- $P(A) = \frac{1}{2}$ (орел или решка).
- $P(B) = \frac{3}{6} = \frac{1}{2}$ (четные числа: 2, 4, 6).

**Проверка независимости:**
Теперь найдем вероятность одновременного наступления событий A и B, то есть $P(A \cap B)$. Поскольку эти события независимы:

$$
P(A \cap B) = P(A) \cdot P(B) = \frac{1}{2} \cdot \frac{1}{2} = \frac{1}{4}
$$

**Интерпретация:**
Вероятность того, что при броске монеты выпадет орел и при броске кубика выпадет четное число, составляет $\frac{1}{4}$ или 25%.

### Пример 2: Успех на экзаменах

**События:**
- **C:** Студент сдает экзамен по математике.
- **D:** Студент сдает экзамен по физике.

**Вероятности:**
- $P(C) = 0.85$ (85% вероятность, что студент сдаст математику).
- $P(D) = 0.90$ (90% вероятность, что студент сдаст физику).

**Проверка независимости:**
События C и D независимы, если:

$$
P(C \cap D) = P(C) \cdot P(D) = 0.85 \cdot 0.90
$$

Теперь вычислим:

$$
P(C \cap D) = 0.85 \cdot 0.90 = 0.765
$$

**Интерпретация:**
Вероятность того, что студент сдаст оба экзамена (математику и физику), составляет 76.5%.

### Пример 3: Зависимые события

**События:**
- **E:** Извлечение карты из колоды, которая является тузом.
- **F:** Извлечение карты из колоды, которая является черной.

**Вероятности:**
- $P(E) = \frac{4}{52} = \frac{1}{13}$ (в колоде 4 туза).
- $P(F) = \frac{26}{52} = \frac{1}{2}$ (в колоде 26 черных карт).

Теперь проверим зависимость. Сначала определим $P(E \cap F)$:

Существует 2 черных туза (туз треф и туз пик).

$$
P(E \cap F) = \frac{2}{52} = \frac{1}{26}
$$

Теперь проверим, равно ли $P(E \cap F)$ произведению $P(E)$ и $P(F)$:

$$
P(E) \cdot P(F) = \frac{1}{13} \cdot \frac{1}{2} = \frac{1}{26}
$$

**Интерпретация:**
Хотя произведение вероятностей равно вероятности одновременного наступления событий, в данном случае события E и F зависят друг от друга, так как вероятность выбора черного туза влияет на общее количество черных карт.

### Пример 4: Три независимых события

**События:**
- **G:** Погода будет солнечной (вероятность $P(G) = 0.7$).
- **H:** Сосед будет дома (вероятность $P(H) = 0.6$).
- **I:** Вы получите бонус на работе (вероятность $P(I) = 0.8$).

**Проверка независимости:**
Так как события G, H и I независимы, то:

$$
P(G \cap H \cap I) = P(G) \cdot P(H) \cdot P(I)
$$

Теперь вычислим:

$$
P(G \cap H \cap I) = 0.7 \cdot 0.6 \cdot 0.8
$$

Сначала умножим $P(G)$ и $P(H)$:

$$
0.7 \cdot 0.6 = 0.42
$$

Теперь умножим результат на $P(I)$:

$$
0.42 \cdot 0.8 = 0.336
$$

**Интерпретация:**
Вероятность того, что погода будет солнечной, сосед будет дома и вы получите бонус на работе, составляет 33.6%.



# Полная группа событий и формула полной вероятности

## Введение

В теории вероятностей существует множество понятий, которые помогают анализировать случайные явления и предсказывать их поведение. Одним из таких понятий является полная группа событий и формула полной вероятности. Эти концепции помогают нам разбивать сложные вероятностные задачи на более простые и решать их более эффективно. В этой лекции мы подробно рассмотрим, что такое полная группа событий, как она формируется, и как применять формулу полной вероятности.

## Полная группа событий

### Определение

Полной группой событий (или полным событием) называется набор событий, которые совместно образуют все возможные результаты какого-либо случайного эксперимента. Эти события должны быть:
1. **Взаимно исключающимися**: два события не могут произойти одновременно.
2. **Полными**: одно из событий обязательно произойдет.

### Пример

Рассмотрим случай с броском честной игральной кости. Возможные исходы - это числа от 1 до 6. Мы можем определить полную группу событий следующим образом:

- $A_1:$ выпало 1
- $A_2:$ выпало 2
- $A_3:$ выпало 3
- $A_4:$ выпало 4
- $A_5:$ выпало 5
- $A_6:$ выпало 6

В этом примере события $A_1, A_2, A_3, A_4, A_5, A_6$ составляют полную группу, поскольку:
- Они не пересекаются (взаимно исключающие).
- Одно из этих событий произойдет при каждом броске кости (полное покрытие).

### Связь с вероятностями

Если $A_1, A_2, \ldots, A_n$ — это полная группа событий, то выполняется следующее равенство:

$$
P(A_1 \cup A_2 \cup \ldots \cup A_n) = P(\Omega) = 1
$$

где $\Omega$ — это пространство элементарных исходов.

## Формула полной вероятности

### Определение

Формула полной вероятности позволяет нам находить вероятность события, которое может произойти через несколько разных, но взаимно исключающих способов. Она выражает вероятность события через вероятность условных событий и вероятности полных событий.

### Формулировка

Если $B$ — событие, вероятность которого мы хотим найти, а $A_1, A_2, \ldots, A_n$ — это полная группа событий, то формула полной вероятности записывается следующим образом:

$$
P(B) = \sum_{i=1}^{n} P(B | A_i) P(A_i)
$$

где:
- $P(B | A_i)$ — условная вероятность события $B$ при условии, что произошло событие $A_i$.
- $P(A_i)$ — вероятность события $A_i$.

### Пример

Рассмотрим ситуацию, в которой у нас есть две фабрики, производящие детали, и мы хотим узнать вероятность того, что случайно выбранная деталь будет бракованной.

- Фабрика 1 производит 70% всех деталей и имеет 2% бракованных деталей.
- Фабрика 2 производит 30% всех деталей и имеет 5% бракованных деталей.

Здесь у нас есть две группы событий:
- $A_1$: деталь произведена на фабрике 1.
- $A_2$: деталь произведена на фабрике 2.
- $B$: деталь бракованная.

Теперь мы можем найти вероятности:
- $P(A_1) = 0.7$
- $P(A_2) = 0.3$
- $P(B | A_1) = 0.02$ (вероятность брака от фабрики 1)
- $P(B | A_2) = 0.05$ (вероятность брака от фабрики 2)

Теперь подставим значения в формулу полной вероятности:

$$
P(B) = P(B | A_1) P(A_1) + P(B | A_2) P(A_2)
$$

$$
P(B) = (0.02 \cdot 0.7) + (0.05 \cdot 0.3)
$$

### Вычисление

Теперь выполним вычисления:

1. Для фабрики 1:
   $$
   P(B | A_1) P(A_1) = 0.02 \cdot 0.7 = 0.014
   $$

2. Для фабрики 2:
   $$
   P(B | A_2) P(A_2) = 0.05 \cdot 0.3 = 0.015
   $$

Теперь сложим результаты:
$$
P(B) = 0.014 + 0.015 = 0.029
$$

Таким образом, вероятность того, что случайно выбранная деталь будет бракованной, составляет 0.029 или 2.9%.

## Заключение

В этой лекции мы рассмотрели, что такое полная группа событий, как она формируется, и как применять формулу полной вероятности для решения задач. Это мощный инструмент в теории вероятностей, который помогает разбивать сложные вероятностные ситуации на более простые части и находить решение. Правильное применение данной формулы позволяет анализировать различные сценарии и принимать обоснованные решения на основе вероятностных оценок.


# Конкретные примеры полной группы событий и формулы полной вероятности

Чтобы лучше понять концепции полной группы событий и формулы полной вероятности, рассмотрим несколько конкретных примеров из различных областей. Эти примеры помогут закрепить теоретические знания на практике и показать, как применять формулу полной вероятности в реальных ситуациях.

## Пример 1: Медицинский тест на заболевание

### Условие задачи

Предположим, существует заболевание, которое встречается у 1% населения. Разработан медицинский тест для этого заболевания, который имеет следующие характеристики:

- **Чувствительность (вероятность положительного теста при наличии заболевания):** 99%
- **Специфичность (вероятность отрицательного теста при отсутствии заболевания):** 95%

Необходимо найти вероятность того, что человек болен, если у него положительный результат теста.

### Решение

Для решения задачи используем формулу Байеса, которая является частным случаем формулы полной вероятности.

1. **Определим события:**
   - $A$: человек болен заболеванием.
   - $\overline{A}$: человек здоров.
   - $B$: положительный результат теста.

2. **Даны вероятности:**
   - $P(A) = 0.01$ (1% населения болеет).
   - $P(\overline{A}) = 0.99$.
   - $P(B | A) = 0.99$ (чувствительность).
   - $P(B | \overline{A}) = 0.05$ (ошибка теста, 100% - специфичность).

3. **Применяем формулу полной вероятности для $P(B)$:**

$$
P(B) = P(B | A) P(A) + P(B | \overline{A}) P(\overline{A}) = (0.99 \cdot 0.01) + (0.05 \cdot 0.99) = 0.0099 + 0.0495 = 0.0594
$$

4. **Используем формулу Байеса для нахождения $P(A | B)$:**

$$
P(A | B) = \frac{P(B | A) P(A)}{P(B)} = \frac{0.99 \cdot 0.01}{0.0594} \approx \frac{0.0099}{0.0594} \approx 0.1667
$$

**Итог:** Вероятность того, что человек болен заболеванием при положительном тесте, составляет примерно 16.67%.

### Интерпретация

Несмотря на высокую чувствительность и специфичность теста, из-за низкой базовой вероятности заболевания (1%) вероятность истинного положительного результата относительно невысока. Это показывает важность учета полной вероятности и базовых показателей при оценке результатов тестов.

## Пример 2: Выбор товара из магазина

### Условие задачи

В магазине продаются три категории товаров: электроника, одежда и продукты питания. Известно следующее:

- Электроника составляет 20% ассортимента, одежда — 50%, продукты питания — 30%.
- Вероятность того, что товар из электроники будет дефектным — 5%.
- Вероятность дефекта для одежды — 2%.
- Вероятность дефекта для продуктов питания — 1%.

Необходимо найти общую вероятность того, что случайно выбранный товар будет дефектным.

### Решение

1. **Определим события:**
   - $A_1$: товар электроника.
   - $A_2$: товар одежда.
   - $A_3$: товар продукты питания.
   - $B$: товар дефектный.

2. **Даны вероятности:**
   - $P(A_1) = 0.20$
   - $P(A_2) = 0.50$
   - $P(A_3) = 0.30$
   - $P(B | A_1) = 0.05$
   - $P(B | A_2) = 0.02$
   - $P(B | A_3) = 0.01$

3. **Применяем формулу полной вероятности для $P(B)$:**

$$
P(B) = P(B | A_1) P(A_1) + P(B | A_2) P(A_2) + P(B | A_3) P(A_3) = (0.05 \cdot 0.20) + (0.02 \cdot 0.50) + (0.01 \cdot 0.30) = 0.01 + 0.01 + 0.003 = 0.023
$$

**Итог:** Общая вероятность того, что случайно выбранный товар будет дефектным, составляет 2.3%.

## Пример 3: Вероятность выпадения числа на кубике с несколькими условиями

### Условие задачи

Предположим, у нас есть два шестигранных кубика:

- **Кубик A:** стандартный, вероятности равновероятны ($P(1) = P(2) = \ldots = P(6) = \frac{1}{6}$).
- **Кубик B:** загруженный, вероятность выпадения 6 увеличена до 0.3, а остальных чисел равны ($P(1) = P(2) = \ldots = P(5) = 0.14$).

Выбирается один кубик случайно (равновероятно) и бросается. Найдите вероятность того, что выпадет число 6.

### Решение

1. **Определим события:**
   - $A_1$: выбран кубик A.
   - $A_2$: выбран кубик B.
   - $B$: выпало число 6.

2. **Даны вероятности:**
   - $P(A_1) = 0.5$
   - $P(A_2) = 0.5$
   - $P(B | A_1) = \frac{1}{6} \approx 0.1667$
   - $P(B | A_2) = 0.3$

3. **Применяем формулу полной вероятности для $P(B)$:**

$$
P(B) = P(B | A_1) P(A_1) + P(B | A_2) P(A_2) = \left(\frac{1}{6} \cdot 0.5\right) + \left(0.3 \cdot 0.5\right) = \frac{0.5}{6} + 0.15 \approx 0.0833 + 0.15 = 0.2333
$$

**Итог:** Вероятность того, что выпадет число 6, составляет примерно 23.33%.

### Интерпретация

Загрузка кубика B увеличивает общую вероятность выпадения 6 по сравнению с использованием только стандартного кубика.

## Пример 4: Погода и вероятность прогулки

### Условие задачи

Предположим, вы планируете прогулку, и прогноз погоды на день может быть солнечным, дождливым или снежным. Известны следующие данные:

- Вероятности прогнозов:
  - Солнечно ($P(A_1)$) — 50%
  - Дождь ($P(A_2)$) — 30%
  - Снег ($P(A_3)$) — 20%

- Вероятности того, что вы пойдете гулять при разных погодных условиях:
  - При солнечной погоде — 90%
  - При дожде — 40%
  - При снеге — 30%

Необходимо найти общую вероятность того, что вы пойдете гулять в этот день.

### Решение

1. **Определим события:**
   - $A_1$: солнечно.
   - $A_2$: дождь.
   - $A_3$: снег.
   - $B$: пойти гулять.

2. **Даны вероятности:**
   - $P(A_1) = 0.5$
   - $P(A_2) = 0.3$
   - $P(A_3) = 0.2$
   - $P(B | A_1) = 0.9$
   - $P(B | A_2) = 0.4$
   - $P(B | A_3) = 0.3$

3. **Применяем формулу полной вероятности для $P(B)$:**

$$
P(B) = P(B | A_1) P(A_1) + P(B | A_2) P(A_2) + P(B | A_3) P(A_3) = (0.9 \cdot 0.5) + (0.4 \cdot 0.3) + (0.3 \cdot 0.2) = 0.45 + 0.12 + 0.06 = 0.63
$$

**Итог:** Общая вероятность того, что вы пойдете гулять, составляет 63%.

### Интерпретация

Погодные условия существенно влияют на ваше решение пойти гулять. Солнечная погода значительно повышает вероятность прогулки, тогда как дождь и снег снижают ее.

## Пример 5: Вероятность выбора книги из разных жанров

### Условие задачи

В библиотеке есть три жанра книг: фантастика, детектив и роман. Известно:

- Количество книг в жанрах:
  - Фантастика — 40 книг.
  - Детектив — 35 книг.
  - Роман — 25 книг.

- Известно, что:
  - Из фантастики 10 книг недавно переизданы.
  - Из детектива 5 книг недавно переизданы.
  - Из романа 15 книг недавно переизданы.

Если случайно выбрать книгу, какова вероятность, что она недавно переиздана?

### Решение

1. **Определим события:**
   - $A_1$: книга жанра фантастика.
   - $A_2$: книга жанра детектив.
   - $A_3$: книга жанра роман.
   - $B$: книга недавно переиздана.

2. **Вычислим вероятности:**
   - Общее количество книг: $40 + 35 + 25 = 100$.
   - $P(A_1) = \frac{40}{100} = 0.4$
   - $P(A_2) = \frac{35}{100} = 0.35$
   - $P(A_3) = \frac{25}{100} = 0.25$
   - $P(B | A_1) = \frac{10}{40} = 0.25$
   - $P(B | A_2) = \frac{5}{35} \approx 0.1429$
   - $P(B | A_3) = \frac{15}{25} = 0.6$

3. **Применяем формулу полной вероятности для $P(B)$:**

$$
P(B) = P(B | A_1) P(A_1) + P(B | A_2) P(A_2) + P(B | A_3) P(A_3) = (0.25 \cdot 0.4) + (0.1429 \cdot 0.35) + (0.6 \cdot 0.25) = 0.10 + 0.05 + 0.15 = 0.30
$$

**Итог:** Вероятность того, что случайно выбранная книга недавно переиздана, составляет 30%.

### Интерпретация

Жанровая принадлежность влияет на вероятность того, что книга была недавно переиздана. Наиболее высокая вероятность переиздания наблюдается в жанре романа.

## Пример 6: Качество продукции на трех заводах

### Условие задачи

Компания производит продукцию на трех заводах: Завод 1, Завод 2 и Завод 3. Известно:

- Доля продукции от каждого завода:
  - Завод 1 — 50%
  - Завод 2 — 30%
  - Завод 3 — 20%

- Вероятности дефекта продукции:
  - Завод 1 — 1%
  - Завод 2 — 2%
  - Завод 3 — 3%

Необходимо найти общую вероятность дефекта продукции и вероятность, что дефект произошел на Заводе 2 при условии, что продукция дефектна.

### Решение

1. **Определим события:**
   - $A_1$: продукция с Завода 1.
   - $A_2$: продукция с Завода 2.
   - $A_3$: продукция с Завода 3.
   - $B$: продукция дефектна.

2. **Даны вероятности:**
   - $P(A_1) = 0.5$
   - $P(A_2) = 0.3$
   - $P(A_3) = 0.2$
   - $P(B | A_1) = 0.01$
   - $P(B | A_2) = 0.02$
   - $P(B | A_3) = 0.03$

3. **Находим общую вероятность дефекта $P(B)$:**

$$
P(B) = P(B | A_1) P(A_1) + P(B | A_2) P(A_2) + P(B | A_3) P(A_3) = (0.01 \cdot 0.5) + (0.02 \cdot 0.3) + (0.03 \cdot 0.2) = 0.005 + 0.006 + 0.006 = 0.017
$$

4. **Находим вероятность, что дефект произошел на Заводе 2 при условии, что продукция дефектна $P(A_2 | B)$:**

$$
P(A_2 | B) = \frac{P(B | A_2) P(A_2)}{P(B)} = \frac{0.02 \cdot 0.3}{0.017} = \frac{0.006}{0.017} \approx 0.3529
$$

**Итог:**
- Общая вероятность дефекта продукции — 1.7%.
- Вероятность того, что дефект произошел на Заводе 2 при условии дефекта — примерно 35.29%.

### Интерпретация

Несмотря на более низкую долю продукции с Завода 2 по сравнению с Заводом 1, из-за более высокой вероятности дефекта, вероятность того, что дефект произошел именно на Заводе 2, существенно возрастает.

Эти инструменты являются основополагающими в теории вероятностей и находят широкое применение в статистике, инженерии, медицине, экономике и многих других областях.
