## Введение

В теории вероятностей и статистике важную роль играют условные распределения, которые позволяют анализировать случайные величины в контексте дополнительной информации. Понимание условных распределений, условных ожиданий и условных вероятностей, а также концепции независимости, является ключевым для изучения вероятностных моделей, включая байесовские сети. В данной лекции мы подробно рассмотрим эти темы.

## 1. Условные распределения

### 1.1. Условное распределение

Условное распределение случайной величины $X$ при условии, что другая случайная величина $Y$ приняла значение $y$, обозначается как $P(X | Y = y)$ для дискретных величин и как $f_{X|Y}(x|y)$ для непрерывных величин.

**Для дискретных случайных величин:**

Условная вероятность $P(X = x | Y = y)$ определяется как:

$$
P(X = x | Y = y) = \frac{P(X = x \cap Y = y)}{P(Y = y)}, \quad P(Y = y) > 0
$$

где $P(X = x \cap Y = y)$ — совместная вероятность.

**Для непрерывных случайных величин:**

Условная плотность вероятности $f_{X|Y}(x|y)$ определяется как:

$$
f_{X|Y}(x|y) = \frac{f_{X,Y}(x,y)}{f_Y(y)}, \quad f_Y(y) > 0
$$

где $f_{X,Y}(x,y)$ — совместная плотность вероятности.

### 1.2. Условное математическое ожидание

Условное математическое ожидание случайной величины $X$ при условии, что $Y = y$, обозначается как $E[X | Y = y]$.

**Для дискретных случайных величин:**

$$
E[X | Y = y] = \sum_{x} x \cdot P(X = x | Y = y)
$$

**Для непрерывных случайных величин:**

$$
E[X | Y = y] = \int_{-\infty}^{\infty} x \cdot f_{X|Y}(x|y) \, dx
$$

### 1.3. Условная дисперсия

Условная дисперсия случайной величины $X$ при условии, что $Y = y$, обозначается как $\text{Var}(X | Y = y)$.

$$
\text{Var}(X | Y = y) = E[X^2 | Y = y] - (E[X | Y = y])^2
$$

### Пример 1

Рассмотрим две дискретные случайные величины: $X$ — количество часов, проведенных за учебой, и $Y$ — оценка на экзамене. Пусть совместное распределение выглядит так:

| $X$ (часы) | $Y=2$ | $Y=3$ | $Y=4$ |
|--------------|---------|---------|---------|
| 1            | 0.1     | 0.2     | 0.3     |
| 2            | 0.1     | 0.1     | 0.1     |
| 3            | 0.0     | 0.1     | 0.1     |

Совместная вероятность для $Y=3$:

$$
P(Y = 3) = 0.2 + 0.1 + 0.1 = 0.4
$$

Условная вероятность $P(X = 1 | Y = 3)$:

$$
P(X = 1 | Y = 3) = \frac{P(X = 1 \cap Y = 3)}{P(Y = 3)} = \frac{0.2}{0.4} = 0.5
$$

Условное ожидание $E[X | Y = 3]$:

$$
E[X | Y = 3] = 1 \cdot 0.5 + 2 \cdot \frac{0.1}{0.4} + 3 \cdot \frac{0.1}{0.4} = 1.5 + 0.5 + 0.75 = 2.75
$$

## 2. Условная вероятность

### 2.1. Условная вероятность для дискретных случайных величин

Как уже упоминалось, условная вероятность может быть рассчитана с помощью формулы:

$$
P(X = x | Y = y) = \frac{P(X = x \cap Y = y)}{P(Y = y)}
$$

### 2.2. Условная вероятность для непрерывных случайных величин

Для непрерывных случайных величин условная вероятность представляется через плотности вероятностей:

$$
P(a < X < b | Y = y) = \int_{a}^{b} f_{X|Y}(x|y) \, dx
$$

### Пример 2

Рассмотрим непрерывные случайные величины $X$ и $Y$ с совместной плотностью $f_{X,Y}(x,y) = 6xy$ для $0 < x < 1$ и $0 < y < 1$. Найдем условную плотность $f_{X|Y}(x|y)$.

Сначала найдем $f_Y(y)$:

$$
f_Y(y) = \int_0^1 f_{X,Y}(x,y) \, dx = \int_0^1 6xy \, dx = 6y \cdot \frac{1}{2} = 3y
$$

Теперь можем найти условную плотность:

$$
f_{X|Y}(x|y) = \frac{f_{X,Y}(x,y)}{f_Y(y)} = \frac{6xy}{3y} = 2x
$$

## 3. Независимость случайных величин

### 3.1. Определение независимости

Случайные величины $X$ и $Y$ независимы, если:

$$
P(X = x \cap Y = y) = P(X = x) P(Y = y) \quad \text{для всех } x, y
$$

Для непрерывных величин:

$$
f_{X,Y}(x,y) = f_X(x) f_Y(y) \quad \text{для всех } x, y
$$

### Пример 3

Пусть $X$ и $Y$ независимы, где $P(X = 1) = 0.5$, $P(X = 2) = 0.5$, и $P(Y = 1) = 0.4$, $P(Y = 2) = 0.6$. Тогда совместная вероятность:

$$
P(X = 1, Y = 1) = P(X = 1) P(Y = 1) = 0.5 \cdot 0.4 = 0.2
$$

### 3.2. Условные распределения и независимость

Если $X$ и $Y$ независимы, то:

$$
P(X = x | Y = y) = P(X = x) \quad \text{и} \quad f_{X|Y}(x|y) = f_X(x)
$$

## 4. Байесовские сети

Байесовские сети — это графические модели, которые представляют вероятностные зависимости между множеством случайных величин. Узлы графа представляют случайные величины, а ребра — зависимости между ними. Каждый узел имеет условное распределение, которое описывает его зависимость от родительских узлов.

### Пример 4

Рассмотрим простую байесовскую сеть с тремя узлами: $A$, $B$, и $C$, где $A$ — родитель для $B$ и $C$. Плотности могут быть следующими:

1. $P(A)$
2. $P(B | A)$
3. $P(C | A)$

Общая вероятность может быть вычислена как:

$$
P(A, B, C) = P(A) P(B | A) P(C | A)
$$

### Пример числового расчета

Рассмотрим байесовскую сеть с следующими вероятностями:

- $P(A = 1) = 0.6$
- $P(B = 1 | A = 1) = 0.7$, $P(B = 1 | A = 0) = 0.2$
- $P(C = 1 | A = 1) = 0.8$, $P(C = 1 | A = 0) = 0.1$

Сначала находим $P(B = 1)$:

$$
P(B = 1) = P(B = 1 | A = 1) P(A = 1) + P(B = 1

 | A = 0) P(A = 0) = 0.7 \cdot 0.6 + 0.2 \cdot 0.4 = 0.42 + 0.08 = 0.5
$$

Затем находим $P(C = 1)$:

$$
P(C = 1) = P(C = 1 | A = 1) P(A = 1) + P(C = 1 | A = 0) P(A = 0) = 0.8 \cdot 0.6 + 0.1 \cdot 0.4 = 0.48 + 0.04 = 0.52
$$

### Задача 1: Условные вероятности

**Условие:**
Даны две случайные величины $X$ и $Y$ с совместным распределением:

| $X$ | $Y=1$ | $Y=2$ |
|-------|---------|---------|
| 1     | 0.3     | 0.2     |
| 2     | 0.1     | 0.4     |

1. Найдите $P(X=1 | Y=1)$.
2. Найдите $P(X=2 | Y=1)$.

#### Решение:

**Шаг 1:** Находим $P(Y=1)$.

Используем закон полной вероятности:

$$
P(Y=1) = P(X=1, Y=1) + P(X=2, Y=1)
$$

Считаем:

$$
P(Y=1) = P(X=1) P(Y=1 | X=1) + P(X=2) P(Y=1 | X=2)
$$

Найдем $P(X=1)$ и $P(X=2)$:

$$
P(X=1) = P(X=1, Y=1) + P(X=1, Y=2) = 0.3 + 0.2 = 0.5
$$
$$
P(X=2) = P(X=2, Y=1) + P(X=2, Y=2) = 0.1 + 0.4 = 0.5
$$

Теперь можем найти $P(Y=1)$:

$$
P(Y=1) = 0.3 + 0.1 = 0.4
$$

**Шаг 2:** Находим $P(X=1 | Y=1)$.

Используем формулу условной вероятности:

$$
P(X=1 | Y=1) = \frac{P(X=1, Y=1)}{P(Y=1)} = \frac{P(X=1) P(Y=1 | X=1)}{P(Y=1)} = \frac{0.3}{0.4} = 0.75
$$

**Шаг 3:** Находим $P(X=2 | Y=1)$.

Сначала находим $P(X=2, Y=1)$:

$$
P(X=2 | Y=1) = \frac{P(X=2, Y=1)}{P(Y=1)} = \frac{0.1}{0.4} = 0.25
$$

Таким образом, получили:

$$
P(X=1 | Y=1) = 0.75, \quad P(X=2 | Y=1) = 0.25
$$



### Задача 2: Условное ожидание и дисперсия

**Условие:**
Случайная величина $X$ имеет следующее распределение:

| $X$ | $P(X)$ |
|-------|----------|
| 1     | 0.5      |
| 2     | 0.5      |

Пусть $Y = 2X + 1$. Найдите:

1. $E[Y | X=1]$
2. $\text{Var}(Y | X=2)$

#### Решение:

**Шаг 1:** Найдем $E[Y | X=1]$.

Используем выражение для $Y$:

$$
Y = 2X + 1 \implies Y | X=1 = 2 \cdot 1 + 1 = 3
$$

Следовательно:

$$
E[Y | X=1] = 3
$$

**Шаг 2:** Найдем $\text{Var}(Y | X=2)$.

Сначала находим $Y | X=2$:

$$
Y | X=2 = 2 \cdot 2 + 1 = 5
$$

В данном случае, так как $Y$ принимает только одно значение при $X=2$, дисперсия равна нулю:

$$
\text{Var}(Y | X=2) = 0
$$

Таким образом, получили:

$$
E[Y | X=1] = 3, \quad \text{Var}(Y | X=2) = 0
$$



### Задача 3: Байесовская сеть

**Условие:**
Рассмотрим байесовскую сеть с узлами $A$, $B$ и $C$. Заданы следующие вероятности:

- $P(A=1) = 0.4$
- $P(B=1 | A=1) = 0.7$
- $P(B=1 | A=0) = 0.3$
- $P(C=1 | B=1) = 0.6$
- $P(C=1 | B=0) = 0.1$

Найдите $P(C=1)$.

#### Решение:

**Шаг 1:** Найдем $P(B=1)$.

Используем закон полной вероятности:

$$
P(B=1) = P(B=1 | A=1) P(A=1) + P(B=1 | A=0) P(A=0)
$$

Сначала найдем $P(A=0)$:

$$
P(A=0) = 1 - P(A=1) = 1 - 0.4 = 0.6
$$

Теперь подставляем значения:

$$
P(B=1) = 0.7 \cdot 0.4 + 0.3 \cdot 0.6 = 0.28 + 0.18 = 0.46
$$

**Шаг 2:** Найдем $P(C=1)$.

Используем аналогичный закон полной вероятности:

$$
P(C=1) = P(C=1 | B=1) P(B=1) + P(C=1 | B=0) P(B=0)
$$

Сначала найдем $P(B=0)$:

$$
P(B=0) = 1 - P(B=1) = 1 - 0.46 = 0.54
$$

Теперь подставляем значения:

$$
P(C=1) = 0.6 \cdot 0.46 + 0.1 \cdot 0.54 = 0.276 + 0.054 = 0.33
$$

Таким образом, получили:

$$
P(C=1) = 0.33
$$

Вот код, который реализует эти три задачи и добавляет визуализации:



```python
import numpy as np
import matplotlib.pyplot as plt

# Задача 1: Условные вероятности
def conditional_probabilities():
    # Задаем вероятности
    P_XY = np.array([[0.3, 0.2],  # P(X=1, Y=1), P(X=1, Y=2)
                     [0.1, 0.4]])  # P(X=2, Y=1), P(X=2, Y=2)

    # Считаем P(Y=1)
    P_Y1 = np.sum(P_XY[:, 0])  # Y=1

    # Считаем условные вероятности
    P_X1_given_Y1 = P_XY[0, 0] / P_Y1
    P_X2_given_Y1 = P_XY[1, 0] / P_Y1

    print(f"P(X=1 | Y=1) = {P_X1_given_Y1:.2f}")
    print(f"P(X=2 | Y=1) = {P_X2_given_Y1:.2f}")

    # Визуализация
    labels = ['P(X=1 | Y=1)', 'P(X=2 | Y=1)']
    probabilities = [P_X1_given_Y1, P_X2_given_Y1]

    plt.figure(figsize=(8, 5))
    plt.bar(labels, probabilities, color=['blue', 'orange'])
    plt.ylim(0, 1)
    plt.ylabel('Вероятность')
    plt.title('Условные вероятности P(X | Y=1)')
    plt.axhline(y=0.5, color='r', linestyle='--', label='0.5 линия')
    plt.legend()
    plt.show()

# Задача 2: Условное ожидание и дисперсия
def conditional_expectation_variance():
    # Задаем распределение X
    X = np.array([1, 2])
    P_X = np.array([0.5, 0.5])

    # Y = 2X + 1
    Y = 2 * X + 1
    E_Y_given_X1 = Y[0]  # при X=1
    Var_Y_given_X2 = 0    # дисперсия при X=2

    print(f"E[Y | X=1] = {E_Y_given_X1:.2f}")
    print(f"Var(Y | X=2) = {Var_Y_given_X2:.2f}")

    # Визуализация
    plt.figure(figsize=(8, 5))
    plt.bar(['E[Y | X=1]', 'Var(Y | X=2)'], [E_Y_given_X1, Var_Y_given_X2], color=['green', 'purple'])
    plt.ylabel('Значение')
    plt.title('Условное ожидание и дисперсия')
    plt.axhline(y=0, color='r', linestyle='--', label='0 линия')
    plt.legend()
    plt.show()

# Задача 3: Байесовская сеть
def bayesian_network():
    # Вероятности
    P_A1 = 0.4
    P_B1_given_A1 = 0.7
    P_B1_given_A0 = 0.3
    P_C1_given_B1 = 0.6
    P_C1_given_B0 = 0.1

    # Вычисляем P(B=1)
    P_A0 = 1 - P_A1
    P_B1 = P_B1_given_A1 * P_A1 + P_B1_given_A0 * P_A0

    # Вычисляем P(C=1)
    P_B0 = 1 - P_B1
    P_C1 = P_C1_given_B1 * P_B1 + P_C1_given_B0 * P_B0

    print(f"P(B=1) = {P_B1:.2f}")
    print(f"P(C=1) = {P_C1:.2f}")

    # Визуализация
    labels = ['P(B=1)', 'P(C=1)']
    probabilities = [P_B1, P_C1]

    plt.figure(figsize=(8, 5))
    plt.bar(labels, probabilities, color=['cyan', 'magenta'])
    plt.ylim(0, 1)
    plt.ylabel('Вероятность')
    plt.title('Вероятности в байесовской сети')
    plt.axhline(y=0.5, color='r', linestyle='--', label='0.5 линия')
    plt.legend()
    plt.show()

# Вызов функций для задач
conditional_probabilities()
conditional_expectation_variance()
bayesian_network()
```
