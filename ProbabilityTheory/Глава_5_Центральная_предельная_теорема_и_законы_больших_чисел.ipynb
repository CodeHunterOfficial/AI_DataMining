## Глава 5. Центральная предельная теорема и законы больших чисел

### Введение

Центральная предельная теорема (ЦПТ) и законы больших чисел (ЗБЧ) являются важнейшими концепциями в теории вероятностей и статистике. Они помогают нам понять, как ведут себя выборки из больших популяций и как приближаются распределения средних выборок к нормальному распределению. Эти теоремы находят широкое применение в статистике, экономике, инженерии и машинном обучении.

### 1. Законы больших чисел

Законы больших чисел описывают поведение средних выборок по мере увеличения объема выборки. Существуют два основных типа законов больших чисел: слабый и сильный.

#### 1.1. Слабый закон больших чисел

**Определение:** Слабый закон больших чисел (СЗБЧ) утверждает, что при увеличении объема выборки среднее значение выборки сходится по вероятности к математическому ожиданию (матожиданию) генеральной совокупности.

Формально, пусть $X_1, X_2, \ldots, X_n$ — независимые и одинаково распределенные случайные величины с математическим ожиданием $\mu = \mathbb{E}[X_i]$ и конечной дисперсией $\sigma^2 = \text{Var}(X_i) < \infty$. Тогда для любого $\epsilon > 0$:

$$
\lim_{n \to \infty} P\left( \left| \frac{1}{n} \sum_{i=1}^n X_i - \mu \right| > \epsilon \right) = 0
$$

**Пример:** Пусть $X_1, X_2, \ldots, X_n$ — результаты бросков игральной кости. Математическое ожидание $\mu$ равно:

$$
\mu = \frac{1 + 2 + 3 + 4 + 5 + 6}{6} = 3.5
$$

Согласно слабому закону, при увеличении $n$ среднее значение $\frac{1}{n} \sum_{i=1}^n X_i$ будет с высокой вероятностью приближаться к 3.5.

#### 1.2. Сильный закон больших чисел

**Определение:** Сильный закон больших чисел (СЗБЧ) утверждает, что среднее значение выборки почти обязательно сходится к математическому ожиданию. Это более строгий результат, чем слабый закон.

Формально, для независимых и одинаково распределенных случайных величин $X_1, X_2, \ldots, X_n$:

$$
P\left( \lim_{n \to \infty} \frac{1}{n} \sum_{i=1}^n X_i = \mu \right) = 1
$$

**Пример:** Рассмотрим те же броски игральной кости. Тогда, согласно сильному закону, среднее значение $\frac{1}{n} \sum_{i=1}^n X_i$ будет почти обязательно равным 3.5 при $n \to \infty$.

### 2. Центральная предельная теорема

Центральная предельная теорема описывает распределение суммы независимых случайных величин. Это основа для многих статистических методов.

**Формулировка:** Пусть $X_1, X_2, \ldots, X_n$ — независимые и одинаково распределенные случайные величины с конечным математическим ожиданием $\mu$ и конечной дисперсией $\sigma^2$. Тогда, если $S_n = \sum_{i=1}^n X_i$, то распределение стандартизированной суммы $Z_n = \frac{S_n - n\mu}{\sigma\sqrt{n}}$ при $n \to \infty$ стремится к стандартному нормальному распределению:

$$
Z_n \xrightarrow{d} N(0, 1)
$$

где $\xrightarrow{d}$ обозначает сходимость по распределению.

#### Пример:

Предположим, мы имеем 100 бросков игральной кости. Математическое ожидание:

$$
\mu = 3.5
$$

Дисперсия:

$$
\sigma^2 = \frac{(1-3.5)^2 + (2-3.5)^2 + (3-3.5)^2 + (4-3.5)^2 + (5-3.5)^2 + (6-3.5)^2}{6} = \frac{17.5}{6} \approx 2.91667
$$

Теперь, если мы посчитаем среднее значение бросков, то:

$$
S_{100} = \sum_{i=1}^{100} X_i
$$

Стандартизированное значение:

$$
Z_{100} = \frac{S_{100} - 100 \cdot 3.5}{\sqrt{100 \cdot 2.91667}} = \frac{S_{100} - 350}{17.1}
$$

При $n \to \infty$, $Z_n$ будет приближаться к нормальному распределению $N(0, 1)$.

### 3. Аппроксимации распределений и их применение в машинном обучении

В машинном обучении часто используется нормальное распределение для аппроксимации различных распределений данных. Центральная предельная теорема позволяет нам предполагать, что распределение выборочного среднего стремится к нормальному распределению, даже если исходное распределение не является нормальным.

#### Пример использования в машинном обучении

Рассмотрим задачу классификации. Пусть у нас есть выборка из $n$ объектов с некоторыми признаками, и мы хотим оценить среднюю точность классификатора.

1. **Обработка данных:** Мы собираем $n$ наблюдений, например, результаты тестирования модели на различных выборках.
2. **Вычисление средней точности:** Мы вычисляем среднюю точность $\hat{p} = \frac{1}{n} \sum_{i=1}^n p_i$, где $p_i$ — точность на $i$-й выборке.
3. **Построение доверительного интервала:** Мы можем построить доверительный интервал для средней точности, используя нормальное распределение:

$$
\hat{p} \pm z \cdot \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}
$$

где $z$ — значение из таблицы стандартного нормального распределения, соответствующее выбранному уровню доверия (например, 1.96 для 95% доверительного интервала).

### 4. Конкретные числовые примеры

#### Пример 1: Слабый закон больших чисел

Предположим, мы бросаем игральную кость 30 раз и получаем следующие результаты:

$$
1, 2, 3, 4, 5, 6, 1, 3, 4, 2, 6, 5, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6
$$

Сумма этих результатов:

$$
S_{30} = 1 + 2 + 3 + 4 + 5 + 6 + 1 + 3 + 4 + 2 + 6 + 5 + 1 + 2 + 3 + 4 + 5 + 6 + 1 + 2 + 3 + 4 + 5 + 6 + 1 + 2 + 3 + 4 + 5 + 6 = 90
$$

Среднее значение:

$$
\frac{S_{30}}{30} = \frac{90}{30} = 3
$$

В соответствии со слабым законом больших чисел, если бы мы продолжили бросать кости, среднее значение бы сближалось с 3.5.

#### Пример 2: Сильный закон больших чисел

Пусть у нас есть 100 бросков кости, и их результаты:

$$
1, 2, 3, 4, 5, 6, 1, 3, 4, 2, 6, 5, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4,

5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6
$$

Сумма результатов:

$$
S_{100} = 350
$$

Среднее значение:

$$
\frac{S_{100}}{100} = \frac{350}{100} = 3.5
$$

Согласно сильному закону, при $n \to \infty$, это значение будет почти обязательно приближаться к 3.5.

#### Пример 3: Центральная предельная теорема

Рассмотрим 50 независимых бросков кости. Предположим, что математическое ожидание и дисперсия остались теми же:

$$
\mu = 3.5, \quad \sigma^2 = \frac{17.5}{6} \approx 2.91667
$$

Стандартизированное значение:

$$
Z_{50} = \frac{S_{50} - 50 \cdot 3.5}{\sqrt{50 \cdot 2.91667}}
$$

Если $S_{50} = 180$, тогда:

$$
Z_{50} = \frac{180 - 175}{\sqrt{145.8335}} \approx \frac{5}{12.1} \approx 0.413
$$

При этом $Z_{50}$ стремится к нормальному распределению $N(0, 1)$.

### Заключение

Центральная предельная теорема и законы больших чисел — это краеугольные камни статистики и теории вероятностей. Они позволяют делать выводы о больших популяциях на основе выборок, а также имеют важное значение для алгоритмов в машинном обучении. Понимание этих теорем и их применения открывает новые горизонты для анализа данных и построения предсказательных моделей.

### Задача 1: Среднее значение бросков игральной кости

**Условие:** Предположим, что мы бросаем игральную кость 60 раз. Найдите среднее значение и стандартное отклонение выборочного среднего. Также найдите 95% доверительный интервал для математического ожидания.

**Решение:**

1. **Определим математическое ожидание и дисперсию:**  
   Для игральной кости, математическое ожидание $\mu$ и дисперсия $\sigma^2$ рассчитываются следующим образом:

   $$
   \mu = \frac{1 + 2 + 3 + 4 + 5 + 6}{6} = \frac{21}{6} = 3.5
   $$

   Дисперсия:

   $$
   \sigma^2 = \frac{(1-3.5)^2 + (2-3.5)^2 + (3-3.5)^2 + (4-3.5)^2 + (5-3.5)^2 + (6-3.5)^2}{6} = \frac{17.5}{6} \approx 2.91667
   $$

   Таким образом, стандартное отклонение $\sigma$:

   $$
   \sigma = \sqrt{2.91667} \approx 1.7078
   $$

2. **Расчёт стандартного отклонения выборочного среднего:**  
   Для выборки из $n = 60$:

   $$
   \sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}} = \frac{1.7078}{\sqrt{60}} \approx \frac{1.7078}{7.746} \approx 0.2204
   $$

3. **Нахождение 95% доверительного интервала:**  
   Для 95% доверительного интервала используем значение $z \approx 1.96$ для нормального распределения. Формула для доверительного интервала:

   $$
   \text{Доверительный интервал} = \bar{X} \pm z \cdot \sigma_{\bar{X}}
   $$

   Так как у нас нет реальных данных о бросках, примем $\bar{X} = 3.5$:

   $$
   \text{Доверительный интервал} = 3.5 \pm 1.96 \cdot 0.2204
   $$

   Расчёт:

   $$
   \text{Доверительный интервал} = 3.5 \pm 0.432
   $$

   Это даёт:

   $$
   (3.068, 3.932)
   $$

### Задача 2: Доверительный интервал для средней зарплаты

**Условие:** В компании 100 сотрудников. Из них выбраны 30 для исследования зарплат. Средняя зарплата выборки составила 50,000 рублей, а стандартное отклонение — 5,000 рублей. Найдите 90% доверительный интервал для средней зарплаты сотрудников компании.

**Решение:**

1. **Определим размеры выборки и стандартное отклонение:**  
   Размер выборки $n = 30$ и стандартное отклонение $s = 5000$ рублей.

2. **Нахождение стандартного отклонения выборочного среднего:**

   $$
   \sigma_{\bar{X}} = \frac{s}{\sqrt{n}} = \frac{5000}{\sqrt{30}} \approx \frac{5000}{5.477} \approx 913.97
   $$

3. **Выбор уровня доверия:**  
   Для 90% доверительного интервала $z \approx 1.645$.

4. **Расчёт доверительного интервала:**

   $$
   \text{Доверительный интервал} = \bar{X} \pm z \cdot \sigma_{\bar{X}}
   $$

   Подставляем значения:

   $$
   \text{Доверительный интервал} = 50000 \pm 1.645 \cdot 913.97
   $$

   Расчёт:

   $$
   \text{Доверительный интервал} = 50000 \pm 1500.5
   $$

   Это даёт:

   $$
   (48500.5, 51499.5)
   $$

### Задача 3: Применение центральной предельной теоремы

**Условие:** В игре, где выбрасывают два игральные кости, найдите распределение средней суммы бросков, если кости бросают 40 раз. Определите вероятность того, что средняя сумма будет больше 7.

**Решение:**

1. **Определим математическое ожидание и дисперсию для одной кости:**  
   Математическое ожидание:

   $$
   \mu = \frac{1 + 2 + 3 + 4 + 5 + 6}{6} = 3.5
   $$

   Дисперсия:

   $$
   \sigma^2 = \frac{17.5}{6} \approx 2.91667
   $$

   Для двух костей:

   $$
   \mu_{sum} = 3.5 + 3.5 = 7
   $$
   $$
   \sigma^2_{sum} = 2.91667 + 2.91667 = 5.83334
   $$

   Стандартное отклонение:

   $$
   \sigma_{sum} = \sqrt{5.83334} \approx 2.414
   $$

2. **Нахождение стандартного отклонения выборочного среднего для $n = 40$:**

   $$
   \sigma_{\bar{X}} = \frac{\sigma_{sum}}{\sqrt{n}} = \frac{2.414}{\sqrt{40}} \approx \frac{2.414}{6.325} \approx 0.381
   $$

3. **Нахождение вероятности, что среднее значение будет больше 7:**  
   Стандартизируем значение:

   $$
   Z = \frac{7 - \mu_{sum}}{\sigma_{\bar{X}}} = \frac{7 - 7}{0.381} = 0
   $$

   Используя таблицу стандартного нормального распределения, находим:

   $$
   P(Z > 0) = 0.5
   $$

   Таким образом, вероятность того, что среднее значение будет больше 7, составляет 50%.

Давайте реализуем каждую из трех задач на Python с использованием библиотек NumPy и Matplotlib для визуализации. Я предоставлю код для каждой задачи, включая создание гистограмм и графиков.

### Задача 1: Среднее значение бросков игральной кости



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Параметры задачи
n = 60  # количество бросков
mu = 3.5  # математическое ожидание
sigma = np.sqrt(17.5 / 6)  # стандартное отклонение

# Генерация данных
np.random.seed(0)  # для воспроизводимости
rolls = np.random.randint(1, 7, size=n)

# Расчет выборочного среднего
sample_mean = np.mean(rolls)
sample_std = np.std(rolls)

# 95% доверительный интервал
z = stats.norm.ppf(0.975)  # значение z для 95% доверительного интервала
margin_of_error = z * (sample_std / np.sqrt(n))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Визуализация
plt.hist(rolls, bins=10, alpha=0.7, color='blue', edgecolor='black', density=True)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=1, label='Выборочное среднее')
plt.axvline(confidence_interval[0], color='green', linestyle='dashed', linewidth=1, label='95% ДИ (нижний)')
plt.axvline(confidence_interval[1], color='orange', linestyle='dashed', linewidth=1, label='95% ДИ (верхний)')
plt.title('Распределение бросков игральной кости')
plt.xlabel('Результат броска')
plt.ylabel('Плотность')
plt.legend()
plt.grid()
plt.show()

print("Выборочное среднее:", sample_mean)
print("95% доверительный интервал:", confidence_interval)
```


### Задача 2: Доверительный интервал для средней зарплаты


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Параметры задачи
n = 30  # размер выборки
sample_mean = 50000  # средняя зарплата
sample_std = 5000  # стандартное отклонение

# 90% доверительный интервал
z = stats.norm.ppf(0.95)  # значение z для 90% доверительного интервала
margin_of_error = z * (sample_std / np.sqrt(n))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Визуализация
x = np.linspace(46000, 54000, 100)
y = stats.norm.pdf(x, sample_mean, sample_std/np.sqrt(n))

plt.plot(x, y, color='blue', label='Нормальное распределение выборочного среднего')
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=1, label='Выборочное среднее')
plt.axvline(confidence_interval[0], color='green', linestyle='dashed', linewidth=1, label='90% ДИ (нижний)')
plt.axvline(confidence_interval[1], color='orange', linestyle='dashed', linewidth=1, label='90% ДИ (верхний)')
plt.title('Доверительный интервал для средней зарплаты')
plt.xlabel('Зарплата')
plt.ylabel('Плотность')
plt.legend()
plt.grid()
plt.show()

print("90% доверительный интервал:", confidence_interval)
```


### Задача 3: Применение центральной предельной теоремы



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Параметры задачи
n = 40  # количество бросков двух костей
mu = 7  # математическое ожидание суммы
sigma = np.sqrt(5.83334)  # стандартное отклонение суммы

# Генерация данных
np.random.seed(0)  # для воспроизводимости
rolls = np.random.randint(1, 7, size=(n, 2)).sum(axis=1)

# Расчет выборочного среднего и стандартного отклонения
sample_mean = np.mean(rolls)
sample_std = np.std(rolls)

# Нахождение Z-значения для средних сумм
z = (7 - mu) / (sample_std / np.sqrt(n))
p_value = 1 - stats.norm.cdf(z)  # вероятность того, что среднее значение больше 7

# Визуализация
plt.hist(rolls, bins=10, alpha=0.7, color='blue', edgecolor='black', density=True)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=1, label='Выборочное среднее')
plt.axvline(7, color='green', linestyle='dashed', linewidth=1, label='Целевое значение (7)')
plt.title('Распределение сумм бросков двух игральных костей')
plt.xlabel('Сумма бросков')
plt.ylabel('Плотность')
plt.legend()
plt.grid()
plt.show()

print("Выборочное среднее:", sample_mean)
print("Вероятность, что среднее значение больше 7:", p_value)
```
