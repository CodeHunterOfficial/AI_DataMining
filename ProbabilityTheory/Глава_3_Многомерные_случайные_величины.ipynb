# Глава 3. Многомерные случайные величины

Многомерные случайные величины – это векторные величины, состоящие из нескольких компонент, каждая из которых является случайной величиной. Если $X_1, X_2, \ldots, X_n$ — случайные величины, то вектор $\mathbf{X} = (X_1, X_2, \ldots, X_n)$ называется многомерной случайной величиной. Основная цель анализа таких величин заключается в изучении их совместного распределения, зависимостей между ними и их свойств.

#### Совместное распределение и маргинальные распределения

1. **Совместное распределение**

   Совместная функция распределения (функция распределения вероятностей) $F_{X_1, \ldots, X_n}(x_1, \ldots, x_n)$ многомерной случайной величины $\mathbf{X} = (X_1, \ldots, X_n)$ определяется как вероятность того, что все компоненты $X_i$ принимают значения, меньшие или равные заданным числам:
   $$
   F_{X_1, \ldots, X_n}(x_1, \ldots, x_n) = P(X_1 \leq x_1, \ldots, X_n \leq x_n).
   $$
   
   Если случайные величины $X_1, \ldots, X_n$ являются непрерывными, то совместная плотность вероятности $f_{X_1, \ldots, X_n}(x_1, \ldots, x_n)$ задается как производная функции распределения:
   $$
   f_{X_1, \ldots, X_n}(x_1, \ldots, x_n) = \frac{\partial^n}{\partial x_1 \ldots \partial x_n} F_{X_1, \ldots, X_n}(x_1, \ldots, x_n).
   $$

2. **Маргинальные распределения**

   Маргинальные распределения — это распределения отдельных компонент многомерной случайной величины. Они вычисляются путем интегрирования совместной плотности по всем переменным, кроме интересующей нас:
   $$
   f_{X_i}(x_i) = \int_{-\infty}^{\infty} \ldots \int_{-\infty}^{\infty} f_{X_1, \ldots, X_n}(x_1, \ldots, x_n) \, dx_1 \ldots dx_{i-1} \, dx_{i+1} \ldots dx_n.
   $$
   
   В случае дискретных случайных величин маргинальная вероятность вычисляется путем суммирования совместной вероятности по всем другим компонентам:
   $$
   P(X_i = x_i) = \sum_{x_j, j \neq i} P(X_1 = x_1, \ldots, X_n = x_n).
   $$

#### Ковариация и корреляция

1. **Ковариация**

   Ковариация между двумя случайными величинами $X$ и $Y$ измеряет степень, в которой они изменяются вместе, и определяется формулой:
   $$
   \text{Cov}(X, Y) = \mathbb{E}[(X - \mathbb{E}[X])(Y - \mathbb{E}[Y])],
   $$
   где $\mathbb{E}[X]$ и $\mathbb{E}[Y]$ — математические ожидания величин $X$ и $Y$.

   Если ковариация положительная, это означает, что величины имеют тенденцию изменяться в одном направлении (при увеличении $X$ также увеличивается $Y$), а если отрицательная — то величины изменяются в противоположных направлениях.

2. **Корреляция**

   Корреляция нормирует ковариацию, делая её безразмерной величиной, что позволяет измерять степень линейной зависимости двух случайных величин:
   $$
   \rho_{XY} = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y},
   $$
   где $\sigma_X$ и $\sigma_Y$ — стандартные отклонения величин $X$ и $Y$.

   Корреляция лежит в диапазоне от -1 до 1. Значение 1 означает полную положительную линейную зависимость, -1 — полную отрицательную линейную зависимость, а 0 — отсутствие линейной зависимости.

#### Независимость случайных величин

Две случайные величины $X$ и $Y$ называются независимыми, если их совместное распределение равно произведению их маргинальных распределений:
$$
f_{X, Y}(x, y) = f_X(x) \cdot f_Y(y).
$$

Для дискретных случайных величин аналогичное свойство выражается в виде:
$$
P(X = x, Y = y) = P(X = x) \cdot P(Y = y).
$$

Если величины независимы, то ковариация между ними равна нулю. Однако нулевая ковариация не обязательно означает независимость — это верно только в случае нормального распределения.

#### Распределение многомерной нормальной (Гауссовой) величины

Многомерное нормальное распределение является обобщением одномерного нормального распределения на случай векторных случайных величин.

1. **Определение**

   Вектор случайных величин $\mathbf{X} = (X_1, \ldots, X_n)^T$ имеет многомерное нормальное распределение с математическим ожиданием $\boldsymbol{\mu} = (\mu_1, \ldots, \mu_n)^T$ и ковариационной матрицей $\Sigma$, если его плотность вероятности задается формулой:
   $$
   f(\mathbf{x}) = \frac{1}{(2\pi)^{n/2} |\Sigma|^{1/2}} \exp \left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu})^T \Sigma^{-1} (\mathbf{x} - \boldsymbol{\mu}) \right),
   $$
   где $|\Sigma|$ — детерминант матрицы $\Sigma$, а $\Sigma^{-1}$ — её обратная матрица.

2. **Свойства многомерного нормального распределения**

   - Линейные комбинации компонент вектора $\mathbf{X}$ также имеют нормальное распределение.
   - Если случайные величины независимы и нормально распределены, то их совместное распределение является многомерным нормальным.
   - Компоненты вектора $\mathbf{X}$ могут быть некоррелированными и всё же зависимыми, в отличие от одномерного случая.

#### Примеры

1. **Пример совместного распределения и маргинальных распределений**

   Пусть совместная плотность двух случайных величин $X$ и $Y$ задана как
   $$
   f_{X,Y}(x, y) = \frac{1}{2\pi} e^{-\frac{1}{2}(x^2 + y^2)}.
   $$
   Это соответствует двум независимым нормально распределенным величинам с математическими ожиданиями ноль и дисперсиями единица. Маргинальные плотности:
   $$
   f_X(x) = \int_{-\infty}^{\infty} f_{X,Y}(x, y) \, dy = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}},
   $$
   $$
   f_Y(y) = \int_{-\infty}^{\infty} f_{X,Y}(x, y) \, dx = \frac{1}{\sqrt{2\pi}} e^{-\frac{y^2}{2}}.
   $$
   Оба маргинальных распределения являются нормальными.

2. **Пример ковариации и корреляции**

   Пусть имеются данные о росте $X$ и весе $Y$ людей, причём:
   $$
   \mathbb{E}[X] = 170 \text{ см}, \quad \mathbb{E}[Y] = 70 \text{ кг},
   $$
   $$
   \sigma_X = 10 \text{ см}, \quad \sigma_Y = 15 \text{ кг},
   $$
   $$
   \text{Cov}(X, Y) = 100 \text{ см·кг}.
   $$
   Тогда корреляция равна
   $$
   \rho_{XY} = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y} = \frac{100}{10 \cdot 15} = \frac{2}{3} \approx 0.67.
   $$

3. **Пример многомерного нормального распределения**

   Пусть $\mathbf{X}= (X_1, X_2)^T$ имеет многомерное нормальное распределение с
   $$
   \boldsymbol{\mu} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \quad \Sigma = \begin{pmatrix} 1 & 0.5 \\ 0.5 & 1 \end{pmatrix}.
   $$
   Тогда его плотность вероятности будет
   $$
   f(\mathbf{x}) = \frac{1}{2\pi \sqrt{1 - 0.25}} \exp\left(-\frac{1}{2(1 - 0.25)} \left( x_1^2 - 0.5 x_1 x_2 + x_2^2 \right)\right).
   $$

В заключение, многомерные случайные величины играют ключевую роль в статистике и вероятности. Изучение их свойств и взаимосвязей предоставляет важные инструменты для анализа данных в различных областях.

### Задача 1: Совместное распределение

**Условие**: Пусть случайные величины $X$ и $Y$ имеют совместное распределение:
$$
f_{X,Y}(x,y) = \begin{cases}
6(1 - x - y), & 0 < x < 1, \, 0 < y < 1, \, x + y < 1, \\
0, & \text{иначе}.
\end{cases}
$$
Найдите маргинальные распределения $f_X(x)$ и $f_Y(y)$.

#### Решение:

1. **Найдите маргинальную плотность $f_X(x)$**:
   $$
   f_X(x) = \int_{-\infty}^{+\infty} f_{X,Y}(x,y) \, dy = \int_0^{1-x} 6(1 - x - y) \, dy.
   $$

2. **Вычислите интеграл**:
   $$
   \int_0^{1-x} 6(1 - x - y) \, dy = 6 \left[ (1 - x)y - \frac{y^2}{2} \right]_0^{1-x} = 6 \left[ (1 - x)(1 - x) - \frac{(1 - x)^2}{2} \right].
   $$
   Подставляем пределы:
   $$
   = 6 \left[ (1 - x)^2 - \frac{(1 - x)^2}{2} \right] = 6 \left[ \frac{(1 - x)^2}{2} \right] = 3(1 - x)^2.
   $$
   Следовательно,
   $$
   f_X(x) = 3(1 - x)^2, \quad 0 < x < 1.
   $$

3. **Найдите маргинальную плотность $f_Y(y)$** аналогичным образом:
   $$
   f_Y(y) = \int_0^{1-y} 6(1 - x - y) \, dx.
   $$

4. **Вычислите интеграл**:
   $$
   \int_0^{1-y} 6(1 - x - y) \, dx = 6 \left[ (1 - y)x - \frac{x^2}{2} \right]_0^{1-y} = 6 \left[ (1 - y)(1 - y) - \frac{(1 - y)^2}{2} \right].
   $$
   Подставляем пределы:
   $$
   = 6 \left[ (1 - y)^2 - \frac{(1 - y)^2}{2} \right] = 6 \left[ \frac{(1 - y)^2}{2} \right] = 3(1 - y)^2.
   $$
   Следовательно,
   $$
   f_Y(y) = 3(1 - y)^2, \quad 0 < y < 1.
   $$

### Ответ:
Маргинальные распределения:
$$
f_X(x) = 3(1 - x)^2, \quad 0 < x < 1,
$$
$$
f_Y(y) = 3(1 - y)^2, \quad 0 < y < 1.
$$



### Задача 2: Ковариация и корреляция

**Условие**: Пусть $X$ и $Y$ — случайные величины с математическими ожиданиями $\mathbb{E}[X] = 2$, $\mathbb{E}[Y] = 3$, дисперсиями $\text{Var}(X) = 4$, $\text{Var}(Y) = 9$ и ковариацией $\text{Cov}(X,Y) = 2$. Найдите коэффициент корреляции $\rho_{XY}$.

#### Решение:

1. **Вспомним формулу для коэффициента корреляции**:
   $$
   \rho_{XY} = \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X) \cdot \text{Var}(Y)}}.
   $$

2. **Подставим известные значения**:
   $$
   \rho_{XY} = \frac{2}{\sqrt{4 \cdot 9}} = \frac{2}{\sqrt{36}} = \frac{2}{6} = \frac{1}{3}.
   $$

### Ответ:
Коэффициент корреляции $\rho_{XY} = \frac{1}{3}$.



### Задача 3: Многомерная нормальная величина

**Условие**: Пусть вектор случайных величин $\mathbf{X} = (X_1, X_2)^T$ имеет многомерное нормальное распределение с параметрами:
$$
\boldsymbol{\mu} = \begin{pmatrix} 1 \\ 2 \end{pmatrix}, \quad \Sigma = \begin{pmatrix} 3 & 1 \\ 1 & 2 \end{pmatrix}.
$$
Найдите вероятность того, что $X_1 < 2$ и $X_2 < 3$.

#### Решение:

1. **Нормализуем величины**:
   Определим новый вектор:
   $$
   Z = \begin{pmatrix} Z_1 \\ Z_2 \end{pmatrix} = \begin{pmatrix} \frac{X_1 - 1}{\sqrt{3}} \\ \frac{X_2 - 2}{\sqrt{2}} \end{pmatrix}.
   $$
   Тогда $Z$ также имеет многомерное нормальное распределение с параметрами:
   $$
   \boldsymbol{\mu}_Z = \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \quad \Sigma_Z = \begin{pmatrix} 1 & \frac{1}{\sqrt{6}} \\ \frac{1}{\sqrt{6}} & 1 \end{pmatrix}.
   $$

2. **Определите область**:
   Нужно найти:
   $$
   P(X_1 < 2, X_2 < 3) = P\left(Z_1 < \frac{2 - 1}{\sqrt{3}}, Z_2 < \frac{3 - 2}{\sqrt{2}}\right) = P(Z_1 < \frac{1}{\sqrt{3}}, Z_2 < \frac{1}{\sqrt{2}}).
   $$
   То есть, необходимо найти вероятность в области:
   $$
   P(Z_1 < 0.577, Z_2 < 0.707.
   $$

3. **Используем таблицы многомерного нормального распределения**:
   Можно найти $P(Z_1 < 0.577)$ и $P(Z_2 < 0.707)$ из таблицы стандартного нормального распределения или используя программное обеспечение.

4. **Находим совместную вероятность**:
   Для многомерного нормального распределения с известной ковариационной матрицей, можно использовать численные методы или специальное программное обеспечение для нахождения вероятности.

5. **Приблизительный расчет**:
   Если предположить, что $Z_1$ и $Z_2$ независимы (это не совсем точно, так как у них есть ковариация), то:
   $$
   P(Z_1 < 0.577) \approx 0.718 \quad \text{и} \quad P(Z_2 < 0.707) \approx 0.760.
   $$
   Поэтому,
   $$
   P(X_1 < 2, X_2 < 3) \approx 0.718 \cdot 0.760 \approx 0.546.
   $$

### Ответ:
Вероятность того, что $X_1 < 2$ и $X_2 < 3$ приблизительно равна 0.546.

### Задача 1: Совместное распределение

Мы создадим визуализацию совместной плотности вероятности $f_{X,Y}(x,y)$ и маргинальных распределений $f_X(x)$ и $f_Y(y)$.



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import dblquad

# Определяем функцию совместной плотности
def joint_pdf(x, y):
    return 6 * (1 - x - y) if 0 < x < 1 and 0 < y < 1 and x + y < 1 else 0

# Создаем сетку для x и y
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

# Вычисляем значения совместной плотности
Z = np.vectorize(joint_pdf)(X, Y)

# Визуализация совместной плотности
plt.figure(figsize=(10, 6))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(label='Плотность вероятности')
plt.title('Совместная плотность вероятности $f_{X,Y}(x,y)$')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
````


### Задача 2: Ковариация и корреляция

Для этой задачи мы просто напечатаем коэффициент корреляции и визуализируем его с помощью гистограммы.


````python
import numpy as np
import matplotlib.pyplot as plt

# Данные
mu_x = 2
mu_y = 3
var_x = 4
var_y = 9
cov_xy = 2

# Вычисление коэффициента корреляции
rho_xy = cov_xy / np.sqrt(var_x * var_y)
print(f'Коэффициент корреляции: {rho_xy:.2f}')

# Генерация случайных величин
np.random.seed(0)
n_samples = 1000
X = np.random.normal(mu_x, np.sqrt(var_x), n_samples)
Y = np.random.normal(mu_y, np.sqrt(var_y), n_samples)

# Визуализация
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, alpha=0.5)
plt.title('Случайные величины $X$ и $Y$')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.axis('equal')
plt.grid(True)
plt.show()
````




### Задача 3: Многомерная нормальная величина

Мы создадим визуализацию многомерного нормального распределения и выведем вероятность того, что $X_1 < 2$ и $X_2 < 3$.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Параметры многомерного нормального распределения
mu = [1, 2]
cov = [[3, 1], [1, 2]]

# Создание сетки для x1 и x2
x1 = np.linspace(-1, 5, 100)
x2 = np.linspace(-1, 5, 100)
X1, X2 = np.meshgrid(x1, x2)

# Вычисление значений плотности
pos = np.dstack((X1, X2))
rv = multivariate_normal(mu, cov)
Z = rv.pdf(pos)

# Визуализация многомерного нормального распределения
plt.figure(figsize=(10, 6))
plt.contourf(X1, X2, Z, levels=50, cmap='viridis')
plt.colorbar(label='Плотность вероятности')
plt.title('Многомерное нормальное распределение')
plt.xlabel('$X_1$')
plt.ylabel('$X_2$')
plt.axis('equal')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.show()

# Вычисление вероятности
from scipy.stats import norm

# Нормализация
z1 = (2 - mu[0]) / np.sqrt(cov[0][0])  # X1 < 2
z2 = (3 - mu[1]) / np.sqrt(cov[1][1])  # X2 < 3

# Найдем вероятность P(X1 < 2, X2 < 3) используя многомерное нормальное распределение
probability = rv.cdf([2, 3])
print(f'Вероятность того, что X1 < 2 и X2 < 3: {probability:.4f}')
````
