#Глава 10. Стохастические процессы

## Введение в случайные процессы

**Определение стохастического процесса**

Стохастический процесс — это семейство случайных переменных, индексированное по времени (или по пространству), представляющее собой модель динамических систем, где неопределенность и случайность играют важную роль. Формально, стохастический процесс можно определить как последовательность случайных величин $\{X(t)\}_{t \in T}$, где $T$ — это индексное множество (например, время).

### Примеры стохастических процессов

1. **Броуновское движение** (или Винеровский процесс).
2. **Марковские процессы** — процессы, в которых будущее зависит только от настоящего, но не от прошлого.
3. **Процессы Леви** — обобщение броуновского движения с дискретными скачками.
4. **Арима и Архимедовы процессы** — используются для моделирования временных рядов.

### Основные характеристики стохастических процессов

1. **Независимость**: Для процесса $\{X(t)\}$ произвольны два времени $t_1 < t_2$ такие, что $X(t_1)$ и $X(t_2)$ независимы.
2. **Стационарность**: Процесс называется стационарным, если его статистические свойства не зависят от времени.
3. **Корреляция**: Определяет степень взаимосвязи между значениями процесса в разные моменты времени.

## Винеровский процесс (броуновское движение)

**Определение Винеровского процесса**

Винеровский процесс $\{W(t), t \geq 0\}$ является стохастическим процессом, который удовлетворяет следующим условиям:

1. $W(0) = 0$ с вероятностью 1.
2. Для любых $0 \leq s < t$, разность $W(t) - W(s)$ имеет нормальное распределение с параметрами:
   $$
   W(t) - W(s) \sim \mathcal{N}(0, t-s)
   $$
3. Для любых $0 \leq s < t < u$ значения $W(t) - W(s)$ и $W(u) - W(t)$ независимы.

### Свойства Винеровского процесса

1. **Непрерывность**: $W(t)$ с вероятностью 1 непрерывен для всех $t \geq 0$.
2. **Постоянная дисперсия**: Дисперсия $\text{Var}(W(t)) = t$.
3. **Центральная предельная теорема**: Винеровский процесс описывает предельное распределение сумм независимых случайных величин.

### Пример: Винеровский процесс

Рассмотрим Винеровский процесс $W(t)$ с параметрами $t = 0, 1, 2, 3$.

- $W(0) = 0$
- $W(1) \sim \mathcal{N}(0, 1)$
- $W(2) - W(1) \sim \mathcal{N}(0, 1)$, $W(2) \sim \mathcal{N}(0, 2)$
- $W(3) - W(2) \sim \mathcal{N}(0, 1)$, $W(3) \sim \mathcal{N}(0, 3)$

Можно ожидать, что значения $W(1)$, $W(2)$ и $W(3)$ будут случайными числами, которые могут принимать как положительные, так и отрицательные значения, но в среднем будут стремиться к нулю.

## Процессы Леви и их использование в финансовом моделировании

**Определение процессов Леви**

Процесс Леви $\{X(t), t \geq 0\}$ является общим типом стохастического процесса, который может описывать как непрерывные изменения (броуновское движение), так и дискретные скачки. Процесс определяется следующими свойствами:

1. **Начальное значение**: $X(0) = 0$ с вероятностью 1.
2. **Стационарные приращения**: Для $0 \leq s < t$ разность $X(t) - X(s)$ имеет одинаковое распределение для всех временных интервалов.
3. **Леви-курва**: Процесс имеет независимые приращения, которые могут быть нормальными или распределенными по закону Леви.

### Пример процессов Леви

Процесс Леви включает в себя следующие компоненты:

1. **Броуновское движение** $W(t)$.
2. **Дискретные скачки** $J(t)$, распределенные по некоторому закону (например, пуассоновскому).

Формально, процесс можно записать как:
$$
X(t) = W(t) + J(t)
$$
где $J(t)$ — это сумма независимых случайных величин.

### Применение в финансовом моделировании

Процессы Леви часто используются в моделировании цен на активы. Например, модель Блэка-Шоулза предполагает, что логарифм цен активов следует нормальному распределению.

**Пример финансовой модели**

Предположим, что цена актива $S(t)$ подчиняется следующему стохастическому дифференциальному уравнению (СДУ):
$$
dS(t) = \mu S(t) dt + \sigma S(t) dW(t) + J(t)
$$
где:
- $\mu$ — ожидаемая доходность,
- $\sigma$ — волатильность,
- $W(t)$ — Винеровский процесс,
- $J(t)$ — процесс Леви, представляющий скачки.

## Примеры численных значений

1. **Винеровский процесс**:
   - Пусть $\sigma = 1$, тогда:
     - $W(1) = 0 + Z_1$, где $Z_1 \sim \mathcal{N}(0, 1)$ может быть равен 0.5.
     - $W(2) = W(1) + Z_2$, где $Z_2 \sim \mathcal{N}(0, 1)$ может быть равен -0.3.
     - $W(3) = W(2) + Z_3$, где $Z_3 \sim \mathcal{N}(0, 1)$ может быть равен 0.8.

     Итого:
     - $W(1) = 0.5$
     - $W(2) = 0.2$
     - $W(3) = 1.0$

2. **Процесс Леви**:
   - Пусть $W(t)$ — стандартное броуновское движение, и дискретный скачок $J(t)$ распределен по Пуассону с параметром $\lambda = 0.5$.
     - На временном интервале $t=1$ с вероятностью 0.2 произойдет скачок в 3 единицы.
     - Таким образом, можно получить:
     - $X(1) = W(1) + J(1) = 0.5 + 3 = 3.5$.

## Заключение

Стохастические процессы, такие как Винеровский процесс и процессы Леви, играют важную роль в моделировании динамических систем, особенно в финансах и экономике. Понимание этих процессов позволяет создавать более точные модели, которые учитывают случайность и неопределенность в реальных данных.

### Задача 1: Вычисление математического ожидания и дисперсии Винеровского процесса

**Задача:** Пусть $ W(t) $ — Винеровский процесс. Найдите математическое ожидание и дисперсию процесса $ X(t) = 2W(t) + 3t $ для любого $ t \geq 0 $.

#### Решение:

1. **Математическое ожидание**:
   - Известно, что для Винеровского процесса $ W(t) $ выполняется:
     $$
     E[W(t)] = 0
     $$
   - Теперь найдем математическое ожидание $ X(t) $:
     $$
     E[X(t)] = E[2W(t) + 3t] = 2E[W(t)] + 3t = 2 \cdot 0 + 3t = 3t
     $$

2. **Дисперсия**:
   - Дисперсия Винеровского процесса $ W(t) $ равна:
     $$
     \text{Var}(W(t)) = t
     $$
   - Найдем дисперсию $ X(t) $:
     $$
     \text{Var}(X(t)) = \text{Var}(2W(t) + 3t) = \text{Var}(2W(t)) = 4 \cdot \text{Var}(W(t)) = 4t
     $$
   - Так как $ 3t $ является детерминированным слагаемым, его дисперсия равна 0.

**Ответ**:
- Математическое ожидание $ E[X(t)] = 3t $
- Дисперсия $ \text{Var}(X(t)) = 4t $



### Задача 2: Найти ковариацию для стохастического процесса

**Задача:** Пусть $ X(t) = W(t) + 2t $, где $ W(t) $ — Винеровский процесс. Найдите ковариацию $ \text{Cov}(X(t_1), X(t_2)) $ для $ t_1 < t_2 $.

#### Решение:

1. **Выражение для ковариации**:
   - Ковариация между $ X(t_1) $ и $ X(t_2) $:
     $$
     \text{Cov}(X(t_1), X(t_2)) = E[X(t_1)X(t_2)] - E[X(t_1)]E[X(t_2)]
     $$

2. **Найдем математические ожидания**:
   - Мы уже нашли, что $ E[X(t)] = 3t $, следовательно:
     $$
     E[X(t_1)] = 3t_1, \quad E[X(t_2)] = 3t_2
     $$

3. **Вычислим произведение математических ожиданий**:
   $$
   E[X(t_1)]E[X(t_2)] = (3t_1)(3t_2) = 9t_1 t_2
   $$

4. **Найдем $ E[X(t_1)X(t_2)] $**:
   - Выразим $ X(t_1)X(t_2) $:
     $$
     X(t_1)X(t_2) = (W(t_1) + 2t_1)(W(t_2) + 2t_2) = W(t_1)W(t_2) + 2t_1 W(t_2) + 2t_2 W(t_1) + 4t_1 t_2
     $$
   - Найдем математическое ожидание:
     $$
     E[X(t_1)X(t_2)] = E[W(t_1)W(t_2)] + 2t_1 E[W(t_2)] + 2t_2 E[W(t_1)] + 4t_1 t_2
     $$
   - Известно, что $ E[W(t)] = 0 $ и $ E[W(t_1)W(t_2)] = \text{Cov}(W(t_1), W(t_2)) = \min(t_1, t_2) = t_1 $:
     $$
     E[X(t_1)X(t_2)] = t_1 + 2t_1 \cdot 0 + 2t_2 \cdot 0 + 4t_1 t_2 = t_1 + 4t_1 t_2
     $$

5. **Теперь найдем ковариацию**:
   $$
   \text{Cov}(X(t_1), X(t_2)) = (t_1 + 4t_1 t_2) - 9t_1 t_2 = t_1 + 4t_1 t_2 - 9t_1 t_2 = t_1 - 5t_1 t_2
   $$

**Ответ**:
- Ковариация $ \text{Cov}(X(t_1), X(t_2)) = t_1 - 5t_1 t_2 $


### Задача 3: Проверка стационарности процесса

**Задача:** Пусть $ X(t) = \sin(t) + W(t) $, где $ W(t) $ — Винеровский процесс. Является ли процесс $ X(t) $ стационарным?

#### Решение:

1. **Определение стационарности**:
   - Процесс является стационарным, если его математическое ожидание и ковариационная функция не зависят от времени.

2. **Найдем математическое ожидание**:
   - $ E[X(t)] = E[\sin(t) + W(t)] = \sin(t) + E[W(t)] = \sin(t) + 0 = \sin(t) $

3. **Заключение о математическом ожидании**:
   - Математическое ожидание зависит от времени, так как $\sin(t)$ — функция времени.

4. **Найдем ковариационную функцию**:
   - Ковариация $ X(t_1) $ и $ X(t_2) $:
     $$
     \text{Cov}(X(t_1), X(t_2)) = E[X(t_1)X(t_2)] - E[X(t_1)]E[X(t_2)]
     $$
   - Мы можем записать:
     $$
     E[X(t_1)] = \sin(t_1), \quad E[X(t_2)] = \sin(t_2)
     $$

5. **Вычислим $ E[X(t_1)X(t_2)] $**:
   $$
   E[X(t_1)X(t_2)] = E[(\sin(t_1) + W(t_1))(\sin(t_2) + W(t_2))]
   $$
   $$
   = \sin(t_1)\sin(t_2) + E[W(t_1)]\sin(t_2) + E[W(t_2)]\sin(t_1) + E[W(t_1)W(t_2)]
   $$
   $$
   = \sin(t_1)\sin(t_2) + 0 + 0 + \min(t_1, t_2) = \sin(t_1)\sin(t_2) + \min(t_1, t_2)
   $$

6. **Ковариация**:
   $$
   \text{Cov}(X(t_1), X(t_2)) = (\sin(t_1)\sin(t_2) + \min(t_1, t_2)) - \sin(t_1)\sin(t_2) = \min(t_1, t_2)
   $$
   - Эта ковариация зависит от $t_1$ и $t_2$, что подтверждает нестационарность.

**Ответ**:
- Процесс $ X(t) = \sin(t) + W(t) $ **не является стационарным**, так как его математическое ожидание зависит от времени.



Давайте реализуем предложенные задачи на Python, включая визуализацию для каждой из них.

### Задача 1: Математическое ожидание и дисперсия Винеровского процесса

Для этой задачи мы создадим график функции $ X(t) = 2W(t) + 3t $ и визуализируем математическое ожидание и дисперсию.

### Задача 2: Ковариация стохастического процесса

Мы построим график ковариационной функции $ \text{Cov}(X(t_1), X(t_2)) = t_1 - 5t_1 t_2 $.

### Задача 3: Стационарность процесса

Для этой задачи мы построим график $ X(t) = \sin(t) + W(t) $ и покажем, как его математическое ожидание и ковариация зависят от времени.


```python
import numpy as np
import matplotlib.pyplot as plt

# Установим параметры
np.random.seed(0)
t = np.linspace(0, 5, 500)  # Временной интервал

# Задача 1: X(t) = 2W(t) + 3t
# Сгенерируем Винеровский процесс
W = np.random.normal(0, 1, size=t.shape)  # Нормальное распределение
W = np.cumsum(W) * np.sqrt(0.01)  # Интегрирование с шагом 0.01

# Вычислим X(t)
X = 2 * W + 3 * t

# Визуализация
plt.figure(figsize=(12, 6))
plt.plot(t, X, label=r'$X(t) = 2W(t) + 3t$', color='blue')
plt.title('Процесс $X(t) = 2W(t) + 3t$')
plt.xlabel('t')
plt.ylabel('$X(t)$')
plt.plot(t, 3 * t, color='red', linestyle='--', label='Математическое ожидание $E[X(t)] = 3t$')
plt.fill_between(t, (3*t - 4*t), (3*t + 4*t), color='yellow', alpha=0.2, label='Дисперсия $Var(X(t)) = 4t$')
plt.legend()
plt.grid()
plt.show()

# Задача 2: Ковариация X(t) = W(t) + 2t
def cov_X(t1, t2):
    return t1 - 5 * t1 * t2

# Сгенерируем t1 и t2 для вычисления ковариации
t1 = np.linspace(0, 5, 100)
t2 = np.linspace(0, 5, 100)
T1, T2 = np.meshgrid(t1, t2)
Cov = cov_X(T1, T2)

# Визуализация
plt.figure(figsize=(10, 8))
plt.contourf(T1, T2, Cov, levels=50, cmap='viridis')
plt.colorbar(label='Ковариация')
plt.title('Ковариация $Cov(X(t_1), X(t_2)) = t_1 - 5t_1 t_2$')
plt.xlabel('$t_1$')
plt.ylabel('$t_2$')
plt.grid()
plt.show()

# Задача 3: Стационарность процесса X(t) = sin(t) + W(t)
X_stationary = np.sin(t) + W

# Визуализация
plt.figure(figsize=(12, 6))
plt.plot(t, X_stationary, label=r'$X(t) = \sin(t) + W(t)$', color='purple')
plt.title('Процесс $X(t) = \sin(t) + W(t)$')
plt.xlabel('t')
plt.ylabel('$X(t)$')
plt.axhline(y=0, color='green', linestyle='--', label='Ожидание $E[X(t)]$ зависит от времени')
plt.fill_between(t, (np.sin(t) - 0.1), (np.sin(t) + 0.1), color='orange', alpha=0.2, label='Влияние $W(t)$ на $X(t)$')
plt.legend()
plt.grid()
plt.show()
```
