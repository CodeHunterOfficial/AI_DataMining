### Математика многослойного перцептрона

Рассмотрим математические основы простой нейронной сети, которая имеет два входа, два выхода и один скрытый слой с тремя нейронами.

#### Структура нейронной сети:

1. **Входной слой**:  
   Наша сеть имеет два входа, которые обозначим как $a_0^0$ и $a_1^0$.

2. **Скрытый слой**:  
   В скрытом слое находятся три нейрона, и их выходы обозначим как $a_0^1$, $a_1^1$ и $a_2^1$.

3. **Выходной слой**:  
   Наша сеть имеет два выхода, которые обозначим как $a_0^2$ и $a_1^2$.

### Параметры нейронной сети

Веса и смещения нейронной сети могут быть представлены в виде матриц и векторов:

1. **Веса**:
   - Для слоя с $n$ входами и $m$ нейронами матрица весов имеет размер $m \times n$.
   - Каждый элемент матрицы весов $W_{ij}^l$ представляет вес, соединяющий $j$-й вход с $i$-м нейроном $l$-го слоя.

2. **Смещения**:
   - Для слоя с $m$ нейронами вектор смещений имеет размер $m \times 1$.
   - Каждый элемент вектора смещений $b_i^l$ представляет смещение для $i$-го нейрона в слое $l$.

### Прямой проход (Forward Pass)

**Прямой проход** — это процесс, в котором входные данные передаются через сеть от входного слоя к выходному. Нейроны на каждом слое принимают значения от предыдущего слоя, обрабатывают их и передают результаты на следующий слой. Давайте пошагово разберем вычисления для каждого слоя.

#### Скрытый слой (первый слой):

Каждый нейрон скрытого слоя получает на вход взвешенную сумму входных значений и смещений. Это значение обозначим как $Z$, и затем к нему применим функцию активации.

Для первого нейрона скрытого слоя:
$$
Z_0^1 = W_{00}^1 \cdot a_0^0 + W_{01}^1 \cdot a_1^0 + b_0^1
$$
$$
a_0^1 = \text{act}(Z_0^1)
$$

Для второго нейрона:
$$
Z_1^1 = W_{10}^1 \cdot a_0^0 + W_{11}^1 \cdot a_1^0 + b_1^1
$$
$$
a_1^1 = \text{act}(Z_1^1)
$$

Для третьего нейрона:
$$
Z_2^1 = W_{20}^1 \cdot a_0^0 + W_{21}^1 \cdot a_1^0 + b_2^1
$$
$$
a_2^1 = \text{act}(Z_2^1)
$$

Здесь $\text{act}(Z)$ — это функция активации, которую можно выбрать из следующих вариантов:
- **Сигмоида**: $\sigma(x) = \frac{1}{1 + e^{-x}}$
- **ReLU**: $f(x) = \max(0, x)$
- **Гиперболический тангенс**: $f(x) = \tanh(x)$
- **Softmax** (для выходного слоя): $\sigma(z_i) = \frac{e^{z_i}}{\sum_{k=1}^{N} e^{z_k}}$

#### Матричная запись для скрытого слоя:

Представим веса, входы и смещения в виде матриц и векторов:
$$
W^1 = \begin{pmatrix}
W_{00}^1 & W_{01}^1 \\
W_{10}^1 & W_{11}^1 \\
W_{20}^1 & W_{21}^1
\end{pmatrix}, \quad
a^0 = \begin{pmatrix}
a_0^0 \\
a_1^0
\end{pmatrix}, \quad
b^1 = \begin{pmatrix}
b_0^1 \\
b_1^1 \\
b_2^1
\end{pmatrix}
$$

Теперь можем записать выходы скрытого слоя:
$$
Z^1 = W^1 \cdot a^0 + b^1
$$
$$
a^1 = \text{act}(Z^1)
$$
где $a^1 = \begin{pmatrix} a_0^1 \\ a_1^1 \\ a_2^1 \end{pmatrix}$ — вектор выходов скрытого слоя.

#### Выходной слой (второй слой):

Для второго (выходного) слоя процесс аналогичен. Матрица весов теперь имеет размер $2 \times 3$, поскольку на вход подаются три значения (выходы скрытого слоя), а на выходе два нейрона.

$$
Z_0^2 = W_{00}^2 \cdot a_0^1 + W_{01}^2 \cdot a_1^1 + W_{02}^2 \cdot a_2^1 + b_0^2
$$
$$
Z_1^2 = W_{10}^2 \cdot a_0^1 + W_{11}^2 \cdot a_1^1 + W_{12}^2 \cdot a_2^1 + b_1^2
$$

Применяем функцию активации к каждому выходу:
$$
a_0^2 = \text{act}(Z_0^2), \quad a_1^2 = \text{act}(Z_1^2)
$$

#### Матричная запись для выходного слоя:

$$
W^2 = \begin{pmatrix}
W_{00}^2 & W_{01}^2 & W_{02}^2 \\
W_{10}^2 & W_{11}^2 & W_{12}^2
\end{pmatrix}, \quad
a^1 = \begin{pmatrix}
a_0^1 \\
a_1^1 \\
a_2^1
\end{pmatrix}, \quad
b^2 = \begin{pmatrix}
b_0^2 \\
b_1^2
\end{pmatrix}
$$

Тогда выходы выходного слоя:
$$
Z^2 = W^2 \cdot a^1 + b^2
$$
$$
a^2 = \text{act}(Z^2)
$$
где $a^2 = \begin{pmatrix} a_0^2 \\ a_1^2 \end{pmatrix}$ — вектор выходов сети.

### Функция стоимости (Cost Function)

Функция стоимости — это мера ошибки или расхождения между предсказанными значениями и реальными результатами. В машинном обучении целью является минимизация функции стоимости, чтобы улучшить производительность модели.

Одной из наиболее популярных функций стоимости является **среднеквадратичная ошибка** (Mean Squared Error, MSE):
$$
MSE = \frac{1}{n} \sum_{i=0}^{n-1} (O_i - Y_i)^2
$$
где:
- $n$ — количество примеров;
- $O_i$ — фактические значения;
- $Y_i$ — предсказанные значения.

Чем меньше значение MSE, тем лучше модель.

### Оптимизация: стохастический градиентный спуск (Stochastic Gradient Descent)

Для минимизации функции стоимости используем алгоритм **стохастического градиентного спуска** (SGD). Этот алгоритм обновляет параметры модели на основе градиентов функции стоимости. Формула обновления параметра $P_i$ выглядит следующим образом:
$$
P_i = P_i - \alpha \cdot \frac{\partial C}{\partial P_i}
$$
где:
- $P_i$ — значение $i$-го параметра (вес или смещение);
- $\alpha$ — скорость обучения;
- $\frac{\partial C}{\partial P_i}$ — частная производная функции стоимости по параметру $P_i$.

#### Векторная форма градиентного спуска

Процесс обновления весов можно записать в векторной форме. Пусть мы изменяем вес $W_{00}^1$. Тогда его обновление будет происходить по формуле:
$$
W_{00}^1 \leftarrow W_{00}^1 - \alpha \cdot \frac{\partial C}{\partial W_{00}^1}
$$
где $\frac{\partial C}{\partial W_{00}^1}$ — это градиент функции стоимости по весу $W_{00}^1$.

Если $\frac{\partial C}{\partial W_{00}^1}$ положительна, то вес $W_{00}^1$ уменьшится, что приведет к уменьшению значения функции стоимости $C$. Если градиент отрицателен, то вес увеличится, что также приведет к снижению стоимости.

### Алгоритм обучения нейронной сети

Алгоритм обучения нейронной сети включает следующие шаги:

1. **Прямой проход (Forward Pass)**: Входные данные передаются через слои сети, вычисляются взвешенные суммы и применяются функции активации.

2. **Вычесление функции стоимости**: После получения предсказанных значений на выходе сети сравниваем их с фактическими значениями, используя функцию стоимости (например, MSE).

3. **Обратный проход (Backward Pass)**: Вычисляем градиенты функции стоимости по каждому параметру сети с помощью правила цепочки (backpropagation).

4. **Обновление параметров**: С помощью стохастического градиентного спуска корректируем веса и смещения, чтобы минимизировать функцию стоимости.

Процесс повторяется до тех пор, пока сеть не достигнет желаемой точности предсказаний или не выполнится заданное количество итераций.

### Bringing the Cost Function into the Picture with an Example

Теперь давайте более детально разберем процесс вычисления градиентов и применения их для обучения нейронной сети.

#### Forward Pass and Cost Calculation

Предположим, у нас есть нейронная сеть с двумя входами и двумя выходами. Мы подаем данные на вход сети, которая, исходя из своих весов и смещений, вычисляет промежуточные результаты и, в конечном счете, выдает выходные значения $O_0$ и $O_1$. Эти выходы сравниваются с истинными значениями $Y_0$ и $Y_1$, что позволяет рассчитать функцию стоимости.

Функция стоимости $C$ для данной задачи может быть определена, например, как среднеквадратичная ошибка (MSE):

$$
C = \frac{1}{2} \left( (O_0 - Y_0)^2 + (O_1 - Y_1)^2 \right)
$$

#### Gradient Calculation for Each Parameter

Чтобы оптимизировать сеть, мы должны минимизировать функцию стоимости. Это достигается с помощью градиентного спуска, для чего нам необходимо вычислить производные функции стоимости $C$ по каждому параметру сети, включая веса и смещения. Начнем с вычисления градиента для одного из весов, скажем $W_{00}^2$, который связывает первый выходной нейрон второго слоя с первым нейроном предыдущего слоя.

Для вычисления частной производной $\frac{\partial C}{\partial W_{00}^2}$ применим цепное правило:

$$
\frac{\partial C}{\partial W_{00}^2} = \frac{\partial C}{\partial O_0} \cdot \frac{\partial O_0}{\partial z_0^2} \cdot \frac{\partial z_0^2}{\partial W_{00}^2}
$$

Теперь последовательно найдем каждую из этих производных.

1. **Частная производная функции стоимости по выходу**:

$$
\frac{\partial C}{\partial O_0} = \frac{1}{2} \cdot \frac{\partial \left( (O_0 - Y_0)^2 + (O_1 - Y_1)^2 \right)}{\partial O_0} = O_0 - Y_0
$$

2. **Частная производная выхода $O_0$ по активации**:

$$
\frac{\partial O_0}{\partial z_0^2} = \frac{\partial a_0^2}{\partial z_0^2} = \frac{\partial \text{act}_2(z_0^2)}{\partial z_0^2} = \text{act}_2'(z_0^2)
$$

где $\text{act}_2'$ — производная функции активации второго слоя.

3. **Частная производная взвешенной суммы $z_0^2$ по весу $W_{00}^2$**:

$$
\frac{\partial z_0^2}{\partial W_{00}^2} = \frac{\partial \left( W_{00}^2 a_0^1 + W_{01}^2 a_1^1 + W_{02}^2 a_2^1 + b_0^2 \right)}{\partial W_{00}^2} = a_0^1
$$

Теперь, подставляя все найденные значения, мы получим:

$$
\frac{\partial C}{\partial W_{00}^2} = (O_0 - Y_0) \cdot \text{act}_2'(z_0^2) \cdot a_0^1
$$

#### Обобщение для других весов

Аналогично мы можем вычислить частные производные для других весов:

$$
\frac{\partial C}{\partial W_{01}^2} = (O_0 - Y_0) \cdot \text{act}_2'(z_0^2) \cdot a_1^1 = a_1^1 \cdot \partial E_0^2
$$
$$
\frac{\partial C}{\partial W_{02}^2} = (O_0 - Y_0) \cdot \text{act}_2'(z_0^2) \cdot a_2^1 = a_2^1 \cdot \partial E_0^2
$$

где $\partial E_0^2 = (O_0 - Y_0) \cdot \text{act}_2'(z_0^2)$.

#### Градиенты для смещений

Для смещения $b_0^2$ вычислим градиент следующим образом:

$$
\frac{\partial z_0^2}{\partial b_0^2} = 1
$$
$$
\frac{\partial C}{\partial b_0^2} = (O_0 - Y_0) \cdot \text{act}_2'(z_0^2) = \partial E_0^2
$$

Таким образом, для смещения мы получаем:

$$
\frac{\partial b_0^2}{\partial C} = \partial E_0^2
$$

#### Матричная форма для второго слоя

В результате мы можем записать производные по всем весам второго слоя в матричной форме:

$$
\frac{\partial W^2}{\partial C} = \begin{bmatrix} \partial E_0^2 a_0^1 & \partial E_0^2 a_1^1 & \partial E_0^2 a_2^1 \\ \partial E_1^2 a_0^1 & \partial E_1^2 a_1^1 & \partial E_1^2 a_2^1 \end{bmatrix}
$$

Это выражение можно записать компактно как:

$$
\frac{\partial W^2}{\partial C} = \partial E^2 (a^1)^T
$$

где $\partial E^2$ — это вектор градиентов по выходам второго слоя, а $(a^1)^T$ — транспонированный вектор активаций первого слоя.

#### Градиенты для смещений второго слоя

Смещению $b^2$ соответствуют частные производные:

$$
\frac{\partial b^2}{\partial C} = \partial E^2
$$

В итоге общее выражение для градиентов смещений второго слоя:

$$
\frac{\partial b^2}{\partial C} = \begin{bmatrix} \partial E_0^2 \\ \partial E_1^2 \end{bmatrix}
$$



Давайте рассмотрим весь процесс шаг за шагом, чтобы вычислить конкретный числовой пример для многослойного перцептрона, включая прямой проход, вычисление функции стоимости и градиентов для весов и смещений.

### Пример

Предположим, у нас есть следующие параметры сети:

- Входы: $a_0^0 = 0.5$, $a_1^0 = -0.2$.
- Веса скрытого слоя:  
  $$
  W^1 = \begin{pmatrix} 0.4 & -0.3 \\ 0.1 & 0.5 \\ -0.2 & 0.3 \end{pmatrix}
  $$
- Смещения скрытого слоя: $b^1 = \begin{pmatrix} 0.1 \\ -0.2 \\ 0.2 \end{pmatrix}$.
- Активационная функция скрытого слоя: сигмоида $\sigma(x) = \frac{1}{1 + e^{-x}}$.
- Веса выходного слоя:  
  $$
  W^2 = \begin{pmatrix} 0.3 & -0.1 & 0.4 \\ -0.5 & 0.2 & -0.3 \end{pmatrix}
  $$
- Смещения выходного слоя: $b^2 = \begin{pmatrix} 0.2 \\ -0.1 \end{pmatrix}$.
- Активационная функция выходного слоя: сигмоида $\sigma(x)$.
- Истинные значения на выходе: $Y_0 = 0.7$, $Y_1 = 0.3$.

### Шаг 1: Прямой проход

#### Скрытый слой:

Для первого нейрона скрытого слоя:
$$
Z_0^1 = W_{00}^1 \cdot a_0^0 + W_{01}^1 \cdot a_1^0 + b_0^1 = 0.4 \cdot 0.5 + (-0.3) \cdot (-0.2) + 0.1 = 0.2 + 0.06 + 0.1 = 0.36
$$
$$
a_0^1 = \sigma(0.36) = \frac{1}{1 + e^{-0.36}} \approx 0.589
$$

Для второго нейрона:
$$
Z_1^1 = W_{10}^1 \cdot a_0^0 + W_{11}^1 \cdot a_1^0 + b_1^1 = 0.1 \cdot 0.5 + 0.5 \cdot (-0.2) + (-0.2) = 0.05 - 0.1 - 0.2 = -0.25
$$
$$

$$
\frac{\partial C}{\partial a_0^1} = (O_0 - Y_0) \cdot act_2'(z_0^2) \cdot W_00^2 + (O_1 - Y_1) \cdot act_2'(z_1^2) \cdot W_10^2

Тогда:

\frac{\partial C}{\partial W_{00}^1} = (O_0 - Y_0) \cdot act_2'(z_0^2) \cdot W_{00}^2 \cdot act_1'(z_0^1) \cdot a_0^0 + (O_1 - Y_1) \cdot act_2'(z_1^2) \cdot W_{10}^2 \cdot act_1'(z_0^1) \cdot a_0^0

Группируя выражения, получаем итоговую формулу:
$$
\frac{\partial C}{\partial W_{00}^1} = \left[ (O_0 - Y_0) \cdot act_2'(z_0^2) \cdot W_{00}^2 + (O_1 - Y_1) \cdot act_2'(z_1^2) \cdot W_{10}^2 \right] \cdot act_1'(z_0^1) \cdot a_0^0
$$

### Итоговая формула для весов первого слоя
Таким образом, для весов первого слоя в общем виде:

Это выражение можно упростить через матричное умножение. Заметим, что каждая строка в матрице весов представляет произведение вектора ошибок $\partial E^1$ на транспонированный вектор входов $a^0$. Таким образом, итоговая формула будет:

$$
E_1^2 = O_1 - Y_1 = 0.3 - 0.0 = 0.3

Теперь у нас есть ошибки для последнего слоя, которые мы можем использовать для вычисления частных производных по весам скрытого слоя.

### Шаг 2: Частные производные по весам скрытого слоя

Теперь перейдем к вычислению частных производных по весам скрытого слоя, например, для веса $W_{00}^1$:

$$
\frac{\partial C}{\partial W_{00}^1} = \frac{\partial C}{\partial a_0^1} \cdot \frac{\partial a_0^1}{\partial z_0^1} \cdot \frac{\partial z_0^1}{\partial W_{00}^1}
$$

$$

$$
   act_1(z) = \sigma(z) = \frac{1}{1 + e^{-z}}, \quad act_1'(z) = \sigma(z)(1 - \sigma(z))
$$

   Предположим, что $z_0^1$ соответствует некоторому значению $z_0^1 = 0.4$. Тогда:
$$
   a_0^1 = \sigma(0.4) = 0.6 \quad \Rightarrow \quad act_1'(0.4) = 0.6(1 - 0.6) = 0.24
$$

2. **Производная $z_0^1$ по весу $W_{00}^1$**:
$$
   \frac{\partial z_0^1}{\partial W_{00}^1} = a_0^0 = 0.5
$$

3. **Производная функции потерь по активации**:
$$
   \frac{\partial C}{\partial a_0^1} = \frac{\partial E_0^2}{\partial a_0^1} \cdot W_{00}^2 + \frac{\partial E_1^2}{\partial a_0^1} \cdot W_{10}^2

$$
   E_0^2 = -0.3 \quad \text{и} \quad E_1^2 = 0.3
$$
   Находим:
$$
   \frac{\partial C}{\partial a_0^1} = (-0.3) \cdot 0.4 + (0.3) \cdot (-0.5) = -0.12 - 0.15 = -0.27
$$

Теперь подставим все найденные значения:
$$
\frac{\partial C}{\partial W_{00}^1} = (-0.27) \cdot 0.24 \cdot 0.5 = -0.0324
$$

$$

### Summarization of the Final Expressions

Итак, мы подошли к заключению, где важно свести все формулы, которые мы вывели, и показать, как мы вычисляем производные для обновления весов и смещений в нашей нейронной сети. Мы рассмотрели нейронную сеть с двумя входными нейронами, скрытым слоем из трех нейронов и выходным слоем из двух нейронов. В ходе вычислений мы проводили прямой проход (forward pass) и обратное распространение ошибки (backpropagation) для вычисления градиентов и обновления параметров сети.

#### Прямой проход (Forward Pass)

Прямой проход заключается в последовательном вычислении активаций на каждом слое сети. Для двухслойной нейронной сети:

1. **Вычисление для скрытого слоя**:
$$
   Z^1 = W^1 \cdot a^0 + b^1
$$
$$
   a^1 = act^1(Z^1)
$$
   Здесь $a^0$ — входной вектор (входные данные), $W^1$ — матрица весов для первого слоя, $b^1$ — вектор смещений для первого слоя, а $act^1$ — функция активации первого слоя.

2. **Вычисление для выходного слоя**:
$$
   Z^2 = W^2 \cdot a^1 + b^2
$$
$$
   a^2 = act^2(Z^2)
$$
   Здесь $a^1$ — активации скрытого слоя, $W^2$ — матрица весов для второго (выходного) слоя, $b^2$ — вектор смещений второго слоя, а $act^2$ — функция активации выходного слоя.

#### Обратное распространение ошибки (Backpropagation)

Цель обратного распространения ошибки — вычислить градиенты функции потерь по всем параметрам (весам и смещениям), чтобы обновить их для минимизации ошибки.

1. **Ошибка на выходном слое**:
$$
   \partial E^2 = (O - Y) \cdot act_2'(Z^2)
$$
   Здесь $O$ — фактический выход сети (предсказание), $Y$ — истинный выход (целевая переменная), а $act_2'(Z^2)$ — производная функции активации выходного слоя по его взвешенной сумме входов $Z^2$.

2. **Градиенты по смещениям и весам второго слоя**:
$$
   \partial b^2 = \partial E^2
$$
$$
   \partial W^2 = \partial E^2 \cdot (a^1)^T
$$
   Смещение и веса второго слоя корректируются на основе ошибки, умноженной на активации скрытого слоя.

3. **Ошибка на скрытом слое**:
$$
   \partial a^1 = (W^2)^T \cdot \partial E^2
$$
$$
   \partial E^1 = \partial a^1 \cdot act_1'(Z^1)
$$
   Ошибка на скрытом слое вычисляется через трансформированную ошибку выходного слоя, умноженную на производную функции активации скрытого слоя.

4. **Градиенты по весам и смещениям первого слоя**:
$$
   \partial W^1 = \partial E^1 \cdot (a^0)^T
$$
$$
   \partial b^1 = \partial E^1
$$

#### Обновление весов и смещений

После того как мы вычислили градиенты, мы обновляем веса и смещения с использованием метода градиентного спуска. Обновления выполняются следующим образом:

1. **Обновление весов и смещений первого слоя**:
$$
   W^1 = W^1 - \eta \cdot \partial W^1
$$
$$
   b^1 = b^1 - \eta \cdot \partial b^1
$$

2. **Обновление весов и смещений второго слоя**:
$$
   W^2 = W^2 - \eta \cdot \partial W^2
$$
$$
   b^2 = b^2 - \eta \cdot \partial b^2
$$

Здесь $\eta$ — это скорость обучения (learning rate), которая определяет, на сколько корректируются веса и смещения на каждом шаге.

### Итоговые Формулы

Теперь соберем все ключевые формулы вместе для полного понимания процесса обучения нейронной сети:

1. **Прямой проход**:
$$
   Z^1 = W^1 \cdot a^0 + b^1, \quad a^1 = act^1(Z^1)
$$
$$
   Z^2 = W^2 \cdot a^1 + b^2, \quad a^2 = act^2(Z^2)
$$

2. **Обратное распространение ошибки**:
$$
   \partial E^2 = (O - Y) \cdot act_2'(Z^2)
$$
$$
   \partial W^2 = \partial E^2 \cdot (a^1)^T, \quad \partial b^2 = \partial E^2
$$
$$
   \partial a^1 = (W^2)^T \cdot \partial E^2
$$
$$
   \partial E^1 = \partial a^1 \cdot act_1'(Z^1)
$$
$$
   \partial W^1 = \partial E^1 \cdot (a^0)^T, \quad \partial b^1 = \partial E^1
$$

3. **Обновление весов и смещений**:
$$
   W^1 = W^1 - \eta \cdot \partial W^1, \quad b^1 = b^1 - \eta \cdot \partial b^1
$$
$$
   W^2 = W^2 - \eta \cdot \partial W^2, \quad b^2 = b^2 - \eta \cdot \partial b^2
$$

Этот набор формул описывает полный процесс обучения нейронной сети, начиная с вычисления активаций, затем через обратное распространение ошибки для вычисления градиентов, и, наконец, обновление параметров с помощью градиентного спуска.


```python
import numpy as np
import matplotlib.pyplot as plt

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size, learning_rate=0.01):
        # Инициализация параметров
        self.W1 = np.random.rand(input_size, hidden_size)  # Веса для скрытого слоя
        self.b1 = np.random.rand(hidden_size)               # Смещения для скрытого слоя
        self.W2 = np.random.rand(hidden_size, output_size)  # Веса для выходного слоя
        self.b2 = np.random.rand(output_size)               # Смещения для выходного слоя
        self.learning_rate = learning_rate

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def sigmoid_derivative(self, z):
        return z * (1 - z)

    def forward_pass(self, X):
        self.Z1 = np.dot(X, self.W1) + self.b1
        self.a1 = self.sigmoid(self.Z1)
        self.Z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = self.sigmoid(self.Z2)
        return self.a2

    def backward_pass(self, X, Y):
        # Вычисление градиентов
        self.dE2 = (self.a2 - Y) * self.sigmoid_derivative(self.a2)
        self.dW2 = np.dot(self.a1.T, self.dE2)
        self.db2 = np.sum(self.dE2, axis=0)

        self.dE1 = np.dot(self.dE2, self.W2.T) * self.sigmoid_derivative(self.a1)
        self.dW1 = np.dot(X.T, self.dE1)
        self.db1 = np.sum(self.dE1, axis=0)

        # Обновление весов и смещений
        self.W1 -= self.learning_rate * self.dW1
        self.b1 -= self.learning_rate * self.db1
        self.W2 -= self.learning_rate * self.dW2
        self.b2 -= self.learning_rate * self.db2

    def train(self, X, Y, epochs):
        for epoch in range(epochs):
            self.forward_pass(X)
            self.backward_pass(X, Y)

    def predict(self, new_data):
        return self.forward_pass(new_data)

    def visualize_predictions(self, predictions, new_data):
        plt.figure(figsize=(8, 4))
        plt.bar(range(len(predictions)), predictions.flatten(), color='skyblue')
        plt.title('Предсказания для новых данных')
        plt.xlabel('Новые примеры')
        plt.ylabel('Вероятность выхода')
        plt.xticks(range(len(predictions)), [str(data) for data in new_data])
        plt.show()


# Пример использования класса
if __name__ == "__main__":
    # Обучающие данные (например, для логического И)
    X = np.array([[0, 0],
                  [0, 1],
                  [1, 0],
                  [1, 1]])

    Y = np.array([[0], [0], [0], [1]])  # Ожидаемые выходы для логического И

    # Создание и обучение нейронной сети
    nn = NeuralNetwork(input_size=2, hidden_size=3, output_size=1, learning_rate=0.1)
    nn.train(X, Y, epochs=10000)

    # Тестирование на новых данных
    new_data = np.array([[0, 0],
                         [0, 1],
                         [1, 0],
                         [1, 1],
                         [0.5, 0.5],
                         [1, 0.5],
                         [0.5, 1],
                         [0.2, 0.8]])

    predictions = nn.predict(new_data)

    # Вывод предсказаний и визуализация
    print("Предсказания для новых данных:")
    for i, pred in enumerate(predictions):
        print(f"Вход: {new_data[i]} -> Выход: {pred}")

    nn.visualize_predictions(predictions, new_data)
```
