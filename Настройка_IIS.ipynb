## Часть 2. Настройка IIS

### Введение
Традиционная схема работы с 1С:Предприятием в файловом режиме предполагает общий доступ к файлам информационной базы через протокол SMB. В данной схеме один из компьютеров выступает в роли файлового сервера, а остальные работают с размещенной в общем доступе базой. Однако эта схема имеет ряд недостатков, таких как зависимость от пропускной способности сети и вычислительных мощностей клиентских компьютеров. Особенно это заметно при использовании конфигураций на основе управляемого приложения, которые требуют больше ресурсов, чем предыдущие версии.

С возникновением потребности в удаленном доступе для мобильных устройств, веб-доступ к информационной базе становится важным. Используя управляемое приложение, можно настроить веб-сервер, позволяющий работать с базой как в файловом режиме, так и через веб-интерфейс, что существенно облегчает доступ из различных мест и с разных устройств.

### Установка IIS

1. **Открываем Панель управления.**  
   Перейдите в раздел «Программы и компоненты».

2. **Включаем IIS.**  
   Нажмите на «Включение или отключение компонентов Windows».  
   Ждите загрузки списка компонентов.  

3. **Выбираем службы IIS.**  
   Найдите элемент **Internet Information Services (Службы IIS)** и установите флажок для его включения.

4. **Устанавливаем дополнительные компоненты.**  
   Убедитесь, что установлены галочки около **ASP.NET**, и нажмите «ОК».  
   Если требуется, также выберите компоненты разработки приложений и установите флажок напротив **Расширения ISAPI**.

### Настройка IIS

5. **Открываем Диспетчер служб IIS.**  
   Перейдите в Панель управления → Администрирование → Диспетчер служб IIS.

6. **Настройка пулов приложений.**  
   В дереве слева выберите **Пулы приложений**. Щелкните правой кнопкой мыши на **DefaultAppPool** и выберите **Дополнительные параметры**. Убедитесь, что разрешены 32-разрядные приложения, установив соответствующую опцию в **True**. Эта настройка важна, если вы будете использовать 32-разрядную платформу 1С:Предприятие.

7. **Выдача прав на папки.**  
   Для папки с информационной базой установите права **Изменение** и **Чтение и выполнение** для групп **IUSR** и **IIS_IUSRS**.  
   Для папки **bin** платформы установите права **Чтение и выполнение**.

8. **Запуск 1С:Предприятия.**  
   Запустите 1С:Предприятие от имени администратора и загрузите нужную информационную базу в режиме конфигуратора.

9. **Публикация на веб-сервере.**  
   В меню выберите **Администрирование → Публикация на веб-сервере**. 1С:Предприятие автоматически определит веб-сервер (IIS) и предложит разместить веб-приложение в корне веб-сервера в каталоге с именем папки информационной базы. При необходимости вы можете изменить эти параметры.

10. **Завершение публикации.**  
    Нажмите **Опубликовать** и согласитесь на перезапуск веб-сервера.

Теперь для доступа к базе используйте адрес вида:  
`http://имя_хоста/имя_публикации`  
Например: `http://192.168.3.109/Acc30`. Откройте браузер и проверьте работу.

## Часть 2. Создание C# клиента

### Шаг 1: Создание проекта
1. Откройте Visual Studio и создайте новый проект C# (например, консольное приложение).

### Шаг 2: Установка NuGet-пакетов
Для работы с HTTP-запросами используйте библиотеку `Newtonsoft.Json`. Установите ее с помощью NuGet Package Manager:

```bash
Install-Package Newtonsoft.Json
```

### Шаг 3: Написание кода клиента

Вот пример кода для выполнения CRUD операций с REST API на C#:

```csharp
using System;
using System.Net.Http;
using System.Net.Http.Headers;
using System.Text;
using System.Threading.Tasks;
using Newtonsoft.Json;

namespace StudentApiClient
{
    class Program
    {
        private static readonly HttpClient client = new HttpClient();

        static async Task Main(string[] args)
        {
            client.BaseAddress = new Uri("http://localhost/StudentAPI/");
            client.DefaultRequestHeaders.Accept.Clear();
            client.DefaultRequestHeaders.Accept.Add(new MediaTypeWithQualityHeaderValue("application/json"));

            // Пример добавления студента
            var student = new { ФИО = "Иванов Иван Иванович", Курс = 2, Возраст = 20 };
            await AddStudentAsync(student);

            // Пример получения студента
            var studentData = await GetStudentAsync(1);
            Console.WriteLine($"Студент: {studentData.ФИО}, Курс: {studentData.Курс}, Возраст: {studentData.Возраст}");

            // Пример обновления студента
            studentData.Возраст = 21; // Обновляем возраст
            await UpdateStudentAsync(studentData);

            // Пример удаления студента
            await DeleteStudentAsync(1);
        }

        static async Task AddStudentAsync(object student)
        {
            var json = JsonConvert.SerializeObject(student);
            var content = new StringContent(json, Encoding.UTF8, "application/json");

            var response = await client.PostAsync("ДобавитьСтудента", content);
            response.EnsureSuccessStatusCode();
        }

        static async Task<dynamic> GetStudentAsync(int id)
        {
            var response = await client.GetAsync($"ПолучитьСтудента?id={id}");
            response.EnsureSuccessStatusCode();

            var json = await response.Content.ReadAsStringAsync();
            return JsonConvert.DeserializeObject(json);
        }

        static async Task UpdateStudentAsync(object student)
        {
            var json = JsonConvert.SerializeObject(student);
            var content = new StringContent(json, Encoding.UTF8, "application/json");

            var response = await client.PutAsync("ОбновитьСтудента", content);
            response.EnsureSuccessStatusCode();
        }



        static async Task DeleteStudentAsync(int id)
        {
            var response = await client.DeleteAsync($"УдалитьСтудента?id={id}");
            response.EnsureSuccessStatusCode();
        }
    }
}
```

#### Введение

#### Что такое REST API?

**REST (Representational State Transfer)** — это архитектурный стиль, который использует HTTP для обмена данными между клиентом и сервером. В REST API основные операции представляют собой взаимодействие с ресурсами (данными), которые могут быть запрашиваемы и модифицируемы с помощью методов HTTP, таких как:
- **GET** — получение данных с сервера;
- **POST** — отправка данных на сервер;
- **PUT** — обновление данных;
- **DELETE** — удаление данных.

#### Шаг 2. Взаимодействие с REST API в C#

Теперь, когда REST API на стороне 1С готово, нам нужно написать C# код для отправки запросов и получения ответов.

##### 2.1. Отправка GET-запроса

Для отправки GET-запроса в C# можно использовать библиотеку **HttpClient**. Пример кода:

```csharp
using System;
using System.Net.Http;
using System.Threading.Tasks;

class Program
{
    static async Task Main(string[] args)
    {
        // Создаем HttpClient
        using (HttpClient client = new HttpClient())
        {
            // Указываем базовый URL API
            client.BaseAddress = new Uri("http://localhost/api/");
            
            // Отправляем GET-запрос
            HttpResponseMessage response = await client.GetAsync("IntegrationService/GetClients");
            
            // Проверяем успешность запроса
            if (response.IsSuccessStatusCode)
            {
                // Чтение ответа как строки
                string result = await response.Content.ReadAsStringAsync();
                Console.WriteLine("Полученные данные: " + result);
            }
            else
            {
                Console.WriteLine("Ошибка: " + response.StatusCode);
            }
        }
    }
}
```

##### 2.2. Отправка POST-запроса

Для отправки POST-запроса с данными JSON, можно использовать следующий код:

```csharp
using System;
using System.Net.Http;
using System.Text;
using System.Threading.Tasks;

class Program
{
    static async Task Main(string[] args)
    {
        // Создаем HttpClient
        using (HttpClient client = new HttpClient())
        {
            // Указываем базовый URL API
            client.BaseAddress = new Uri("http://localhost/api/");
            
            // Создаем JSON для отправки
            var json = "{\"Имя\":\"Петр\",\"Фамилия\":\"Петров\",\"ID\":2}";
            
            // Создаем HttpContent с JSON
            StringContent content = new StringContent(json, Encoding.UTF8, "application/json");
            
            // Отправляем POST-запрос
            HttpResponseMessage response = await client.PostAsync("IntegrationService/CreateClient", content);
            
            // Проверяем успешность запроса
            if (response.IsSuccessStatusCode)
            {
                // Чтение ответа как строки
                string result = await response.Content.ReadAsStringAsync();
                Console.WriteLine("Ответ сервера: " + result);
            }
            else
            {
                Console.WriteLine("Ошибка: " + response.StatusCode);
            }
        }
    }
}
```

##### 2.3. Сериализация и десериализация JSON

Для работы с JSON в C# можно использовать библиотеку **Newtonsoft.Json** (Json.NET). Пример кода для сериализации и десериализации:

```csharp
using Newtonsoft.Json;
using System;

class Program
{
    class Client
    {
        public string Имя { get; set; }
        public string Фамилия { get; set; }
        public int ID { get; set; }
    }

    static void Main(string[] args)
    {
        // Пример сериализации объекта в JSON
        Client client = new Client { Имя = "Анна", Фамилия = "Смирнова", ID = 3 };
        string json = JsonConvert.SerializeObject(client);
        Console.WriteLine("Сериализованный JSON: " + json);

        // Пример десериализации JSON обратно в объект
        string jsonResponse = "{\"Имя\":\"Петр\",\"Фамилия\":\"Петров\",\"ID\":2}";
        Client deserializedClient = JsonConvert.DeserializeObject<Client>(jsonResponse);
        Console.WriteLine($"Клиент: {deserializedClient.Имя} {deserializedClient.Фамилия}, ID: {deserializedClient.ID}");
    }
}
```
