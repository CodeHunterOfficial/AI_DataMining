#Глава  5. Функции нескольких переменных



### 1. Введение в функции нескольких переменных

Функции нескольких переменных являются важной частью математического анализа и имеют широкое применение в различных областях, таких как физика, экономика, инженерное дело и многие другие. Функция нескольких переменных — это функция, которая принимает более одной переменной. Например, функция $f(x, y)$ — это функция двух переменных, где $x$ и $y$ — независимые переменные.

### 2. Частные производные

Частная производная функции нескольких переменных — это производная функции по одной переменной, при этом остальные переменные считаются постоянными.

#### 2.1 Определение частной производной

Для функции двух переменных $f(x, y)$ частная производная по $x$ обозначается как $\frac{\partial f}{\partial x}$ и определяется как:

$$
\frac{\partial f}{\partial x} = \lim_{h \to 0} \frac{f(x+h, y) - f(x, y)}{h}
$$

Аналогично, частная производная по $y$ обозначается как $\frac{\partial f}{\partial y}$ и определяется как:

$$
\frac{\partial f}{\partial y} = \lim_{h \to 0} \frac{f(x, y+h) - f(x, y)}{h}
$$

#### 2.2 Свойства частных производных

1. Линейность: Если $f$ и $g$ — функции, и $a$, $b$ — константы, то:

$$
\frac{\partial (af + bg)}{\partial x} = a \frac{\partial f}{\partial x} + b \frac{\partial g}{\partial x}
$$

2. Правило произведения:

$$
\frac{\partial (f g)}{\partial x} = g \frac{\partial f}{\partial x} + f \frac{\partial g}{\partial x}
$$

3. Правило цепи: Если $z = f(x, y)$, $x = g(t)$, $y = h(t)$, то:

$$
\frac{dz}{dt} = \frac{\partial f}{\partial x} \frac{dx}{dt} + \frac{\partial f}{\partial y} \frac{dy}{dt}
$$

#### 2.3 Высшие частные производные

Частные производные могут быть высшими. Например, вторая частная производная функции $f(x, y)$ по $x$ обозначается как:

$$
\frac{\partial^2 f}{\partial x^2} = \frac{\partial}{\partial x} \left( \frac{\partial f}{\partial x} \right)
$$

Существует также смешанная частная производная:

$$
\frac{\partial^2 f}{\partial x \partial y} = \frac{\partial}{\partial y} \left( \frac{\partial f}{\partial x} \right)
$$

### 3. Критические точки и экстремумы функций нескольких переменных

#### 3.1 Определение критических точек

Критическая точка функции $f(x, y)$ — это точка $(x_0, y_0)$, в которой частные производные функции равны нулю или не существуют:

$$
\frac{\partial f}{\partial x}(x_0, y_0) = 0, \quad \frac{\partial f}{\partial y}(x_0, y_0) = 0
$$

#### 3.2 Условия экстремума

Для определения типа критической точки можно использовать второй производный тест. Для функции двух переменных $f(x, y)$ вычисляются следующие величины:

$$
D = \frac{\partial^2 f}{\partial x^2} \frac{\partial^2 f}{\partial y^2} - \left( \frac{\partial^2 f}{\partial x \partial y} \right)^2
$$

1. Если $D > 0$ и $\frac{\partial^2 f}{\partial x^2} > 0$, то $(x_0, y_0)$ — минимум.
2. Если $D > 0$ и $\frac{\partial^2 f}{\partial x^2} < 0$, то $(x_0, y_0)$ — максимум.
3. Если $D < 0$, то $(x_0, y_0)$ — седловая точка.
4. Если $D = 0$, тест не дает информации о характере критической точки.

#### 3.3 Пример

Рассмотрим функцию $f(x, y) = x^2 + y^2$.

1. Найдем частные производные:

$$
\frac{\partial f}{\partial x} = 2x, \quad \frac{\partial f}{\partial y} = 2y
$$

2. Найдем критические точки:

$$
2x = 0 \quad \Rightarrow \quad x = 0, \quad 2y = 0 \quad \Rightarrow \quad y = 0
$$

Критическая точка: $(0, 0)$.

3. Вычислим $D$:

$$
\frac{\partial^2 f}{\partial x^2} = 2, \quad \frac{\partial^2 f}{\partial y^2} = 2, \quad \frac{\partial^2 f}{\partial x \partial y} = 0
$$

$$
D = 2 \cdot 2 - 0^2 = 4 > 0
$$

Так как $D > 0$ и $\frac{\partial^2 f}{\partial x^2} > 0$, то $(0, 0)$ — это минимум функции.

### 4. Градиентный спуск для многомерных функций

Градиентный спуск — это метод оптимизации, используемый для нахождения минимума функции. Он основан на использовании градиента функции.

#### 4.1 Определение градиента

Градиент функции $f(x, y)$ обозначается как $\nabla f$ и определяется как вектор частных производных:

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
$$

#### 4.2 Алгоритм градиентного спуска

1. Инициализация: Выберите начальную точку $(x_0, y_0)$.
2. Обновление: Для каждой итерации обновляем координаты:

$$
x_{n+1} = x_n - \alpha \frac{\partial f}{\partial x}(x_n, y_n)
$$
$$
y_{n+1} = y_n - \alpha \frac{\partial f}{\partial y}(x_n, y_n)
$$

где $\alpha$ — скорость обучения (параметр, определяющий размер шага).

3. Условие остановки: Итерации продолжаются до достижения заданной точности или максимального количества итераций.

#### 4.3 Пример

Рассмотрим функцию $f(x, y) = x^2 + y^2$.

1. Начнем с точки $(x_0, y_0) = (1, 1)$.
2. Вычислим градиент:

$$
\nabla f(1, 1) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = (2, 2)
$$

3. Обновим точки:

$$
x_1 = 1 - \alpha \cdot 2, \quad y_1 = 1 - \alpha \cdot 2
$$

Если выбрать $\alpha = 0.1$, то:

$$
x_1 = 1 - 0.2 = 0.8, \quad y_1 = 1 - 0.2 = 0.8
$$

4. Продолжаем процесс до тех пор, пока не достигнем минимума.

## Пример 1: Частные производные

Рассмотрим функцию двух переменных:

$$
f(x, y) = x^2y + 3xy^2
$$

### Шаг 1: Найти частные производные

#### 1.1 Частная производная по $x$:

Используем определение частной производной:

$$
\frac{\partial f}{\partial x} = \lim_{h \to 0} \frac{f(x+h, y) - f(x, y)}{h}
$$

Вычисляем $f(x+h, y)$:

$$
f(x+h, y) = (x+h)^2y + 3(x+h)y^2 = (x^2 + 2xh + h^2)y + 3(x+h)y^2
$$

Раскроем скобки:

$$
= x^2y + 2xhy + h^2y + 3xy^2 + 3hy^2
$$

Теперь подставим в формулу для производной:

$$
\frac{\partial f}{\partial x} = \lim_{h \to 0} \frac{(x^2y + 2xhy + h^2y + 3xy^2 + 3hy^2) - (x^2y + 3xy^2)}{h}
$$

Сократим:

$$
= \lim_{h \to 0} \frac{(2xhy + h^2y + 3hy^2)}{h} = \lim_{h \to 0} (2xy + hy + 3y^2)
$$

Подставим $h \to 0$:

$$
\frac{\partial f}{\partial x} = 2xy + 3y^2
$$

#### 1.2 Частная производная по $y$:

Аналогично находим частную производную:

$$
\frac{\partial f}{\partial y} = \lim_{h \to 0} \frac{f(x, y+h) - f(x, y)}{h}
$$

Вычисляем $f(x, y+h)$:

$$
f(x, y+h) = x^2(y+h) + 3x(y+h)^2 = x^2y + x^2h + 3x(y^2 + 2yh + h^2)
$$

Раскроем скобки:

$$
= x^2y + x^2h + 3xy^2 + 6xyh + 3xh^2
$$

Теперь подставим в формулу для производной:

$$
\frac{\partial f}{\partial y} = \lim_{h \to 0} \frac{(x^2y + x^2h + 3xy^2 + 6xyh + 3xh^2) - (x^2y + 3xy^2)}{h}
$$

Сократим:

$$
= \lim_{h \to 0} \frac{(x^2h + 6xyh + 3xh^2)}{h} = \lim_{h \to 0} (x^2 + 6xy + 3h)
$$

Подставим $h \to 0$:

$$
\frac{\partial f}{\partial y} = x^2 + 6xy
$$

### Ответ:

$$
\frac{\partial f}{\partial x} = 2xy + 3y^2
$$
$$
\frac{\partial f}{\partial y} = x^2 + 6xy
$$



## Пример 2: Критические точки и экстремумы

Рассмотрим функцию:

$$
f(x, y) = x^2 + y^2 - 4x - 6y
$$

### Шаг 1: Найти частные производные

#### 1.1 Частная производная по $x$:

$$
\frac{\partial f}{\partial x} = 2x - 4
$$

#### 1.2 Частная производная по $y$:

$$
\frac{\partial f}{\partial y} = 2y - 6
$$

### Шаг 2: Найти критические точки

Приравняем частные производные к нулю:

$$
2x - 4 = 0 \quad \Rightarrow \quad x = 2
$$
$$
2y - 6 = 0 \quad \Rightarrow \quad y = 3
$$

Таким образом, критическая точка:

$$
(2, 3)
$$

### Шаг 3: Использовать второй производный тест

#### 3.1 Найти вторые производные:

$$
\frac{\partial^2 f}{\partial x^2} = 2, \quad \frac{\partial^2 f}{\partial y^2} = 2, \quad \frac{\partial^2 f}{\partial x \partial y} = 0
$$

#### 3.2 Вычислить $D$:

$$
D = \frac{\partial^2 f}{\partial x^2} \cdot \frac{\partial^2 f}{\partial y^2} - \left( \frac{\partial^2 f}{\partial x \partial y} \right)^2
$$

Подставим значения:

$$
D = 2 \cdot 2 - 0^2 = 4 > 0
$$

Так как $D > 0$ и $\frac{\partial^2 f}{\partial x^2} > 0$, значит, в точке $(2, 3)$ — минимум.

### Шаг 4: Найти значение функции в критической точке

$$
f(2, 3) = 2^2 + 3^2 - 4 \cdot 2 - 6 \cdot 3 = 4 + 9 - 8 - 18 = -13
$$

### Ответ:

Критическая точка: $(2, 3)$ — минимум с значением $-13$.



## Пример 3: Градиентный спуск

Рассмотрим функцию:

$$
f(x, y) = x^2 + y^2
$$

### Шаг 1: Найти градиент

Градиент функции:

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
$$

Где:

$$
\frac{\partial f}{\partial x} = 2x, \quad \frac{\partial f}{\partial y} = 2y
$$

Таким образом:

$$
\nabla f(x, y) = (2x, 2y)
$$

### Шаг 2: Инициализация и обновление

Предположим, что мы начинаем с точки $(x_0, y_0) = (3, 4)$ и выберем скорость обучения $\alpha = 0.1$.

### Шаг 3: Первый шаг

1. Вычисляем градиент в текущей точке:

$$
\nabla f(3, 4) = (2 \cdot 3, 2 \cdot 4) = (6, 8)
$$

2. Обновляем координаты:

$$
x_1 = x_0 - \alpha \cdot \frac{\partial f}{\partial x}(3, 4) = 3 - 0.1 \cdot 6 = 3 - 0.6 = 2.4
$$
$$
y_1 = y_0 - \alpha \cdot \frac{\partial f}{\partial y}(3, 4) = 4 - 0.1 \cdot 8 = 4 - 0.8 = 3.2
$$

### Шаг 4: Второй шаг

Теперь с новой точкой $(2.4, 3.2)$:

1. Вычисляем градиент:

$$
\nabla f(2.4, 3.2) = (2 \cdot 2.4, 2 \cdot 3.2) = (4.8, 6.4)
$$

2. Обновляем координаты:

$$
x_2 = 2.4 - 0.1 \cdot 4.8 = 2.4 - 0.48 = 1.92
$$
$$
y_2 = 3.2 - 0.1 \cdot 6.4 = 3.2 - 0.64 = 2.56
$$

### Шаг 5: Продолжение

Этот процесс повторяется, обновляя $x$ и $y$, пока не достигнем требуемой точности.

### Ответ:

Градиентный спуск позволяет находить минимум функции $f(x, y) = x^2 + y^2$, начиная с произвольной точки и постепенно обновляя координаты.



Эти примеры иллюстрируют, как находить частные производные, критические точки и использовать градиентный спуск для оптимизации многомерных функций.


Давайте реализуем примеры с визуализацией на Python, используя библиотеки NumPy и Matplotlib. Мы создадим графики для функции нескольких переменных, найдем критические точки и покажем процесс градиентного спуска.

### Пример 1: Частные производные

### Пример 2: Критические точки и экстремумы

Для функции $f(x, y) = x^2 + y^2 - 4x - 6y$ мы найдем критическую точку и визуализируем ее.

### Пример 3: Градиентный спуск

Наконец, мы применим градиентный спуск к функции $f(x, y) = x^2 + y^2$ и визуализируем путь спуска.

Вот код для реализации всех трех примеров с визуализацией:


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Функция 1: f(x, y) = x^2 * y + 3 * x * y^2
def f1(x, y):
    return x**2 * y + 3 * x * y**2

# Частные производные для функции 1
def partial_x1(x, y):
    return 2 * x * y + 3 * y**2

def partial_y1(x, y):
    return x**2 + 6 * x * y  # Исправлено: x2 на x**2

# Задаем диапазон значений для первой функции
x1 = np.linspace(-3, 3, 100)
y1 = np.linspace(-3, 3, 100)
X1, Y1 = np.meshgrid(x1, y1)
Z1 = f1(X1, Y1)

# Функция 2: f(x, y) = x^2 + y^2 - 4x - 6y
def f2(x, y):
    return x**2 + y**2 - 4 * x - 6 * y  # Исправлено: x2 и y2 на x**2 и y**2

# Задаем диапазон значений для второй функции
x2 = np.linspace(-1, 5, 100)
y2 = np.linspace(-1, 5, 100)
X2, Y2 = np.meshgrid(x2, y2)
Z2 = f2(X2, Y2)

# Градиентный спуск для функции f(x, y) = x^2 + y^2
def gradient_descent(start, learning_rate, num_iterations):
    x, y = start
    trajectory = [(x, y)]

    for _ in range(num_iterations):
        grad_x = 2 * x
        grad_y = 2 * y

        x -= learning_rate * grad_x
        y -= learning_rate * grad_y
        trajectory.append((x, y))

    return trajectory

# Параметры градиентного спуска
start_point = (3, 4)
learning_rate = 0.1
num_iterations = 10

# Получение траектории
trajectory = gradient_descent(start_point, learning_rate, num_iterations)
trajectory = np.array(trajectory)

# Визуализация
fig = plt.figure(figsize=(18, 8))

# Подграфик 1: f(x, y) = x^2*y + 3xy^2
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X1, Y1, Z1, cmap='viridis', alpha=0.8)
ax1.set_title('Surface of f(x, y) = x^2*y + 3xy^2')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')
ax1.scatter(2, 3, f1(2, 3), color='red', s=100)  # Критическая точка
ax1.text(2, 3, f1(2, 3), 'Critical Point (2, 3)', color='red')

# Подграфик 2: f(x, y) = x^2 + y^2 - 4x - 6y с градиентным спуском
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X2, Y2, Z2, cmap='plasma', alpha=0.8)
ax2.set_title('Surface of f(x, y) = x^2 + y^2 - 4x - 6y with Gradient Descent')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('f(x, y)')

# Отображаем траекторию градиентного спуска
ax2.plot(trajectory[:, 0], trajectory[:, 1], f2(trajectory[:, 0], trajectory[:, 1]),
         marker='o', color='red', markersize=5, label='Gradient Descent Path')
ax2.legend()

plt.tight_layout()
plt.show()
```
