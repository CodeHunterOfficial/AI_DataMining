#Глава 3. Интегралы


#### 1. Введение в ряды

Ряд — это выражение вида
$$
\sum_{n=1}^{\infty} a_n = a_1 + a_2 + a_3 + \ldots,
$$
где $a_n$ — члены ряда.

Если последовательность частичных сумм ряда
$$
S_N = \sum_{n=1}^{N} a_n
$$
стремится к конечному пределу $S$ при $N \to \infty$, то ряд называется сходящимся, и пишут
$$
\sum_{n=1}^{\infty} a_n = S.
$$
Если предела нет, то ряд называется расходящимся.

#### 2. Сходимость числовых рядов

Для исследования сходимости числовых рядов разработаны различные тесты и признаки.

##### 2.1 Признак Коши

Ряд $\sum_{n=1}^{\infty} a_n$ сходится тогда и только тогда, когда для любого $\epsilon > 0$ существует такое натуральное число $N$, что для всех $n > m \geq N$ выполняется неравенство
$$
|a_{m+1} + a_{m+2} + \ldots + a_n| < \epsilon.
$$

##### 2.2 Необходимый признак сходимости

Если ряд $\sum_{n=1}^{\infty} a_n$ сходится, то его общий член стремится к нулю:
$$
\lim_{n \to \infty} a_n = 0.
$$
Однако обратное утверждение неверно: если $a_n \to 0$, это не гарантирует сходимости ряда.

##### 2.3 Признак сравнения

Пусть даны два ряда $\sum_{n=1}^{\infty} a_n$ и $\sum_{n=1}^{\infty} b_n$ с положительными членами:

- Если существует постоянная $C > 0$, такая что для всех $n$ выполняется $0 \leq a_n \leq C b_n$, и ряд $\sum_{n=1}^{\infty} b_n$ сходится, то сходится и ряд $\sum_{n=1}^{\infty} a_n$.
- Если $b_n \leq C a_n$ и ряд $\sum_{n=1}^{\infty} b_n$ расходится, то расходится и ряд $\sum_{n=1}^{\infty} a_n$.

##### 2.4 Признак Д’Аламбера (признак Лагранжа)

Если существует предел
$$
\lim_{n \to \infty} \left|\frac{a_{n+1}}{a_n}\right| = L,
$$
то:

- При $L < 1$ ряд $\sum_{n=1}^{\infty} a_n$ сходится.
- При $L > 1$ ряд расходится.
- При $L = 1$ признак не дает ответа о сходимости.

##### 2.5 Признак Коши — Маклорена (признак корня)

Если существует предел
$$
\lim_{n \to \infty} \sqrt[n]{|a_n|} = L,
$$
то:

- При $L < 1$ ряд $\sum_{n=1}^{\infty} a_n$ сходится.
- При $L > 1$ ряд расходится.
- При $L = 1$ признак не дает ответа.

#### 3. Функциональные ряды

Функциональным рядом называется ряд вида
$$
\sum_{n=0}^{\infty} f_n(x),
$$
где $f_n(x)$ — функции, зависящие от переменной $x$.

##### 3.1 Поточечная сходимость

Ряд $\sum_{n=0}^{\infty} f_n(x)$ поточечно сходится на множестве $X$, если для любого фиксированного $x \in X$ ряд чисел $\sum_{n=0}^{\infty} f_n(x)$ сходится.

##### 3.2 Равномерная сходимость

Ряд $\sum_{n=0}^{\infty} f_n(x)$ равномерно сходится на множестве $X$, если для любого $\epsilon > 0$ существует такое $N$, что для всех $n > N$ и всех $x \in X$ выполняется
$$
\left| \sum_{k=n+1}^{\infty} f_k(x) \right| < \epsilon.
$$

##### 3.3 Признак Вейерштрасса

Если существует последовательность положительных чисел $M_n$, такая что $|f_n(x)| \leq M_n$ для всех $x \in X$, и ряд $\sum_{n=0}^{\infty} M_n$ сходится, то ряд $\sum_{n=0}^{\infty} f_n(x)$ равномерно сходится на $X$.

#### 4. Ряды Тейлора и их применение

Рядом Тейлора функции $f(x)$ в точке $a$ называется ряд вида
$$
\sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!} (x - a)^n,
$$
где $f^{(n)}(a)$ — $n$-я производная функции $f(x)$ в точке $a$.

##### 4.1 Ряд Маклорена

Если $a = 0$, то ряд Тейлора принимает вид
$$
\sum_{n=0}^{\infty} \frac{f^{(n)}(0)}{n!} x^n,
$$
и называется рядом Маклорена.

##### 4.2 Остаточный член ряда Тейлора

Если $f(x)$ имеет $n+1$ производную, то можно записать остаточный член в форме Лагранжа:
$$
R_n(x) = \frac{f^{(n+1)}(\xi)}{(n+1)!} (x - a)^{n+1},
$$
где $\xi$ лежит между $a$ и $x$.

##### 4.3 Примеры разложения в ряд Тейлора

1. Экспоненциальная функция $e^x$:
 $$
   e^x = \sum_{n=0}^{\infty} \frac{x^n}{n!}.
 $$

2. Синус и косинус:
 $$
   \sin x = \sum_{n=0}^{\infty} (-1)^n \frac{x^{2n+1}}{(2n+1)!},
 $$
 $$
   \cos x = \sum_{n=0}^{\infty} (-1)^n \frac{x^{2n}}{(2n)!}.
 $$

3. Логарифмическая функция:
 $$
   \ln(1+x) = \sum_{n=1}^{\infty} (-1)^{n+1} \frac{x^n}{n}, \quad |x| < 1.
 $$

##### 4.4 Применение рядов Тейлора

- Приближенные вычисления значений функций.
- Решение дифференциальных уравнений.
- Аппроксимация функций в окрестностях точек.
- Оценка погрешности при вычислении функций.

#### 5. Свойства и теоремы, связанные с рядами

##### Теорема Абеля

Если степенной ряд
$$
\sum_{n=0}^{\infty} a_n x^n
$$
сходится при $x = R$, то он равномерно сходится на отрезке $[0, R]$.

##### Теорема о почленном дифференцировании и интегрировании

Если функциональный ряд $\sum_{n=0}^{\infty} f_n(x)$ равномерно сходится, то его можно почленно интегрировать и дифференцировать:
$$
\left( \sum_{n=0}^{\infty} f_n(x) \right)' = \sum_{n=0}^{\infty} f_n'(x),
$$
$$
\int \left( \sum_{n=0}^{\infty} f_n(x) \right) dx = \sum_{n=0}^{\infty} \int f_n(x) dx.
$$

### Пример 1: Исследование сходимости ряда

Рассмотрим ряд:
$$
\sum_{n=1}^{\infty} \frac{1}{n^2}.
$$

Шаг 1: Признак сравнения.

Для исследования сходимости мы можем воспользоваться признаком сравнения с известным рядом. Известно, что ряд
$$
\sum_{n=1}^{\infty} \frac{1}{n^p}
$$
сходится при $p > 1$.

Шаг 2: Определяем $p$.

В нашем случае $p = 2$, и поскольку $2 > 1$, мы предполагаем, что ряд $\sum_{n=1}^{\infty} \frac{1}{n^2}$ должен сходиться.

Шаг 3: Применяем признак сравнения.

Сравниваем $a_n = \frac{1}{n^2}$ с $b_n = \frac{1}{n^p}$, где $p = 2$. Так как
$$
0 \leq \frac{1}{n^2} \leq \frac{1}{n^2} \text{ для всех } n \geq 1,
$$
и ряд $\sum_{n=1}^{\infty} b_n$ сходится, то по признаку сравнения ряд $\sum_{n=1}^{\infty} \frac{1}{n^2}$ также сходится.

Ответ:
Ряд $\sum_{n=1}^{\infty} \frac{1}{n^2}$ сходится.



### Пример 2: Разложение функции в ряд Тейлора

Рассмотрим функцию $f(x) = e^x$ и разложим её в ряд Тейлора в точке $a = 0$.

Шаг 1: Находим производные.

Вычисляем производные функции:
$$
f^{(n)}(x) = e^x \implies f^{(n)}(0) = e^0 = 1.
$$

Шаг 2: Записываем ряд Тейлора.

Используем формулу ряда Тейлора:
$$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(0)}{n!} x^n = \sum_{n=0}^{\infty} \frac{1}{n!} x^n.
$$

Шаг 3: Убедимся в сходимости.

Ряд
$$
\sum_{n=0}^{\infty} \frac{x^n}{n!}
$$
является степенным рядом, и мы можем использовать признак Д’Аламбера для его сходимости. Рассмотрим предел:
$$
\lim_{n \to \infty} \left| \frac{a_{n+1}}{a_n} \right| = \lim_{n \to \infty} \frac{x^{n+1}/(n+1)!}{x^n/n!} = \lim_{n \to \infty} \frac{|x|}{n+1} = 0.
$$
Поскольку предел равен 0, ряд сходится для всех $x \in \mathbb{R}$.

Ответ:
Ряд Тейлора для функции $e^x$:
$$
e^x = \sum_{n=0}^{\infty} \frac{x^n}{n!}.
$$



### Пример 3: Исследование сходимости функционального ряда

Рассмотрим ряд:
$$
\sum_{n=1}^{\infty} \frac{x^n}{n^2}.
$$

Шаг 1: Определяем область сходимости.

Это степенной ряд, и мы можем использовать признак Д’Аламбера. Находим
$$
a_n = \frac{x^n}{n^2}.
$$
Вычисляем
$$
\lim_{n \to \infty} \left| \frac{a_{n+1}}{a_n} \right| = \lim_{n \to \infty} \frac{|x|^{n+1}/(n+1)^2}{|x|^n/n^2} = \lim_{n \to \infty} |x| \cdot \frac{n^2}{(n+1)^2} = |x| \cdot 1 = |x|.
$$

Шаг 2: Условия сходимости.

Ряд сходится, если $|x| < 1$, а расходится, если $|x| > 1$. Если $|x| = 1$, нужно проверить отдельно.

Шаг 3: Проверка на границе.

- При $x = 1$:
$$
\sum_{n=1}^{\infty} \frac{1}{n^2},
$$
как мы уже установили, сходится.

- При $x = -1$:
$$
\sum_{n=1}^{\infty} \frac{(-1)^n}{n^2},
$$
это знакочередующийся ряд, который также сходится по критерию Лейбница.

Ответ:
Ряд $\sum_{n=1}^{\infty} \frac{x^n}{n^2}$ сходится для $|x| \leq 1$.



### Пример 4: Разложение функции в ряд Тейлора (логарифм)

Рассмотрим функцию $f(x) = \ln(1+x)$ и разложим её в ряд Тейлора в точке $a = 0$.

Шаг 1: Находим производные.

Вычисляем производные:
- $f'(x) = \frac{1}{1+x}$,
- $f''(x) = -\frac{1}{(1+x)^2}$,
- $f^{(n)}(x) = (-1)^{n-1} \frac{(n-1)!}{(1+x)^n}$.

Шаг 2: Находим значение производных в точке $a = 0$.

$$
f^{(n)}(0) = (-1)^{n-1} (n-1)!.
$$

Шаг 3: Записываем ряд Тейлора.

Используем формулу:
$$
f(x) = \sum_{n=1}^{\infty} \frac{f^{(n)}(0)}{n!} x^n = \sum_{n=1}^{\infty} \frac{(-1)^{n-1} (n-1)!}{n!} x^n = \sum_{n=1}^{\infty} (-1)^{n-1} \frac{x^n}{n}.
$$

Шаг 4: Определяем область сходимости.

Этот ряд сходится для $|x| < 1$ и на границе $x = 1$ (сходится, как $\sum_{n=1}^{\infty} \frac{1}{n}$ и расходится на $x = -1$ (так как $\ln(1+x)$ определен только для $x > -1$).

Ответ:
Ряд Тейлора для функции $\ln(1+x)$:
$$
\ln(1+x) = \sum_{n=1}^{\infty} (-1)^{n-1} \frac{x^n}{n}, \quad |x| < 1.
$$

### Пример 1: Исследование сходимости ряда $\sum_{n=1}^{\infty} \frac{1}{n^2}$

В этом примере мы будем вычислять частичные суммы ряда и визуализировать их.


```python
import numpy as np
import matplotlib.pyplot as plt

# Количество членов ряда
N = 1000

# Вычисляем частичные суммы
n = np.arange(1, N + 1)
partial_sums = np.cumsum(1 / n**2)

# Настройка графика
plt.figure(figsize=(10, 6))
plt.plot(n, partial_sums, label='Частичные суммы', color='blue')
plt.axhline(y=np.pi**2 / 6, color='red', linestyle='--', label='Предел (π²/6)')
plt.title('Сходимость ряда $\\sum_{n=1}^{\\infty} \\frac{1}{n^2}$')
plt.xlabel('N')
plt.ylabel('Сумма')
plt.legend()
plt.grid()
plt.show()
```



### Пример 2: Разложение функции $e^x$ в ряд Тейлора

Мы будем вычислять значения функции $e^x$ и её разложения в ряд Тейлора, затем визуализируем их.


```python
# Функция e^x
def exp_taylor(x, n_terms):
    return sum(x**n / np.math.factorial(n) for n in range(n_terms))

# Параметры
x_values = np.linspace(-2, 2, 100)
actual_values = np.exp(x_values)
taylor_values = [exp_taylor(x, 10) for x in x_values]  # 10 членов ряда

# Настройка графика
plt.figure(figsize=(10, 6))
plt.plot(x_values, actual_values, label='$e^x$', color='blue')
plt.plot(x_values, taylor_values, label='Ряд Тейлора (10 членов)', color='orange', linestyle='--')
plt.title('Сравнение $e^x$ и его ряда Тейлора')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```



### Пример 3: Исследование сходимости функционального ряда $\sum_{n=1}^{\infty} \frac{x^n}{n^2}$

В этом примере мы будем визуализировать поведение ряда для различных значений $x$.



```python
# Количество членов ряда
N = 100

# Функция для вычисления частичных сумм ряда
def functional_series_sum(x, n_terms):
    return sum(x**n / n**2 for n in range(1, n_terms + 1))

# Значения x
x_values = np.linspace(-1, 1, 100)
partial_sums = np.array([functional_series_sum(x, N) for x in x_values])

# Настройка графика
plt.figure(figsize=(10, 6))
plt.plot(x_values, partial_sums, label='Частичные суммы', color='green')
plt.title('Сходимость ряда $\\sum_{n=1}^{\\infty} \\frac{x^n}{n^2}$')
plt.xlabel('x')
plt.ylabel('Сумма')
plt.legend()
plt.grid()
plt.show()
```




### Пример 4: Разложение функции $\ln(1+x)$ в ряд Тейлора

Мы будем вычислять значения функции $\ln(1+x)$ и её разложения в ряд Тейлора.



```python
import numpy as np
import matplotlib.pyplot as plt

# Функция ln(1+x)
def ln_taylor(x, n_terms):
    return sum((-1)**(n - 1) * x**n / n for n in range(1, n_terms + 1))

# Значения x для ln(1+x)
x_values = np.linspace(0, 0.9, 100)  # Только положительные x
actual_values = np.log(1 + x_values)
taylor_values = [ln_taylor(x, 10) for x in x_values]  # 10 членов ряда

# Настройка графика
plt.figure(figsize=(10, 6))
plt.plot(x_values, actual_values, label='$\ln(1+x)$', color='blue')
plt.plot(x_values, taylor_values, label='Ряд Тейлора (10 членов)', color='orange', linestyle='--')
plt.title('Сравнение $\\ln(1+x)$ и его ряда Тейлора')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```
