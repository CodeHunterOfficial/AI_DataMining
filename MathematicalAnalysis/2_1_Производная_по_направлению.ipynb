# 2.1. Производная по направлению

## Введение

Производная по направлению — это обобщение понятия производной функции нескольких переменных, которое позволяет изучать изменение функции в произвольном направлении в данной точке. В этой лекции мы рассмотрим основные определения, свойства, теоремы и примеры, связанные с производной по направлению.

## Определение

Рассмотрим функцию $f: \mathbb{R}^n \to \mathbb{R}$, определённую в некоторой окрестности точки $\mathbf{x_0} \in \mathbb{R}^n$. Пусть $\mathbf{u}$ — это единичный вектор направления, т.е. $\|\mathbf{u}\| = 1$.

### Определение производной по направлению

Производная функции $f$ в точке $\mathbf{x_0}$ по направлению $\mathbf{u}$ определяется как предел:

$$
D_{\mathbf{u}} f(\mathbf{x_0}) = \lim_{h \to 0} \frac{f(\mathbf{x_0} + h\mathbf{u}) - f(\mathbf{x_0})}{h}
$$

Если этот предел существует, мы говорим, что $f$ дифференцируема в направлении $\mathbf{u}$ в точке $\mathbf{x_0}$.

## Связь с градиентом

Если функция $f$ имеет градиент $\nabla f(\mathbf{x_0})$, то производная по направлению связана с градиентом следующим образом:

$$
D_{\mathbf{u}} f(\mathbf{x_0}) = \nabla f(\mathbf{x_0}) \cdot \mathbf{u}
$$

где $\cdot$ обозначает скалярное произведение.

### Свойства градиента

1. **Линейность**: Для любых $\alpha, \beta \in \mathbb{R}$ и векторов $\mathbf{u}, \mathbf{v} \in \mathbb{R}^n$:

$$
\nabla (\alpha f + \beta g) = \alpha \nabla f + \beta \nabla g
$$

2. **Правило произведения**: Если $f$ и $g$ — функции, то:

$$
\nabla (fg) = f \nabla g + g \nabla f
$$

## Теоремы

### Теорема 1: Существование производной по направлению

Если функция $f$ непрерывна в точке $\mathbf{x_0}$ и дифференцируема в окрестности этой точки, то производная по направлению $\mathbf{u}$ существует для любого направления.

### Теорема 2: Формула производной по направлению

Если $f$ имеет градиент в точке $\mathbf{x_0}$, то производная по направлению $\mathbf{u}$ выражается через градиент:

$$
D_{\mathbf{u}} f(\mathbf{x_0}) = \nabla f(\mathbf{x_0}) \cdot \mathbf{u}
$$

### Теорема 3: Критические точки

Если $D_{\mathbf{u}} f(\mathbf{x_0}) = 0$ для всех направлений $\mathbf{u}$, то точка $\mathbf{x_0}$ является критической точкой функции $f$.

## Примеры

### Пример 1

Рассмотрим функцию $f(x, y) = x^2 + y^2$. Найдём производную по направлению $\mathbf{u} = (1/\sqrt{2}, 1/\sqrt{2})$ в точке $(1, 1)$.

1. Найдём градиент:

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = (2x, 2y)
$$

2. Вычислим градиент в точке $(1, 1)$:

$$
\nabla f(1, 1) = (2 \cdot 1, 2 \cdot 1) = (2, 2)
$$

3. Найдём производную по направлению:

$$
D_{\mathbf{u}} f(1, 1) = \nabla f(1, 1) \cdot \mathbf{u} = (2, 2) \cdot \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right) = 2\sqrt{2}
$$

### Пример 2

Рассмотрим функцию $f(x, y) = e^{xy}$. Найдём производную по направлению $\mathbf{u} = (1, 0)$ в точке $(0, 1)$.

1. Найдём градиент:

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = \left( y e^{xy}, x e^{xy} \right)
$$

2. Вычислим градиент в точке $(0, 1)$:

$$
\nabla f(0, 1) = (1 e^{0}, 0 e^{0}) = (1, 0)
$$

3. Найдём производную по направлению:

$$
D_{\mathbf{u}} f(0, 1) = \nabla f(0, 1) \cdot \mathbf{u} = (1, 0) \cdot (1, 0) = 1
$$

## Заключение

Производная по направлению является важным понятием в математическом анализе функций нескольких переменных. Она позволяет оценивать, как функция изменяется в разных направлениях и играет ключевую роль в оптимизации и исследовании свойств функций.

### Резюме основных формул

1. Производная по направлению:

$$
D_{\mathbf{u}} f(\mathbf{x_0}) = \lim_{h \to 0} \frac{f(\mathbf{x_0} + h\mathbf{u}) - f(\mathbf{x_0})}{h}
$$

2. Связь с градиентом:

$$
D_{\mathbf{u}} f(\mathbf{x_0}) = \nabla f(\mathbf{x_0}) \cdot \mathbf{u}
$$

3. Условия существования производной по направлению.

Производная по направлению находит применение в различных областях науки и техники, включая физику, экономику и инженерные науки, что делает её неотъемлемой частью современного математического аппарата.

### Пример 1: Производная функции $f(x, y) = x^2 + y^2$

**Задача**: Найдите производную функции $f(x, y) = x^2 + y^2$ в точке $(1, 2)$ по направлению вектора $\mathbf{u} = \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right)$.

#### Решение:

1. **Найти градиент** функции $f$:

   $$
   \nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = \left( 2x, 2y \right)
   $$

2. **Вычислить градиент** в точке $(1, 2)$:

   $$
   \nabla f(1, 2) = (2 \cdot 1, 2 \cdot 2) = (2, 4)
   $$

3. **Найти производную по направлению** $\mathbf{u}$:

   $$
   D_{\mathbf{u}} f(1, 2) = \nabla f(1, 2) \cdot \mathbf{u} = (2, 4) \cdot \left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right)
   $$

   $$
   D_{\mathbf{u}} f(1, 2) = 2 \cdot \frac{1}{\sqrt{2}} + 4 \cdot \frac{1}{\sqrt{2}} = \frac{2 + 4}{\sqrt{2}} = \frac{6}{\sqrt{2}} = 3\sqrt{2}
   $$

**Ответ**: Производная функции $f(x, y) = x^2 + y^2$ в точке $(1, 2)$ по направлению $\mathbf{u}$ равна $3\sqrt{2}$.



### Пример 2: Производная функции $f(x, y) = e^{xy}$

**Задача**: Найдите производную функции $f(x, y) = e^{xy}$ в точке $(0, 1)$ по направлению вектора $\mathbf{u} = (1, 0)$.

#### Решение:

1. **Найти градиент** функции $f$:

   $$
   \nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
   $$
   Сначала вычислим частные производные:

   $$
   \frac{\partial f}{\partial x} = y e^{xy}, \quad \frac{\partial f}{\partial y} = x e^{xy}
   $$

   Таким образом,

   $$
   \nabla f(x, y) = (y e^{xy}, x e^{xy})
   $$

2. **Вычислить градиент** в точке $(0, 1)$:

   $$
   \nabla f(0, 1) = (1 \cdot e^{0}, 0 \cdot e^{0}) = (1, 0)
   $$

3. **Найти производную по направлению** $\mathbf{u} = (1, 0)$:

   $$
   D_{\mathbf{u}} f(0, 1) = \nabla f(0, 1) \cdot \mathbf{u} = (1, 0) \cdot (1, 0)
   $$

   $$
   D_{\mathbf{u}} f(0, 1) = 1 \cdot 1 + 0 \cdot 0 = 1
   $$

**Ответ**: Производная функции $f(x, y) = e^{xy}$ в точке $(0, 1)$ по направлению $\mathbf{u}$ равна $1$.



### Пример 3: Производная функции $f(x, y) = \sin(xy)$

**Задача**: Найдите производную функции $f(x, y) = \sin(xy)$ в точке $(1, \frac{\pi}{2})$ по направлению вектора $\mathbf{u} = \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right)$.

#### Решение:

1. **Найти градиент** функции $f$:

   $$
   \nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
   $$
   Сначала вычислим частные производные:

   $$
   \frac{\partial f}{\partial x} = y \cos(xy), \quad \frac{\partial f}{\partial y} = x \cos(xy)
   $$

   Таким образом,

   $$
   \nabla f(x, y) = \left( y \cos(xy), x \cos(xy) \right)
   $$

2. **Вычислить градиент** в точке $(1, \frac{\pi}{2})$:

   $$
   \nabla f\left(1, \frac{\pi}{2}\right) = \left(\frac{\pi}{2} \cos(1 \cdot \frac{\pi}{2}), 1 \cdot \cos(1 \cdot \frac{\pi}{2})\right) = \left(\frac{\pi}{2} \cdot 0, 1 \cdot 0\right) = (0, 0)
   $$

3. **Найти производную по направлению** $\mathbf{u} = \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right)$:

   $$
   D_{\mathbf{u}} f\left(1, \frac{\pi}{2}\right) = \nabla f\left(1, \frac{\pi}{2}\right) \cdot \mathbf{u} = (0, 0) \cdot \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right) = 0
   $$

**Ответ**: Производная функции $f(x, y) = \sin(xy)$ в точке $(1, \frac{\pi}{2})$ по направлению $\mathbf{u}$ равна $0$.



### Пример 4: Производная функции $f(x, y) = x^3 - 3xy^2$

**Задача**: Найдите производную функции $f(x, y) = x^3 - 3xy^2$ в точке $(1, 1)$ по направлению вектора $\mathbf{u} = (1, -1)$.

#### Решение:

1. **Найти градиент** функции $f$:

   $$
   \nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
   $$

   Вычисляем частные производные:

   $$
   \frac{\partial f}{\partial x} = 3x^2 - 3y^2, \quad \frac{\partial f}{\partial y} = -6xy
   $$

   Таким образом,

   $$
   \nabla f(x, y) = \left( 3x^2 - 3y^2, -6xy \right)
   $$

2. **Вычислить градиент** в точке $(1, 1)$:

   $$
   \nabla f(1, 1) = \left( 3 \cdot 1^2 - 3 \cdot 1^2, -6 \cdot 1 \cdot 1 \right) = (3 - 3, -6) = (0, -6)
   $$

3. **Найти производную по направлению** $\mathbf{u} = (1, -1)$:

   Сначала нормализуем $\mathbf{u}$:

   $$
   \|\mathbf{u}\| = \sqrt{1^2 + (-1)^2} = \sqrt{2} \quad \Rightarrow \quad \mathbf{u} = \left( \frac{1}{\sqrt{2}}, \frac{-1}{\sqrt{2}} \right)
   $$

   Затем:

   $$
   D_{\mathbf{u}} f(1, 1) = \nabla f(1, 1) \cdot \mathbf{u} = (0, -6) \cdot \left( \frac{1}{\sqrt{2}}, \frac{-1}{\sqrt{2}} \right)
   $$

   Вычисляем скалярное произведение:

   $$
   D_{\mathbf{u}} f(1, 1) = 0 \cdot \frac{1}{\sqrt{2}} + (-6) \cdot \left(-\frac{1}{\sqrt{2}}\right) = 0 + \frac{6}{\sqrt{2}} = \frac{6\sqrt{2}}{2} = 3\sqrt{2}
   $$

Таким образом, производная функции $f(x, y)$ в точке $(1, 1)$ по направлению вектора $\mathbf{u} = (1, -1)$ равна $3\sqrt{2}$.



Давайте реализуем указанные примеры на Python, включая визуализацию результатов. Для визуализации мы будем использовать библиотеку `matplotlib` и `numpy`. Убедитесь, что у вас установлены эти библиотеки. Если они не установлены, вы можете установить их с помощью:





### Пример 1: Производная функции $f(x, y) = x^2 + y^2$



```python
import numpy as np
import matplotlib.pyplot as plt

# Функция
def f(x, y):
    return x**2 + y**2

# Градиент функции
def gradient_f(x, y):
    return np.array([2*x, 2*y])

# Точка и направление
point = np.array([1, 2])
direction = np.array([1/np.sqrt(2), 1/np.sqrt(2)])

# Вычисление градиента в точке
grad = gradient_f(point[0], point[1])
direction = direction / np.linalg.norm(direction)  # Нормализация направления

# Производная по направлению
D_u = np.dot(grad, direction)

# Визуализация
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.figure(figsize=(10, 8))
plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(label='f(x, y)')
plt.scatter(point[0], point[1], color='red', label='Точка (1, 2)')
plt.quiver(point[0], point[1], direction[0], direction[1],
           angles='xy', scale_units='xy', scale=1, color='blue', label='Направление')
plt.title('Контуры функции f(x, y) = x² + y²')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()

print(f"Производная по направлению: D_u = {D_u:.2f}")
```



### Пример 2: Производная функции $f(x, y) = e^{xy}$



```python
# Функция
def f_exp(x, y):
    return np.exp(x * y)

# Градиент функции
def gradient_f_exp(x, y):
    return np.array([y * np.exp(x * y), x * np.exp(x * y)])

# Точка и направление
point_exp = np.array([0, 1])
direction_exp = np.array([1, 0])

# Вычисление градиента в точке
grad_exp = gradient_f_exp(point_exp[0], point_exp[1])

# Производная по направлению
D_u_exp = np.dot(grad_exp, direction_exp)

# Визуализация
x_exp = np.linspace(-2, 2, 100)
y_exp = np.linspace(0, 2, 100)
X_exp, Y_exp = np.meshgrid(x_exp, y_exp)
Z_exp = f_exp(X_exp, Y_exp)

plt.figure(figsize=(10, 8))
plt.contour(X_exp, Y_exp, Z_exp, levels=20, cmap='plasma')
plt.colorbar(label='f(x, y)')
plt.scatter(point_exp[0], point_exp[1], color='red', label='Точка (0, 1)')
plt.quiver(point_exp[0], point_exp[1], direction_exp[0], direction_exp[1],
           angles='xy', scale_units='xy', scale=1, color='blue', label='Направление')
plt.title('Контуры функции f(x, y) = e^(xy)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.xlim(-2, 2)
plt.ylim(0, 2)
plt.show()

print(f"Производная по направлению: D_u = {D_u_exp:.2f}")
```
