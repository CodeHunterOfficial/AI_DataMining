#2.2. Метод градиентного спуска

Метод градиентного спуска является фундаментальным алгоритмом оптимизации, используемым для минимизации функций. Он широко применяется в машинном обучении, численных методах и других областях, где требуется найти минимум функции.

### Основные понятия и формулировки

#### Определение

Пусть дана дифференцируемая функция $f: \mathbb{R}^n \rightarrow \mathbb{R}$. Цель метода градиентного спуска — найти такую точку $\mathbf{x}^*$, что $f(\mathbf{x}^*)$ является локальным минимумом функции $f$.

#### Градиент функции

Градиент функции $f$ в точке $\mathbf{x} \in \mathbb{R}^n$ определяется как вектор частных производных:

$$ \nabla f(\mathbf{x}) = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \ldots, \frac{\partial f}{\partial x_n} \right)^T $$

#### Алгоритм градиентного спуска

1. **Инициализация**: Выбирается начальная точка $\mathbf{x}^{(0)} \in \mathbb{R}^n$.

2. **Шаги итерации**:
   - На каждом шаге $k$ вычисляется градиент $\nabla f(\mathbf{x}^{(k)})$.
   - Выбирается направление спуска $-\nabla f(\mathbf{x}^{(k)})$.
   - Выполняется шаг в направлении антиградиента с шаговым размером $\alpha_k$:
     $$ \mathbf{x}^{(k+1)} = \mathbf{x}^{(k)} - \alpha_k \nabla f(\mathbf{x}^{(k)}) $$

3. **Условие остановки**: Алгоритм продолжает итерации до выполнения критерия остановки, например, до достижения заданной точности или после заданного числа итераций.

#### Выбор шага $\alpha_k$

Выбор правильного шага $\alpha_k$ играет важную роль:
- **Фиксированный шаг**: $\alpha_k = \alpha$, где $\alpha$ — константа.
- **Линейный поиск**: $\alpha_k$ выбирается на каждой итерации с использованием линейного поиска или метода золотого сечения.

#### Условия сходимости

Метод градиентного спуска сходится к стационарной точке (где $\nabla f(\mathbf{x}^*) = 0$) при выполнении условий:
- $f$ выпуклая или сильно выпуклая функция.
- Правильный выбор шага $\alpha_k$.

### Примеры и приложения

Метод градиентного спуска широко используется в машинном обучении для обучения моделей, таких как линейная регрессия, логистическая регрессия, нейронные сети и другие алгоритмы оптимизации параметров.

### Заключение

Метод градиентного спуска — мощный инструмент для оптимизации функций, обладающий различными модификациями и условиями применимости. Правильная настройка параметров и выбор шага играют ключевую роль в его эффективности и скорости сходимости.


Давайте рассмотрим метод градиентного спуска на конкретных примерах, включая линейную регрессию и простую функцию с двумя переменными.

### Пример 1: Линейная регрессия

**Задача:** Найти параметры линейной модели для предсказания зависимой переменной $y$ от независимой переменной $x$ по данным, представленным в виде пар $(x_i, y_i)$.

**Модель:** Линейная модель описывается уравнением:

$$
y = w_0 + w_1 x
$$

где $w_0$ — смещение (intercept), $w_1$ — коэффициент (slope).

**Функция ошибки:** Для оценки качества модели используем среднеквадратичную ошибку (MSE):

$$
J(w_0, w_1) = \frac{1}{m} \sum_{i=1}^{m} (y_i - (w_0 + w_1 x_i))^2
$$

где $m$ — количество наблюдений.

**Градиенты:** Для применения метода градиентного спуска вычислим градиенты:

$$
\frac{\partial J}{\partial w_0} = -\frac{2}{m} \sum_{i=1}^{m} (y_i - (w_0 + w_1 x_i))
$$

$$
\frac{\partial J}{\partial w_1} = -\frac{2}{m} \sum_{i=1}^{m} (y_i - (w_0 + w_1 x_i)) x_i
$$

**Алгоритм градиентного спуска:**

1. **Инициализация:** Начальные значения $w_0$ и $w_1$ (например, 0).
2. **Итерация:** Обновление параметров с использованием градиентов:

$$
w_0^{(new)} = w_0^{(old)} - \alpha \frac{\partial J}{\partial w_0}
$$
$$
w_1^{(new)} = w_1^{(old)} - \alpha \frac{\partial J}{\partial w_1}
$$

3. **Повторяйте** шаги, пока не достигнете желаемой точности или максимального числа итераций.


```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация данных
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Гиперпараметры
alpha = 0.01  # Скорость обучения
iterations = 1000
m = len(y)

# Инициализация параметров
w0, w1 = 0, 0

# Градиентный спуск
for _ in range(iterations):
    y_pred = w0 + w1 * X
    w0 -= alpha * (-2/m) * np.sum(y - y_pred)
    w1 -= alpha * (-2/m) * np.sum((y - y_pred) * X)

# Визуализация
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Данные')
plt.plot(X, w0 + w1 * X, color='red', label='Линия регрессии')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Линейная регрессия с методом градиентного спуска')
plt.legend()
plt.grid(True)
plt.show()

print(f'Обнаруженные параметры: w0 = {w0}, w1 = {w1}')
```

###Пример 2: Минимизация функции двух переменных
Рассмотрим функцию:

$$
f(x, y) = x^2 + y^2 + 2xy
$$

### 1. Анализ функции

#### 1.1. Выпуклость функции

Перед тем как минимизировать функцию, стоит проверить ее выпуклость. Функция $f$ является выпуклой, если ее Гессиан положительно определен.

**Гессиан** — это матрица вторых производных:

$$
H = \begin{bmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{bmatrix}
$$

#### 1.2. Вычисление производных

Вычислим первые производные функции $f$:

$$
\frac{\partial f}{\partial x} = 2x + 2y
$$

$$
\frac{\partial f}{\partial y} = 2y + 2x
$$

Теперь вычислим вторые производные:

$$
\frac{\partial^2 f}{\partial x^2} = 2
$$

$$
\frac{\partial^2 f}{\partial y^2} = 2
$$

$$
\frac{\partial^2 f}{\partial x \partial y} = 2
$$

Теперь составим матрицу Гессиана:

$$
H = \begin{bmatrix}
2 & 2 \\
2 & 2
\end{bmatrix}
$$

#### 1.3. Проверка положительной определенности

Чтобы проверить, положительно ли определена матрица $H$, вычислим ее определитель:

$$
\text{det}(H) = 2 \cdot 2 - 2 \cdot 2 = 0
$$

Поскольку определитель равен нулю, это означает, что матрица не является положительно определенной, и функция может быть не строго выпуклой. Тем не менее, мы можем продолжить минимизацию, чтобы найти критические точки.

### 2. Нахождение критических точек

Критические точки находятся там, где градиент функции равен нулю:

$$
\nabla f(x, y) = \begin{bmatrix}
\frac{\partial f}{\partial x} \\
\frac{\partial f}{\partial y}
\end{bmatrix} = \begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

Система уравнений:

1. $2x + 2y = 0$
2. $2y + 2x = 0$

Эти уравнения эквивалентны, и их можно упростить:

$$
x + y = 0 \implies y = -x
$$

Подставляя $y = -x$ в функцию $f(x, y)$:

$$
f(x, -x) = x^2 + (-x)^2 + 2x(-x) = x^2 + x^2 - 2x^2 = 0
$$

Таким образом, $f(0, 0) = 0$.

### 3. Определение типа критической точки

Чтобы определить, является ли точка $(0, 0)$ минимумом, максимумом или седловой точкой, необходимо изучить свойства Гессиана.

- Если $H$ положительно определен, то $(0, 0)$ — локальный минимум.
- Если $H$ отрицательно определен, то $(0, 0)$ — локальный максимум.
- Если $H$ имеет собственные значения разных знаков, то $(0, 0)$ — седловая точка.

Поскольку $H$ имеет нулевой определитель, чтобы понять, является ли $(0, 0)$ минимумом, мы можем исследовать поведение функции в окрестности этой точки.

### 4. Исследование поведения функции

Взглянем на значение функции в точках, близких к нулю:

- Для $(1, 0)$:

$$
f(1, 0) = 1^2 + 0^2 + 2 \cdot 1 \cdot 0 = 1
$$

- Для $(0, 1)$:

$$
f(0, 1) = 0^2 + 1^2 + 2 \cdot 0 \cdot 1 = 1
$$

- Для $(-1, 0)$:

$$
f(-1, 0) = (-1)^2 + 0^2 + 2 \cdot (-1) \cdot 0 = 1
$$

- Для $(0, -1)$:

$$
f(0, -1) = 0^2 + (-1)^2 + 2 \cdot 0 \cdot (-1) = 1
$$

Из этих значений видно, что функция принимает значения больше нуля в окрестности точки $(0, 0)$.

### 5. Вывод

Таким образом, точка $(0, 0)$ является локальным минимумом функции $f(x, y)$, а также глобальным минимумом, так как все исследованные значения функции в окрестности этой точки больше нуля.

### 6. Заключение

Мы провели полный анализ функции двух переменных, используя аналитические методы. Были найдены критические точки, исследовано поведение функции в окрестности этих точек и сделан вывод о том, что точка $(0, 0)$ является минимумом функции $f(x, y)$.




```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию и её градиенты
def f(x, y):
    return x**2 + y**2 + 2 * x * y

def df_dx(x, y):
    return 2 * x + 2 * y

def df_dy(x, y):
    return 2 * y + 2 * x

# Гиперпараметры
alpha = 0.1  # Скорость обучения
iterations = 50

# Инициализация параметров
x, y = 1, 1
trajectory = [(x, y)]

# Градиентный спуск
for _ in range(iterations):
    x -= alpha * df_dx(x, y)
    y -= alpha * df_dy(x, y)
    trajectory.append((x, y))

# Визуализация
x_vals = np.linspace(-3, 3, 100)
y_vals = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

plt.figure(figsize=(10, 8))
contour = plt.contour(X, Y, Z, levels=30, cmap='viridis')
plt.clabel(contour, inline=True, fontsize=8)
plt.colorbar(label='f(x, y)')
trajectory = np.array(trajectory)
plt.plot(trajectory[:, 0], trajectory[:, 1], marker='o', color='red', label='Траектория спуска')
plt.scatter(trajectory[-1, 0], trajectory[-1, 1], color='blue', label='Минимум')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Минимизация функции двух переменных с методом градиентного спуска')
plt.legend()
plt.grid(True)
plt.show()

print(f'Минимум достигается в точке: x = {x}, y = {y}')
```

##Затухание и взрыв градиента
Затухание и взрыв градиента — это две проблемы, с которыми можно столкнуться при обучении нейронных сетей и использовании метода градиентного спуска. Давайте рассмотрим оба этих явления, приведем примеры и решим их.

## Пример 1: Затухание градиента

**Описание:**
Затухание градиента происходит, когда значения градиента становятся слишком маленькими, что делает обновления весов очень незначительными. Это особенно актуально для глубоких нейронных сетей с активирующими функциями, такими как сигмоидальная или гиперболический тангенс.

### 1.1. Пример затухания градиента

Рассмотрим простую нейронную сеть с сигмоидальной функцией активации:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

При вычислении градиента весов через несколько слоев, производные сигмоидальной функции будут значительно меньше 1, что приводит к затуханию.

**Обучающая функция:**

Для иллюстрации этой проблемы, давайте представим, что мы обучаем простую сеть с одним скрытым слоем.

### 1.2. Симуляция затухания градиента

Реализуем симуляцию, которая покажет, как градиенты уменьшаются при обучении:


```python
import numpy as np
import matplotlib.pyplot as plt

# Функция активации и её производная
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Генерация случайных данных
np.random.seed(0)
X = np.random.randn(100, 10)
y = np.random.randint(0, 2, size=(100, 1))

# Инициализация весов
weights = np.random.randn(10, 5)  # 10 входов, 5 нейронов в скрытом слое
learning_rate = 0.1
epochs = 200

# Список для хранения значений градиента
gradient_values = []

# Обучение
for epoch in range(epochs):
    # Прямое распространение
    z = X.dot(weights)
    a = sigmoid(z)

    # Вычисление ошибки
    error = y - a
    # Вычисление градиента
    gradient = error * sigmoid_derivative(z)

    # Сохранение максимального значения градиента
    gradient_values.append(np.max(np.abs(gradient)))

    # Обновление весов
    weights += learning_rate * X.T.dot(gradient)

# Визуализация
plt.plot(gradient_values)
plt.title('Затухание градиента')
plt.xlabel('Эпохи')
plt.ylabel('Максимальное значение градиента')
plt.yscale('log')  # Логарифмическая шкала
plt.grid()
plt.show()
```

### 1.3. Решение проблемы затухания градиента

**Методы решения:**
1. **Использование других функций активации:** Применение функций, таких как ReLU (Rectified Linear Unit), помогает избежать затухания градиента, так как их производные равны 1 для положительных значений.
   
2. **Нормализация весов:** Инициализация весов с помощью методов, таких как Xavier или He, может помочь избежать проблем с затуханием градиента.

3. **Использование оптимизаторов:** Использование адаптивных оптимизаторов, таких как Adam или RMSprop, также может помочь справиться с проблемой.

## Пример 2: Взрыв градиента

**Описание:**
Взрыв градиента происходит, когда значения градиента становятся очень большими, что приводит к чрезмерным обновлениям весов и, как следствие, к нестабильности обучения. Это часто случается в глубоких нейронных сетях, особенно при использовании рекуррентных нейронных сетей (RNN).

### 2.1. Пример взрыва градиента

Рассмотрим, как взрыв градиента может возникнуть при обучении рекуррентной нейронной сети.

### 2.2. Симуляция взрыва градиента

Создадим простую RNN, где будет легко наблюдать за взрывом градиента.


```python
import numpy as np
import matplotlib.pyplot as plt

# Гиперпараметры
timesteps = 20
input_dim = 10
hidden_dim = 10
learning_rate = 0.01
epochs = 50

# Инициализация весов
Wxh = np.random.randn(input_dim, hidden_dim) * 0.01  # Входные веса
Whh = np.random.randn(hidden_dim, hidden_dim) * 0.01  # Веса скрытого слоя
Why = np.random.randn(hidden_dim, input_dim) * 0.01  # Выходные веса

# Генерация случайных данных
np.random.seed(0)
X = np.random.randn(timesteps, input_dim)

# Список для хранения значений градиента
gradient_values = []

# Обучение
for epoch in range(epochs):
    # Скрытое состояние
    h = np.zeros((timesteps, hidden_dim))

    # Прямое распространение
    for t in range(timesteps):
        h[t] = np.tanh(X[t].dot(Wxh) + h[t - 1].dot(Whh) if t > 0 else X[t].dot(Wxh))

    # Ошибка
    error = X - h
    # Инициализация градиента
    dWhy = np.zeros_like(Why)
    dWhh = np.zeros_like(Whh)
    dWxh = np.zeros_like(Wxh)

    # Обратное распространение
    for t in reversed(range(timesteps)):
        dh = error[t] * (1 - h[t] ** 2)  # Производная tanh
        dWhy += np.outer(h[t], error[t])
        if t > 0:
            dWhh += np.outer(h[t-1], dh)
            dWxh += np.outer(X[t], dh)

    # Сохранение максимального значения градиента
    max_gradient = np.max([np.max(np.abs(dWxh)), np.max(np.abs(dWhh)), np.max(np.abs(dWhy))])
    gradient_values.append(max_gradient)

    # Обновление весов
    Wxh += learning_rate * dWxh
    Whh += learning_rate * dWhh
    Why += learning_rate * dWhy

# Визуализация
plt.plot(gradient_values)
plt.title('Взрыв градиента')
plt.xlabel('Эпохи')
plt.ylabel('Максимальное значение градиента')
plt.yscale('log')  # Логарифмическая шкала
plt.grid()
plt.show()
```



### 2.3. Решение проблемы взрыва градиента

**Методы решения:**
1. **Ограничение градиента:** Ограничение градиента (gradient clipping) позволяет предотвратить слишком большие обновления весов, устанавливая максимальные значения для градиентов.
   
2. **Использование других архитектур:** Использование архитектур, таких как LSTM (Long Short-Term Memory) или GRU (Gated Recurrent Unit), позволяет лучше справляться с длинными последовательностями, так как они имеют механизмы, предотвращающие взрыв градиента.

3. **Нормализация данных:** Нормализация входных данных также может помочь избежать этой проблемы.

## Заключение

Затухание и взрыв градиента — это важные проблемы при обучении нейронных сетей, которые могут значительно ухудшить процесс обучения. Понимание причин их возникновения и применение соответствующих методов решения может значительно повысить эффективность обучения.
