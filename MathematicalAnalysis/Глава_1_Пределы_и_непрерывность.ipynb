#Глава 1. Пределы и непрерывность

### 1. Пределы функций

#### 1.1 Определение предела функции

Пусть функция $f(x)$ определена в некоторой окрестности точки $a$, за исключением, возможно, самой точки $a$. Число $L$ называется пределом функции $f(x)$ при $x \to a$, если для любого положительного числа $\varepsilon > 0$ существует такое положительное число $\delta > 0$, что для всех $x$, удовлетворяющих неравенству $0 < |x - a| < \delta$, выполняется неравенство $|f(x) - L| < \varepsilon$.

Запись этого утверждения:
$$
\lim_{x \to a} f(x) = L.
$$

#### 1.2 Односторонние пределы

Предел функции может существовать слева и справа:

- Левосторонний предел (предел при $x \to a^-$):
$$
\lim_{x \to a^-} f(x) = L,
$$
если для любого $\varepsilon > 0$ существует $\delta > 0$, такое что при $a - \delta < x < a$ выполняется $|f(x) - L| < \varepsilon$.

- Правосторонний предел (предел при $x \to a^+$):
$$
\lim_{x \to a^+} f(x) = L,
$$
если для любого $\varepsilon > 0$ существует $\delta > 0$, такое что при $a < x < a + \delta$ выполняется $|f(x) - L| < \varepsilon$.

Для существования предела функции в точке $a$ необходимо и достаточно, чтобы левосторонний и правосторонний пределы совпадали:
$$
\lim_{x \to a} f(x) = L \iff \lim_{x \to a^-} f(x) = \lim_{x \to a^+} f(x) = L.
$$

#### 1.3 Свойства пределов

Если функции $f(x)$ и $g(x)$ имеют пределы в точке $a$, то справедливы следующие свойства:

1. Линейность предела:
 $$
   \lim_{x \to a} (c \cdot f(x)) = c \cdot \lim_{x \to a} f(x),
 $$
   где $c$ — константа.

2. Сумма пределов:
 $$
   \lim_{x \to a} (f(x) + g(x)) = \lim_{x \to a} f(x) + \lim_{x \to a} g(x).
 $$

3. Разность пределов:
 $$
   \lim_{x \to a} (f(x) - g(x)) = \lim_{x \to a} f(x) - \lim_{x \to a} g(x).
 $$

4. Произведение пределов:
 $$
   \lim_{x \to a} (f(x) \cdot g(x)) = \left( \lim_{x \to a} f(x) \right) \cdot \left( \lim_{x \to a} g(x) \right).
 $$

5. Частное пределов:
 $$
   \lim_{x \to a} \frac{f(x)}{g(x)} = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)},
 $$
   при условии, что $\lim_{x \to a} g(x) \neq 0$.

#### 1.4 Бесконечно малые и бесконечно большие функции

Функция $\alpha(x)$ называется бесконечно малой при $x \to a$, если $\lim_{x \to a} \alpha(x) = 0$.

Функция $\beta(x)$ называется бесконечно большой при $x \to a$, если $\lim_{x \to a} \beta(x) = \infty$ или $\lim_{x \to a} \beta(x) = -\infty$.

#### 1.5 Теорема о пределе сложной функции

Если функция $y = f(u)$ имеет предел $L$ при $u \to b$, а функция $u = g(x)$ имеет предел $b$ при $x \to a$, то сложная функция $y = f(g(x))$ имеет предел $L$ при $x \to a$:
$$
\lim_{x \to a} f(g(x)) = \lim_{u \to b} f(u) = L.
$$

### 2. Непрерывные функции и их свойства

#### 2.1 Определение непрерывности

Функция $f(x)$ называется непрерывной в точке $a$, если выполняются следующие условия:

1. $f(a)$ определено.
2. Существует предел $\lim_{x \to a} f(x)$.
3. Предел функции в этой точке равен значению функции: $\lim_{x \to a} f(x) = f(a)$.

Если функция $f(x)$ непрерывна в каждой точке некоторого интервала, то она называется непрерывной на этом интервале.

#### 2.2 Свойства непрерывных функций

1. Арифметические операции над непрерывными функциями:
   Если функции $f(x)$ и $g(x)$ непрерывны в точке $a$, то функции $f(x) + g(x)$, $f(x) - g(x)$, $f(x) \cdot g(x)$ и $\frac{f(x)}{g(x)}$ (при $g(a) \neq 0$) также непрерывны в точке $a$.

2. Непрерывность сложной функции:
   Если функция $y = f(u)$ непрерывна в точке $u = b$, а функция $u = g(x)$ непрерывна в точке $x = a$, то сложная функция $y = f(g(x))$ также непрерывна в точке $x = a$.

#### 2.3 Классификация точек разрыва

Точка $a$ называется точкой разрыва функции $f(x)$, если функция не является непрерывной в этой точке. Точки разрыва делятся на:

1. Точки разрыва первого рода:
   Если существуют конечные левосторонний и правосторонний пределы функции в точке $a$, но они не равны, то точка разрыва называется разрывом первого рода (устранимым разрывом).

2. Точки разрыва второго рода:
   Если хотя бы один из односторонних пределов не существует или равен бесконечности, то точка называется разрывом второго рода (неустранимым разрывом).

#### 2.4 Теорема о промежуточном значении

Если функция $f(x)$ непрерывна на отрезке $[a, b]$ и принимает значения разных знаков в концах отрезка (т.е. $f(a) \cdot f(b) < 0$), то существует хотя бы одна точка $c \in (a, b)$, в которой функция принимает нулевое значение: $f(c) = 0$.

#### 2.5 Теорема Больцано-Коши о непрерывности

Пусть функция $f(x)$ непрерывна на отрезке $[a, b]$. Тогда для любого числа $y$, лежащего между $f(a)$ и $f(b)$, существует точка $c \in [a, b]$, такая что $f(c) = y$.

#### 2.6 Теорема Вейерштрасса

Если функция $f(x)$ непрерывна на отрезке $[a, b]$, то она достигает на этом отрезке своего максимума и минимума. То есть существуют такие точки $c$ и $d$ на отрезке $[a, b]$, что:
$$
f(c) = \max_{x \in [a, b]} f(x) \quad \text{и} \quad f(d) = \min_{x \in [a, b]} f(x).
$$

#### 2.7 Непрерывность элементарных функций

Элементарные функции, такие как многочлены, рациональные функции (за исключением точек, где знаменатель равен нулю), тригонометрические функции, показательные и логарифмические функции, непрерывны

 на соответствующих областях определения.

### Примеры

#### Пример 1: Найти предел функции
Рассмотрим функцию:
$$
f(x) = \frac{2x^2 - 3x + 1}{x - 1}.
$$
Найдём предел этой функции при $x \to 1$:
$$
\lim_{x \to 1} \frac{2x^2 - 3x + 1}{x - 1}.
$$

Решение:
1. Проверка вида выражения.
   Подставим $x = 1$:
 $$
   \frac{2(1)^2 - 3(1) + 1}{1 - 1} = \frac{2 - 3 + 1}{0} = \frac{0}{0}.
 $$
   Мы получили неопределённость вида $\frac{0}{0}$. Для её устранения воспользуемся разложением числителя на множители.

2. Разложение числителя.
   Разложим числитель $2x^2 - 3x + 1$ на множители. Найдём корни квадратного уравнения:
 $$
   2x^2 - 3x + 1 = 0.
 $$
   Найдём дискриминант:
 $$
   D = (-3)^2 - 4 \cdot 2 \cdot 1 = 9 - 8 = 1.
 $$
   Корни:
 $$
   x = \frac{-(-3) \pm \sqrt{1}}{2 \cdot 2} = \frac{3 \pm 1}{4} = \frac{4}{4} = 1, \quad \frac{2}{4} = \frac{1}{2}.
 $$
   Значит, числитель можно представить в виде:
 $$
   2x^2 - 3x + 1 = 2(x - 1)\left(x - \frac{1}{2}\right).
 $$

3. Сокращение дроби.
   Упростим исходное выражение:
 $$
   \frac{2(x - 1)\left(x - \frac{1}{2}\right)}{x - 1}.
 $$
   При $x \neq 1$ можно сократить на $x - 1$, получаем:
 $$
   \lim_{x \to 1} 2 \left(x - \frac{1}{2}\right) = 2 \left(1 - \frac{1}{2}\right) = 2 \cdot \frac{1}{2} = 1.
 $$
   Таким образом,
 $$
   \lim_{x \to 1} \frac{2x^2 - 3x + 1}{x - 1} = 1.
 $$

#### Пример 2: Проверка непрерывности функции
Рассмотрим функцию:
$$
f(x) =
\begin{cases}
x^2 - 1, & \text{если } x < 2, \\
3, & \text{если } x = 2, \\
2x - 1, & \text{если } x > 2.
\end{cases}
$$
Проверим непрерывность этой функции в точке $x = 2$.

Решение:
1. Проверим существование функции в точке $x = 2$.
 $$
   f(2) = 3.
 $$

2. Найдём левосторонний предел.
   Рассмотрим предел функции при $x \to 2^-$:
 $$
   \lim_{x \to 2^-} f(x) = \lim_{x \to 2^-} (x^2 - 1) = 2^2 - 1 = 3.
 $$

3. Найдём правосторонний предел.
   Рассмотрим предел функции при $x \to 2^+$:
 $$
   \lim_{x \to 2^+} f(x) = \lim_{x \to 2^+} (2x - 1) = 2 \cdot 2 - 1 = 3.
 $$

4. Сравним пределы и значение функции.
   Левосторонний и правосторонний пределы равны $3$, а также значение функции в этой точке равно $3$:
 $$
   \lim_{x \to 2^-} f(x) = \lim_{x \to 2^+} f(x) = f(2) = 3.
 $$
   Следовательно, функция $f(x)$ непрерывна в точке $x = 2$.

#### Пример 3: Использование теоремы о промежуточном значении
Пусть функция $f(x) = x^3 - 4x + 1$ непрерывна на отрезке $[0, 2]$. Докажем, что существует хотя бы одна точка $c \in [0, 2]$, такая что $f(c) = 0$.

Решение:
1. Найдём значения функции на концах отрезка.
 $$
   f(0) = 0^3 - 4 \cdot 0 + 1 = 1,
 $$
 $$
   f(2) = 2^3 - 4 \cdot 2 + 1 = 8 - 8 + 1 = 1.
 $$

2. Проверим знак функции на концах интервала.
   На концах интервала значения функции имеют разные знаки:
 $$
   f(0) = -1 < 0 \quad \text{и} \quad f(2) = 1 > 0.
 $$
   
3. Применяем теорему о промежуточном значении.
   Так как функция $f(x)$ непрерывна на $[0, 2]$ и принимает отрицательное значение в точке $x = 0$ и положительное значение в точке $x = 2$, то согласно теореме о промежуточном значении существует хотя бы одна точка $c \in [0, 2]$, такая что:
 $$
   f(c) = 0.
 $$

Теперь напишем код для решения этих примеров.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Пример 1: Найти предел функции
x = sp.symbols('x')
f1 = (2*x**2 - 3*x + 1) / (x - 1)

# Находим предел при x -> 1
limit_f1 = sp.limit(f1, x, 1)
print(f"Предел функции f1 при x -> 1: {limit_f1}")

# Визуализация функции f1
x_vals = np.linspace(0, 2, 400)
f1_func = sp.lambdify(x, f1, 'numpy')
y_vals = f1_func(x_vals)

plt.figure(figsize=(12, 6))
plt.plot(x_vals, y_vals, label='f(x) = (2x^2 - 3x + 1)/(x - 1)', color='blue')
plt.axvline(x=1, color='red', linestyle='--', label='x = 1 (предел)')
plt.axhline(y=limit_f1, color='green', linestyle='--', label='Предел = 1')
plt.title('График функции f1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-10, 10)
plt.xlim(0, 2)
plt.grid()
plt.legend()
plt.show()

# Пример 2: Проверка непрерывности функции
def f2(x):
    if x < 2:
        return x**2 - 1
    elif x == 2:
        return 3
    else:
        return 2*x - 1

# Находим левосторонний и правосторонний пределы
left_limit = f2(2 - 1e-10)
right_limit = f2(2 + 1e-10)
value_at_2 = f2(2)
print(f"Левосторонний предел при x -> 2: {left_limit}")
print(f"Правосторонний предел при x -> 2: {right_limit}")
print(f"f(2): {value_at_2}")

# Визуализация функции f2
x_vals_f2 = np.linspace(0, 4, 400)
y_vals_f2 = np.array([f2(x) for x in x_vals_f2])

plt.figure(figsize=(12, 6))
plt.plot(x_vals_f2, y_vals_f2, label='f(x)', color='blue')
plt.axvline(x=2, color='red', linestyle='--', label='x = 2')
plt.axhline(y=3, color='green', linestyle='--', label='f(2) = 3')
plt.title('График функции f2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-5, 5)
plt.xlim(0, 4)
plt.grid()
plt.legend()
plt.show()

# Пример 3: Использование теоремы о промежуточном значении
f3 = x**3 - 4*x + 1

# Находим значения функции на концах отрезка
value_at_0 = f3.subs(x, 0)
value_at_2 = f3.subs(x, 2)
print(f"f(0) = {value_at_0}, f(2) = {value_at_2}")

# Визуализация функции f3
x_vals_f3 = np.linspace(0, 2, 400)
f3_func = sp.lambdify(x, f3, 'numpy')
y_vals_f3 = f3_func(x_vals_f3)

plt.figure(figsize=(12, 6))
plt.plot(x_vals_f3, y_vals_f3, label='f(x) = x^3 - 4x + 1', color='blue')
plt.axhline(y=0, color='red', linestyle='--', label='f(c) = 0')
plt.title('График функции f3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-5, 5)
plt.xlim(0, 2)
plt.grid()
plt.legend()
plt.show()
```
