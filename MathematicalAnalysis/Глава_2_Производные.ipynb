## Лекция: Производные

### 1. Введение в производные

Производная — это одно из ключевых понятий математического анализа, которое используется для исследования изменения функций. В общем случае, производная функции в точке представляет собой предел отношения приращения функции к приращению аргумента, когда последнее стремится к нулю. Это понятие позволяет описать скорость изменения функции в данной точке, направление её изменения и многие другие важные свойства.



### 2. Определение производной

#### Определение 1. (Классическое определение)
Пусть $f(x)$ — функция, определенная на некотором интервале, содержащем точку $x_0$. Производная функции $f$ в точке $x_0$, если она существует, определяется следующим образом:
$$
f'(x_0) = \lim_{h \to 0} \frac{f(x_0 + h) - f(x_0)}{h}.
$$
Здесь $h$ — приращение аргумента, а разность $f(x_0 + h) - f(x_0)$ — приращение функции.

Если предел существует, то говорят, что функция $f$ дифференцируема в точке $x_0$, и значение этого предела называется производной функции в точке $x_0$.

#### Геометрический смысл
Производная функции в точке представляет собой угловой коэффициент касательной к графику функции в этой точке. Если функция $f$ задана графически, то $f'(x_0)$ характеризует наклон касательной к графику функции в точке $x_0$.

#### Физический смысл
Производная функции $f$, описывающей закон движения материальной точки $s = f(t)$, по времени $t$, определяет скорость точки в момент времени $t_0$:
$$
v(t_0) = \lim_{\Delta t \to 0} \frac{f(t_0 + \Delta t) - f(t_0)}{\Delta t}.
$$



### 3. Правила дифференцирования

#### 3.1 Линейность
Для двух дифференцируемых функций $u(x)$ и $v(x)$ и констант $a$ и $b$ верно:
$$
\frac{d}{dx} (au(x) + bv(x)) = a \frac{du}{dx} + b \frac{dv}{dx}.
$$

#### 3.2 Правило произведения (Правило Лейбница)
Для функций $u(x)$ и $v(x)$ выполнено:
$$
\frac{d}{dx} (u(x) v(x)) = u'(x) v(x) + u(x) v'(x).
$$

#### 3.3 Правило частного
Если $u(x)$ и $v(x)$ — дифференцируемые функции, причём $v(x) \neq 0$, то производная частного задаётся формулой:
$$
\frac{d}{dx} \left( \frac{u(x)}{v(x)} \right) = \frac{u'(x) v(x) - u(x) v'(x)}{v^2(x)}.
$$

#### 3.4 Правило цепочки
Для композитной функции $y = f(g(x))$, где $f$ и $g$ — дифференцируемые функции, справедливо:
$$
\frac{dy}{dx} = f'(g(x)) \cdot g'(x).
$$

### 4. Частные производные и градиенты

#### Определение 2. Частная производная
Пусть $f(x_1, x_2, \ldots, x_n)$ — функция нескольких переменных. Частной производной функции $f$ по переменной $x_i$ в точке $(x_1, x_2, \ldots, x_n)$ называется предел
$$
\frac{\partial f}{\partial x_i} = \lim_{h \to 0} \frac{f(x_1, \ldots, x_i + h, \ldots, x_n) - f(x_1, \ldots, x_i, \ldots, x_n)}{h},
$$
если этот предел существует.

#### Градиент
Градиент функции $f$, зависящей от $n$ переменных, это вектор, состоящий из её частных производных:
$$
\nabla f = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \ldots, \frac{\partial f}{\partial x_n} \right).
$$
Градиент показывает направление наибольшего возрастания функции и его величина равна скорости наибольшего изменения функции в данном направлении.



### 5. Производная по направлению

#### Определение 3. Производная функции по направлению
Производной функции $f$ в точке $\mathbf{a} = (a_1, \ldots, a_n)$ по направлению единичного вектора $\mathbf{u} = (u_1, \ldots, u_n)$ называется предел
$$
D_{\mathbf{u}} f(\mathbf{a}) = \lim_{h \to 0} \frac{f(\mathbf{a} + h \mathbf{u}) - f(\mathbf{a})}{h},
$$
если этот предел существует.

Связь производной по направлению с градиентом:
$$
D_{\mathbf{u}} f(\mathbf{a}) = \nabla f(\mathbf{a}) \cdot \mathbf{u},
$$
где «$\cdot$» обозначает скалярное произведение векторов.



### 6. Применение производных в оптимизации

#### Условие экстремума
Для функции одной переменной $f(x)$ необходимым условием экстремума (минимума или максимума) является равенство её производной нулю:
$$
f'(x_0) = 0.
$$

#### Достаточное условие
Для функции одной переменной, если в точке $x_0$ выполнено $f'(x_0) = 0$, то:
- Если $f''(x_0) > 0$, то в точке $x_0$ функция имеет локальный минимум.
- Если $f''(x_0) < 0$, то в точке $x_0$ функция имеет локальный максимум.

#### Метод градиентного спуска
Для функции нескольких переменных градиентный спуск используется для нахождения минимумов. Алгоритм обновляет значения переменных в направлении отрицательного градиента:
$$
\mathbf{x}_{k+1} = \mathbf{x}_k - \alpha \nabla f(\mathbf{x}_k),
$$
где $\alpha$ — шаг обучения (скорость изменения).



### Заключение

Производные являются мощным инструментом для анализа функций, исследования их поведения, поиска экстремумов, оптимизации и решения различных прикладных задач в физике, инженерии и экономике.

### Пример 1: Нахождение производной простой функции

Задача: Найти производную функции $f(x) = 3x^2 - 5x + 4$.

#### Шаг 1: Применение правила дифференцирования

Используем правило дифференцирования для полиномов:
$$
f'(x) = \frac{d}{dx}(3x^2 - 5x + 4) = \frac{d}{dx}(3x^2) - \frac{d}{dx}(5x) + \frac{d}{dx}(4).
$$

#### Шаг 2: Дифференцирование каждого члена отдельно

1. Производная от $3x^2$ по правилу $\frac{d}{dx}(x^n) = n x^{n-1}$:
 $$
   \frac{d}{dx}(3x^2) = 3 \cdot 2x^{2-1} = 6x.
 $$

2. Производная от $-5x$ по правилу $\frac{d}{dx}(ax) = a$:
 $$
   \frac{d}{dx}(-5x) = -5.
 $$

3. Производная от константы $4$ равна нулю:
 $$
   \frac{d}{dx}(4) = 0.
 $$

#### Шаг 3: Сложение результатов

Складываем найденные производные:
$$
f'(x) = 6x - 5.
$$

Ответ: $f'(x) = 6x - 5$.



### Пример 2: Применение правила цепочки

Задача: Найти производную функции $f(x) = (2x^3 - x)^4$.

#### Шаг 1: Применение правила цепочки

Пусть $u = 2x^3 - x$, тогда функция может быть переписана как $f(x) = u^4$.

Теперь найдем производные по отдельности:
1. Производная от внешней функции $u^4$:
 $$
   \frac{d}{du}(u^4) = 4u^3.
 $$

2. Производная внутренней функции $2x^3 - x$:
 $$
   \frac{d}{dx}(2x^3 - x) = 6x^2 - 1.
 $$

#### Шаг 2: Применение правила цепочки

Теперь, по правилу цепочки, производная $f'(x)$ равна:
$$
f'(x) = \frac{d}{du}(u^4) \cdot \frac{d}{dx}(2x^3 - x) = 4(2x^3 - x)^3 \cdot (6x^2 - 1).
$$

Ответ: $f'(x) = 4(2x^3 - x)^3 (6x^2 - 1)$.



### Пример 3: Нахождение градиента функции

Задача: Найти градиент функции $f(x, y) = 3x^2 y + 2y^3$ в точке $(1, 2)$.

#### Шаг 1: Нахождение частных производных

1. Частная производная по $x$:
 $$
   \frac{\partial f}{\partial x} = \frac{\partial}{\partial x}(3x^2 y + 2y^3) = 6xy.
 $$

2. Частная производная по $y$:
 $$
   \frac{\partial f}{\partial y} = \frac{\partial}{\partial y}(3x^2 y + 2y^3) = 3x^2 + 6y^2.
 $$

#### Шаг 2: Вычисление градиента в точке $(1, 2)$

Подставим $x = 1$ и $y = 2$ в частные производные:
$$
\frac{\partial f}{\partial x}(1, 2) = 6 \cdot 1 \cdot 2 = 12,
$$
$$
\frac{\partial f}{\partial y}(1, 2) = 3 \cdot 1^2 + 6 \cdot 2^2 = 3 + 24 = 27.
$$

Градиент функции в точке $(1, 2)$ равен:
$$
\nabla f(1, 2) = (12, 27).
$$

Ответ: Градиент в точке $(1, 2)$ равен $(12, 27)$.



### Пример 4: Производная по направлению

Задача: Найти производную функции $f(x, y) = x^2 + y^2$ в точке $(1, 1)$ по направлению вектора $\mathbf{u} = (3, 4)$.

#### Шаг 1: Нахождение единичного вектора направления

Вектор $\mathbf{u} = (3, 4)$ не является единичным, поэтому сначала нормируем его:
$$
|\mathbf{u}| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = 5.
$$
Единичный вектор:
$$
\hat{\mathbf{u}} = \left( \frac{3}{5}, \frac{4}{5} \right).
$$

#### Шаг 2: Нахождение градиента функции

Градиент функции $f(x, y) = x^2 + y^2$ равен:
$$
\nabla f = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = (2x, 2y).
$$
В точке $(1, 1)$:
$$
\nabla f(1, 1) = (2 \cdot 1, 2 \cdot 1) = (2, 2).
$$

#### Шаг 3: Вычисление производной по направлению

Производная по направлению $\hat{\mathbf{u}}$ определяется как скалярное произведение градиента на единичный вектор направления:
$$
D_{\hat{\mathbf{u}}} f(1, 1) = \nabla f(1, 1) \cdot \hat{\mathbf{u}} = (2, 2) \cdot \left( \frac{3}{5}, \frac{4}{5} \right).
$$
Скалярное произведение:
$$
D_{\hat{\mathbf{u}}} f(1, 1) = 2 \cdot \frac{3}{5} + 2 \cdot \frac{4}{5} = \frac{6}{5} + \frac{8}{5} = \frac{14}{5}.
$$

Ответ: Производная функции по направлению вектора $(3, 4)$ в точке $(1, 1)$ равна $\frac{14}{5}$.



### Пример 5: Применение производных в оптимизации

Задача: Найти экстремумы функции $f(x) = x^3 - 3x^2 + 2$.

#### Шаг 1: Нахождение критических точек

Находим первую производную функции:
$$
f'(x) = 3x^2 - 6x.
$$
Приравниваем производную к нулю для нахождения критических точек:
$$
3x^2 - 6x = 0 \implies 3x(x - 2) = 0.
$$
Отсюда получаем $x = 0$ и $x = 2$.

#### Шаг 2: Проверка второй производной

Вторая производная функции:
$$
f''(x) = 6x - 6.
$$
Проверим знак второй производной в точках $x = 0$ и $x = 2$:
1. $f''(0) = 6 \cdot 0 - 6 = -6$ (отрицательная), значит, $x = 0$ — точка локального максимума.
2. $f''(2) = 6 \cdot 2 - 6 = 6$ (положительная), значит, $x = 2$ — точка локального минимума.

#### Шаг 3: Нахождение значений функции в экстремумах

1. $f(0) = 0^3 - 3 \cdot 0^2 + 2 = 2$.
2. $f(2) = 2^3 - 3  \cdot 2^2 + 2 = 8 - 12 + 2 = -2$.

Ответ: Функция имеет локальный максимум в точке $(0, 2)$ и локальный минимум в точке $(2, -2)$.



Эти примеры иллюстрируют различные аспекты производных, включая нахождение производных, градиентов, производных по направлению и применение производных в задачах оптимизации.

### Пример 1: Нахождение производной функции

Задача: Найти производную функции $f(x) = 4x^3 - 2x^2 + 7x - 5$.

#### Шаг 1: Применение правила дифференцирования

Используем правила дифференцирования для каждого члена функции:
$$
f'(x) = \frac{d}{dx}(4x^3) - \frac{d}{dx}(2x^2) + \frac{d}{dx}(7x) - \frac{d}{dx}(5).
$$

#### Шаг 2: Находим производные каждого члена

1. Производная от $4x^3$:
 $$
   \frac{d}{dx}(4x^3) = 4 \cdot 3x^{3-1} = 12x^2.
 $$

2. Производная от $-2x^2$:
 $$
   \frac{d}{dx}(-2x^2) = -2 \cdot 2x^{2-1} = -4x.
 $$

3. Производная от $7x$:
 $$
   \frac{d}{dx}(7x) = 7.
 $$

4. Производная от константы $-5$ равна $0$:
 $$
   \frac{d}{dx}(-5) = 0.
 $$

#### Шаг 3: Сложение результатов

Теперь складываем найденные производные:
$$
f'(x) = 12x^2 - 4x + 7.
$$

Ответ: $f'(x) = 12x^2 - 4x + 7$.



### Пример 2: Использование правила цепочки

Задача: Найти производную функции $f(x) = \sin(3x^2 + 1)$.

#### Шаг 1: Определение внешней и внутренней функции

Пусть $u = 3x^2 + 1$. Тогда $f(x) = \sin(u)$.

#### Шаг 2: Находим производные

1. Производная внешней функции:
 $$
   \frac{d}{du}(\sin(u)) = \cos(u).
 $$

2. Производная внутренней функции:
 $$
   \frac{d}{dx}(3x^2 + 1) = 6x.
 $$

#### Шаг 3: Применение правила цепочки

По правилу цепочки:
$$
f'(x) = \frac{d}{du}(\sin(u)) \cdot \frac{du}{dx} = \cos(3x^2 + 1) \cdot 6x.
$$

Ответ: $f'(x) = 6x \cos(3x^2 + 1)$.



### Пример 3: Нахождение градиента функции двух переменных

Задача: Найти градиент функции $f(x, y) = x^2y + 3xy^2$ в точке $(1, 1)$.

#### Шаг 1: Нахождение частных производных

1. Частная производная по $x$:
 $$
   \frac{\partial f}{\partial x} = \frac{\partial}{\partial x}(x^2y + 3xy^2) = 2xy + 3y^2.
 $$

2. Частная производная по $y$:
 $$
   \frac{\partial f}{\partial y} = \frac{\partial}{\partial y}(x^2y + 3xy^2) = x^2 + 6xy.
 $$

#### Шаг 2: Вычисление градиента в точке $(1, 1)$

Подставим $x = 1$ и $y = 1$:
1. Для $\frac{\partial f}{\partial x}$:
 $$
   \frac{\partial f}{\partial x}(1, 1) = 2 \cdot 1 \cdot 1 + 3 \cdot 1^2 = 2 + 3 = 5.
 $$

2. Для $\frac{\partial f}{\partial y}$:
 $$
   \frac{\partial f}{\partial y}(1, 1) = 1^2 + 6 \cdot 1 \cdot 1 = 1 + 6 = 7.
 $$

Таким образом, градиент функции в точке $(1, 1)$ равен:
$$
\nabla f(1, 1) = (5, 7).
$$

Ответ: Градиент в точке $(1, 1)$ равен $(5, 7)$.



### Пример 4: Производная по направлению

Задача: Найти производную функции $f(x, y) = x^2 + y^2$ в точке $(1, 1)$ по направлению вектора $\mathbf{u} = (3, 4)$.

#### Шаг 1: Нормализация вектора

Вектор $\mathbf{u} = (3, 4)$ не является единичным, поэтому нормируем его:
$$
|\mathbf{u}| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = 5.
$$
Единичный вектор:
$$
\hat{\mathbf{u}} = \left( \frac{3}{5}, \frac{4}{5} \right).
$$

#### Шаг 2: Нахождение градиента функции

Градиент функции $f(x, y) = x^2 + y^2$ равен:
$$
\nabla f = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = (2x, 2y).
$$
В точке $(1, 1)$:
$$
\nabla f(1, 1) = (2 \cdot 1, 2 \cdot 1) = (2, 2).
$$

#### Шаг 3: Вычисление производной по направлению

Производная по направлению $\hat{\mathbf{u}}$ определяется как скалярное произведение градиента на единичный вектор направления:
$$
D_{\hat{\mathbf{u}}} f(1, 1) = \nabla f(1, 1) \cdot \hat{\mathbf{u}} = (2, 2) \cdot \left( \frac{3}{5}, \frac{4}{5} \right).
$$
Скалярное произведение:
$$
D_{\hat{\mathbf{u}}} f(1, 1) = 2 \cdot \frac{3}{5} + 2 \cdot \frac{4}{5} = \frac{6}{5} + \frac{8}{5} = \frac{14}{5}.
$$

Ответ: Производная функции по направлению вектора $(3, 4)$ в точке $(1, 1)$ равна $\frac{14}{5}$.



### Пример 5: Применение производных в оптимизации

Задача: Найти экстремумы функции $f(x) = x^3 - 3x^2 + 2$.

#### Шаг 1: Нахождение критических точек

Находим первую производную функции:
$$
f'(x) = 3x^2 - 6x.
$$
Приравниваем производную к нулю:
$$
3x^2 - 6x = 0 \implies 3x(x - 2) = 0.
$$
Отсюда получаем критические точки $x = 0$ и $x = 2$.

#### Шаг 2: Проверка второй производной

Находим вторую производную:
$$
f''(x) = 6x - 6.
$$
Теперь проверим знак второй производной в критических точках:
1. Для $x = 0$:
 $$
   f''(0) = 6 \cdot 0 - 6 = -6 \quad (\text{отрицательная, значит локальный максимум}).
 $$

2. Для $x = 2$:
 $$
   f''(2) = 6 \cdot 2 - 6 = 6 \quad (\text{положительная, значит локальный минимум}).
 $$

#### Шаг 3: Нахождение значений функции в экстремумах

1. Для $x = 0$:
 $$
   f(0) = 0^3 - 3 \

cdot 0^2 + 2 = 2.
 $$

2. Для $x = 2$:
 $$
   f(2) = 2^3 - 3 \cdot 2^2 + 2 = 8 - 12 + 2 = -2.
 $$

Ответ: Функция имеет локальный максимум в точке $(0, 2)$ и локальный минимум в точке $(2, -2)$.



Эти примеры иллюстрируют шаг за шагом различные аспекты производных, включая их нахождение, использование в задачах оптимизации и нахождение градиентов и производных по направлению. Если у вас есть дополнительные вопросы или нужны более сложные примеры, не стесняйтесь спрашивать!





Для визуализации примеров, связанных с производными и градиентами, мы будем использовать библиотеку Matplotlib на языке Python. Это позволит наглядно представить графики функций, их производные и другие элементы.

Сначала реализуем каждый пример шаг за шагом с графическим отображением.

### Пример 1: Нахождение производной функции $f(x) = 4x^3 - 2x^2 + 7x - 5$

1. Построим график функции $f(x)$.
2. Построим график производной $f'(x) = 12x^2 - 4x + 7$, чтобы показать, как функция изменяется.

### Пример 2: Использование правила цепочки для функции $f(x) = \sin(3x^2 + 1)$

1. Построим график функции $f(x) = \sin(3x^2 + 1)$.
2. Построим график её производной.

### Пример 3: Градиент функции $f(x, y) = x^2y + 3xy^2$ в точке $(1, 1)$

1. Отобразим поверхность функции $f(x, y)$.
2. Покажем вектор градиента в точке $(1, 1)$.

### Пример 4: Производная по направлению функции $f(x, y) = x^2 + y^2$ в точке $(1, 1)$

1. Отобразим поверхность функции $f(x, y)$.
2. Покажем направление вектора $\hat{\mathbf{u}}$ и производную по этому направлению.

### Пример 5: Оптимизация функции $f(x) = x^3 - 3x^2 + 2$

1. Построим график функции.
2. Покажем точки локального максимума и минимума.

Реализуем это на Python.


```python
import numpy as np
import matplotlib.pyplot as plt

# Настройка визуализации
plt.style.use('seaborn-darkgrid')

# 1. Пример 1: Производная функции f(x) = 4x^3 - 2x^2 + 7x - 5
def f1(x):
    return 4*x**3 - 2*x**2 + 7*x - 5

def f1_prime(x):
    return 12*x**2 - 4*x + 7

x_vals = np.linspace(-2, 2, 400)
f1_vals = f1(x_vals)
f1_prime_vals = f1_prime(x_vals)

# Визуализация функции и её производной
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f1_vals, label='$f(x) = 4x^3 - 2x^2 + 7x - 5$', color='b')
plt.plot(x_vals, f1_prime_vals, label="$f'(x) = 12x^2 - 4x + 7$", color='r', linestyle='--')
plt.title('График функции и её производной')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axhline(0, color='black', linewidth=0.7, linestyle=':')
plt.axvline(0, color='black', linewidth=0.7, linestyle=':')
plt.show()

# 2. Пример 2: Производная от функции f(x) = sin(3x^2 + 1)
def f2(x):
    return np.sin(3*x**2 + 1)

def f2_prime(x):
    return 6*x * np.cos(3*x**2 + 1)

f2_vals = f2(x_vals)
f2_prime_vals = f2_prime(x_vals)

# Визуализация функции и её производной
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f2_vals, label='$f(x) = \sin(3x^2 + 1)$', color='b')
plt.plot(x_vals, f2_prime_vals, label="$f'(x)$", color='r', linestyle='--')
plt.title('График функции и её производной')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axhline(0, color='black', linewidth=0.7, linestyle=':')
plt.axvline(0, color='black', linewidth=0.7, linestyle=':')
plt.show()

# 3. Пример 3: Градиент функции f(x, y) = x^2y + 3xy^2
def f3(x, y):
    return x**2 * y + 3 * x * y**2

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f3(X, Y)

# Визуализация поверхности функции
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)
ax.quiver(1, 1, f3(1, 1), 5, 7, 0, color='r', length=0.5) # Градиент в точке (1, 1)
ax.set_title('Поверхность функции и градиент в точке (1, 1)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()

# 4. Пример 4: Производная по направлению для f(x, y) = x^2 + y^2
def f4(x, y):
    return x**2 + y**2

Z = f4(X, Y)

# Визуализация поверхности функции и направление вектора u
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma', alpha=0.7)
ax.quiver(1, 1, f4(1, 1), 3/5, 4/5, 0, color='g', length=0.5) # Направление вектора u
ax.set_title('Поверхность функции и направление вектора u в точке (1, 1)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()

# 5. Пример 5: Оптимизация функции f(x) = x^3 - 3x^2 + 2
def f5(x):
    return x**3 - 3*x**2 + 2

f5_vals = f5(x_vals)

# Визуализация функции и её экстремумов
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f5_vals, label='$f(x) = x^3 - 3x^2 + 2$', color='b')
plt.scatter([0, 2], [f5(0), f5(2)], color='red', label='Экстремумы', zorder=5)
plt.title('График функции с отмеченными экстремумами')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axhline(0, color='black', linewidth=0.7, linestyle=':')
plt.axvline(0, color='black', linewidth=0.7, linestyle=':')
plt.show()
```
