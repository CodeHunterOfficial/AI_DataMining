# FastText

## Введение

FastText — это библиотека для обработки естественного языка (NLP), разработанная Facebook AI Research. Она основана на методах векторизации слов и предоставляет высокоэффективные способы для представления текстовых данных. FastText позволяет работать с задачами классификации текстов и извлечения смыслового значения слов, при этом учитывая контекст, в котором слова используются.

### Основные преимущества FastText:

1.  Скорость : FastText оптимизирован для быстрого обучения и предсказания.
2.  Качество : FastText показывает высокое качество представления слов, учитывая морфологические аспекты.
3.  Генерация векторов для незнакомых слов : В отличие от других моделей, таких как Word2Vec, FastText может создавать векторы для слов, которые не встречались в обучающей выборке, благодаря использованию подслов.

## Основные концепции

### 1. Векторное представление слов

FastText использует векторные представления слов, которые являются основой для большинства задач NLP. Векторное представление слова — это n-мерный вектор, который фиксирует семантическую информацию о слове.

#### Формула для векторного представления слова

Для слова $w$ в FastText, вектор представления $V(w)$ может быть вычислен как сумма векторов всех его подслов:

$$
V(w) = \sum_{s \in S(w)} V(s)
$$

где:
- $S(w)$ — множество подслов слова $w$.
- $V(s)$ — вектор подслова $s$.

### 2. Обучение модели

FastText обучается на больших объемах текстовых данных с использованием методов стохастического градиентного спуска (SGD). Модель обновляет веса векторов слов и подслов для минимизации функции потерь.

#### Функция потерь

FastText использует подход с негативной выборкой. Функция потерь $L$ для задачи классификации может быть записана как:

$$
L = - \sum_{i=1}^{N} \log P(y_i | x_i) + \sum_{j=1}^{K} \log P(\bar{y_j} | x_i)
$$

где:
- $N$ — количество примеров в выборке.
- $y_i$ — истинный класс для примера $x_i$.
- $\bar{y_j}$ — негативные классы.
- $K$ — количество негативных примеров.

### 3. Генерация подслов

FastText использует подсловные векторы, что позволяет ему учитывать морфологию слов. Подслова создаются с использованием "n-gram" представления.

#### Формула для подсловных векторов

Для слова $w$ длины $L$ можно создать подслова длины $n$ (например, 3-grams) следующим образом:

$$
S(w) = \{w[i:i+n] | 1 \leq i \leq L - n + 1\} \cup \{prefix(w, n)\} \cup \{suffix(w, n)\}
$$

где:
- $prefix(w, n)$ — первые $n$ символов слова $w$.
- $suffix(w, n)$ — последние $n$ символов слова $w$.

Допустим, у нас есть слово "FastText". Для длины подслов $n = 3$ мы можем сгенерировать подслова:

-  Подслова : "Fas", "ast", "stT", "Tex", "ext".

### Шаг 2: Векторное представление подслов

Предположим, что векторы подслов уже известны:

$$
\begin{align*}
V(\text{"Fas"}) & = [0.2, 0.1, 0.4] \\
V(\text{"ast"}) & = [0.3, 0.3, 0.2] \\
V(\text{"stT"}) & = [0.4, 0.1, 0.3] \\
V(\text{"Tex"}) & = [0.5, 0.2, 0.1] \\
V(\text{"ext"}) & = [0.3, 0.4, 0.3] \\
\end{align*}
$$

### Шаг 3: Вычисление вектора слова

Вектор слова "FastText" можно вычислить следующим образом:

$$
\begin{align*}
V(\text{"FastText"}) & = V(\text{"Fas"}) + V(\text{"ast"}) + V(\text{"stT"}) + V(\text{"Tex"}) + V(\text{"ext"}) \\
& = [0.2, 0.1, 0.4] + [0.3, 0.3, 0.2] + [0.4, 0.1, 0.3] + [0.5, 0.2, 0.1] + [0.3, 0.4, 0.3] \\
& = [1.7, 1.1, 1.3] \\
\end{align*}
$$

Таким образом, вектор представления слова "FastText" равен $[1.7, 1.1, 1.3]$.

## Пример 1: Генерация подслов и векторизация слова

### Задача

Пусть у нас есть слово "apple". Нам нужно сгенерировать подслова длиной 3 и вычислить вектор слова на основе известных векторов подслов.

### Шаг 1: Генерация подслов

Для слова "apple" с длиной подслов $n = 3$ подслова будут:

- "app"
- "ppl"
- "ple"

### Шаг 2: Векторное представление подслов

Предположим, что у нас есть следующие векторы подслов:

$$
\begin{align*}
V(\text{"app"}) & = [0.1, 0.2, 0.3] \\
V(\text{"ppl"}) & = [0.4, 0.1, 0.5] \\
V(\text{"ple"}) & = [0.3, 0.4, 0.2] \\
\end{align*}
$$

### Шаг 3: Вычисление вектора слова

Вектор слова "apple" вычисляется как сумма векторов его подслов:

$$
\begin{align*}
V(\text{"apple"}) & = V(\text{"app"}) + V(\text{"ppl"}) + V(\text{"ple"}) \\
& = [0.1, 0.2, 0.3] + [0.4, 0.1, 0.5] + [0.3, 0.4, 0.2] \\
& = [0.1 + 0.4 + 0.3, 0.2 + 0.1 + 0.4, 0.3 + 0.5 + 0.2] \\
& = [0.8, 0.7, 1.0] \\
\end{align*}
$$

Таким образом, вектор представления слова "apple" равен $[0.8, 0.7, 1.0]$.

## Пример 2: Классификация текста

### Задача

У нас есть текст "I love programming" с меткой класса "Positive". Мы хотим обучить модель FastText для классификации текста на позитивные и негативные отзывы.

### Шаг 1: Подготовка данных

Рассмотрим три текста с метками классов:

- "I love programming" (Positive)
- "I hate bugs" (Negative)
- "Coding is fun" (Positive)

### Шаг 2: Векторизация слов

Сначала мы генерируем векторы для каждого слова:

$$
\begin{align*}
V(\text{"I"}) & = [0.1, 0.2, 0.1] \\
V(\text{"love"}) & = [0.5, 0.1, 0.4] \\
V(\text{"programming"}) & = [0.3, 0.4, 0.2] \\
V(\text{"hate"}) & = [0.4, 0.5, 0.3] \\
V(\text{"bugs"}) & = [0.2, 0.3, 0.4] \\
V(\text{"Coding"}) & = [0.6, 0.1, 0.2] \\
V(\text{"is"}) & = [0.2, 0.2, 0.2] \\
V(\text{"fun"}) & = [0.5, 0.4, 0.1] \\
\end{align*}
$$

### Шаг 3: Вычисление вектора текста

Для текста "I love programming":

$$
\begin{align*}
V(\text{"I love programming"}) & = V(\text{"I"}) + V(\text{"love"}) + V(\text{"programming"}) \\
& = [0.1, 0.2, 0.1] + [0.5, 0.1, 0.4] + [0.3, 0.4, 0.2] \\
& = [0.1 + 0.5 + 0.3, 0.2 + 0.1 + 0.4, 0.1 + 0.4 + 0.2] \\
& = [0.9, 0.7, 0.7] \\
\end{align*}
$$

Таким образом, вектор текста "I love programming" равен $[0.9, 0.7, 0.7]$.

### Шаг 4: Обучение модели

Предположим, что мы используем метод негативной выборки для обучения и классифицируем текст. Мы можем использовать вектор текста для предсказания метки класса.

## Пример 3: Обучение модели FastText с негативной выборкой

### Задача

Предположим, у нас есть обучающий набор из трех текстов:

1. "I love programming" (Positive)
2. "I hate bugs" (Negative)
3. "Coding is fun" (Positive)

### Шаг 1: Определение функции потерь

Используем функцию потерь, основанную на негативной выборке. Мы будем использовать 1 негативный пример для каждого положительного.

Пусть $y_i$ — это истинный класс, а $\bar{y_j}$ — негативные классы.

Функция потерь:

$$
L = - \sum_{i=1}^{N} \log P(y_i | x_i) + \sum_{j=1}^{K} \log P(\bar{y_j} | x_i)
$$

### Шаг 2: Применение функции потерь

Для текста "I love programming" с истинной меткой "Positive" и негативным примером "I hate bugs", мы можем вычислить вероятность:

Предположим:

$$
P(\text{"Positive"} | x) = 0.7, \quad P(\text{"Negative"} | x) = 0.3
$$

Подставляем в функцию потерь:

$$
\begin{align*}
L & = - \log(0.7) + \log(0.3) \\
& \approx -(-0.3567) - (-1.2039) \\
& = 0.3567 + 1.2039 \\
& \approx 1.5606 \\
\end{align*}
$$

### Шаг 3: Обучение модели

Используем значение функции потерь для обновления весов векторов слов. Таким образом, мы можем продолжить обучение модели FastText, оптимизируя векторы на основе ошибок, которые мы получили.

## Заключение

В этих примерах мы рассмотрели конкретные числовые сценарии, связанные с FastText, включая генерацию подслов, векторизацию слов и обучение модели с использованием функции потерь на основе негативной выборки. Это помогает лучше понять, как FastText работает на практике.



Вот пример реализации трех упомянутых числовых примеров на Python с использованием библиотек `numpy` и `matplotlib` для визуализации. Мы будем использовать matplotlib для графического отображения векторов слов.


```python
import numpy as np
import matplotlib.pyplot as plt

# Пример 1: Генерация подслов и векторизация слова

def generate_subwords(word, n):
    """Генерирует подслова заданной длины."""
    return [word[i:i+n] for i in range(len(word)-n+1)]

def vectorize_word(word, subword_vectors):
    """Вычисляет вектор слова на основе векторов подслов."""
    subwords = generate_subwords(word, 3)
    word_vector = sum(subword_vectors[sub] for sub in subwords)
    return word_vector

# Векторы подслов
subword_vectors = {
    "app": np.array([0.1, 0.2, 0.3]),
    "ppl": np.array([0.4, 0.1, 0.5]),
    "ple": np.array([0.3, 0.4, 0.2])
}

word = "apple"
word_vector = vectorize_word(word, subword_vectors)

print(f"Вектор слова '{word}': {word_vector}")

# Визуализация вектора слова
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, word_vector[0], word_vector[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.title(f"Вектор слова '{word}'")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.show()

# Пример 2: Классификация текста

texts = [
    ("I love programming", np.array([0.9, 0.7, 0.7])),
    ("I hate bugs", np.array([0.5, 0.3, 0.4])),
    ("Coding is fun", np.array([0.8, 0.6, 0.5]))
]

# Визуализация векторов текстов
plt.figure(figsize=(8, 8))
for text, vector in texts:
    plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, label=text)

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.title("Векторы текстов")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.legend()
plt.show()

# Пример 3: Обучение модели FastText с негативной выборкой

def compute_loss(true_prob, negative_prob):
    """Вычисляет функцию потерь для классификации."""
    return -np.log(true_prob) + np.log(negative_prob)

# Вероятности
true_prob = 0.7  # Вероятность позитивного класса
negative_prob = 0.3  # Вероятность негативного класса

loss = compute_loss(true_prob, negative_prob)

print(f"Функция потерь: {loss:.4f}")
```
