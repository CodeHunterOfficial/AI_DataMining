#Математическое ожидание, дисперсия и стандартное отклонение случайных величин

## Введение

В области обработки естественного языка (NLP) математическое ожидание, дисперсия и стандартное отклонение играют важную роль в статистическом анализе данных, оценке моделей и интерпретации результатов. Эти концепции помогают понять распределение случайных величин и принимать обоснованные решения на основе статистических данных.

## 1. Математическое ожидание

### 1.1 Определение

Математическое ожидание (или среднее значение) случайной величины $X$ обозначается как $E(X)$ и определяется как:

- Для дискретной случайной величины:
$$
E(X) = \sum_{i=1}^{n} x_i P(X = x_i)
$$

- Для непрерывной случайной величины:
$$
E(X) = \int_{-\infty}^{\infty} x f_X(x) \, dx
$$
где $f_X(x)$ — функция плотности вероятности.

### 1.2 Свойства математического ожидания

1. **Линейность**:
   $$
   E(aX + bY) = aE(X) + bE(Y)
   $$
   для любых констант $a$ и $b$.

2. **Несмещенность**:
   Если $X$ — выборка, то
   $$
   E\left(\frac{1}{n} \sum_{i=1}^{n} X_i\right) = E(X)
   $$

## 2. Дисперсия

### 2.1 Определение

Дисперсия случайной величины $X$ обозначается как $Var(X)$ и измеряет разброс значений относительно математического ожидания. Дисперсия определяется как:

$$
Var(X) = E\left((X - E(X))^2\right)
$$

### 2.2 Вычисление дисперсии

Для дискретной случайной величины:
$$
Var(X) = \sum_{i=1}^{n} (x_i - E(X))^2 P(X = x_i)
$$

Для непрерывной случайной величины:
$$
Var(X) = \int_{-\infty}^{\infty} (x - E(X))^2 f_X(x) \, dx
$$

### 2.3 Свойства дисперсии

1. **Неотрицательность**:
   $$
   Var(X) \geq 0
   $$

2. **Линейность**:
   $$
   Var(aX + b) = a^2 Var(X)
   $$
   для любой константы $b$.

## 3. Стандартное отклонение

### 3.1 Определение

Стандартное отклонение $σ$ — это квадратный корень из дисперсии:
$$
\sigma(X) = \sqrt{Var(X)}
$$

### 3.2 Интерпретация

Стандартное отклонение указывает на среднее расстояние значений случайной величины от ее математического ожидания. Чем меньше стандартное отклонение, тем более сгруппированы данные.

## 4. Применение в NLP

В NLP математическое ожидание, дисперсия и стандартное отклонение могут использоваться для:

1. **Оценки частоты слов**: Для анализа распределения частот слов в тексте.
2. **Анализа производительности моделей**: Например, для оценки ошибок моделей классификации.
3. **Нормализации данных**: Стандартизация данных перед обучением моделей.

## 5. Примеры

### Пример 1: Дискретная случайная величина

Рассмотрим случайную величину $X$ с возможными значениями и вероятностями:

| $x_i$ | 1 | 2 | 3 |
|   | | | |
| $P(X = x_i)$ | 0.2 | 0.5 | 0.3 |

#### Шаг 1: Вычисление математического ожидания

$$
E(X) = 1 \cdot 0.2 + 2 \cdot 0.5 + 3 \cdot 0.3 = 0.2 + 1.0 + 0.9 = 2.1
$$

#### Шаг 2: Вычисление дисперсии

$$
Var(X) = (1 - 2.1)^2 \cdot 0.2 + (2 - 2.1)^2 \cdot 0.5 + (3 - 2.1)^2 \cdot 0.3
$$
$$
= (1.1)^2 \cdot 0.2 + (0.1)^2 \cdot 0.5 + (0.9)^2 \cdot 0.3
$$
$$
= 1.21 \cdot 0.2 + 0.01 \cdot 0.5 + 0.81 \cdot 0.3
$$
$$
= 0.242 + 0.005 + 0.243 = 0.490
$$

#### Шаг 3: Вычисление стандартного отклонения

$$
\sigma(X) = \sqrt{Var(X)} = \sqrt{0.490} \approx 0.70
$$

### Пример 2: Непрерывная случайная величина

Рассмотрим непрерывную случайную величину $Y$ с функцией плотности вероятности $f_Y(y) = 2y$ для $0 < y < 1$.

#### Шаг 1: Вычисление математического ожидания

$$
E(Y) = \int_0^1 y \cdot 2y \, dy = \int_0^1 2y^2 \, dy = \left[\frac{2y^3}{3}\right]_0^1 = \frac{2}{3}
$$

#### Шаг 2: Вычисление дисперсии

$$
Var(Y) = E(Y^2) - (E(Y))^2
$$
Сначала найдем $E(Y^2)$:
$$
E(Y^2) = \int_0^1 y^2 \cdot 2y \, dy = \int_0^1 2y^3 \, dy = \left[\frac{2y^4}{4}\right]_0^1 = \frac{1}{2}
$$
Теперь найдем дисперсию:
$$
Var(Y) = \frac{1}{2} - \left(\frac{2}{3}\right)^2 = \frac{1}{2} - \frac{4}{9} = \frac{9}{18} - \frac{8}{18} = \frac{1}{18}
$$

#### Шаг 3: Вычисление стандартного отклонения

$$
\sigma(Y) = \sqrt{Var(Y)} = \sqrt{\frac{1}{18}} \approx 0.24
$$

### Пример 1: Математическое ожидание дискретной случайной величины

#### Условие:
Рассмотрим случайную величину $X$, которая принимает следующие значения: $1$, $2$, $3$, $4$ с вероятностями: $P(X = 1) = 0.1$, $P(X = 2) = 0.3$, $P(X = 3) = 0.4$, $P(X = 4) = 0.2$.

Найти математическое ожидание $E(X)$.

#### Решение:

Математическое ожидание дискретной случайной величины определяется как:

$$
E(X) = \sum_{i=1}^{n} x_i P(X = x_i)
$$

Где $x_i$ — возможные значения случайной величины, а $P(X = x_i)$ — соответствующие вероятности.

Подставим значения из задачи:

$$
E(X) = 1 \cdot 0.1 + 2 \cdot 0.3 + 3 \cdot 0.4 + 4 \cdot 0.2
$$

Выполним вычисления:

$$
E(X) = 0.1 + 0.6 + 1.2 + 0.8 = 2.7
$$

Таким образом, математическое ожидание случайной величины $X$ равно $2.7$.



### Пример 2: Дисперсия дискретной случайной величины

#### Условие:
Используем ту же случайную величину $X$ с возможными значениями $1$, $2$, $3$, $4$ и вероятностями $P(X = 1) = 0.1$, $P(X = 2) = 0.3$, $P(X = 3) = 0.4$, $P(X = 4) = 0.2$. Найти дисперсию $Var(X)$.

#### Решение:

Дисперсия случайной величины $X$ вычисляется как:

$$
Var(X) = E\left((X - E(X))^2\right) = \sum_{i=1}^{n} (x_i - E(X))^2 P(X = x_i)
$$

Из предыдущего примера мы знаем, что $E(X) = 2.7$.

Теперь вычислим $Var(X)$, подставив все значения:

$$
Var(X) = (1 - 2.7)^2 \cdot 0.1 + (2 - 2.7)^2 \cdot 0.3 + (3 - 2.7)^2 \cdot 0.4 + (4 - 2.7)^2 \cdot 0.2
$$

Выполним вычисления:

$$
= (1.7)^2 \cdot 0.1 + (0.7)^2 \cdot 0.3 + (0.3)^2 \cdot 0.4 + (1.3)^2 \cdot 0.2
$$
$$
= 2.89 \cdot 0.1 + 0.49 \cdot 0.3 + 0.09 \cdot 0.4 + 1.69 \cdot 0.2
$$
$$
= 0.289 + 0.147 + 0.036 + 0.338 = 0.81
$$

Таким образом, дисперсия случайной величины $X$ равна $0.81$.



### Пример 3: Стандартное отклонение непрерывной случайной величины

#### Условие:
Пусть случайная величина $Y$ имеет функцию плотности вероятности $f_Y(y) = 3y^2$ для $0 \leq y \leq 1$. Найти математическое ожидание, дисперсию и стандартное отклонение $Y$.

#### Решение:

1. **Математическое ожидание**

Математическое ожидание для непрерывной случайной величины определяется как:

$$
E(Y) = \int_{0}^{1} y \cdot f_Y(y) \, dy = \int_{0}^{1} y \cdot 3y^2 \, dy = \int_{0}^{1} 3y^3 \, dy
$$

Выполним интегрирование:

$$
E(Y) = \left[\frac{3y^4}{4}\right]_0^1 = \frac{3}{4}
$$

2. **Дисперсия**

Для нахождения дисперсии сначала вычислим $E(Y^2)$:

$$
E(Y^2) = \int_{0}^{1} y^2 \cdot f_Y(y) \, dy = \int_{0}^{1} y^2 \cdot 3y^2 \, dy = \int_{0}^{1} 3y^4 \, dy
$$

Выполним интегрирование:

$$
E(Y^2) = \left[\frac{3y^5}{5}\right]_0^1 = \frac{3}{5}
$$

Теперь найдём дисперсию:

$$
Var(Y) = E(Y^2) - (E(Y))^2 = \frac{3}{5} - \left(\frac{3}{4}\right)^2 = \frac{3}{5} - \frac{9}{16}
$$

Приведём к общему знаменателю:

$$
Var(Y) = \frac{48}{80} - \frac{45}{80} = \frac{3}{80}
$$

3. **Стандартное отклонение**

Стандартное отклонение — это квадратный корень из дисперсии:

$$
\sigma(Y) = \sqrt{Var(Y)} = \sqrt{\frac{3}{80}} \approx 0.1936
$$

Таким образом, стандартное отклонение $Y$ равно $0.1936$.



### Заключение

Мы рассмотрели три примера, которые показывают, как вычисляются математическое ожидание, дисперсия и стандартное отклонение для дискретных и непрерывных случайных величин. Эти вычисления играют важную роль в статистическом анализе данных и помогают понять структуру распределений в задачах NLP и других областях.







Давайте реализуем три примера, которые мы рассмотрели ранее, с использованием Python и библиотеки Matplotlib для визуализации. Мы создадим код для вычисления математического ожидания, дисперсии и стандартного отклонения, а также визуализируем результаты, когда это будет необходимо.


```python
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt

# Пример 1: Дискретная случайная величина
x_values = np.array([1, 2, 3, 4])
probabilities = np.array([0.1, 0.3, 0.4, 0.2])

# Вычисление математического ожидания
E_X = np.sum(x_values * probabilities)
print(f"Математическое ожидание E(X): {E_X:.2f}")

# Вычисление дисперсии
Var_X = np.sum((x_values - E_X) ** 2 * probabilities)
print(f"Дисперсия Var(X): {Var_X:.2f}")

# Вычисление стандартного отклонения
std_dev_X = np.sqrt(Var_X)
print(f"Стандартное отклонение σ(X): {std_dev_X:.2f}")

# Пример 2: Непрерывная случайная величина
# Функция плотности вероятности
def f_Y(y):
    return 3 * y**2

# Математическое ожидание
E_Y, _ = spi.quad(lambda y: y * f_Y(y), 0, 1)
print(f"Математическое ожидание E(Y): {E_Y:.2f}")

# E(Y^2)
E_Y2, _ = spi.quad(lambda y: y**2 * f_Y(y), 0, 1)

# Дисперсия
Var_Y = E_Y2 - E_Y**2
print(f"Дисперсия Var(Y): {Var_Y:.2f}")

# Стандартное отклонение
std_dev_Y = np.sqrt(Var_Y)
print(f"Стандартное отклонение σ(Y): {std_dev_Y:.2f}")

# Визуализация функции плотности вероятности
y_values = np.linspace(0, 1, 100)
plt.plot(y_values, f_Y(y_values), label='f_Y(y) = 3y^2')
plt.title('Функция плотности вероятности')
plt.xlabel('y')
plt.ylabel('f_Y(y)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Исходный текст
text = ("Курс для начинающих о создании микросервисов на .NET. "
        "Инструктор Джулио Гасаль помогает понять основы архитектуры микросервисов, "
        "приводя пример создания игры с различными магазинами и системой валюты. "
        "Курс включает работу с MongoDB, Docker и RabbitMQ.")

# Разделение текста на предложения
sentences = text.split('. ')
sentences = [s.strip() for s in sentences if s]

# Вычисление длины каждого предложения
lengths = [len(s.split()) for s in sentences]

# Математическое ожидание
mean_length = np.mean(lengths)

# Дисперсия
variance_length = np.var(lengths)

# Стандартное отклонение
std_dev_length = np.std(lengths)

# Вывод результатов
print(f"Длина предложений: {lengths}")
print(f"Средняя длина предложения: {mean_length:.2f} слов")
print(f"Дисперсия длины предложений: {variance_length:.2f}")
print(f"Стандартное отклонение длины предложений: {std_dev_length:.2f}")

# Визуализация
plt.figure(figsize=(10, 6))
plt.bar(range(len(lengths)), lengths, color='skyblue')
plt.axhline(mean_length, color='red', linestyle='--', label=f'Средняя длина: {mean_length:.2f}')
plt.title('Длина предложений в тексте')
plt.xlabel('Номер предложения')
plt.ylabel('Количество слов')
plt.xticks(range(len(lengths)), [f'Предложение {i+1}' for i in range(len(lengths))])
plt.legend()
plt.grid(axis='y')
plt.show()
```
