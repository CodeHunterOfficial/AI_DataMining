### 1. Векторное представление слов

#### 1.1. Модели представления слов

Существуют несколько способов векторизации слов:

- **Модель "мешок слов" (Bag of Words, BoW)**: Каждый уникальный токен в тексте представляет собой отдельное измерение в векторе. Вектор для документа формируется путем подсчета частоты каждого слова.

- **TF-IDF (Term Frequency-Inverse Document Frequency)**: Эта модель учитывает не только частоту слова в документе, но и его редкость в корпусе текстов. Формула TF-IDF для слова $w$ в документе $d$ представляется как:

$$
\text{TF-IDF}(w, d) = \text{TF}(w, d) \times \log\left(\frac{N}{\text{DF}(w)}\right)
$$

где:
- $N$ — общее количество документов,
- $\text{DF}(w)$ — количество документов, содержащих слово $w$.

- **Словарные векторы (Word Embeddings)**: Векторные представления слов, такие как Word2Vec, GloVe и FastText, обучаются на больших текстовых корпусах, позволяя улавливать семантические связи между словами.

#### 1.2. Пример векторизации

Рассмотрим текст: "Кошка сидит на ковре". В модели BoW получим следующее векторное представление:

| Слово     | Частота | Вектор   |
|   --|   |   -|
| Кошка     | 1       | (1, 0, 0, 0) |
| Сидит     | 1       | (0, 1, 0, 0) |
| На        | 1       | (0, 0, 1, 0) |
| Ковре     | 1       | (0, 0, 0, 1) |



### 2. Операции с векторами

Векторное представление позволяет проводить различные математические операции:

#### 2.1. Сложение векторов

Сложение векторов позволяет объединять их значения. Например, если у нас есть два вектора:

$$
\mathbf{v_1} = \begin{pmatrix} 1 \\ 2 \\ 3 \end{pmatrix}, \quad \mathbf{v_2} = \begin{pmatrix} 4 \\ 5 \\ 6 \end{pmatrix}
$$

Сложение этих векторов:

$$
\mathbf{v} = \mathbf{v_1} + \mathbf{v_2} = \begin{pmatrix} 1+4 \\ 2+5 \\ 3+6 \end{pmatrix} = \begin{pmatrix} 5 \\ 7 \\ 9 \end{pmatrix}
$$

#### 2.2. Умножение векторов

Умножение вектора на скаляр также является распространенной операцией. Например, для вектора:

$$
\mathbf{v} = \begin{pmatrix} 1 \\ 2 \\ 3 \end{pmatrix}
$$

умножение на скаляр 2 дает:

$$
2\mathbf{v} = \begin{pmatrix} 2 \\ 4 \\ 6 \end{pmatrix}
$$



### 3. Метрики расстояния

Для измерения схожести между векторами в NLP используют различные метрики расстояния:

#### 3.1. Евклидово расстояние

Евклидово расстояние между двумя векторами $\mathbf{a}$ и $\mathbf{b}$ определяется как:

$$
d(\mathbf{a}, \mathbf{b}) = \sqrt{\sum_{i=1}^{n} (a_i - b_i)^2}
$$

#### 3.2. Косинусное расстояние

Косинусное расстояние используется для оценки угла между двумя векторами. Формула выглядит так:

$$
\text{cosine similarity} = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\| \|\mathbf{b}\|}
$$

где $\cdot$ — скалярное произведение, а $\|\mathbf{a}\|$ и $\|\mathbf{b}\|$ — длины векторов.



### 4. Примеры и числовые расчеты

#### Пример 1: Вычисление TF-IDF

Рассмотрим два документа:

- Документ 1: "Кошка сидит на ковре"
- Документ 2: "Собака лежит на полу"

**Шаг 1: Подсчет частоты слов**

| Слово     | Документ 1 | Документ 2 |
|   --|    -|    -|
| Кошка     | 1           | 0           |
| Сидит     | 1           | 0           |
| На        | 1           | 1           |
| Ковре     | 1           | 0           |
| Собака    | 0           | 1           |
| Лежит     | 0           | 1           |
| Полу      | 0           | 1           |

**Шаг 2: Вычисление TF**

TF для слова "кошка" в документе 1:

$$
\text{TF}(\text{кошка}, \text{документ 1}) = \frac{1}{4} = 0.25
$$

**Шаг 3: Подсчет DF**

DF для слова "на":

$$
\text{DF}(\text{на}) = 2 \quad (\text{поскольку слово "на" встречается в обоих документах})
$$

**Шаг 4: Подсчет TF-IDF**

Общее количество документов $N = 2$:

$$
\text{TF-IDF}(\text{кошка}, \text{документ 1}) = 0.25 \times \log\left(\frac{2}{1}\right) = 0.25 \times 0.3010 \approx 0.07525
$$

#### Пример 2: Вычисление косинусной схожести

Пусть $\mathbf{a} = \begin{pmatrix} 1 \\ 2 \\ 3 \end{pmatrix}$ и $\mathbf{b} = \begin{pmatrix} 4 \\ 5 \\ 6 \end{pmatrix}$.

**Шаг 1: Вычисление скалярного произведения**

$$
\mathbf{a} \cdot \mathbf{b} = 1 \cdot 4 + 2 \cdot 5 + 3 \cdot 6 = 4 + 10 + 18 = 32
$$

**Шаг 2: Вычисление норм**

$$
\|\mathbf{a}\| = \sqrt{1^2 + 2^2 + 3^2} = \sqrt{1 + 4 + 9} = \sqrt{14} \approx 3.74
$$
$$
\|\mathbf{b}\| = \sqrt{4^2 + 5^2 + 6^2} = \sqrt{16 + 25 + 36} = \sqrt{77} \approx 8.77
$$

**Шаг 3: Вычисление косинусной схожести**

$$
\text{cosine similarity} = \frac{32}{3.74 \cdot 8.77} \approx \frac{32}{32.81} \approx 0.974
$$

### Пример 1: Вычисление TF-IDF

**Задача:** У нас есть два документа:

- Документ 1: "Кошка сидит на ковре"
- Документ 2: "Собака лежит на полу"

Необходимо вычислить TF-IDF для слова "кошка" в документе 1 и "собака" в документе 2.

#### Шаг 1: Подсчет частоты слов

| Слово     | Документ 1 | Документ 2 |
|   --|    -|    -|
| Кошка     | 1           | 0           |
| Сидит     | 1           | 0           |
| На        | 1           | 1           |
| Ковре     | 1           | 0           |
| Собака    | 0           | 1           |
| Лежит     | 0           | 1           |
| Полу      | 0           | 1           |

#### Шаг 2: Подсчет TF

Формула TF (частота термина) для слова $w$ в документе $d$:

$$
\text{TF}(w, d) = \frac{\text{Количество } w \text{ в } d}{\text{Общее количество слов в } d}
$$

Для документа 1:

- Общее количество слов = 5
- TF для "кошка":

$$
\text{TF}(\text{кошка}, \text{документ 1}) = \frac{1}{5} = 0.2
$$

Для документа 2:

- Общее количество слов = 5
- TF для "собака":

$$
\text{TF}(\text{собака}, \text{документ 2}) = \frac{1}{5} = 0.2
$$

#### Шаг 3: Подсчет DF

Формула DF (частота документа) для слова $w$:

$$
\text{DF}(w) = \text{Количество документов, содержащих } w
$$

- DF для "кошка" = 1 (только в документе 1)
- DF для "собака" = 1 (только в документе 2)

#### Шаг 4: Подсчет TF-IDF

Общее количество документов $N = 2$:

Формула TF-IDF:

$$
\text{TF-IDF}(w, d) = \text{TF}(w, d) \times \log\left(\frac{N}{\text{DF}(w)}\right)
$$

Для "кошка" в документе 1:

$$
\text{TF-IDF}(\text{кошка}, \text{документ 1}) = 0.2 \times \log\left(\frac{2}{1}\right) = 0.2 \times 0.3010 \approx 0.0602
$$

Для "собака" в документе 2:

$$
\text{TF-IDF}(\text{собака}, \text{документ 2}) = 0.2 \times \log\left(\frac{2}{1}\right) = 0.2 \times 0.3010 \approx 0.0602
$$



### Пример 2: Вычисление косинусной схожести

**Задача:** Найдите косинусную схожесть между двумя векторами:

$$
\mathbf{a} = \begin{pmatrix} 3 \\ 4 \\ 0 \end{pmatrix}, \quad \mathbf{b} = \begin{pmatrix} 4 \\ 0 \\ 3 \end{pmatrix}
$$

#### Шаг 1: Вычисление скалярного произведения

Формула скалярного произведения:

$$
\mathbf{a} \cdot \mathbf{b} = a_1 \cdot b_1 + a_2 \cdot b_2 + a_3 \cdot b_3
$$

Расчет:

$$
\mathbf{a} \cdot \mathbf{b} = 3 \cdot 4 + 4 \cdot 0 + 0 \cdot 3 = 12 + 0 + 0 = 12
$$

#### Шаг 2: Вычисление норм

Формула нормы вектора:

$$
\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + v_3^2}
$$

Расчет норм:

$$
\|\mathbf{a}\| = \sqrt{3^2 + 4^2 + 0^2} = \sqrt{9 + 16 + 0} = \sqrt{25} = 5
$$
$$
\|\mathbf{b}\| = \sqrt{4^2 + 0^2 + 3^2} = \sqrt{16 + 0 + 9} = \sqrt{25} = 5
$$

#### Шаг 3: Вычисление косинусной схожести

Формула косинусной схожести:

$$
\text{cosine similarity} = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\| \|\mathbf{b}\|}
$$

Расчет:

$$
\text{cosine similarity} = \frac{12}{5 \cdot 5} = \frac{12}{25} = 0.48
$$

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer

# Документы
documents = [
    "Кошка сидит на ковре",
    "Собака лежит на полу"
]

# Векторизация TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# Получение слов и их значений TF-IDF
feature_names = vectorizer.get_feature_names_out()
dense = tfidf_matrix.todense().A
df_tfidf = pd.DataFrame(dense, columns=feature_names)

# Вывод значений TF-IDF
print("TF-IDF:\n", df_tfidf)

# Визуализация
df_tfidf.plot(kind='bar', figsize=(10, 5))
plt.title('TF-IDF Values')
plt.xlabel('Documents')
plt.ylabel('TF-IDF Value')
plt.xticks(ticks=range(len(documents)), labels=[f'Doc {i+1}' for i in range(len(documents))], rotation=0)
plt.legend(title='Words')
plt.show()
````

### Пример 2: Косинусная схожесть


````python
import numpy as np
import matplotlib.pyplot as plt

# Векторы
a = np.array([3, 4, 0])
b = np.array([4, 0, 3])

# Вычисление скалярного произведения
dot_product = np.dot(a, b)

# Вычисление норм
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

# Вычисление косинусной схожести
cosine_similarity = dot_product / (norm_a * norm_b)

print("Косинусная схожесть:", cosine_similarity)

# Визуализация
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a (3, 4)')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b (4, 3)')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.title('Vectors a and b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
````

#Word Эмбеддинги

#### Введение

В последние годы обработка естественного языка (NLP) получила значительное развитие благодаря внедрению различных методов представления слов. Одним из таких методов являются **word embeddings** (эмбеддинги слов), которые позволяют представлять слова в виде векторов в непрерывном пространстве. Это упрощает анализ текста и улучшает результаты машинного обучения.

#### Что такое Word Эмбеддинги?

Word эмбеддинги – это способ представления слов в виде плотных векторов фиксированной длины, которые сохраняют семантическую и синтаксическую информацию о словах. В отличие от традиционных методов, таких как one-hot encoding, где каждое слово представляется вектором с нулями и единицами, эмбеддинги обеспечивают более компактное и информативное представление.

**Формально**, пусть $V$ – это множество слов в словаре. Каждое слово $w \in V$ представляется вектором $\mathbf{e}(w) \in \mathbb{R}^d$, где $d$ – это размерность вектора эмбеддинга.


Эмбеддинги могут быть обучены с использованием различных алгоритмов.



##2 Word2Vec

## Введение в Word2Vec

Word2Vec — это алгоритм для представления слов в виде векторов, который был предложен в 2013 году командой Google, возглавляемой Томасом Миколовым. Основная идея Word2Vec заключается в том, чтобы преобразовать слова в векторное пространство таким образом, чтобы семантически похожие слова находились близко друг к другу. Это позволяет эффективно использовать векторные представления для различных задач обработки естественного языка (NLP), таких как классификация текста, анализ тональности и машинный перевод.

### Принципы работы Word2Vec

Word2Vec работает на основе нейронных сетей и использует два основных подхода для обучения векторов слов: Continuous Bag of Words (CBOW) и Skip-gram.

1. CBOW: предсказывает текущее слово по его контексту. Например, в предложении "собака лежит на ковре" модель предсказывает слово "лежит" по словам "собака", "на", "ковре".

2. Skip-gram: наоборот, предсказывает контекстные слова по данному слову. В том же предложении модель будет использовать слово "лежит", чтобы предсказать слова "собака", "на", "ковре".

## Математическая модель Word2Vec

### 1. CBOW (Continuous Bag of Words)

#### Архитектура

В CBOW модель принимает набор контекстных слов и пытается предсказать целевое слово. Обозначим:

- $C$ — множество контекстных слов.
- $W$ — целевое слово.

#### Функция стоимости

Функция стоимости (loss function) для CBOW может быть выражена как:

$$
L = -\sum_{(C, W) \in D} \log P(W | C)
$$

где $D$ — это обучающий набор данных. Вероятность $P(W | C)$ может быть представлена как:

$$
P(W | C) = \frac{e^{v_w \cdot v_c}}{\sum_{w' \in V} e^{v_{w'} \cdot v_c}}
$$

где:
- $v_w$ — вектор целевого слова.
- $v_c$ — вектор контекстных слов, усредненный по всем словам в контексте.
- $V$ — весь словарь.

### 2. Skip-gram

#### Архитектура

В Skip-gram модель принимает одно целевое слово и предсказывает контекстные слова. Обозначим:

- $W$ — целевое слово.
- $C$ — множество контекстных слов.

#### Функция стоимости

Функция стоимости для Skip-gram выражается как:

$$
L = -\sum_{(W, C) \in D} \sum_{c \in C} \log P(c | W)
$$

где вероятность $P(c | W)$ может быть представлена как:

$$
P(c | W) = \frac{e^{v_c \cdot v_w}}{\sum_{c' \in V} e^{v_{c'} \cdot v_w}}
$$

### Обучение моделей

Word2Vec обычно обучается с помощью стохастического градиентного спуска (SGD) и метода негативной выборки (negative sampling) для эффективного обновления весов векторов. Негативная выборка позволяет ограничить количество обновлений весов, что значительно ускоряет обучение.

#### Негативная выборка

Метод негативной выборки подразумевает выбор небольшого количества "негативных" примеров (то есть слов, которые не являются контекстными) для каждой итерации. Таким образом, общая функция стоимости принимает форму:

$$
L = -\sum_{(W, C) \in D} \left( \log P(c | W) + \sum_{n=1}^k \log P(n | W) \right)
$$

где $k$ — количество негативных выборок.

## Примеры числовых решений

### Пример 1: CBOW

Рассмотрим простое предложение "кошка сидит на ковре".

- Контекстные слова: "кошка", "на", "ковре".
- Целевое слово: "сидит".

#### Шаг 1: Вычисление векторов

Допустим, векторы слов следующие:

- $v_{\text{кошка}} = [0.2, 0.3]$
- $v_{\text{на}} = [0.1, 0.4]$
- $v_{\text{ковре}} = [0.3, 0.2]$

Усредним векторы контекстных слов:

$$
v_C = \frac{1}{3} (v_{\text{кошка}} + v_{\text{на}} + v_{\text{ковре}}) = \frac{1}{3} \left( [0.2, 0.3] + [0.1, 0.4] + [0.3, 0.2] \right) = \frac{1}{3} [0.6, 0.9] = [0.2, 0.3]
$$

#### Шаг 2: Расчет вероятности

Предположим, вектор целевого слова "сидит" равен $v_{\text{сидит}} = [0.4, 0.5]$.

Вычислим:

$$
P(\text{сидит} | C) = \frac{e^{v_{\text{сидит}} \cdot v_C}}{\sum_{w' \in V} e^{v_{w'} \cdot v_C}}
$$

Сначала найдем скалярное произведение:

$$
v_{\text{сидит}} \cdot v_C = 0.4 \cdot 0.2 + 0.5 \cdot 0.3 = 0.08 + 0.15 = 0.23
$$

Теперь подставим в формулу:

$$
P(\text{сидит} | C) = \frac{e^{0.23}}{\sum_{w' \in V} e^{v_{w'} \cdot v_C}}
$$

Пусть для простоты, $\sum_{w' \in V} e^{v_{w'} \cdot v_C} = 3$. Тогда:

$$
P(\text{сидит} | C) = \frac{e^{0.23}}{3} \approx \frac{1.259}{3} \approx 0.4197
$$

### Пример 2: Skip-gram

Рассмотрим то же самое предложение "кошка сидит на ковре".

- Целевое слово: "сидит".
- Контекстные слова: "кошка", "на", "ковре".

#### Шаг 1: Вычисление вероятностей

Предположим, векторы контекстных слов:

- $v_{\text{кошка}} = [0.2, 0.3]$
- $v_{\text{на}} = [0.1, 0.4]$
- $v_{\text{ковре}} = [0.3, 0.2]$

#### Шаг 2: Расчет вероятности

Вычислим вероятности для каждого контекстного слова:

$$
P(\text{кошка} | \text{сидит}) = \frac{e^{v_{\text{кошка}} \cdot v_{\text{сидит}}}}{\sum_{c' \in V} e^{v_{c'} \cdot v_{\text{сидит}}}}
$$

Сначала найдем вектор для "сидит":

$$
v_{\text{сидит}} = [0.4, 0.5]
$$

Скалярное произведение:

$$
v_{\text{кошка}} \cdot v_{\text{сидит}} = 0.2 \cdot 0.4 + 0.3 \cdot 0.5 = 0.08 + 0.15 = 0.23
$$

Теперь подставим в формулу:

$$
P(\text{кошка} | \text{сидит}) = \frac{e^{0.23}}{\sum_{c' \in V} e^{v_{c'} \cdot v_{\text{сидит}}}}
$$

Для упрощения $\sum_{c' \in V} e^{v_{c'} \cdot v_{\text{сидит}}} = 3$:

$$
P(\text{кошка} | \text{сидит}) = \frac{e^{0.23}}{3} \approx 0.4197
$$

### Пример 1: Continuous Bag of Words (CBOW)

Допустим, у нас есть предложение: "собака лежит на ковре".

- Целевое слово: "лежит"
- Контекстные слова: "собака", "на", "ковре"

#### Шаг 1: Представление слов в виде векторов

Пусть следующие вектора соответствуют словам:
- $v_{\text{собака}} = [0.2, 0.3]$
- $v_{\text{на}} = [0.1, 0.4]$
- $v_{\text{ковре}} = [0.3, 0.2]$
- $v_{\text{лежит}} = [0.4, 0.5]$

#### Шаг 2: Средний вектор контекста

Нужно усреднить векторы контекстных слов $v_{\text{собака}}$, $v_{\text{на}}$, $v_{\text{ковре}}$:

$$
v_C = \frac{1}{3} \left( v_{\text{собака}} + v_{\text{на}} + v_{\text{ковре}} \right) = \frac{1}{3} \left( [0.2, 0.3] + [0.1, 0.4] + [0.3, 0.2] \right) = \frac{1}{3} [0.6, 0.9] = [0.2, 0.3]
$$

#### Шаг 3: Вероятность предсказания целевого слова

Теперь вычислим вероятность того, что целевое слово "лежит" предсказывается данным контекстом. Для этого используется функция softmax:

$$
P(\text{лежит} | C) = \frac{e^{v_{\text{лежит}} \cdot v_C}}{\sum_{w' \in V} e^{v_{w'} \cdot v_C}}
$$

Вначале найдем скалярное произведение векторов $v_{\text{лежит}}$ и $v_C$:

$$
v_{\text{лежит}} \cdot v_C = 0.4 \cdot 0.2 + 0.5 \cdot 0.3 = 0.08 + 0.15 = 0.23
$$

Теперь подставим это в формулу softmax. Предположим, сумма по всем словам в словаре $\sum_{w' \in V} e^{v_{w'} \cdot v_C} = 3$:

$$
P(\text{лежит} | C) = \frac{e^{0.23}}{3} = \frac{1.259}{3} \approx 0.4197
$$

Таким образом, вероятность того, что "лежит" будет предсказано контекстом, равна примерно 41.97%.



### Пример 2: Skip-gram

Возьмем то же самое предложение "собака лежит на ковре".

- Целевое слово: "лежит"
- Контекстные слова: "собака", "на", "ковре"

#### Шаг 1: Представление слов в виде векторов

Допустим, вектора для контекстных слов такие же, как в Примере 1:
- $v_{\text{собака}} = [0.2, 0.3]$
- $v_{\text{на}} = [0.1, 0.4]$
- $v_{\text{ковре}} = [0.3, 0.2]$
- $v_{\text{лежит}} = [0.4, 0.5]$

#### Шаг 2: Вычисление вероятности для каждого контекстного слова

Для каждого контекстного слова вычислим вероятность его появления с целевым словом "лежит".

##### 1. Для слова "собака":

$$
P(\text{собака} | \text{лежит}) = \frac{e^{v_{\text{собака}} \cdot v_{\text{лежит}}}}{\sum_{c' \in V} e^{v_{c'} \cdot v_{\text{лежит}}}}
$$

Скалярное произведение $v_{\text{собака}} \cdot v_{\text{лежит}}$:

$$
v_{\text{собака}} \cdot v_{\text{лежит}} = 0.2 \cdot 0.4 + 0.3 \cdot 0.5 = 0.08 + 0.15 = 0.23
$$

Подставим в формулу softmax, предположив, что $\sum_{c' \in V} e^{v_{c'} \cdot v_{\text{лежит}}} = 3$:

$$
P(\text{собака} | \text{лежит}) = \frac{e^{0.23}}{3} \approx \frac{1.259}{3} = 0.4197
$$

##### 2. Для слова "на":

Аналогично вычислим:

$$
v_{\text{на}} \cdot v_{\text{лежит}} = 0.1 \cdot 0.4 + 0.4 \cdot 0.5 = 0.04 + 0.2 = 0.24
$$

$$
P(\text{на} | \text{лежит}) = \frac{e^{0.24}}{3} \approx \frac{1.271}{3} = 0.4237
$$

##### 3. Для слова "ковре":

$$
v_{\text{ковре}} \cdot v_{\text{лежит}} = 0.3 \cdot 0.4 + 0.2 \cdot 0.5 = 0.12 + 0.1 = 0.22
$$

$$
P(\text{ковре} | \text{лежит}) = \frac{e^{0.22}}{3} \approx \frac{1.246}{3} = 0.4153
$$

Таким образом, мы получили вероятности появления каждого контекстного слова с целевым словом "лежит".



### Пример 3: Негативная выборка (Negative Sampling)

Предположим, мы снова берем предложение "собака лежит на ковре", но используем метод негативной выборки, где положительные примеры — это слова из контекста, а отрицательные примеры — случайные слова из словаря.

- Целевое слово: "лежит"
- Контекстные слова: "собака", "на", "ковре"
- Отрицательные примеры: "кошка", "дерево"

#### Шаг 1: Представление слов в виде векторов

Пусть следующие вектора для всех слов:
- $v_{\text{собака}} = [0.2, 0.3]$
- $v_{\text{на}} = [0.1, 0.4]$
- $v_{\text{ковре}} = [0.3, 0.2]$
- $v_{\text{кошка}} = [0.5, 0.6]$ (негативное слово)
- $v_{\text{дерево}} = [0.6, 0.1]$ (негативное слово)
- $v_{\text{лежит}} = [0.4, 0.5]$

#### Шаг 2: Функция стоимости для позитивных примеров

Для положительных примеров рассчитываем вероятность, что они предсказываются целевым словом. Например, для слова "собака":

$$
P(\text{собака} | \text{лежит}) = \sigma(v_{\text{собака}} \cdot v_{\text{лежит}}) = \frac{1}{1 + e^{-v_{\text{собака}} \cdot v_{\text{лежит}}}}
$$

Скалярное произведение $v_{\text{собака}} \cdot v_{\text{лежит}} = 0.23$, тогда:

$$
P(\text{собака} | \text{лежит}) = \frac{1}{1 + e^{-0.23}} \approx 0.5572
$$

#### Шаг 3: Функция стоимости для негативных примеров

Для отрицательных примеров используем следующую формулу для вероятности:

$$
P(\text{кошка} | \text{лежит}) = \sigma(-v_{\text{кошка}} \cdot v_{\text{лежит}})
$$

Для слова "кошка" скалярное произведение $v_{\text{кошка}} \cdot v_{\text{лежит}} = 0.44$, тогда:

$$
P(\text{кошка} | \text{лежит}) = \frac{1}{1 + e^{0.44}} \approx 0.3932
$$

#### Шаг 4: Полная функция стоимости

Общая функция стоимости для одной итерации будет выглядеть следующим образом:

$$
L = -\log(P(\text{собака} | \text{лежит})) - \log(1 - P(\text{кошка} | \text{лежит}))
$$

Подставляем значения:

$$
L = -\log(0.5572) - \log(1 - 0.3932) = -(-0.5813) - (-0.4186) = 0.5813 + 0.4186 = 1.0
$$

Таким образом, общая функция стоимости равна 1.0 для одной итерации.



Эти примеры показывают, как работают модели Word2Vec (CBOW и Skip-gram), а также использование негативной выборки для оптимизации функций стоимости в обучении.

#3. GloVe (Global Vectors for Word Representation)

## Введение в GloVe

GloVe — это метод для получения векторных представлений слов, который основывается на статистике текста. Он был разработан в 2014 году командой, возглавляемой Юджином Вашингтоном и О. В. Г. Мансуром. Главная идея GloVe заключается в том, что отношения между словами в векторном пространстве могут быть выведены из их контекстов. Это представление позволяет моделям NLP эффективно захватывать семантические и синтаксические отношения между словами.

### Основные понятия

1. Слово и контекст: В GloVe каждое слово рассматривается в контексте других слов. Например, если мы рассматриваем слово "кошка", его контекст может включать слова "дом", "животное", "ласкавое".

2. Матрица соотношений: Для GloVe строится матрица соотношений, где элемент $X_{ij}$ представляет собой частоту со-occurrence слова $w_i$ и слова $w_j$.

## Математическая основа GloVe

### Построение матрицы со-occurrence

Для начала, определим матрицу со-occurrence:

$$
X_{ij} = \text{count}(w_i, w_j)
$$

где $\text{count}(w_i, w_j)$ — это количество раз, когда слово $w_i$ встречается в контексте слова $w_j$.

### Статистика и вероятности

Вероятность появления слова $w_j$ в контексте слова $w_i$ можно представить как:

$$
P_{ij} = \frac{X_{ij}}{X_{i \cdot}}
$$

где $X_{i \cdot}$ — это сумма всех со-occurrence для слова $w_i$:

$$
X_{i \cdot} = \sum_j X_{ij}
$$

### Функция потерь GloVe

GloVe обучает векторы слов так, чтобы их скалярные произведения могли предсказывать логарифмические вероятности со-occurrence. Формула потерь GloVe выглядит следующим образом:

$$
J = \sum_{i,j=1}^V f(X_{ij}) (u_i^T v_j + b_i + b_j - \log(X_{ij}))^2
$$

где:
- $V$ — размер словаря,
- $u_i$ и $v_j$ — векторные представления слов $w_i$ и $w_j$ соответственно,
- $b_i$ и $b_j$ — смещения для слов $w_i$ и $w_j$,
- $f(X_{ij})$ — весовая функция, которая обычно имеет вид:

$$
f(x) = \begin{cases}
x^\alpha & \text{if } x < x_{\text{max}} \\
1 & \text{otherwise}
\end{cases}
$$

### Оптимизация

Оптимизация функции потерь может быть выполнена с использованием методов градиентного спуска. В процессе обучения мы обновляем векторы слов, минимизируя функцию потерь.

## Пример числового решения

Предположим, у нас есть три слова: "кошка", "собака" и "животное". Мы можем построить следующую матрицу со-occurrence:

$$
X = \begin{bmatrix}
0 & 2 & 3 \\
2 & 0 & 4 \\
3 & 4 & 0
\end{bmatrix}
$$

### Шаг 1: Построение вероятностей

Рассчитаем вероятности со-occurrence:

- Для слова "кошка" ($w_1$):
  $$
  X_{1\cdot} = 2 + 3 = 5
  $$
  $$
  P_{11} = 0, \quad P_{12} = \frac{2}{5} = 0.4, \quad P_{13} = \frac{3}{5} = 0.6
  $$

- Для слова "собака" ($w_2$):
  $$
  X_{2\cdot} = 2 + 4 = 6
  $$
  $$
  P_{21} = \frac{2}{6} = 0.33, \quad P_{22} = 0, \quad P_{23} = \frac{4}{6} \approx 0.67
  $$

- Для слова "животное" ($w_3$):
  $$
  X_{3\cdot} = 3 + 4 = 7
  $$
  $$
  P_{31} = \frac{3}{7} \approx 0.43, \quad P_{32} = \frac{4}{7} \approx 0.57, \quad P_{33} = 0
  $$

### Шаг 2: Вычисление потерь

Теперь мы можем использовать эти вероятности для вычисления функции потерь GloVe. Предположим, мы инициализируем векторы слов случайным образом:

- $u_1, u_2, u_3$ — векторы для "кошка", "собака" и "животное".
- Скажем, они равны $[0.1, 0.2]$, $[0.2, 0.3]$ и $[0.4, 0.5]$.

Подставим в формулу потерь:

$$
J = f(X_{12}) (u_1^T u_2 + b_1 + b_2 - \log(X_{12}))^2
$$

Предположим, $f(X_{12}) = 1$:

$$
u_1^T u_2 = 0.1 \cdot 0.2 + 0.2 \cdot 0.3 = 0.02 + 0.06 = 0.08
$$

Таким образом, подставив значения, мы можем вычислить:

$$
J = (0.08 + b_1 + b_2 - \log(2))^2
$$

### Заключение

Метод GloVe позволяет эффективно представлять слова в виде векторов, захватывая их семантические отношения. Он основан на статистике со-occurrence и позволяет обучать векторы, которые могут быть использованы для различных задач NLP, таких как классификация текста, машинный перевод и анализ тональности.

### Применение GloVe

1. Классификация текста: Векторы слов могут быть использованы в качестве входных данных для классификаторов.
2. Машинный перевод: Они могут быть интегрированы в системы машинного перевода для лучшего понимания контекста.
3. Анализ тональности: GloVe позволяет захватывать эмоциональную окраску слов, что важно для анализа тональности.

Используя GloVe, мы можем эффективно обрабатывать текстовые данные и извлекать из них ценную информацию.

#4. ELMo (Embeddings from Language Models)

## Введение

ELMo (Embeddings from Language Models) представляет собой метод создания контекстуализированных эмбеддингов слов для задач обработки естественного языка (NLP). В отличие от традиционных методов, таких как Word2Vec и GloVe, которые предоставляют фиксированные векторные представления для каждого слова, ELMo генерирует эмбеддинги, зависящие от контекста, что значительно улучшает качество выполнения различных задач NLP, таких как классификация текстов, анализ настроений и извлечение информации.

## Теоретическая основа

### 1. Архитектура ELMo

ELMo использует двухнаправленные LSTM (Long Short-Term Memory) для генерации эмбеддингов, учитывающих как левую, так и правую контекстуальные зависимости. Основная идея состоит в том, чтобы использовать слои нейронной сети для генерации векторов, которые адаптируются к контексту предложения.

#### 1.1. Двухнаправленные LSTM

Двухнаправленный LSTM состоит из двух отдельных LSTM: один обрабатывает последовательность слева направо, а другой — справа налево. Обозначим входную последовательность как $X = (x_1, x_2, \ldots, x_n)$. Для каждого слова $x_t$ мы получаем два скрытых состояния:

- **Состояние слева направо**: $h_t^{\text{forward}}$
- **Состояние справа налево**: $h_t^{\text{backward}}$

Скрытые состояния вычисляются следующим образом:

$$
h_t^{\text{forward}} = LSTM_{\text{forward}}(h_{t-1}^{\text{forward}}, x_t)
$$

$$
h_t^{\text{backward}} = LSTM_{\text{backward}}(h_{t+1}^{\text{backward}}, x_t)
$$

### 2. Генерация эмбеддингов ELMo

Эмбеддинги для слова $x_t$ формируются путем объединения двух скрытых состояний:

$$
\text{ELMo}(x_t) = W_1 h_t^{\text{forward}} + W_2 h_t^{\text{backward}}
$$

где $W_1$ и $W_2$ — обучаемые параметры, которые регулируются во время тренировки модели.

### 3. Обучение модели

Обучение ELMo осуществляется с помощью языковой модели, которая предсказывает вероятности следующего слова в последовательности. Функция потерь может быть определена как:

$$
L = -\sum_{t=1}^{n} \log P(x_t | x_1, x_2, \ldots, x_{t-1})
$$

### 4. Применение ELMo в NLP

Эмбеддинги ELMo могут быть использованы как входные данные для различных моделей, включая полносвязные нейронные сети, CNN и трансформеры. Важно отметить, что ELMo можно легко интегрировать в существующие архитектуры, добавляя контекстуализированные эмбеддинги к фиксированным представлениям слов.

## Математические формулы

1. **Скрытые состояния LSTM**:

$$
h_t = f(W_h h_{t-1} + W_x x_t + b)
$$

где $f$ — активационная функция (обычно tanh), $W_h$ и $W_x$ — матрицы весов, а $b$ — вектор смещения.

2. **Обновление состояния ячейки**:

$$
C_t = f(C_{t-1} \cdot i_t + \tilde{C}_t \cdot f_t)
$$

где:
- $i_t$ — входной сигнал,
- $f_t$ — забывающая ячейка,
- $\tilde{C}_t$ — новая ячейка.

3. **Генерация эмбеддингов**:

$$
\text{ELMo}(x_t) = W_1 h_t^{\text{forward}} + W_2 h_t^{\text{backward}}
$$

где $W_1$ и $W_2$ — обучаемые параметры.

### Пример 1: Генерация эмбеддингов ELMo

Рассмотрим простую последовательность:


Для слова "play" будем вычислять эмбеддинги с использованием ELMo.

1. **Входные данные**:
   - $X = (I, love, to, play, football)$

   - $W_1 = [0.1, 0.2, 0.3]$
   - $W_2 = [0.3, 0.4, 0.5]$

5. **Вычисление эмбеддинга**:

$$
\text{ELMo}(\text{play}) = W_1 h_{\text{play}}^{\text{forward}} + W_2 h_{\text{play}}^{\text{backward}}

$$
= [0.1, 0.2, 0.3] \cdot [0.2, 0.3, 0.4] + [0.3, 0.4, 0.5] \cdot [0.5, 0.6, 0.7]

- $h_{\text{sat}}^{\text{backward}} = [0.2, 0.4, 0.5]$

- $h_{\text{the}}^{\text{backward}} = [0.2, 0.1, 0.4]$

Для реализации примеров, приведённых выше, на Python мы можем использовать библиотеки numpy для вычислений и matplotlib для визуализации. Мы создадим функции для вычисления эмбеддингов ELMo и визуализируем результаты для каждого примера.

Приведем код, который реализует все три примера и визуализирует результаты:


````python
import numpy as np
import matplotlib.pyplot as plt

# Функция для вычисления эмбеддинга ELMo
def elmo_embedding(forward, backward, W1, W2):
    forward_part = np.dot(W1, forward)
    backward_part = np.dot(W2, backward)
    return forward_part + backward_part

# Пример 1: Генерация эмбеддингов для слова
def example_1():
    # Векторы скрытых состояний для слова "sat"
    h_forward = np.array([0.4, 0.5, 0.6])  # Вектор прямого прохода
    h_backward = np.array([0.3, 0.4, 0.2])  # Вектор обратного прохода
    W1 = np.array([0.4, 0.3, 0.2])
    W2 = np.array([0.2, 0.4, 0.3])

    # Вычисляем ELMo
    elmo_value = elmo_embedding(h_forward, h_backward, W1, W2)
    print(f"ELMo embedding for 'sat': {elmo_value:.2f}")

# Пример 2: Классификация текста
def example_2():
    # Эмбеддинги для первого отзыва
    avg_positive = np.mean([0.5, 0.8, 0.6, 0.9])
    # Эмбеддинги для второго отзыва
    avg_negative = np.mean([0.3, 0.4, 0.2, 0.5, 0.4, 0.3])

    print(f"Average embedding for positive review: {avg_positive:.2f}")
    print(f"Average embedding for negative review: {avg_negative:.2f}")

    # Визуализация
    labels = ['Positive Review', 'Negative Review']
    values = [avg_positive, avg_negative]

    plt.bar(labels, values, color=['green', 'red'])
    plt.ylim(0, 1)
    plt.ylabel('Average Embedding')
    plt.title('Text Classification Based on ELMo Embeddings')
    plt.show()

# Пример 3: Сравнение эмбеддингов для контекста
def example_3():
    # Векторы скрытых состояний для слова "bank"
    h_forward_1 = np.array([0.6, 0.7, 0.5])
    h_backward_1 = np.array([0.3, 0.4, 0.2])

    h_forward_2 = np.array([0.8, 0.5, 0.6])
    h_backward_2 = np.array([0.4, 0.2, 0.3])

    W1 = np.array([0.5, 0.4, 0.3])
    W2 = np.array([0.2, 0.5, 0.4])

    # Вычисляем ELMo для первого контекста
    elmo_value_1 = elmo_embedding(h_forward_1, h_backward_1, W1, W2)
    # Вычисляем ELMo для второго контекста
    elmo_value_2 = elmo_embedding(h_forward_2, h_backward_2, W1, W2)

    print(f"ELMo embedding for 'bank' in context 1: {elmo_value_1:.2f}")
    print(f"ELMo embedding for 'bank' in context 2: {elmo_value_2:.2f}")

    # Визуализация
    contexts = ['Context 1', 'Context 2']
    embeddings = [elmo_value_1, elmo_value_2]

    plt.bar(contexts, embeddings, color=['blue', 'orange'])
    plt.ylim(0, 2)
    plt.ylabel('ELMo Embedding Value')
    plt.title('Comparison of ELMo Embeddings for "bank" in Different Contexts')
    plt.show()

# Выполнение примеров
example_1()
example_2()
example_3()
````

#5. Skip-Gram

### Введение

Skip-Gram — это один из основных алгоритмов представления слов, разработанный для обучения векторных представлений слов в NLP (Natural Language Processing). Этот метод позволяет предсказывать контекстные слова по данному целевому слову, основываясь на контекстном окружении. Алгоритм был представлен в статье **"Distributed Representations of Words and Phrases and their Compositionality"** (Mikolov et al., 2013).

### Цели и Задачи

- Обучить векторные представления слов.
- Обеспечить высокое качество представления слов в контексте.
- Использовать контекстные данные для предсказания слов.

### Основные Концепции

1. **Контекст и Целевое Слово**:
   - Для данного слова $w_t$ в предложении $S$ мы можем определить набор контекстных слов $C$.
   - Если $C$ включает $n$ слов до и после $w_t$, то длина окна контекста равна $2n$.

2. **Представление Слов**:
   - Каждое слово $w$ представляется в виде вектора $\mathbf{v}_w$.
   - Для простоты обозначим словарь слов как $V$.

### Математическая Модель

#### Целевая Функция

Алгоритм Skip-Gram нацелен на максимизацию вероятности появления контекстных слов $C$, учитывая целевое слово $w_t$:

$$
P(C|w_t) = \prod_{w_j \in C} P(w_j|w_t)
$$

где $w_j$ — контекстные слова.

#### Вероятностная Модель

Используется логистическая регрессия для предсказания вероятности $P(w_j|w_t)$:

$$
P(w_j|w_t) = \frac{\exp(\mathbf{v}_{w_j} \cdot \mathbf{v}_{w_t'})}{\sum_{w \in V} \exp(\mathbf{v}_{w} \cdot \mathbf{v}_{w_t'})}
$$

где $\mathbf{v}_{w_j}$ и $\mathbf{v}_{w_t'}$ — векторы представления соответствующих слов.

### Градиентный Спуск

Для оптимизации целевой функции можно использовать метод градиентного спуска. Для этого необходимо вычислить градиенты.

#### Вычисление Градиентов

1. **Функция Потерь**:

Используем логарифмическую функцию потерь:

$$
L = -\sum_{(w_t, w_j) \in D} \log P(w_j|w_t)
$$

где $D$ — набор пар (целевое слово, контекстное слово).

2. **Градиенты**:

Чтобы вычислить градиенты, нужно взять производные от функции потерь по вектору $\mathbf{v}_{w_t}$ и $\mathbf{v}_{w_j}$:

$$
\frac{\partial L}{\partial \mathbf{v}_{w_t}} = -\sum_{w_j \in C} \left(1 - P(w_j|w_t)\right) \mathbf{v}_{w_j}
$$

$$
\frac{\partial L}{\partial \mathbf{v}_{w_j}} = -\sum_{w_t} \left(1 - P(w_j|w_t)\right) \mathbf{v}_{w_t}
$$

#### Обновление Векторов

Для обновления векторов используем следующие формулы:

$$
\mathbf{v}_{w_t} \leftarrow \mathbf{v}_{w_t} - \eta \frac{\partial L}{\partial \mathbf{v}_{w_t}}
$$

$$
\mathbf{v}_{w_j} \leftarrow \mathbf{v}_{w_j} - \eta \frac{\partial L}{\partial \mathbf{v}_{w_j}}
$$

где $\eta$ — скорость обучения.

### Пример

Рассмотрим предложение: "Кошка сидит на коврике". Если $n=1$, целевое слово $w_t = "сидит"$, а контекстные слова будут $C = \{ "Кошка", "на", "коврике" \}$.

1. **Выбор Целевого Слова**: $w_t = "сидит"$.
2. **Контекст**: $C = \{ "Кошка", "на", "коврике" \}$.
3. **Вероятность**: Рассчитываем $P("Кошка" | "сидит")$, $P("на" | "сидит")$, $P("коврике" | "сидит")$.
4. **Обновление Векторов**: На основе вычисленных градиентов обновляем векторы для "сидит", "Кошка", "на", и "коврике".

### Заключение

Skip-Gram является мощным методом для обучения векторных представлений слов. Этот алгоритм обеспечивает контекстуальную информацию, которая может быть использована в различных задачах обработки естественного языка. Оптимизация с помощью градиентного спуска и его производных позволяет эффективно обучать модели, улучшая качество представления слов.

### Пример 1: Обучение модели Skip-Gram

#### Данные
Рассмотрим простое предложение:
- "Кошка сидит на коврике".

Для этого примера мы выберем контекстное окно $n = 1$. Таким образом, целевое слово и его контекст будут:

- Целевое слово $w_t = \text{"сидит"}$
- Контекстные слова $C = \{ \text{"Кошка"}, \text{"на"} \}$

#### Инициализация
Предположим, что у нас есть случайные векторы представления для каждого слова:

- $\mathbf{v}_{\text{"Кошка"}} = [0.2, 0.3]$
- $\mathbf{v}_{\text{"сидит"}} = [0.5, 0.1]$
- $\mathbf{v}_{\text{"на"}} = [0.4, 0.6]$
- $\mathbf{v}_{\text{"коврике"}} = [0.1, 0.2]$

#### Шаг 1: Вычисление вероятностей

Мы будем использовать логистическую регрессию для вычисления вероятности контекстных слов. Для простоты используем Softmax для двух контекстных слов.

##### Вычисление $P(\text{"Кошка"}|\text{"сидит"})$:

$$
P(\text{"Кошка"}|\text{"сидит"}) = \frac{\exp(\mathbf{v}_{\text{"Кошка"}} \cdot \mathbf{v}_{\text{"сидит"}})}{\sum_{w \in V} \exp(\mathbf{v}_{w} \cdot \mathbf{v}_{\text{"сидит"}})}
$$

Давайте реализуем примеры Skip-Gram на Python с использованием библиотеки NumPy для вычислений и Matplotlib для визуализации. Мы создадим простую реализацию, которая будет включать в себя инициализацию векторов слов, вычисление вероятностей и обновление векторов, а затем визуализируем результаты.


````python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    """Вычисление Softmax для вектора."""
    exp_x = np.exp(x - np.max(x))  # Вычитаем max для численной стабильности
    return exp_x / exp_x.sum(axis=0)

def compute_probabilities(target_vector, context_vectors):
    """Вычисление вероятностей контекстных слов."""
    scores = np.dot(context_vectors, target_vector)
    probabilities = softmax(scores)
    return probabilities

def update_vectors(target_vector, context_vectors, probabilities, learning_rate):
    """Обновление векторов на основе градиентов."""
    grad = np.zeros_like(target_vector)
    for i, p in enumerate(probabilities):
        grad += (1 - p) * context_vectors[i]
    target_vector -= learning_rate * grad
    return target_vector

def visualize_vectors(vectors, words):
    """Визуализация векторов слов."""
    plt.figure(figsize=(8, 8))
    for i, word in enumerate(words):
        plt.quiver(0, 0, vectors[i][0], vectors[i][1], angles='xy', scale_units='xy', scale=1, label=word)
        plt.text(vectors[i][0], vectors[i][1], word, fontsize=12)
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    plt.grid()
    plt.axhline(0, color='black',linewidth=0.5, ls='--')
    plt.axvline(0, color='black',linewidth=0.5, ls='--')
    plt.title("Word Vectors")
    plt.legend()
    plt.show()

# Пример 1
words_1 = ["Кошка", "сидит", "на", "коврике"]
vectors_1 = np.array([[0.2, 0.3], [0.5, 0.1], [0.4, 0.6], [0.1, 0.2]])
target_word_idx_1 = 1  # "сидит"
context_word_indices_1 = [0, 2]  # "Кошка" и "на"

# Параметры
learning_rate = 0.01
epochs = 10

# Обучение модели для Примера 1
for epoch in range(epochs):
    target_vector_1 = vectors_1[target_word_idx_1]
    context_vectors_1 = vectors_1[context_word_indices_1]
    probabilities_1 = compute_probabilities(target_vector_1, context_vectors_1)
    target_vector_1 = update_vectors(target_vector_1, context_vectors_1, probabilities_1, learning_rate)
    vectors_1[target_word_idx_1] = target_vector_1

# Визуализация результатов Примера 1
visualize_vectors(vectors_1, words_1)

# Пример 2
words_2 = ["Собака", "гуляет", "в", "парке"]
vectors_2 = np.array([[0.6, 0.2], [0.1, 0.9], [0.4, 0.5], [0.3, 0.8]])
target_word_idx_2 = 1  # "гуляет"
context_word_indices_2 = [0, 2]  # "Собака" и "в"

# Обучение модели для Примера 2
for epoch in range(epochs):
    target_vector_2 = vectors_2[target_word_idx_2]
    context_vectors_2 = vectors_2[context_word_indices_2]
    probabilities_2 = compute_probabilities(target_vector_2, context_vectors_2)
    target_vector_2 = update_vectors(target_vector_2, context_vectors_2, probabilities_2, learning_rate)
    vectors_2[target_word_idx_2] = target_vector_2

# Визуализация результатов Примера 2
visualize_vectors(vectors_2, words_2)
````
