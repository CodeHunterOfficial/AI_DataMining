## 1. Основные понятия деревьев решений

### 1.1 Что такое дерево решений?

Дерево решений — это модель, состоящая из узлов и ветвей, которая используется для принятия решений на основе входных данных. Каждый узел представляет собой тест на атрибут, каждая ветвь — результат теста, а листья — классы или значения, которые мы предсказываем.

### 1.2 Структура дерева решений

- Корень дерева: Начальный узел, который содержит все данные.
- Внутренние узлы: Узлы, представляющие тесты на атрибуты.
- Листовые узлы: Узлы, которые представляют конечные результаты.

### 1.3 Алгоритм построения дерева решений

Для построения дерева решений используется алгоритм, который включает следующие шаги:

1. Выбор атрибута: На каждом узле выбирается атрибут, который наилучшим образом разделяет данные. Для этого часто используется критерия Gini impurity или энтропия.
2. Разделение данных: Данные разбиваются на подмножества на основе выбранного атрибута.
3. Рекурсия: Процесс повторяется для каждого подмножества, пока не будет достигнут критерий остановки (например, максимальная глубина дерева или минимальное количество объектов в листе).

## 2. Критерии оценки разделения

### 2.1 Энтропия

Энтропия используется для измерения неопределенности в данных. Для множества $S$ с классами $C_1, C_2, \ldots, C_k$ она определяется как:

$$
H(S) = -\sum_{i=1}^{k} p(C_i) \log_2 p(C_i)
$$

где $p(C_i)$ — вероятность появления класса $C_i$.

### 2.2 Уменьшение энтропии

Когда мы разделяем данные, мы можем вычислить, как изменяется энтропия. Для подмножества $S_t$ с атрибутом $A$:

$$
H(S | A) = \sum_{t} \frac{|S_t|}{|S|} H(S_t)
$$

где $|S_t|$ — размер подмножества $S_t$.

### 2.3 Индекс Джини

Индекс Джини также используется для оценки чистоты узла и вычисляется как:

$$
Gini(S) = 1 - \sum_{i=1}^{k} p(C_i)^2
$$

Где $p(C_i)$ — доля объектов класса $C_i$ в множестве $S$.

### 2.4 Выбор лучшего атрибута

Выбор атрибута $A$ основан на максимизации уменьшения энтропии или минимизации индекса Джини:

$$
\Delta H = H(S) - H(S | A)
$$

или

$$
\Delta Gini = Gini(S) - Gini(S | A)
$$

## 3. Применение деревьев решений в NLP

- Идентификация тем: Классификация документов по темам.

## 4. Математические основы

### 4.1 Обучение дерева решений

Процесс обучения включает минимизацию функции потерь. Для классификации это может быть:

$$
L(y, \hat{y}) = -\sum_{i=1}^{n} y_i \log(\hat{y}_i)
$$

где $y$ — истинные метки, а $\hat{y}$ — предсказанные вероятности.

### 4.2 Вычисление градиентов

Для обновления параметров в модели, использующей градиентный спуск, необходимо вычислить градиенты функции потерь. Например, для функции потерь в контексте классификации:

$$
\frac{\partial L}{\partial \hat{y}} = -\frac{y}{\hat{y}} + \frac{1 - y}{1 - \hat{y}}
$$

Это значение используется для обновления предсказанных вероятностей.

### 4.3 Регуляризация

Для предотвращения переобучения деревьев решений можно использовать методы регуляризации, такие как обрезка (pruning). Обрезка позволяет удалить ветви, которые имеют низкую значимость, уменьшая сложность модели.

### 5.2 Пример: Анализ тональности

- Выходные данные: метки (положительный/отрицательный).

Мы можем использовать различные атрибуты, такие как:

- Частота появления негативных слов.

### 5.3 Оценка производительности

Для оценки производительности модели можно использовать метрики, такие как точность, полнота и F1-мера.

$$
Precision = \frac{TP}{TP + FP}
$$

$$
Recall = \frac{TP}{TP + FN}
$$

$$
F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}
$$

где $TP$ — истинно положительные, $FP$ — ложно положительные, а $FN$ — ложно отрицательные.

### Пример 1: Классификация текстов

Задача: Классификация отзывов о фильмах на положительные и отрицательные на основе некоторых признаков.

#### Данные

Предположим, у нас есть следующий набор отзывов, которые мы представили в виде таблицы:

| Отзыв                  | Длина (слов) | Кол-во позитивных слов | Кол-во негативных слов | Класс       |
||--|--|--|-|
| Отличный фильм!        | 3            | 2                     | 0                     | Положительный |
| Плохо, не рекомендую.  | 5            | 0                     | 3                     | Отрицательный  |
| Замечательная работа!  | 3            | 3                     | 0                     | Положительный |
| Ужасный сюжет.         | 3            | 0                     | 2                     | Отрицательный  |
| Неплохой, но могло быть лучше.| 7 | 1                     | 1                     | Положительный |

#### Шаг 1: Выбор атрибута

Мы можем использовать атрибуты: длина отзыва, количество позитивных слов и количество негативных слов. Чтобы выбрать наилучший атрибут, вычислим энтропию для каждого атрибута.

##### Вычисление общей энтропии:

Сначала найдем общую энтропию $H(S)$:

- Число положительных отзывов $= 3$
- Число отрицательных отзывов $= 2$
- Всего отзывов $= 5$

Вероятности классов:

$$
p(\text{Положительный}) = \frac{3}{5} = 0.6
$$
$$
p(\text{Отрицательный}) = \frac{2}{5} = 0.4
$$

Теперь подставим в формулу для энтропии:

$$
H(S) = -\left(0.6 \cdot \log_2(0.6) + 0.4 \cdot \log_2(0.4)\right) \approx 0.971
$$

##### Разделение по количеству позитивных слов

Рассмотрим разделение по атрибуту "Количество позитивных слов". Получим следующие подмножества:

1. Подмножество 1 (0 позитивных слов):
   - Отзывы: "Плохо, не рекомендую", "Ужасный сюжет"
   - Классы: 2 отрицательных

2. Подмножество 2 (1 позитивное слово):
   - Отзывы: "Неплохой, но могло быть лучше"
   - Класс: 1 положительный

3. Подмножество 3 (2 позитивных слов):
   - Отзывы: "Отличный фильм!"
   - Класс: 1 положительный

4. Подмножество 4 (3 позитивных слов):
   - Отзывы: "Замечательная работа!"
   - Класс: 1 положительный

Теперь рассчитаем энтропию для каждого подмножества:

- Для подмножества 1 (2 отзыва):
$$
H(S_1) = -\left(1 \cdot \log_2(1)\right) = 0
$$

- Для подмножества 2 (1 отзыв):
$$
H(S_2) = -\left(1 \cdot \log_2(1)\right) = 0
$$

- Для подмножества 3 (1 отзыв):
$$
H(S_3) = -\left(1 \cdot \log_2(1)\right) = 0
$$

- Для подмножества 4 (1 отзыв):
$$
H(S_4) = -\left(1 \cdot \log_2(1)\right) = 0
$$

##### Расчет уменьшения энтропии

Теперь рассчитаем уменьшение энтропии:

$$
H(S | \text{позитивные слова}) = \frac{2}{5} \cdot 0 + \frac{1}{5} \cdot 0 + \frac{1}{5} \cdot 0 + \frac{1}{5} \cdot 0 = 0
$$

Теперь можем вычислить уменьшение энтропии:

$$
\Delta H = H(S) - H(S | \text{позитивные слова}) = 0.971 - 0 = 0.971
$$

#### Шаг 2: Анализ других атрибутов

Аналогично, мы можем рассчитать уменьшение энтропии для других атрибутов (длина отзыва и количество негативных слов). Предположим, что для длины отзыва мы получили уменьшение энтропии $0.321$, а для негативных слов $0.630$.

#### Шаг 3: Выбор лучшего атрибута

Лучший атрибут для разделения — это "Количество позитивных слов" (уменьшение энтропии = 0.971). Это значит, что мы будем делить данные по этому атрибуту.

#### Шаг 4: Построение дерева

1. Разделяем данные по "Количество позитивных слов".
2. Для каждого подмножества повторяем процесс, пока не достигнем максимальной глубины или листьев.

Таким образом, на первом уровне мы делим по "Количество позитивных слов", а дальше продолжаем делить по оставшимся атрибутам, пока не достигнем конечных классов.



### Пример 2: Анализ тональности

Задача: Определить тональность текстов на основе некоторых характеристик.

#### Данные

Допустим, у нас есть следующий набор текстов с атрибутами:

| Текст                  | Длина (слов) | Частота позитивных слов | Частота негативных слов | Тональность   |
||--||||
| Это было замечательно! | 4            | 3                      | 0                      | Положительная  |
| Ужасно, не рекомендую. | 5            | 0                      | 3                      | Отрицательная  |
| В целом неплохо.       | 3            | 1                      | 1                      | Нейтральная    |
| Это худший фильм.      | 4            | 0                      | 2                      | Отрицательная  |
| Прекрасная работа!     | 3            | 3                      | 0                      | Положительная  |

#### Шаг 1: Выбор атрибута

Для начала рассчитаем энтропию $H(S)$ для всего множества.

- Положительные: 2
- Отрицательные: 2
- Нейтральные: 1

Общее количество отзывов: 5.

Вероятности классов:

$$
p(\text{Положительная}) = \frac{2}{5} = 0.4, \quad p(\text{Отрицательная}) = \frac{2}{5} = 0.4, \quad p(\text{Нейтральная}) = \frac{1}{5} = 0.2
$$

Теперь можем найти общую энтропию:

$$
H(S) = -\left(0.4 \cdot \log_2(0.4) + 0.4 \cdot \log_2(0.4) + 0.2 \cdot \log_2(0.2)\right) \approx 1.485
$$

#### Шаг 2: Разделение по частоте позитивных слов

Теперь разделим на подмножества по "Частота позитивных слов":

1. Подмножество 1 (0 позитивных слов):
   - Тексты: "Ужасно, не рекомендую", "Это худший фильм."
   - Классы: 2 отрицательных

2. Подмножество 2 (1 позитивное слово):
   - Текст: "В целом неплохо."
   - Класс: Нейтральный

3. Подмножество 3 (2 позитивных слова):
   - Нет таких текстов.

4. Подмножество 4 (3 позитивных слова):
   - Тексты: "Это было замечательно!", "Прекрасная работа!"
   - Классы: 2 положительных

Теперь можем вычислить энтропию для каждого подмножества:

- Для подмножества 1 (2 отзыва):
$$
H(S_1) = -\left(1 \cdot \log_2(1)\right) = 0
$$

- Для подмножества 2 (1 отзыв):
$$
H(S_2) = -\left(1 \cdot \log_2(1)\right) = 0
$$

- Для подмножества 3 (0 слов):
$$
H(S_3) = 0
$$

- Для подмножества 4 (2 отзыва):
$$
H(S_4) = -\left(0.5 \cdot \log_2(0.5) + 0.5 \cdot \log_2(0.5)\right) = 1
$$

#### Шаг 3: Расчет уменьшения энтропии

Теперь рассчитаем уменьшение энтропии для данного атрибута:

$$
H(S | \text{позитивные слова}) = \frac{2}{5} \cdot H(S_1) + \frac{1}{5} \cdot H(S_2) + \frac{0}{5} \cdot H(S_3) + \frac{2}{5} \cdot H(S_4)
$$

Подставляя значения, получаем:

$$
H(S | \text{позитивные слова}) = \frac{2}{5} \cdot 0 + \frac{1}{5} \cdot 0 + \frac{0}{5} \cdot 0 + \frac{2}{5} \cdot 1 = \frac{2}{5} = 0.4
$$

Теперь вычислим уменьшение энтропии:

$$
\Delta H = H(S) - H(S | \text{позитивные слова}) = 1.485 - 0.4 = 1.085
$$

#### Шаг 4: Анализ других атрибутов

Аналогично, мы можем рассчитать уменьшение энтропии для других атрибутов. Предположим, что мы получили следующие значения:

- Уменьшение энтропии по "Длина (слов)" = $0.5$
- Уменьшение энтропии по "Частота негативных слов" = $0.321$

#### Шаг 5: Выбор лучшего атрибута

Лучший атрибут для разделения — это "Частота позитивных слов" (уменьшение энтропии = 1.085). Это значит, что мы будем делить данные по этому атрибуту.

#### Шаг 6: Построение дерева

1. Разделяем данные по "Частота позитивных слов".
2. Для каждого подмножества повторяем процесс, пока не достигнем максимальной глубины или листья.

Таким образом, на первом уровне мы делим по "Частота позитивных слов", а дальше продолжаем делить по оставшимся атрибутам, пока не достигнем конечных классов.


### Пример 1: Классификация текстов

В этом примере мы классифицируем отзывы о фильмах как положительные или отрицательные.


```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Данные
data = {
    'Длина (слов)': [3, 5, 3, 3, 7],
    'Количество позитивных слов': [2, 0, 3, 0, 1],
    'Количество негативных слов': [0, 3, 0, 2, 1],
    'Класс': ['Положительный', 'Отрицательный', 'Положительный', 'Отрицательный', 'Положительный']
}

# Создаем DataFrame
df = pd.DataFrame(data)

# Преобразование классов в числовые значения
df['Класс'] = df['Класс'].map({'Положительный': 1, 'Отрицательный': 0})

# Определяем признаки и целевую переменную
X = df[['Длина (слов)', 'Количество позитивных слов', 'Количество негативных слов']]
y = df['Класс']

# Создаем и обучаем модель дерева решений
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X, y)

# Визуализация дерева решений
plt.figure(figsize=(10, 6))
plot_tree(model, feature_names=X.columns, class_names=['Отрицательный', 'Положительный'], filled=True)
plt.title('Дерево решений для классификации отзывов о фильмах')
plt.show()
```

### Пример 2: Анализ тональности

Во втором примере мы определяем тональность текстов на основе некоторых характеристик.



```python
# Данные
data_tone = {
    'Длина (слов)': [4, 5, 3, 4, 3],
    'Частота позитивных слов': [3, 0, 1, 0, 3],
    'Частота негативных слов': [0, 3, 1, 2, 0],
    'Тональность': ['Положительная', 'Отрицательная', 'Нейтральная', 'Отрицательная', 'Положительная']
}

# Создаем DataFrame
df_tone = pd.DataFrame(data_tone)

# Преобразование тональности в числовые значения
df_tone['Тональность'] = df_tone['Тональность'].map({'Положительная': 1, 'Отрицательная': 0, 'Нейтральная': 2})

# Определяем признаки и целевую переменную
X_tone = df_tone[['Длина (слов)', 'Частота позитивных слов', 'Частота негативных слов']]
y_tone = df_tone['Тональность']

# Создаем и обучаем модель дерева решений
model_tone = DecisionTreeClassifier(criterion='entropy')
model_tone.fit(X_tone, y_tone)

# Визуализация дерева решений
plt.figure(figsize=(10, 6))
plot_tree(model_tone, feature_names=X_tone.columns, class_names=['Отрицательная', 'Нейтральная', 'Положительная'], filled=True)
plt.title('Дерево решений для анализа тональности текстов')
plt.show()
```
