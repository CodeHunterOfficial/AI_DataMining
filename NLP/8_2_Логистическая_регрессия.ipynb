#8.2. Логистическая регрессия

Логистическая регрессия - это статистический метод для бинарной классификации, который широко используется в задачах обработки естественного языка (Natural Language Processing, NLP). Он подходит для моделирования вероятности принадлежности некоторого наблюдения к одному из двух классов, используя набор входных признаков.

#### 1. Введение в логистическую регрессию

Логистическая регрессия - это модель классификации, основанная на линейной регрессии, но с сигмоидной функцией активации для преобразования выхода в вероятность. Основная идея состоит в том, чтобы найти линейную комбинацию входных признаков, которая максимально отделяет два класса.

Рассмотрим модель логистической регрессии:

$$ P(y = 1 \mid \mathbf{x}) = \sigma(\mathbf{w}^T \mathbf{x} + b), $$

где:
- $ y \in \{0, 1\}$ - целевая переменная.
- $ \mathbf{x} = [x_1, x_2, \ldots, x_n]$ - вектор входных признаков.
- $ \mathbf{w} = [w_1, w_2, \ldots, w_n]$ - вектор весов модели.
- $ b$ - смещение (bias).
- $ \sigma(z)$ - сигмоидная функция, определяемая как $ \sigma(z) = \frac{1}{1 + e^{-z}}$.

Функция активации $ \sigma(z)$ преобразует линейную комбинацию входных признаков $ z = \mathbf{w}^T \mathbf{x} + b$ в вероятность принадлежности к классу 1.

#### 2. Сигмоидная функция

Сигмоидная функция определяет, как линейная комбинация признаков преобразуется в вероятности:

$$ \sigma(z) = \frac{1}{1 + e^{-z}}, $$

где $ z = \mathbf{w}^T \mathbf{x} + b$.

График сигмоидной функции имеет S-образную форму и принимает значения в интервале $ (0, 1)$, что удобно для моделирования вероятности.

#### 3. Логистическая функция потерь (Log-Loss)

Для обучения модели логистической регрессии используется логистическая функция потерь, также называемая кросс-энтропийной функцией потерь:

$$ L(\mathbf{w}, b) = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log(\hat{y}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{y}^{(i)}) \right], $$

где:
- $ m$ - количество обучающих примеров.
- $ y^{(i)}$ - истинное значение метки для $ i$-го примера (0 или 1).
- $ \hat{y}^{(i)} = \sigma(\mathbf{w}^T \mathbf{x}^{(i)} + b)$ - предсказанная вероятность для $ i$-го примера.

Эта функция минимизируется с помощью методов оптимизации, таких как градиентный спуск.

#### 4. Градиентный спуск для логистической регрессии

Метод градиентного спуска используется для минимизации функции потерь. Основная идея заключается в обновлении параметров модели в направлении отрицательного градиента функции потерь.

Градиенты для параметров $ \mathbf{w}$ и $ b$ вычисляются следующим образом:

1. Для $ \mathbf{w}$:

   $$ \frac{\partial L}{\partial w_j} = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)}) x_j^{(i)}, $$

   где $ j$ - индекс признака.

2. Для $ b$:

   $$ \frac{\partial L}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)}). $$

Обновление параметров выполняется по правилам:

$$ w_j := w_j - \alpha \frac{\partial L}{\partial w_j}, $$

$$ b := b - \alpha \frac{\partial L}{\partial b}, $$

где $ \alpha$ - коэффициент обучения.

#### 5. Применение логистической регрессии в задачах NLP

Логистическая регрессия используется в NLP для решения множества задач:

- Классификация текста: Примеры включают определение спама, анализ тональности (положительный/отрицательный отзыв) и классификацию новостей.
- Тематическое моделирование: Определение темы документа на основе набора меток.
- Распознавание именованных сущностей (NER): Классификация слов в тексте как именованных сущностей, таких как имена людей, организации и т.д.

В этих задачах текст сначала преобразуется в численные признаки, такие как частоты слов (bag-of-words), TF-IDF или векторные представления (например, word2vec или embeddings).

#### 6. Пример применения логистической регрессии в классификации текста

Рассмотрим задачу классификации текстов по тональности (положительная или отрицательная оценка). Для решения этой задачи мы можем использовать следующие шаги:

1. Предобработка текста:
   - Токенизация: разбиение текста на отдельные слова.
   - Удаление стоп-слов (например, "и", "на", "в").
   - Преобразование слов к базовой форме (лемматизация или стемминг).

2. Преобразование текста в численные признаки:
   - Использование подхода "мешок слов" (bag-of-words).
   - Преобразование в TF-IDF признаки.

3. Обучение модели логистической регрессии:
   - Обучение с использованием градиентного спуска для минимизации логистической функции потерь.

4. Оценка качества модели:
   - Метрики: точность, полнота, F1-оценка.

#### 7. Вычисление градиентов и обновление весов на примере

Рассмотрим пример с двумя признаками $ x_1$ и $ x_2$ для бинарной классификации. Пусть $ \mathbf{x} = [x_1, x_2]$, $ \mathbf{w} = [w_1, w_2]$, $ b = 0$, и известен один обучающий пример $ (x_1, x_2, y) = (0.5, -1.2, 1)$.

1. Предсказание вероятности:

   $$ z = w_1 \cdot 0.5 + w_2 \cdot (-1.2), $$
   $$ \hat{y} = \sigma(z). $$

2. Вычисление функции потерь:

   $$ L = -[y \log(\hat{y}) + (1 - y) \log(1 - \hat{y})]. $$

3. Вычисление градиентов:

   $$ \frac{\partial L}{\partial w_1} = (\hat{y} - y) \cdot 0.5, $$
   $$ \frac{\partial L}{\partial w_2} = (\hat{y} - y) \cdot (-1.2). $$

4. Обновление весов:

   $$ w_1 := w_1 - \alpha \frac{\partial L}{\partial w_1}, $$
   $$ w_2 := w_2 - \alpha \frac{\partial L}{\partial w_2}. $$

#### 8. Регуляризация в логистической регрессии

Чтобы предотвратить переобучение, часто добавляется регуляризация:

- L2-регуляризация (ридж-регрессия): добавление квадрата нормы весов к функции потерь.

   $$ L_{\text{рег}} = L + \frac{\lambda}{2} \sum_{j=1}^{n} w_j^2, $$

   где $ \lambda$ - параметр регуляризации.

- L1-регуляризация (лассо-регрессия): добавление суммы модулей весов.

   $$ L_{\text{рег}} = L + \lambda \sum_{j=1}^{n} |w_j|. $$

Регуляризация помогает улучшить обобщающую способность модели, уменьшая влияние отдельных признаков.

#### 9. Преимущества и недостатки логистической регрессии в NLP

Преимущества:
- Простота реализации и интерпретации.
- Хорошо работает на линейно разделимых данных.
- Подходит для задач, где требуется предсказание вероятностей.

Недостатки:
- Ограниченные возможности для нелинейных зависимостей.
- Чувствительность к выбросам и несбалансированным классам.
- При недостаточном количестве данных может переобучаться.

Понимание математических основ и методов оптимизации, таких как градиентный спуск, позволяет эффективно применять логистическую регрессию к реальным задачам обработки естественного языка.


Давайте рассмотрим два числовых примера применения логистической регрессии в задачах обработки естественного языка (NLP). В первом примере мы будем классифицировать отзывы о фильмах по тональности (положительные или отрицательные), а во втором — определим, является ли электронное письмо спамом или нет.

### Пример 1: Классификация отзывов о фильмах

#### Задача

У нас есть небольшой набор данных с отзывами о фильмах, и мы хотим классифицировать их как положительные (1) или отрицательные (0). Пусть у нас есть три отзыва с двумя признаками:

- Признаки:
  - $ x_1$: Частота слова "хороший".
  - $ x_2$: Частота слова "плохой".

| Отзыв | $ x_1$ | $ x_2$ | Метка $ y$ |
|-|-|-||
| 1     | 3        | 0        | 1             |
| 2     | 1        | 2        | 0             |
| 3     | 0        | 3        | 0             |

#### Шаг 1: Инициализация параметров

Сначала инициализируем веса и смещение. Пусть:

- $ w_1 = 0.5$
- $ w_2 = -0.5$
- $ b = 0.0$

#### Шаг 2: Прогнозирование

Для каждого отзыва мы будем предсказывать вероятность $ \hat{y}$ с использованием формулы логистической регрессии:

$$ \hat{y} = \sigma(w_1 \cdot x_1 + w_2 \cdot x_2 + b), $$

где $ \sigma(z) = \frac{1}{1 + e^{-z}}$.

Для отзыва 1:
- $ x_1 = 3$, $ x_2 = 0$
- Вычисляем $ z = 0.5 \cdot 3 + (-0.5) \cdot 0 + 0 = 1.5$
- $ \hat{y} = \sigma(1.5) = \frac{1}{1 + e^{-1.5}} \approx 0.8176$

Для отзыва 2:
- $ x_1 = 1$, $ x_2 = 2$
- Вычисляем $ z = 0.5 \cdot 1 + (-0.5) \cdot 2 + 0 = -0.5$
- $ \hat{y} = \sigma(-0.5) = \frac{1}{1 + e^{0.5}} \approx 0.3775$

Для отзыва 3:
- $ x_1 = 0$, $ x_2 = 3$
- Вычисляем $ z = 0.5 \cdot 0 + (-0.5) \cdot 3 + 0 = -1.5$
- $ \hat{y} = \sigma(-1.5) = \frac{1}{1 + e^{1.5}} \approx 0.1824$

#### Шаг 3: Вычисление функции потерь

Теперь вычислим логистическую функцию потерь для каждого отзыва. Используем формулу потерь:

$$ L = -[y \log(\hat{y}) + (1 - y) \log(1 - \hat{y})]. $$

Для отзыва 1:
- $ y = 1$
- $ L_1 = -[1 \cdot \log(0.8176) + 0 \cdot \log(1 - 0.8176)] \approx 0.203$

Для отзыва 2:
- $ y = 0$
- $ L_2 = -[0 \cdot \log(0.3775) + 1 \cdot \log(1 - 0.3775)] \approx 0.974$

Для отзыва 3:
- $ y = 0$
- $ L_3 = -[0 \cdot \log(0.1824) + 1 \cdot \log(1 - 0.1824)] \approx 0.200$

#### Шаг 4: Обновление параметров

Теперь мы можем обновить параметры $ w_1, w_2, b$ с помощью градиентного спуска. Для этого нам нужно вычислить градиенты:

1. Градиенты для весов:

   $$ \frac{\partial L_1}{\partial w_1} = (\hat{y_1} - y_1) \cdot x_1^{(1)} = (0.8176 - 1) \cdot 3 \approx -0.5452 $$
   
   $$ \frac{\partial L_1}{\partial w_2} = (\hat{y_1} - y_1) \cdot x_2^{(1)} = (0.8176 - 1) \cdot 0 \approx 0.0 $$

   $$ \frac{\partial L_2}{\partial w_1} = (\hat{y_2} - y_2) \cdot x_1^{(2)} = (0.3775 - 0) \cdot 1 \approx 0.3775 $$

   $$ \frac{\partial L_2}{\partial w_2} = (\hat{y_2} - y_2) \cdot x_2^{(2)} = (0.3775 - 0) \cdot 2 \approx 0.755 $$

   $$ \frac{\partial L_3}{\partial w_1} = (\hat{y_3} - y_3) \cdot x_1^{(3)} = (0.1824 - 0) \cdot 0 \approx 0.0 $$

   $$ \frac{\partial L_3}{\partial w_2} = (\hat{y_3} - y_3) \cdot x_2^{(3)} = (0.1824 - 0) \cdot 3 \approx 0.5472 $$

2. Обновление параметров:  
Предположим, что мы используем коэффициент обучения $ \alpha = 0.1$.

Обновление для отзыва 1:
- $ w_1 := 0.5 - 0.1 \cdot (-0.5452) \approx 0.5545$
- $ w_2 := -0.5 - 0.1 \cdot 0.0 = -0.5$
- $ b := 0 - 0.1 \cdot (-0.1824) \approx 0.0182$

Обновление для отзыва 2:
- $ w_1 := 0.5545 - 0.1 \cdot 0.3775 \approx 0.5168$
- $ w_2 := -0.5 - 0.1 \cdot 0.755 \approx -0.5755$
- $ b := 0.0182 - 0.1 \cdot (0.3775) \approx -0.0195$

Обновление для отзыва 3:
- $ w_1 := 0.5168 - 0.1 \cdot 0.0 = 0.5168$
- $ w_2 := -0.5755 - 0.1 \cdot 0.5472 \approx -0.6302$
- $ b := -0.0195 - 0.1 \cdot 0.1824 \approx -0.0377$

Теперь у нас есть новые параметры для следующей итерации.



### Пример 2: Классификация электронной почты на спам и не спам

#### Задача

Рассмотрим набор данных с несколькими признаками для классификации электронных писем. Признаки:

- $ x_1$: Количество слов "подарок".
- $ x_2$: Количество слов "срочно".

| Письмо | $ x_1$ | $ x_2$ | Метка $ y$ |
|--|-|-||
| 1      | 5        | 0        | 1             |
| 2      | 0        | 3        | 0             |
| 3      | 2        | 1        | 0             |

#### Шаг 1: Инициализация параметров

Пусть:
- $ w_1 = 0.4$
- $ w_2 = -0.6$
- $ b = 0.1$

#### Шаг 2: Прогнозирование

Теперь рассчитаем предсказанную вероятность $ \hat{y}$ для каждого письма.

Для письма 1:
- $ x_1 = 5$, $ x_2 = 0$


- $ z = 0.4 \cdot 5 + (-0.6) \cdot 0 + 0.1 = 2.1$
- $ \hat{y} = \sigma(2.1) = \frac{1}{1 + e^{-2.1}} \approx 0.8909$

Для письма 2:
- $ x_1 = 0$, $ x_2 = 3$
- $ z = 0.4 \cdot 0 + (-0.6) \cdot 3 + 0.1 = -1.7$
- $ \hat{y} = \sigma(-1.7) = \frac{1}{1 + e^{1.7}} \approx 0.1545$

Для письма 3:
- $ x_1 = 2$, $ x_2 = 1$
- $ z = 0.4 \cdot 2 + (-0.6) \cdot 1 + 0.1 = 0.3$
- $ \hat{y} = \sigma(0.3) = \frac{1}{1 + e^{-0.3}} \approx 0.5744$

#### Шаг 3: Вычисление функции потерь

Для письма 1:
- $ y = 1$
- $ L_1 = -[1 \cdot \log(0.8909) + 0 \cdot \log(1 - 0.8909)] \approx 0.115$

Для письма 2:
- $ y = 0$
- $ L_2 = -[0 \cdot \log(0.1545) + 1 \cdot \log(1 - 0.1545)] \approx 0.168$

Для письма 3:
- $ y = 0$
- $ L_3 = -[0 \cdot \log(0.5744) + 1 \cdot \log(1 - 0.5744)] \approx 0.854$

#### Шаг 4: Обновление параметров

Вычисляем градиенты и обновляем параметры аналогично первому примеру.

1. Градиенты для $ w_1$, $ w_2$, $ b$ вычисляются аналогично, и выполняются обновления с учетом коэффициента обучения $ \alpha$.

Таким образом, каждый шаг градиентного спуска приближает нас к оптимальным параметрам, которые минимизируют функцию потерь.











Давайте реализуем оба примера на Python с использованием библиотеки `matplotlib` для визуализации. Мы будем использовать логистическую регрессию для классификации отзывов о фильмах и электронных писем, а также покажем, как выглядят данные и результаты предсказаний.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Пример 1: Классификация отзывов о фильмах
# Данные
X1 = np.array([[3, 0], [1, 2], [0, 3]])  # Признаки: [Частота "хороший", Частота "плохой"]
y1 = np.array([1, 0, 0])  # Метки: положительный или отрицательный отзыв

# Модель логистической регрессии
model1 = LogisticRegression()
model1.fit(X1, y1)

# Прогнозы
pred_probs1 = model1.predict_proba(X1)[:, 1]

# Визуализация
plt.figure(figsize=(10, 6))
plt.scatter(X1[:, 0], X1[:, 1], c=y1, s=100, edgecolors='k', cmap=plt.cm.Paired)
plt.title("Классификация отзывов о фильмах")
plt.xlabel("Частота слова 'хороший'")
plt.ylabel("Частота слова 'плохой'")

# Границы решения
x_min, x_max = X1[:, 0].min() - 1, X1[:, 0].max() + 1
y_min, y_max = X1[:, 1].min() - 1, X1[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z1 = model1.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z1, alpha=0.3, cmap=plt.cm.Paired)
plt.colorbar(ticks=[0, 1], label='Класс')
plt.show()

# Пример 2: Классификация электронных писем
# Данные
X2 = np.array([[5, 0], [0, 3], [2, 1]])  # Признаки: [Количество "подарок", Количество "срочно"]
y2 = np.array([1, 0, 0])  # Метки: спам или не спам

# Модель логистической регрессии
model2 = LogisticRegression()
model2.fit(X2, y2)

# Прогнозы
pred_probs2 = model2.predict_proba(X2)[:, 1]

# Визуализация
plt.figure(figsize=(10, 6))
plt.scatter(X2[:, 0], X2[:, 1], c=y2, s=100, edgecolors='k', cmap=plt.cm.Paired)
plt.title("Классификация электронных писем")
plt.xlabel("Количество слов 'подарок'")
plt.ylabel("Количество слов 'срочно'")

# Границы решения
x_min, x_max = X2[:, 0].min() - 1, X2[:, 0].max() + 1
y_min, y_max = X2[:, 1].min() - 1, X2[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z2 = model2.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z2, alpha=0.3, cmap=plt.cm.Paired)
plt.colorbar(ticks=[0, 1], label='Класс')
plt.show()
```
