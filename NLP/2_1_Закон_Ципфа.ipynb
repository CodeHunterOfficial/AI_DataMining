## Закон Ципфа

### Введение

**Закон Ципфа** — это эмпирическое наблюдение, которое описывает частотное распределение слов в текстах на естественном языке. Он был впервые предложен американским лингвистом и филологом Джорджем Ципфом в 1935 году и более полно изложен в его книге «Гуманистическая наука и принцип минимальных усилий» в 1949 году. Закон Ципфа утверждает, что частота встречаемости слова обратно пропорциональна его рангу в списке частот слов.

В NLP (Natural Language Processing, обработка естественного языка) закон Ципфа играет важную роль в понимании структуры языка, распределения слов и их значений. В данной лекции мы подробно разберем основы этого закона, приведем его математическое выражение, изучим его следствия и обсудим примеры его применения в NLP.

### Основная идея закона Ципфа

Закон Ципфа утверждает, что для любого языка частота слова $f(r)$, находящегося на $r$-м месте по частоте, пропорциональна величине $\frac{1}{r}$:

$$
f(r) \propto \frac{1}{r}
$$

Где:
- $f(r)$ — частота слова с рангом $r$;
- $r$ — ранг слова (позиция в списке всех слов, отсортированных по убыванию частоты).

Частота встречаемости слова стремительно убывает с увеличением ранга. На практике это означает, что наиболее часто встречающееся слово в корпусе текста будет встречаться гораздо чаще, чем второе по частоте слово, а второе — гораздо чаще, чем третье, и так далее.

### Математическое выражение закона Ципфа

Закон Ципфа в математической форме может быть выражен как:

$$
f(r) = \frac{C}{r^a}
$$

Где:
- $f(r)$ — частота слова с рангом $r$;
- $C$ — нормирующая константа, которая зависит от общего количества слов в тексте;
- $r$ — ранг слова;
- $a$ — параметр закона, который для естественных языков обычно близок к 1 (стандартное значение $a \approx 1$).

#### Нормировка

Чтобы обеспечить корректное распределение вероятностей, важно правильно выбрать нормирующую константу $C$. Нормирующая константа определяется так, чтобы сумма всех частот была равна общему числу слов в корпусе.

Если мы рассматриваем вероятности встречаемости слов, то сумма всех вероятностей для всех рангов должна равняться 1:

$$
\sum_{r=1}^{N} P(r) = 1
$$

Где:
- $P(r)$ — вероятность встретить слово с рангом $r$, что определяется как $P(r) = \frac{f(r)}{T}$, где $T$ — общее количество слов в корпусе текста;
- $N$ — общее количество уникальных слов.

Тогда для вероятности получаем нормированное распределение:

$$
P(r) = \frac{1}{Z(r^a)}
$$

Где $Z$ — это нормировочная константа, которая называется *функцией Зипфа* и определяет нормировку распределения для всех рангов.

### Примеры распределения

Пусть $N = 10^5$ (общее число слов в корпусе), $C = 10^6$, тогда частота первых слов может выглядеть так:
1. Ранг 1: частота $f(1) = \frac{10^6}{1} = 10^6$
2. Ранг 2: частота $f(2) = \frac{10^6}{2} = 5 \times 10^5$
3. Ранг 3: частота $f(3) = \frac{10^6}{3} \approx 3.33 \times 10^5$

И так далее. Это показывает быстрое убывание частоты по мере увеличения ранга.

### Свойства закона Ципфа

1. **Обратная пропорциональность**: Частота слова обратно пропорциональна его рангу. Это одно из ключевых свойств закона Ципфа.
2. **Долгий хвост**: Закон Ципфа характеризуется наличием так называемого «долгого хвоста» (long tail). Большое количество слов имеют очень низкую частоту появления в корпусе.
3. **Эмпирическая универсальность**: Закон Ципфа работает не только для естественных языков, но и для других типов данных, таких как распределение населенных пунктов, экономические данные и даже генетические последовательности.

### Логарифмическое представление

Закон Ципфа часто удобно изображать в логарифмическом масштабе. Возьмем логарифм обеих частей выражения:

$$
\log f(r) = \log \left( \frac{C}{r^a} \right) = \log C - a \log r
$$

Это уравнение линейно относительно $\log r$, что позволяет представить закон Ципфа в виде прямой линии на логарифмическом графике. Наклон этой линии будет равен $-a$, а пересечение с осью $\log f(r)$ — $\log C$.

#### Пример

Для текста с параметром $a = 1$ график зависимости $\log f(r)$ от $\log r$ будет прямой с наклоном -1.

### Закон Хипса

Вместе с законом Ципфа часто рассматривается **закон Хипса** (Heaps' Law), который описывает зависимость количества уникальных слов в тексте от общего объема текста. Согласно закону Хипса:

$$
V(N) = KN^\beta
$$

Где:
- $V(N)$ — количество уникальных слов (словарь) в тексте из $N$ слов;
- $K$ и $\beta$ — эмпирические параметры (для большинства естественных языков $\beta \approx 0.5$).

Закон Хипса связан с законом Ципфа через представление текста: чем больше слов, тем больше уникальных слов добавляется по мере роста текста, но темп этого добавления замедляется по мере роста текста.

### Применение закона Ципфа в NLP

1. **Оптимизация словарей**: В NLP часто используются сокращенные словари для моделей. Закон Ципфа помогает определить, какие слова можно отбросить (слова с низким рангом) без значительной потери информации.
2. **Tokenization и Subword Units**: Методы разбиения текста на токены (как в алгоритмах BPE или WordPiece) активно используют понимание частотности слов, которое следует из закона Ципфа.
3. **Оценка языковых моделей**: При построении языковых моделей, таких как n-граммные модели или трансформеры, распределение вероятностей слов, предсказанных моделью, часто связано с законом Ципфа.
4. **Анализ семантики и прагматики**: Высокочастотные слова часто являются функциональными, тогда как редкие слова более содержательные, что используется для семантического анализа текста.

### Отклонения от закона Ципфа

Хотя закон Ципфа хорошо описывает распределение слов в текстах, существуют ситуации, где наблюдаются отклонения:
- В небольших текстах или узкоспециализированных корпусах распределение может быть более плоским.
- В очень больших текстах возможно появление более сложных закономерностей (смешение нескольких законов распределения).

### Связь с распределением Парето и законом Брэдфорда

Закон Ципфа является частным случаем более общего распределения Парето. Распределение Парето описывает вероятность того, что случайная величина больше некоторого значения, и оно также имеет гиперболическую форму. Закон Брэдфорда, применяемый в библиометрии, также связан с распределением Парето и законом Ципфа, описывая закономерности распределения информации в научных статьях.

### Заключение

Закон Ципфа — это фундаментальный закон, описывающий частотное распределение слов в естественных языках. Он находит широкое применение в NLP, начиная от построения моделей до оптимизации словарей и анализа текстов. Понимание этого закона помогает лучше разрабатывать алгоритмы и инструменты для обработки естественного языка.

#### Основные моменты:
- Закон Ципфа описывает обратную пропорциональность частоты слова его рангу.
- Математическое выражение: $f(r) = \frac{C}{r^a}$, где $a \approx 1$ для естественных языков.
- Важные свойства включают наличие «долгого хвоста» и универсальность закона.
- В логарифмическом масштабе закон Ц

ипфа имеет линейную форму.
- Закон Хипса связан с законом Ципфа и описывает рост словаря текста.

Эти концепции являются важными инструментами для анализа и обработки текстов, особенно в больших корпусах данных.




Давайте разберем конкретные числовые примеры закона Ципфа, чтобы закрепить его понимание на практике.

### Пример 1: Применение закона Ципфа для текста

Предположим, у нас есть небольшой корпус текста, содержащий **1000 слов**. Допустим, параметр $a = 1$, что соответствует стандартному закону Ципфа.

Частота слова с рангом $r$ описывается формулой:

$$
f(r) = \frac{C}{r}
$$

Где:
- $f(r)$ — частота слова с рангом $r$;
- $C$ — нормирующая константа.

Для простоты рассчитаем $C$ из общего числа слов в корпусе, которое равно 1000:

$$
C = 1000
$$

Теперь посчитаем частоты для первых нескольких слов.

#### Частота слова с рангом 1 (самое частое слово):

$$
f(1) = \frac{1000}{1} = 1000
$$

#### Частота слова с рангом 2:

$$
f(2) = \frac{1000}{2} = 500
$$

#### Частота слова с рангом 3:

$$
f(3) = \frac{1000}{3} \approx 333
$$

#### Частота слова с рангом 10:

$$
f(10) = \frac{1000}{10} = 100
$$

#### Частота слова с рангом 100:

$$
f(100) = \frac{1000}{100} = 10
$$

Таким образом, частота встречаемости слова резко уменьшается по мере увеличения ранга.

### Пример 2: Использование нормированной вероятности

Теперь предположим, что мы хотим вычислить не частоту, а вероятность встречаемости слов. Пусть общий объем текста составляет $N = 1000$ слов. Вероятность $P(r)$ слова с рангом $r$ определяется как частота слова, деленная на общее количество слов:

$$
P(r) = \frac{f(r)}{N} = \frac{C}{rN}
$$

Для нашего примера, где $C = 1000$ и $N = 1000$, вероятность для первых слов:

#### Вероятность слова с рангом 1:

$$
P(1) = \frac{1000}{1 \times 1000} = 1
$$

#### Вероятность слова с рангом 2:

$$
P(2) = \frac{1000}{2 \times 1000} = 0.5
$$

#### Вероятность слова с рангом 3:

$$
P(3) = \frac{1000}{3 \times 1000} \approx 0.333
$$

#### Вероятность слова с рангом 10:

$$
P(10) = \frac{1000}{10 \times 1000} = 0.1
$$

#### Вероятность слова с рангом 100:

$$
P(100) = \frac{1000}{100 \times 1000} = 0.01
$$

Из этих расчетов видно, что наиболее частое слово будет встречаться в половине всех случаев, тогда как 10-е по частоте слово встречается только в 10% всех случаев, а 100-е — всего в 1% случаев.

### Пример 3: Закон Хипса

Теперь посмотрим, как работает закон Хипса для роста словаря. Пусть закон Хипса для этого корпуса имеет параметры $K = 10$ и $\beta = 0.5$. Тогда количество уникальных слов $V(N)$ можно вычислить по формуле:

$$
V(N) = K N^\beta
$$

Посчитаем количество уникальных слов для разных размеров текста $N$.

#### Для $N = 1000$:

$$
V(1000) = 10 \times 1000^{0.5} = 10 \times 31.62 \approx 316
$$

#### Для $N = 10,000$:

$$
V(10,000) = 10 \times 10,000^{0.5} = 10 \times 100 = 1000
$$

#### Для $N = 100,000$:

$$
V(100,000) = 10 \times 100,000^{0.5} = 10 \times 316.23 \approx 3162
$$

Таким образом, количество уникальных слов увеличивается по мере роста общего объема текста, но с замедляющимся темпом (логарифмическое замедление).

### Пример 4: Логарифмическая зависимость

Рассмотрим, как распределение частот выглядит на логарифмическом графике. Для этого воспользуемся выражением:

$$
\log f(r) = \log C - \log r
$$

Если $C = 1000$, то частоты для первых пяти слов в логарифмическом масштабе:

- $f(1) = 1000$, $\log f(1) = \log 1000 = 3$
- $f(2) = 500$, $\log f(2) = \log 500 \approx 2.69897$
- $f(3) = 333$, $\log f(3) = \log 333 \approx 2.52244$
- $f(4) = 250$, $\log f(4) = \log 250 \approx 2.39794$
- $f(5) = 200$, $\log f(5) = \log 200 \approx 2.30103$

График зависимости $\log f(r)$ от $\log r$ будет прямой линией с углом наклона -1, что является классической иллюстрацией закона Ципфа.

### Пример 5: Проверка отклонений

Рассмотрим, что происходит с распределением, если параметр $a$ отклоняется от единицы. Например, пусть $a = 1.2$. В этом случае формула для частоты становится:

$$
f(r) = \frac{C}{r^{1.2}}
$$

Для простоты снова возьмем $C = 1000$ и вычислим частоты для первых пяти слов:

- $f(1) = \frac{1000}{1^{1.2}} = 1000$
- $f(2) = \frac{1000}{2^{1.2}} \approx 435$
- $f(3) = \frac{1000}{3^{1.2}} \approx 285$
- $f(4) = \frac{1000}{4^{1.2}} \approx 193$
- $f(5) = \frac{1000}{5^{1.2}} \approx 144$

Частоты слов убывают чуть быстрее, чем в стандартном случае $a = 1$. Это показывает, как изменение параметра $a$ влияет на распределение.

### Заключение

Числовые примеры закона Ципфа наглядно демонстрируют его основные свойства: быстрое убывание частот слов и «долгий хвост» редких слов. Пример использования закона Хипса показывает, как растет количество уникальных слов по мере увеличения объема текста. Логарифмическое представление позволяет легко увидеть линейную зависимость частот от ранга. Закон Ципфа является мощным инструментом для моделирования и анализа текстов в задачах NLP.





```python
# Импортируем необходимые библиотеки
import numpy as np
import matplotlib.pyplot as plt

# Настройки для красивых графиков
plt.style.use('seaborn-darkgrid')

# Реализуем закон Ципфа для параметров C = 1000 и a = 1
def zipf_frequency(rank, C=1000, a=1):
    return C / rank**a

# Генерируем ранги слов
ranks = np.arange(1, 101)  # Ранги от 1 до 100

# Считаем частоты по закону Ципфа для a = 1
frequencies_a1 = zipf_frequency(ranks, C=1000, a=1)

# Считаем частоты по закону Ципфа для a = 1.2
frequencies_a12 = zipf_frequency(ranks, C=1000, a=1.2)

# Визуализация частот в линейном масштабе
plt.figure(figsize=(10, 6))
plt.plot(ranks, frequencies_a1, label='a=1.0', marker='o')
plt.plot(ranks, frequencies_a12, label='a=1.2', marker='o')
plt.title('Закон Ципфа (Линейный масштаб)')
plt.xlabel('Ранг слова')
plt.ylabel('Частота слова')
plt.legend()
plt.grid(True)
plt.show()

# Теперь логарифмическая шкала
plt.figure(figsize=(10, 6))
plt.plot(np.log(ranks), np.log(frequencies_a1), label='a=1.0', marker='o')
plt.plot(np.log(ranks), np.log(frequencies_a12), label='a=1.2', marker='o')
plt.title('Закон Ципфа (Логарифмический масштаб)')
plt.xlabel('log(Ранг слова)')
plt.ylabel('log(Частота слова)')
plt.legend()
plt.grid(True)
plt.show()

# Закон Хипса
def heaps_law(N, K=10, beta=0.5):
    return K * N**beta

# Генерируем массив для размеров текста
text_sizes = np.array([100, 1000, 10000, 100000])

# Количество уникальных слов по закону Хипса
vocab_sizes = heaps_law(text_sizes)

# Визуализация закона Хипса
plt.figure(figsize=(10, 6))
plt.plot(text_sizes, vocab_sizes, marker='o')
plt.title('Закон Хипса')
plt.xlabel('Размер текста (количество слов)')
plt.ylabel('Количество уникальных слов')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.show()
```







На графиках показаны примеры применения закона Ципфа и закона Хипса:

1. **Закон Ципфа (линейный масштаб)** — первый график показывает частоту слов в зависимости от ранга в линейной шкале для двух значений параметра $a$ (1 и 1.2). Видно, что частота слова стремительно убывает по мере увеличения его ранга.

2. **Закон Ципфа (логарифмический масштаб)** — на втором графике тот же закон показан в логарифмическом масштабе. Линейный характер графиков подтверждает, что закон Ципфа можно хорошо описать на логарифмической шкале как прямую линию.

3. **Закон Хипса** — третий график показывает зависимость количества уникальных слов от общего количества слов в тексте. Чем больше текста, тем больше уникальных слов, но их прирост замедляется с увеличением объема текста (что иллюстрирует логарифмический масштаб).

Эти графики наглядно демонстрируют численные примеры действия законов Ципфа и Хипса.
