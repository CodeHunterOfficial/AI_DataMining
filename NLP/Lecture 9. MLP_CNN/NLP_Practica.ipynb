#Многослойный перцептрон (MLP) в задачах обработки естественного языка (NLP)

### Введение

Обработка естественного языка (Natural Language Processing, NLP) — это область искусственного интеллекта, которая фокусируется на взаимодействии между компьютерами и человеческими языками. В последние годы многослойный перцептрон (MLP) стал основой для множества задач NLP благодаря его способности решать сложные задачи классификации, регрессии и генерации данных.

### Основы многослойного перцептрона (MLP)

Многослойный перцептрон — это класс искусственных нейронных сетей, который состоит из следующих компонентов:

1. **Входной слой**: принимает данные (вектор признаков).
2. **Скрытые слои**: нелинейно преобразуют входные данные. Каждый слой состоит из нейронов, которые вычисляют взвешенную сумму входных сигналов, добавляют смещение (bias) и применяют нелинейную активационную функцию.
3. **Выходной слой**: возвращает прогнозы модели (например, вероятности классов для задачи классификации).

#### Математика MLP:
- Входные данные: $\mathbf{x} = [x_1, x_2, ..., x_n]$
- Выход скрытого слоя: $h_i = \sigma(\sum_{j} w_{ij}x_j + b_i)$, где $\sigma$ — активационная функция.
- Итоговый выход модели: $\hat{y} = g(\mathbf{h})$, где $g$ — функция для преобразования выходных сигналов в прогнозы (например, softmax для классификации).


### Применение MLP в NLP

#### 1. **Преобразование текстовых данных**
Чтобы использовать MLP для NLP-задач, необходимо преобразовать текстовые данные в численные представления. Популярные подходы:
- **One-Hot Encoding**: каждое слово представлено вектором, где одна позиция равна 1, остальные 0.
- **Bag-of-Words (BoW)**: векторизация текста на основе частот слов.
- **TF-IDF**: измеряет важность слова в документе относительно корпуса.
- **Word Embeddings**: более сложные методы, такие как Word2Vec, GloVe или FastText, которые создают плотные векторы, отражающие семантику слов.

#### 2. **Типовые задачи NLP**
MLP может быть использован для решения следующих задач:

##### a. **Классификация текста**
- **Пример задачи**: классификация отзывов на положительные/отрицательные.
- **Подход**:
  - Преобразовать текст в численное представление (BoW, TF-IDF или эмбеддинги).
  - Подать вектор на вход MLP.
  - Выходной слой содержит $n$ нейронов, где $n$ — число классов.

##### b. **Распознавание сущностей (NER)**
- Задача: идентификация сущностей в тексте (имена, даты, организации).
- MLP используется для классификации каждого токена текста в соответствующий класс сущности.

##### c. **Анализ тональности текста**
- Цель: определить эмоциональную окраску текста (положительная, отрицательная, нейтральная).
- Часто используется в маркетинге и для анализа пользовательских отзывов.

##### d. **Предсказание пропущенного текста**
- Пример: автозаполнение текста или корректировка ошибок.
- MLP прогнозирует следующее слово или корректный вариант на основе контекста.


### Архитектура MLP в NLP

#### 1. **Входной слой**
Каждый входной текст преобразуется в числовой вектор. Например:
- $\mathbf{x} \in \mathbb{R}^d$, где $d$ — размер эмбеддинга или длина TF-IDF вектора.

#### 2. **Скрытые слои**
- Используются плотные слои (Dense Layers), которые соединяют каждый нейрон с нейронами следующего слоя.
- Популярные активационные функции:
  - ReLU ($\text{max}(0, x)$) — для скрытых слоев.
  - Sigmoid или Softmax — для выходного слоя.

#### 3. **Выходной слой**
Тип выхода зависит от задачи:
- **Классификация**: слой с softmax, возвращающий вероятности классов.
- **Регрессия**: один нейрон с линейной активацией.

### Преимущества MLP для NLP
1. Простота архитектуры и настройки.
2. Хорошо подходит для задач с плотными числовыми векторами (TF-IDF, эмбеддинги).
3. Быстрая обучаемость на малых наборах данных.

### Ограничения MLP в NLP
1. Неэффективность при работе с последовательностями текста (не учитывает порядок слов).
2. Ограниченная способность обрабатывать длинные зависимости (например, смысловая связь между удалёнными словами).
3. Часто уступает архитектурам, специально разработанным для NLP (например, RNN, LSTM, GRU, трансформеры).





#Создание модели многослойного перцептрона (MLP) для классификации текста

В данной лекции мы подробно рассмотрим процесс создания модели многослойного перцептрона (MLP) с использованием библиотеки TensorFlow (Keras) для задачи классификации текста. Мы пошагово разберем каждый элемент, который присутствует в предложенном фрагменте кода.

### 1. Что такое многослойный перцептрон (MLP)?

Многослойный перцептрон (MLP, Multi-Layer Perceptron) — это тип искусственной нейронной сети, состоящий из нескольких слоев нейронов, каждый из которых связан с каждым нейроном предыдущего слоя. MLP состоит из:
- Входного слоя (input layer),
- Один или несколько скрытых слоев (hidden layers),
- Выходного слоя (output layer).

MLP является универсальной архитектурой для решения задач, связанных с классификацией и регрессией.

В контексте обработки текстов MLP часто используется в задачах классификации, где входной текст должен быть преобразован в числовые признаки, которые затем обрабатываются с помощью сети для предсказания класса.

### 2. Создание модели с использованием Keras

Мы будем использовать высокоуровневое API Keras, которое предоставляет удобные методы для создания, обучения и оценки нейронных сетей.

#### Шаг 6: Создание модели MLP

```python
model = tf.keras.Sequential()
```

Здесь мы начинаем создание модели с помощью класса `Sequential` из TensorFlow Keras. Модель `Sequential` подразумевает, что слои будут добавляться последовательно, то есть выход одного слоя будет входом для следующего.

### 3. Добавление слоя эмбеддинга

```python
model.add(tf.keras.layers.Embedding(input_dim=10000, output_dim=embedding_dim,
                                    weights=[embedding_matrix], input_length=maxlen, trainable=False))
```

**Слой эмбеддинга** — это слой, который преобразует текстовые данные (обычно индексы слов) в плотные векторные представления (эмбеддинги). Слой `Embedding` принимает:
- `input_dim`: размер словаря, то есть максимальное количество уникальных слов, которое модель может учитывать. В данном примере это 10000 слов.
- `output_dim`: размерность эмбеддинга — количество признаков, которое будет иметь каждый вектор, соответствующий слову. Обычно используется значение от 50 до 300.
- `weights`: матрица предварительно обученных эмбеддингов. Это может быть матрица, содержащая предобученные эмбеддинги слов, которые будут использованы вместо случайных инициализаций.
- `input_length`: длина последовательности входных данных (в данном примере — `maxlen`), то есть длина каждого входного текста.
- `trainable=False`: означает, что веса эмбеддинга не будут обновляться в процессе обучения (например, если используются предобученные эмбеддинги).

Этот слой преобразует последовательности индексов слов в плотные векторы фиксированной длины, что позволяет нейронной сети работать с текстом.

### 4. Добавление скрытых слоев

```python
model.add(tf.keras.layers.Flatten())
```

Слой `Flatten` используется для преобразования выходных данных (обычно это многомерный тензор) в одномерный вектор. Это необходимо перед подачей данных на полносвязные слои, такие как `Dense`.

```python
model.add(tf.keras.layers.Dense(128, activation='relu'))
```

Слой `Dense` представляет собой полностью связанный слой нейронов. В данном случае:
- 128 — это количество нейронов в слое.
- `activation='relu'` — функция активации ReLU (Rectified Linear Unit). ReLU является одной из самых популярных функций активации, которая помогает избежать проблемы исчезающего градиента и ускоряет обучение.

Этот слой обрабатывает выходные данные из предыдущего слоя и применяет к ним нелинейную функцию активации.

```python
model.add(tf.keras.layers.Dropout(0.5))
```

Слой `Dropout` применяется для регуляризации и предотвращения переобучения. В данном случае вероятность "выключения" нейронов составляет 50% (параметр 0.5). Это помогает модели не слишком сильно полагаться на отдельные нейроны, повышая ее обобщающую способность.

### 5. Добавление выходного слоя

```python
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

Этот слой представляет собой последний слой сети, который генерирует итоговый результат. Он:
- Содержит один нейрон, так как задача классификации является бинарной (например, положительный/отрицательный класс).
- Использует функцию активации `sigmoid`, которая ограничивает выходное значение в диапазоне от 0 до 1. Это делает ее подходящей для бинарной классификации, где значение около 0.5 может означать переход между классами.

### 6. Обзор параметров

#### Размер входа

Размер входа модели зависит от длины текстовых данных (`maxlen`), то есть, если каждое слово представлено индексом, то на вход подается последовательность индексов длиной `maxlen`.

#### Размерность эмбеддинга

Размерность эмбеддинга (`embedding_dim`) определяет, насколько "богато" будет представление каждого слова. Обычно выбираются такие значения, как 100, 200 или 300.

#### Скрытые слои

В предложенной модели добавлен только один скрытый слой с 128 нейронами, но можно добавить и другие слои для улучшения производительности. Количество нейронов и функция активации могут варьироваться в зависимости от задачи.

#### Выходной слой

Выходной слой с функцией активации `sigmoid` подходит для бинарной классификации, например, для задачи определения, является ли текст положительным или отрицательным.


### Отличия модели для классификации от модели для регрессии

Модели для **классификации** и **регрессии** решают разные задачи, и это отражается на структуре выходного слоя и функции активации. Давайте рассмотрим основные различия:

#### 1. **Задача и тип выходного слоя**
- **Классификация**: В задаче классификации модель должна предсказать вероятность принадлежности объекта к одному из классов. В случае бинарной классификации выходной слой часто содержит один нейрон с функцией активации `sigmoid`. В многоклассовой классификации используется выходной слой с количеством нейронов, равным числу классов, и функция активации `softmax`.
- **Регрессия**: Задача регрессии заключается в предсказании непрерывного значения, например, числовой оценки или прогноза. Для регрессии выходной слой обычно содержит один нейрон без функции активации (или с линейной функцией активации), так как результат должен быть непрерывным числом.

**Пример для классификации**:
```python
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))  # Бинарная классификация
```

**Пример для регрессии**:
```python
model.add(tf.keras.layers.Dense(1))  # Регрессия (без активации)
```

#### 2. **Функция потерь**
- **Классификация**: Для задачи бинарной классификации часто используется функция потерь `binary_crossentropy`, а для многоклассовой классификации — `categorical_crossentropy`.
- **Регрессия**: Для задачи регрессии используется функция потерь `mean_squared_error` (среднеквадратичная ошибка), так как цель заключается в минимизации разницы между предсказанными и истинными значениями.

**Пример для классификации**:
```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

**Пример для регрессии**:
```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

#### 3. **Оценка производительности**
- **Классификация**: В задачах классификации оценивают модель с помощью таких метрик, как **точность (accuracy)**, **precision**, **recall**, **F1-score** и другие метрики, которые оценивают качество классификации.
- **Регрессия**: В задачах регрессии применяют такие метрики, как **средняя абсолютная ошибка (MAE)**, **среднеквадратичная ошибка (MSE)** и **коэффициент детерминации R²**, чтобы оценить точность предсказания непрерывных значений.

### Подбор лучших параметров и количества слоев

При создании нейронной сети важно правильно настроить параметры модели, чтобы добиться хороших результатов. Существует несколько факторов, которые влияют на выбор оптимальных параметров.

#### 1. **Количество слоев и нейронов**
- **Количество скрытых слоев**: Обычно для начала используют один или два скрытых слоя. Однако при усложнении задачи (например, при увеличении размерности данных) можно добавить больше слоев. Стоит отметить, что слишком большое количество слоев может привести к переобучению, особенно при небольшом объеме данных.
- **Количество нейронов в слое**: Общее правило — больше нейронов в скрытом слое дает модели больше "способности" для извлечения признаков. Однако это также увеличивает количество параметров и может привести к переобучению. Часто количество нейронов в скрытых слоях выбирается экспериментально. Например, можно попробовать значения от 32 до 512 нейронов.

#### 2. **Функции активации**
- **ReLU** (Rectified Linear Unit) — одна из самых популярных функций активации для скрытых слоев, поскольку она помогает ускорить обучение и предотвращает проблему исчезающего градиента. Однако для выходного слоя для классификации используется `sigmoid` или `softmax`, а для регрессии — линейная активация.
- Другие функции активации, такие как **tanh** или **LeakyReLU**, также могут быть использованы, но они применяются реже.

#### 3. **Регуляризация**
- **Dropout** — помогает предотвратить переобучение, случайным образом "выключая" нейроны во время обучения. Значение параметра dropout (например, 0.2-0.5) выбирается экспериментально.
- **L2-регуляризация** (или weight decay) может быть полезна для ограничения величины весов модели, особенно при использовании больших сетей.

#### 4. **Выбор гиперпараметров**
- **Размер батча (batch size)**: Размер батча влияет на скорость обучения и стабильность градиентов. Типичные значения — 32, 64, 128. Чем больше размер батча, тем быстрее обучение, но это может привести к меньшей стабильности.
- **Скорость обучения (learning rate)**: Важно настроить скорость обучения для оптимизатора. Слишком высокая скорость может привести к расходимости, а слишком низкая — к слишком медленному обучению. Можно использовать методы, такие как **adaptive learning rate**, чтобы автоматизировать этот процесс.

#### 5. **Кросс-валидация и перебор гиперпараметров**
Один из лучших способов подбора параметров — это использование **кросс-валидации** и **поиска по сетке** (Grid Search) или **случайного поиска** (Random Search). Для каждого гиперпараметра можно пробовать разные значения и оценивать производительность модели.

### Пример настройки модели с несколькими слоями для задачи классификации

```python
model = tf.keras.Sequential()

# Слой эмбеддинга
model.add(tf.keras.layers.Embedding(input_dim=10000, output_dim=128, input_length=maxlen))

# Скрытые слои
model.add(tf.keras.layers.Flatten())  # Преобразуем выход из эмбеддинга в одномерный вектор
model.add(tf.keras.layers.Dense(256, activation='relu'))  # Первый скрытый слой с 256 нейронами
model.add(tf.keras.layers.Dropout(0.5))  # Регуляризация

model.add(tf.keras.layers.Dense(128, activation='relu'))  # Второй скрытый слой с 128 нейронами
model.add(tf.keras.layers.Dropout(0.5))  # Регуляризация

# Выходной слой
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))  # Для бинарной классификации

# Компиляция модели
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

### Заключение

- В модели для **классификации** выходной слой использует активацию `sigmoid` или `softmax`, а функция потерь зависит от типа задачи.
- В модели для **регрессии** выходной слой не использует активацию, и используется функция потерь `mean_squared_error` или `mae`.
- Для подбора лучших параметров и количества слоев рекомендуется использовать подходы, такие как кросс-валидация, Grid Search и Random Search.
- Количество слоев и нейронов зависит от сложности задачи, и часто лучше начинать с простой модели и постепенно усложнять ее, добавляя дополнительные слои и регулируя параметры.

Подбор параметров требует экспериментов и анализа, так как оптимальная конфигурация зависит от данных и специфики задачи.


## b. **Распознавание сущностей (NER)**





#### 1. **Введение в NER**
Распознавание именованных сущностей (NER, Named Entity Recognition) — это задача обработки естественного языка (NLP), направленная на выявление и классификацию сущностей в тексте. Например, "Barack Obama" может быть распознан как персона (PER), "Microsoft" — как организация (ORG), а "Paris" — как местоположение (LOC).  
NER используется в таких приложениях, как анализ текстов, информационный поиск и построение чат-ботов.



#### 2. **Основные этапы обработки данных**

**Шаг 1: Создание датасета**  
Данные представлены в виде текстовых предложений и меток сущностей. Например, для предложения `Barack Obama was born in Hawaii` указано, что "Barack Obama" — это персона, а "Hawaii" — местоположение.

```python
data = {
    "sentence": ["Barack Obama was born in Hawaii.", ...],
    "entities": [[('Barack Obama', 'PER'), ('Hawaii', 'LOC')], ...]
}
```



**Шаг 2: Разметка текста**  
Предложения разбиваются на слова, а каждое слово получает метку:  
- `PER`, `ORG`, `LOC` для сущностей.
- `O` для слов, не относящихся к сущностям.

```python
X.append(words)  # Список слов предложения
tags = ['O'] * len(words)  # Инициализация меток "O"

# Присваиваем метки сущностям
for entity, tag in entity_list:
    entity_words = entity.split()
    ...
    tags[i:i+len(entity_words)] = [tag] * len(entity_words)
```



**Шаг 3: Индексация меток**  
Метки преобразуются в индексы для работы с числовыми данными:
```python
tag2idx = {tag: idx for idx, tag in enumerate(all_tags)}
y = [[tag2idx[tag] for tag in tags] for tags in y]
```



**Шаг 4: Паддинг последовательностей**  
Так как предложения имеют разную длину, их выравнивают до максимального размера (`max_len`), добавляя нули (`padding='post'`).

```python
X_padded = pad_sequences([[hash(word) % 10000 for word in sentence] for sentence in X], maxlen=max_len, padding='post')
y_padded = pad_sequences(y, maxlen=max_len, padding='post')
```



#### 3. **Создание модели NER**

Модель для NER строится на основе TensorFlow/Keras.

**Архитектура модели:**
1. **Входной слой** принимает данные с фиксированной длиной последовательности (`max_len`).
2. **Embeddings**: отображают слова в векторное пространство.
3. **Плотные слои**: два слоя `Dense` с активацией ReLU для обучения глубоких представлений.
4. **Dropout**: предотвращает переобучение, случайно исключая нейроны.
5. **Выходной слой**: `softmax` с количеством выходов, равным числу меток (`len(tag2idx)`).

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(max_len,)),
    tf.keras.layers.Embedding(input_dim=10000, output_dim=64, input_length=max_len),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(len(tag2idx), activation='softmax')
])
```


#### 4. **Обучение модели**
Модель компилируется с оптимизатором Adam и функцией потерь `sparse_categorical_crossentropy`, подходящей для задач классификации.

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=20, batch_size=2)
```

Параметры обучения:  
- `epochs=20`: количество эпох (обходов всего набора данных).  
- `batch_size=2`: число примеров в одном шаге обучения.


#### 5. **Оценка модели**
После обучения проводится тестирование на отложенной выборке:
```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Потери: {loss}, Точность: {accuracy}')
```
Модель оценивается по потере (`loss`) и точности (`accuracy`).



#### 6. **Тестирование на новых данных**
Для проверки модели подготавливаются новые предложения:
```python
new_sentences = ["Steve Jobs was the CEO of Apple.", ...]

```

Модель возвращает вероятности для каждой метки. Индекс с наибольшей вероятностью преобразуется обратно в метку:

#### Заключение
Код демонстрирует пошаговое создание модели NER: от обработки данных до обучения и тестирования. Модель показывает, как с помощью глубокого обучения можно эффективно решать задачи обработки текста, используя Keras и TensorFlow.

````python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Создание простого датасета для NER
data = {
    "sentence": [
        "Barack Obama was born in Hawaii.",
        "Microsoft is based in Redmond.",
        "Elon Musk founded Tesla.",
        "New York is a large city.",
        "Amazon is an online retail company.",
        "California is known for its beaches.",
        "Bill Gates is the co-founder of Microsoft.",
        "Facebook is a social media platform.",
        "Tesla is an electric vehicle manufacturer.",
        "The Eiffel Tower is in Paris."
    ],
    "entities": [
        [('Barack Obama', 'PER'), ('Hawaii', 'LOC')],
        [('Microsoft', 'ORG'), ('Redmond', 'LOC')],
        [('Elon Musk', 'PER'), ('Tesla', 'ORG')],
        [('New York', 'LOC')],
        [('Amazon', 'ORG')],
        [('California', 'LOC')],
        [('Bill Gates', 'PER'), ('Microsoft', 'ORG')],
        [('Facebook', 'ORG')],
        [('Tesla', 'ORG')],
        [('Eiffel Tower', 'LOC'), ('Paris', 'LOC')]
    ]
}

# Предварительная обработка текста
sentences = data['sentence']
entities = data['entities']

# Создание меток для сущностей
all_tags = set(tag for entity in entities for _, tag in entity)
all_tags.add('O')  # Добавляем 'O' для не сущностей
tag2idx = {tag: idx for idx, tag in enumerate(all_tags)}
idx2tag = {idx: tag for tag, idx in tag2idx.items()}

# Подготовка данных для обучения
X = []
y = []

for sentence, entity_list in zip(sentences, entities):
    words = sentence.split()
    X.append(words)
    tags = ['O'] * len(words)  # Инициализация меток как 'O'

    for entity, tag in entity_list:
        entity_words = entity.split()
        for i in range(len(words)):
            if words[i:i+len(entity_words)] == entity_words:
                tags[i:i+len(entity_words)] = [tag] * len(entity_words)

    y.append(tags)

# Преобразование меток в индексы
y = [[tag2idx[tag] for tag in tags] for tags in y]

# Паддинг последовательностей
max_len = max(len(sentence) for sentence in X)
X_padded = pad_sequences([[hash(word) % 10000 for word in sentence] for sentence in X], maxlen=max_len, padding='post')
y_padded = pad_sequences(y, maxlen=max_len, padding='post')

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X_padded, y_padded, test_size=0.2, random_state=42)

# Создание улучшенной модели
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(max_len,)),
    tf.keras.layers.Embedding(input_dim=10000, output_dim=64, input_length=max_len),
    tf.keras.layers.Dense(128, activation='relu'),  # Плотный слой с ReLU
    tf.keras.layers.Dense(128, activation='relu'),  # Дополнительный слой для более глубокого обучения
    tf.keras.layers.Dropout(0.5),  # Dropout для регуляризации
    tf.keras.layers.Dense(len(tag2idx), activation='softmax')  # Выходной слой с softmax
])

# Компиляция модели
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Обучение модели
model.fit(X_train, y_train, epochs=20, batch_size=2)  # Увеличьте количество эпох

# Тестирование модели
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Потери: {loss}, Точность: {accuracy}')

# Новый текст для тестирования
new_sentences = [
    "Steve Jobs was the CEO of Apple.",
    "The Great Wall is in China.",
    "Google is a technology company."
]

# Предварительная обработка новых текстов
new_X = []
for sentence in new_sentences:
    words = sentence.split()
    new_X.append(words)

# Преобразование слов в индексы
new_X_padded = pad_sequences([[hash(word) % 10000 for word in sentence] for sentence in new_X], maxlen=max_len, padding='post')

# Получение предсказаний от модели
predictions = model.predict(new_X_padded)

# Преобразование предсказанных индексов обратно в метки
predicted_tags = np.argmax(predictions, axis=-1)

# Вывод результатов
for i, sentence in enumerate(new_sentences):
    print(f"Sentence: {sentence}")
    tags = [idx2tag[tag] for tag in predicted_tags[i]]
    print(f"Predicted Tags: {tags}")
````

## d. **Предсказание пропущенного текста**


Предсказание пропущенного текста (text inpainting) — задача заполнения пропусков в тексте. Мы будем использовать многослойный персептрон (MLP) для этого.



#### 1. **Общая идея**
- **Цель**: Заполнить пропущенное слово в предложении. Например:
  - Вход: "The ___ is blue."
  - Выход: "sky."

- **Процесс**:
  1. Преобразуем текст в числовое представление (векторизация).
  2. Создаем обучающую выборку: вводим пробелы в предложениях и задаем цель — предсказать пропущенные слова.
  3. Строим многослойный персептрон, который принимает вектор входного текста и возвращает индекс пропущенного слова.
  4. Тренируем модель на этих данных.



#### 2. **Подготовка данных**

### Заключение

Многослойный персептрон успешно решает задачу заполнения пропущенного текста, если ему предоставлены качественные данные и достаточное количество эпох обучения. В дальнейшем модель можно улучшить, добавив более сложные архитектуры или обучив её на большем корпусе текстов.

````python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
import numpy as np

# Данные
sentences = [
    "The sky is blue",
    "I love machine learning",
    "Artificial intelligence is the future",
    "The sun is bright",
    "Deep learning is fascinating"
]

# Преобразование текста в формат с пробелами
def create_dataset(sentences):
    input_texts = []
    target_words = []
    vocab = set()

    for sentence in sentences:
        words = sentence.split()
        vocab.update(words)
        for i, word in enumerate(words):
            # Убираем одно слово и оставляем место для пробела
            input_text = words[:i] + ["___"] + words[i+1:]
            input_texts.append(" ".join(input_text))
            target_words.append(word)

    return input_texts, target_words, sorted(vocab)

input_texts, target_words, vocab = create_dataset(sentences)

# Создание словарей для преобразования текста
word2idx = {word: idx for idx, word in enumerate(vocab)}
idx2word = {idx: word for word, idx in word2idx.items()}

# Преобразование входов и целей в числовой формат
X = []
y = []

for text, target in zip(input_texts, target_words):
    words = text.split()
    X.append([word2idx[word] if word in word2idx else 0 for word in words])
    y.append(word2idx[target])

# Паддинг последовательностей
max_len = max(len(seq) for seq in X)
X_padded = pad_sequences(X, maxlen=max_len, padding='post')
y_categorical = to_categorical(y, num_classes=len(vocab))

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X_padded, y_categorical, test_size=0.2, random_state=42)

# Гиперпараметры
embedding_dim = 64
hidden_units = 128

# Модель
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=len(vocab), output_dim=embedding_dim, input_length=max_len),
    tf.keras.layers.Flatten(),  # Переход к плотным слоям
    tf.keras.layers.Dense(hidden_units, activation='relu'),  # Скрытый слой
    tf.keras.layers.Dropout(0.5),  # Регуляризация
    tf.keras.layers.Dense(len(vocab), activation='softmax')  # Выходной слой
])

# Компиляция
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Обучение
model.fit(X_train, y_train, epochs=20, batch_size=4, validation_data=(X_test, y_test))

# Функция для предсказания
def predict_missing_word(model, sentence, word2idx, idx2word, max_len):
    words = sentence.split()
    padded_sentence = pad_sequences([[word2idx[word] if word in word2idx else 0 for word in words]], maxlen=max_len, padding='post')
    prediction = model.predict(padded_sentence)
    predicted_idx = np.argmax(prediction, axis=-1)[0]
    return idx2word[predicted_idx]

# Новое предложение для теста
test_sentence = "The ___ is bright"
predicted_word = predict_missing_word(model, test_sentence, word2idx, idx2word, max_len)

print(f"Input sentence: {test_sentence}")
print(f"Predicted missing word: {predicted_word}")
````
