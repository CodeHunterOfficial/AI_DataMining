#Метрики классификации в задачах машинного обучения

### Введение

Одной из ключевых задач машинного обучения является классификация, целью которой является отнесение объекта к одному из заранее определенных классов. Для оценки качества работы моделей классификации применяются специальные метрики, основанные на сравнении предсказаний модели с истинными значениями классов. Выбор и интерпретация этих метрик имеют решающее значение при анализе производительности алгоритма, особенно в условиях несбалансированных данных.

Настоящая лекция подробно рассматривает наиболее распространенные метрики классификации, включая их математическое описание, интерпретацию, ограничения и области применения. Особое внимание уделяется тому, как выбор метрики влияет на построение и оценку моделей классификации.



## 1. Матрица ошибок как основа метрик классификации

Матрица ошибок (confusion matrix) представляет собой таблицу, в которой отображаются результаты классификации, разделённые по типам:

|                   | **Предсказано: Положительное** | **Предсказано: Отрицательное** |
|-|||
| **Истинное: Положительное (TP)** | True Positive (TP)                 | False Negative (FN)                |
| **Истинное: Отрицательное (TN)** | False Positive (FP)                | True Negative (TN)                 |

### Определения основных категорий:

- **True Positive (TP):** Количество объектов положительного класса, которые модель правильно предсказала как положительные.
- **True Negative (TN):** Количество объектов отрицательного класса, правильно классифицированных моделью как отрицательные.
- **False Positive (FP):** Количество объектов отрицательного класса, ошибочно классифицированных как положительные.
- **False Negative (FN):** Количество объектов положительного класса, ошибочно классифицированных как отрицательные.

Матрица ошибок является основой для большинства метрик классификации, которые определяются на основе сочетаний TP, TN, FP и FN.

### Особенности использования:
Матрица ошибок удобна для визуализации ошибок и анализа точности классификации для каждой категории. Однако, для количественной оценки качества модели применяются производные метрики.



## 2. Метрики для бинарной классификации

### 2.1 Точность (Accuracy)

**Формула:**
$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

**Интерпретация:**
Accuracy измеряет долю правильно классифицированных объектов среди общего числа. Эта метрика удобна для задач, где классы примерно сбалансированы, то есть имеют схожие пропорции.

**Недостатки:**
Точность может вводить в заблуждение в случае несбалансированных классов. Например, если 95% объектов принадлежат одному классу, модель может достичь высокой точности, просто всегда предсказывая этот класс, игнорируя меньшинство.



### 2.2 Полнота (Recall) или Чувствительность (Sensitivity)

**Формула:**
$$
Recall = \frac{TP}{TP + FN}
$$

**Интерпретация:**
Полнота оценивает долю объектов положительного класса, которые модель правильно классифицировала как положительные. Эта метрика критична в задачах, где важно минимизировать пропущенные случаи, например, в медицинской диагностике.



### 2.3 Точность (Precision)

**Формула:**
$$
Precision = \frac{TP}{TP + FP}
$$

**Интерпретация:**
Точность измеряет, какая доля объектов, предсказанных как положительные, действительно принадлежит положительному классу. Это ключевая метрика в задачах, где важно избегать ложных срабатываний (FP). Например, в системах фильтрации спама высокая точность снижает вероятность блокировки обычных писем.



### 2.4 F-мера (F1-score)

**Формула:**
$$
F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}
$$

**Интерпретация:**
F-мера представляет собой гармоническое среднее точности и полноты. Она используется в ситуациях, где необходимо найти баланс между двумя этими метриками.

**Особенности:**
- Значение $F1 = 1$ соответствует идеальной модели.
- Хорошо подходит для задач с несбалансированными классами.






### 2.5 Площадь под кривой ROC (ROC-AUC)

#### ROC-кривая

**ROC-кривая (Receiver Operating Characteristic)** — это график, который показывает производительность бинарной классификационной модели при различных порогах. Она иллюстрирует зависимость между двумя важными метриками:

- **True Positive Rate (TPR)** (доля истинно положительных):
$$
  TPR = \frac{TP}{TP + FN}
$$
  - **TP (True Positives)**: количество правильно предсказанных положительных случаев.
  - **FN (False Negatives)**: количество неправильно предсказанных отрицательных случаев.

- **False Positive Rate (FPR)** (доля ложно положительных):
$$
  FPR = \frac{FP}{FP + TN}
$$
  - **FP (False Positives)**: количество неправильно предсказанных положительных случаев.
  - **TN (True Negatives)**: количество правильно предсказанных отрицательных случаев.

На графике ROC:
- **Ось X**: FPR (ложно положительные).
- **Ось Y**: TPR (истинно положительные).

ROC-кривая строится путем изменения порога классификации и вычисления TPR и FPR для каждого порога. Это позволяет увидеть, как меняется производительность модели в зависимости от выбранного порога.

#### AUC (Area Under Curve)

**AUC (Area Under Curve)** — это площадь под ROC-кривой и важная метрика для оценки качества модели. Значение AUC варьируется от 0 до 1:

- **$AUC = 0.5$**: Модель эквивалентна случайному угадыванию. Это означает, что модель не может различать положительные и отрицательные классы лучше, чем случайный выбор.
  
- **$AUC = 1.0$**: Идеальная классификация. Модель безошибочно классифицирует все случаи, что означает, что TPR равен 1 при FPR равном 0.

- **$0.5 < AUC < 1.0$**: Чем ближе AUC к 1.0, тем лучше модель. Значения AUC, близкие к 1, указывают на высокую способность модели различать классы.

#### Преимущества ROC-AUC

1. **Устойчивость к несбалансированным данным**: ROC-AUC является особенно полезным в задачах с несбалансированными классами, где количество примеров одного класса значительно больше, чем другого. В таких случаях точность может быть вводящей в заблуждение, а AUC предоставляет более надежную метрику.

2. **Независимость от порога**: AUC учитывает все возможные пороги, что делает ее более универсальной по сравнению с другими метриками, которые зависят от конкретного порога.

3. **Сравнение моделей**: AUC позволяет легко сравнивать производительность различных моделей. Модель с более высоким значением AUC будет предпочтительнее.

#### Визуализация AUC

На графике ROC:
- Площадь под ROC-кривой (AUC) визуально представляется как область между ROC-кривой и линией случайного угадывания.
- Чем больше эта площадь, тем лучше модель, так как она указывает на высокую TPR при низком FPR.

Таким образом, ROC-AUC является одной из наиболее универсальных метрик для оценки производительности бинарных классификационных моделей. Она позволяет учитывать различные пороги и предоставляет ясное представление о способности модели различать между классами, особенно в условиях несбалансированных данных.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Генерация случайных данных
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Обучение модели
model = LogisticRegression()
model.fit(X_train, y_train)

# Получение вероятностей предсказания
y_scores = model.predict_proba(X_test)[:, 1]

# Вычисление ROC-кривой и AUC
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

# Визуализация ROC-кривой
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Линия случайного угадывания
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

### 2.6 PR-AUC (Precision-Recall Area Under Curve)

**Описание:**

PR-AUC (площадь под кривой Precision-Recall) — это метрика, используемая для оценки качества бинарных классификаторов, особенно в условиях несбалансированных данных. Она основана на зависимости между точностью (Precision) и полнотой (Recall).

#### Основные понятия:

- **Precision (Точность)**: Доля истинно положительных результатов среди всех положительных предсказаний. Выражается формулой:
  
$$
  \text{Precision} = \frac{TP}{TP + FP}
$$

  где:
  - $TP$ — истинно положительные,
  - $FP$ — ложно положительные.

- **Recall (Полнота)**: Доля истинно положительных результатов среди всех реальных положительных примеров. Выражается формулой:
  
$$
  \text{Recall} = \frac{TP}{TP + FN}
$$

  где:
  - $FN$ — ложно отрицательные.

#### Построение PR-кривой:

1. **Сбор данных**: Для построения PR-кривой необходимо получить значения Precision и Recall на различных порогах классификации.
2. **Построение графика**:
   - По оси X откладывается Recall.
   - По оси Y откладывается Precision.
   - Для каждого порога классификации вычисляются соответствующие значения Precision и Recall, которые затем соединяются для формирования кривой.

#### Преимущества PR-AUC:

- **Чувствительность к классу**: PR-AUC более информативен в ситуациях с несбалансированными данными, так как фокусируется на положительном классе. Это особенно важно в задачах, где важно минимизировать количество ложноположительных или ложноотрицательных результатов.
  
- **Интерпретируемость**: Значение PR-AUC варьируется от 0 до 1, где 1 указывает на идеальный классификатор, а 0.5 — на случайный.

#### Применение:

PR-AUC часто используется в таких областях, как:

- Медицинская диагностика (например, обнаружение заболеваний на ранних стадиях),
- Обнаружение мошенничества (например, в финансовых транзакциях),
- Обработка естественного языка (например, в задачах классификации текстов).

Таким образом, PR-AUC является важным инструментом для оценки эффективности моделей в условиях несбалансированных данных. Понимание и правильное применение этой метрики позволяет разработчикам и исследователям более точно оценивать качество своих моделей и принимать обоснованные решения на основе полученных результатов.




```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, auc
from sklearn.datasets import make_classification

# Генерация искусственного набора данных
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2,
                           n_informative=2, n_redundant=10, random_state=42)

# Предположим, что мы обучили модель и получили вероятности для положительного класса
# Здесь мы просто используем случайные вероятности для примера
y_scores = np.random.rand(len(y))

# Вычисление Precision, Recall и порогов
precision, recall, thresholds = precision_recall_curve(y, y_scores)

# Вычисление PR-AUC
pr_auc = auc(recall, precision)

# Визуализация PR-кривой
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label='PR-кривая (AUC = {:.2f})'.format(pr_auc))
plt.title('Precision-Recall Curve')
plt.xlabel('Полнота (Recall)')
plt.ylabel('Точность (Precision)')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.grid()
plt.legend()
plt.show()
```

### 2.7 Log Loss (Логарифмическая функция потерь)

**Формула:**
$$
LogLoss = -\frac{1}{N} \sum_{i=1}^N \left[ y_i \log(p_i) + (1 - y_i) \log(1 - p_i) \right]
$$

**Где:**
- **$y_i$** — истинное значение метки (0 или 1), которое указывает на принадлежность к классу.
- **$p_i$** — предсказанная вероятность положительного класса (значение от 0 до 1).

#### Пояснение формулы
- **$N$** — общее количество наблюдений в наборе данных.
- Сумма берется по всем наблюдениям, и для каждого наблюдения учитываются два случая:
  - Если истинное значение **$y_i = 1$** (положительный класс), то вклад в Log Loss будет **$-\log(p_i)$**. Это означает, что чем выше предсказанная вероятность **$p_i$**, тем меньший вклад в Log Loss.
  - Если истинное значение **$y_i = 0$** (отрицательный класс), то вклад будет **$-\log(1 - p_i)$**. Здесь также, чем выше вероятность, предсказанная для отрицательного класса, тем меньший вклад в Log Loss.

#### Интерпретация
- **Качество предсказаний:** Log Loss измеряет, насколько хорошо модель предсказывает вероятности классов. Чем ближе предсказанные вероятности к истинным значениям, тем меньше значение Log Loss.
- **Наказание за ошибки:** Если модель предсказывает высокую вероятность для класса, который на самом деле не является истинным (например, предсказывает **$p_i = 0.9$** при **$y_i = 0$**), то Log Loss значительно увеличивается, что указывает на плохое качество предсказания.
- **Нормализация:** Значение Log Loss нормализуется на количество наблюдений **$N$**, что позволяет сравнивать модели на разных наборах данных.

#### Применение
Log Loss часто используется в задачах бинарной классификации, особенно в контексте машинного обучения и статистики. Он позволяет не только оценивать качество предсказаний, но и оптимизировать модели, минимизируя значение функции потерь во время обучения.




```python
import numpy as np
import matplotlib.pyplot as plt

# Функция для расчета Log Loss
def log_loss(y_true, y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# Значения предсказанных вероятностей
p = np.linspace(0.01, 0.99, 100)

# Истинные значения меток
y_true_0 = np.zeros_like(p)  # y = 0
y_true_1 = np.ones_like(p)   # y = 1

# Расчет Log Loss для каждого случая
log_loss_0 = log_loss(y_true_0, p)
log_loss_1 = log_loss(y_true_1, p)

# Визуализация
plt.figure(figsize=(10, 6))

# Логарифмическая функция потерь для y = 0
plt.plot(p, -np.log(1 - p), label='Log Loss (y = 0)', color='blue')

# Логарифмическая функция потерь для y = 1
plt.plot(p, -np.log(p), label='Log Loss (y = 1)', color='red')

# Настройки графика
plt.title('Log Loss в зависимости от предсказанных вероятностей')
plt.xlabel('Предсказанная вероятность (p)')
plt.ylabel('Log Loss')
plt.ylim(0, 5)  # Ограничиваем ось y для лучшей визуализации
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0.5, color='gray', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.show()
```

### 2.8 Коэффициент Каппы Коэна

Коэффициент Каппы Коэна — это статистический метод, используемый для оценки согласованности между двумя классификаторами или наблюдателями. Он позволяет учесть вероятность случайного совпадения, что делает его более надежным инструментом для оценки качества классификации по сравнению с простыми метриками точности.

#### Формула:
$$
\kappa = \frac{P_o - P_e}{1 - P_e}
$$

Где:
- **$P_o$** — наблюдаемая точность (Accuracy), то есть доля верных предсказаний среди всех предсказаний.
- **$P_e$** — ожидаемая точность при случайных предсказаниях, которая рассчитывается на основе вероятностей классов в данных.

#### Расчет $P_o$ и $P_e$:

1. **Наблюдаемая точность ($P_o$)**:
   $$
   P_o = \frac{N_{correct}}{N_{total}}
   $$
   где $N_{correct}$ — количество правильных предсказаний, а $N_{total}$ — общее количество предсказаний.

2. **Ожидаемая точность ($P_e$)**:
   Для вычисления $P_e$ необходимо определить доли каждого класса в данных:
   - $p_A$ — доля класса A.
   - $p_B$ — доля класса B.

   Тогда:
   $$
   P_e = p_A^2 + p_B^2
   $$

#### Интерпретация:

- **Значение Каппы**:
  - $\kappa < 0$ — отсутствие согласия, лучше, чем случайное совпадение.
  - $0 \leq \kappa < 0.20$ — слабое согласие.
  - $0.21 \leq \kappa < 0.40$ — умеренное согласие.
  - $0.41 \leq \kappa < 0.60$ — значительное согласие.
  - $0.61 \leq \kappa < 0.80$ — хорошее согласие.
  - $0.81 \leq \kappa < 1.00$ — практически полное согласие.
  - $\kappa = 1$ — полное согласие.

Коэффициент Каппы Коэна является полезным инструментом в задачах классификации, особенно в медицинских исследованиях, оценке качества аннотаций и других областях, где важно учитывать случайные совпадения.

```python
import numpy as np
import matplotlib.pyplot as plt

# Функция для расчета Каппы Коэна
def cohen_kappa(P_o, P_e):
    return (P_o - P_e) / (1 - P_e)

# Значения наблюдаемой точности (P_o)
P_o_values = np.linspace(0, 1, 100)

# Предположим, что P_e фиксировано на уровне 0.2
P_e = 0.2

# Рассчитываем значения Каппы
kappa_values = cohen_kappa(P_o_values, P_e)

# Создаем график
plt.figure(figsize=(10, 6))
plt.plot(P_o_values, kappa_values, label=f'P_e = {P_e}', color='blue')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.7)
plt.axvline(0.2, color='red', linestyle='--', label='P_o = 0.2')
plt.axvline(0.8, color='green', linestyle='--', label='P_o = 0.8')

# Вычисляем и выводим значения Каппы для P_o = 0.2 и P_o = 0.8
kappa_at_02 = cohen_kappa(0.2, P_e)
kappa_at_08 = cohen_kappa(0.8, P_e)

# Добавляем аннотации на график
plt.annotate(f'κ = {kappa_at_02:.2f}', xy=(0.2, kappa_at_02),
             xytext=(0.25, kappa_at_02 + 0.1),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.annotate(f'κ = {kappa_at_08:.2f}', xy=(0.8, kappa_at_08),
             xytext=(0.75, kappa_at_08 - 0.2),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Настройка графика
plt.title("Коэффициент Каппы Коэна")
plt.xlabel("Наблюдаемая точность (P_o)")
plt.ylabel("Каппа (κ)")
plt.ylim(-1, 1)
plt.xlim(0, 1)
plt.grid()
plt.legend()
plt.show()
```




## 3. Метрики для мультиклассовой классификации

В задачах, где классов больше двух, используют модифицированные метрики:

### 3.1 Micro-averaging и Macro-averaging

- **Micro-averaging:** Среднее значение метрик по всем примерам, независимо от классов. Учитывает дисбаланс.
- **Macro-averaging:** Среднее значение метрик по всем классам, с равным весом для каждого класса.

### 3.2 Matthews Correlation Coefficient (MCC)

**Формула:**
$$
MCC = \frac{TP \cdot TN - FP \cdot FN}{\sqrt{(TP + FP)(TP + FN)(TN + FP)(TN + FN)}}
$$

**Интерпретация:**
MCC измеряет степень корреляции между предсказаниями и истинными метками. Значения варьируются от -1 (анти-корреляция) до 1 (идеальная корреляция).



## 4. Выбор метрик

### 4.1 Учет специфики задачи:
- **Классы сбалансированы:** Accuracy, F1-score.
- **Классы несбалансированы:** ROC-AUC, PR-AUC, F1-score.
- **Критически важна полнота:** Recall.
- **Критически важна точность:** Precision.

### 4.2 Рекомендации:
1. **Используйте несколько метрик:** Например, сочетание ROC-AUC и F1-score.
2. **Анализируйте метрики на тестовых данных:** Это позволяет избежать переобучения.
3. **Визуализируйте кривые ROC и PR для лучшего понимания поведения модели.**







### Примеры применения метрик классификации

Для наглядности разберём примеры из разных областей, используя основные метрики классификации. Мы будем рассматривать задачу бинарной классификации с заданными данными, а затем интерпретировать метрики.



## 1. Бинарная классификация: Медицина

**Задача:** Определить наличие заболевания по результатам теста (1 — заболевание, 0 — здоров).

| **Истинный класс** | **Предсказание** |  
|||  
| 1                   | 1 (TP)          |  
| 1                   | 0 (FN)          |  
| 0                   | 1 (FP)          |  
| 0                   | 0 (TN)          |  

### Пример матрицы ошибок:

|                   | **Предсказано: 1 (Положительное)** | **Предсказано: 0 (Отрицательное)** |  
|-|--|--|  
| **Истинное: 1 (Положительное)** | 50 (TP)                             | 10 (FN)                             |  
| **Истинное: 0 (Отрицательное)** | 5 (FP)                              | 35 (TN)                             |  

### Расчёт метрик:

1. **Точность (Accuracy):**
$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN} = \frac{50 + 35}{50 + 35 + 5 + 10} = 0.85 (85\%)
$$
Модель верно классифицирует 85% случаев.

2. **Полнота (Recall):**
$$
Recall = \frac{TP}{TP + FN} = \frac{50}{50 + 10} = 0.833 (83.3\%)
$$
Модель обнаруживает 83.3% пациентов с заболеванием.

3. **Точность (Precision):**
$$
Precision = \frac{TP}{TP + FP} = \frac{50}{50 + 5} = 0.909 (90.9\%)
$$
Из всех положительных предсказаний 90.9% действительно больны.

4. **F1-мера:**
$$
F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall} = 2 \cdot \frac{0.909 \cdot 0.833}{0.909 + 0.833} \approx 0.869 (86.9\%)
$$

**Вывод:**
- Модель имеет высокую точность и полноту. Однако ложные отрицания (FN = 10) могут быть критичными, так как это означает пропуск заболевания. В данном случае, возможно, стоит повысить Recall, даже если это снизит Precision.



## 2. Несбалансированные данные: Выявление мошенничества

**Задача:** Определить мошеннические транзакции (1 — мошенничество, 0 — нормальная транзакция).

**Данные:** Из 10 000 транзакций только 100 мошеннические (1% положительного класса).

### Пример матрицы ошибок:

|                   | **Предсказано: 1 (Положительное)** | **Предсказано: 0 (Отрицательное)** |  
|-|--|--|  
| **Истинное: 1 (Мошенничество)** | 80 (TP)                              | 20 (FN)                              |  
| **Истинное: 0 (Нормальная транзакция)** | 50 (FP)                              | 9 850 (TN)                           |  

### Расчёт метрик:

1. **Точность (Accuracy):**
$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN} = \frac{80 + 9850}{80 + 9850 + 50 + 20} = 0.993 (99.3\%)
$$
Модель демонстрирует очень высокую точность.

2. **Полнота (Recall):**
$$
Recall = \frac{TP}{TP + FN} = \frac{80}{80 + 20} = 0.8 (80\%)
$$
Модель обнаруживает 80% мошеннических транзакций.

3. **Точность (Precision):**
$$
Precision = \frac{TP}{TP + FP} = \frac{80}{80 + 50} \approx 0.615 (61.5\%)
$$
Из всех обнаруженных мошеннических транзакций только 61.5% действительно мошеннические.

4. **F1-мера:**
$$
F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall} = 2 \cdot \frac{0.615 \cdot 0.8}{0.615 + 0.8} \approx 0.696 (69.6\%)
$$

**Вывод:**
- Высокое значение Accuracy ($99.3\%$) вводит в заблуждение, так как связано с доминированием класса 0.
- В данном случае ключевыми метриками являются Recall и F1. Для задач типа антифрод необходимо минимизировать пропуски (FN), так как они могут привести к серьёзным финансовым потерям.



## 3. Мультиклассовая классификация: Рекомендательная система

**Задача:** Классифицировать жанр фильма (например, комедия, драма, боевик, фантастика). Допустим, у нас 4 класса.

### Данные (упрощённые):
- Истинные классы: [Комедия, Драма, Боевик, Фантастика, Драма].
- Предсказания модели: [Комедия, Драма, Комедия, Фантастика, Фантастика].

### Матрица ошибок:

|                   | **Комедия** | **Драма** | **Боевик** | **Фантастика** |  
|-|-|--||-|  
| **Комедия**       | 1           | 0         | 1          | 0              |  
| **Драма**         | 0           | 1         | 0          | 1              |  
| **Боевик**        | 0           | 0         | 0          | 0              |  
| **Фантастика**    | 0           | 0         | 0          | 1              |  

### Расчёт метрик (Macro и Micro-averaging):

1. **Точность (Micro-averaging):**
$$
Micro = \frac{\text{Общее число верных предсказаний}}{\text{Общее число примеров}} = \frac{3}{5} = 0.6 (60\%)
$$

2. **F1-мера (Macro-averaging):**
Для каждого класса:
- Комедия: $Recall = 0.5, Precision = 1.0, F1 = 0.67$.
- Драма: $Recall = 0.5, Precision = 0.5, F1 = 0.5$.
- Боевик: $Recall = 0.0, Precision = 0.0, F1 = 0.0$.
- Фантастика: $Recall = 1.0, Precision = 0.5, F1 = 0.67$.

Среднее:
$$
Macro\text{-}F1 = \frac{0.67 + 0.5 + 0.0 + 0.67}{4} = 0.46 (46\%)
$$

**Вывод:**
- Micro-averaging подходит, если важен общий успех модели.
- Macro-averaging важен для оценки качества по каждому классу, особенно если классы несбалансированы.
