#8.6. k-ближайшие соседи
### Введение в k-ближайших соседей (k-NN)

k-ближайшие соседи (k-NN) — это простой алгоритм машинного обучения, используемый для классификации и регрессии. В случае классификации, k-NN определяет класс объекта путём голосования большинства из k ближайших обучающих примеров. В задачах NLP, где объектами могут быть тексты, предложения или документы, k-NN также может применяться для определения их класса.

### Основные концепции k-NN

1. Метрика расстояния: Определение расстояния между объектами. В задачах NLP часто используются метрики, такие как косинусное расстояние или евклидово расстояние между векторами, представляющими тексты.

2. Выбор числа соседей (k): Параметр k определяет количество ближайших соседей, участвующих в голосовании. Выбор правильного k влияет на качество классификации.

3. Взвешенное голосование: Иногда используется взвешенное голосование, где соседи вносят различный вклад в голосование в зависимости от их расстояния до классифицируемого объекта.

### Математическая формулировка k-NN

Пусть $ D = \{ (x_i, y_i) \}_{i=1}^{N}$ — обучающая выборка, где $ x_i$ — признаковое представление объекта, $ y_i$ — метка класса объекта.

#### Алгоритм k-NN классификации:

1. Расстояние между объектами: Обычно используется евклидово расстояние или косинусное расстояние.
   
   - Евклидово расстояние:
     $$ d(x_i, x_j) = \sqrt{\sum_{m=1}^{M} (x_{im} - x_{jm})^2} $$
   
   - Косинусное расстояние:
     $$ d(x_i, x_j) = 1 - \frac{\langle x_i, x_j \rangle}{\|x_i\| \|x_j\|} $$

2. Определение класса: Для нового объекта $ x_{new}$, находим k ближайших соседей среди $ \{ x_i \}_{i=1}^{N}$ и используем голосование для определения класса:
   $$ \hat{y}_{new} = \arg \max_{y} \sum_{i \in N_k(x_{new})} I(y_i = y) $$
   где $ N_k(x_{new})$ — множество k ближайших соседей объекта $ x_{new}$, $ I(\cdot)$ — индикаторная функция.

### Пример

Предположим, у нас есть обучающая выборка с двумя классами: спам и не спам. Каждый текст представлен как вектор признаков (например, TF-IDF вектор).

1. Обучение: Вычисляем векторы признаков для всех текстов обучающей выборки.
2. Классификация нового текста: При поступлении нового текста, мы сначала вычисляем его вектор признаков. Затем находим k ближайших текстов из обучающей выборки с помощью выбранной метрики (например, косинусное расстояние).
3. Голосование: Классифицируем новый текст на основе меток ближайших соседей.

### Вывод

k-ближайшие соседи (k-NN) представляют собой простой и эффективный метод классификации, особенно в задачах с текстовыми данными, такими как NLP. Правильный выбор метрики расстояния и числа соседей существенно влияет на результаты классификации.


Давайте рассмотрим два числовых примера использования алгоритма k-ближайших соседей (k-NN) в задачах обработки естественного языка (NLP). Оба примера будут подробно описаны с шагами решения.

### Пример 1: Классификация текстов по категориям

Предположим, у нас есть небольшой набор текстов, которые мы хотим классифицировать на две категории: "Спам" и "Не спам". Для простоты, мы будем использовать векторы TF-IDF (Term Frequency-Inverse Document Frequency) для представления текстов.

#### Данные

- Тексты в виде векторов TF-IDF:

| Текст              | Вектор (TF-IDF) | Класс    |
|--||-|
| "Buy cheap shoes"  | $ (0.7, 0.2)$ | Спам     |
| "Lowest price on books" | $ (0.6, 0.3)$ | Спам     |
| "Meeting at noon"  | $ (0.1, 0.9)$ | Не спам  |
| "Your invoice"     | $ (0.2, 0.8)$ | Не спам  |

- Новый текст для классификации: "Discount shoes available" с вектором $ (0.65, 0.25)$.

#### Шаги решения

1. Вычисление расстояний: Вычислим евклидово расстояние между вектором нового текста и каждым из векторов обучающей выборки.

   - Расстояние до "Buy cheap shoes":
     $$
     d = \sqrt{(0.65 - 0.7)^2 + (0.25 - 0.2)^2} = \sqrt{(-0.05)^2 + (0.05)^2} = \sqrt{0.0025 + 0.0025} = \sqrt{0.005} \approx 0.0707
     $$

   - Расстояние до "Lowest price on books":
     $$
     d = \sqrt{(0.65 - 0.6)^2 + (0.25 - 0.3)^2} = \sqrt{(0.05)^2 + (-0.05)^2} = \sqrt{0.0025 + 0.0025} = \sqrt{0.005} \approx 0.0707
     $$

   - Расстояние до "Meeting at noon":
     $$
     d = \sqrt{(0.65 - 0.1)^2 + (0.25 - 0.9)^2} = \sqrt{(0.55)^2 + (-0.65)^2} = \sqrt{0.3025 + 0.4225} = \sqrt{0.725} \approx 0.851
     $$

   - Расстояние до "Your invoice":
     $$
     d = \sqrt{(0.65 - 0.2)^2 + (0.25 - 0.8)^2} = \sqrt{(0.45)^2 + (-0.55)^2} = \sqrt{0.2025 + 0.3025} = \sqrt{0.505} \approx 0.711
     $$

2. Сортировка расстояний: Теперь отсортируем расстояния от нового текста до всех обучающих примеров:

   | Текст                     | Расстояние |
   |||
   | "Buy cheap shoes"        | 0.0707     |
   | "Lowest price on books"   | 0.0707     |
   | "Your invoice"           | 0.711      |
   | "Meeting at noon"        | 0.851      |

3. Выбор k: Допустим, мы выберем $ k = 2$.

4. Голосование: Теперь определим класс нового текста по ближайшим соседям. У нас два ближайших соседа:

   - "Buy cheap shoes" (Спам)
   - "Lowest price on books" (Спам)

   Оба соседа относятся к классу "Спам". Таким образом, голосование приводит к тому, что новый текст классифицируется как Спам.

### Пример 2: Классификация настроений

Теперь рассмотрим более сложный пример, в котором мы классифицируем тексты по настроению: "Положительное" или "Отрицательное". У нас есть следующие данные:

#### Данные

- Тексты в виде векторов TF-IDF:

| Текст                       | Вектор (TF-IDF) | Класс         |
|--|||
| "I love this product"      | $ (0.8, 0.1)$ | Положительное  |
| "This is amazing"          | $ (0.9, 0.05)$| Положительное  |
| "I hate this"              | $ (0.1, 0.7)$ | Отрицательное  |
| "This is terrible"         | $ (0.2, 0.6)$ | Отрицательное  |

- Новый текст для классификации: "I dislike this item" с вектором $ (0.1, 0.5)$.

#### Шаги решения

1. Вычисление расстояний: Вычислим евклидово расстояние между вектором нового текста и каждым из векторов обучающей выборки.

   - Расстояние до "I love this product":
     $$
     d = \sqrt{(0.1 - 0.8)^2 + (0.5 - 0.1)^2} = \sqrt{(-0.7)^2 + (0.4)^2} = \sqrt{0.49 + 0.16} = \sqrt{0.65} \approx 0.806
     $$

   - Расстояние до "This is amazing":
     $$
     d = \sqrt{(0.1 - 0.9)^2 + (0.5 - 0.05)^2} = \sqrt{(-0.8)^2 + (0.45)^2} = \sqrt{0.64 + 0.2025} = \sqrt{0.8425} \approx 0.918
     $$

   - Расстояние до "I hate this":
     $$
     d = \sqrt{(0.1 - 0.1)^2 + (0.5 - 0.7)^2} = \sqrt{(0)^2 + (-0.2)^2} = \sqrt{0 + 0.04} = 0.2
     $$

   - Расстояние до "This is terrible":
     $$
     d = \sqrt{(0.1 - 0.2)^2 + (0.5 - 0.6)^2} = \sqrt{(-0.1)^2 + (-0.1)^2} = \sqrt{0.01 + 0.01} = \sqrt{0.02} \approx 0.141
     $$

2. Сортировка расстояний: Теперь отсортируем расстояния от нового текста до всех обучающих примеров:

   | Текст                     | Расстояние |
   |||
   | "This is terrible"       | 0.141      |
   | "I hate this"           | 0.2        |
   | "I love this product"    | 0.806      |
   | "This is amazing"        | 0.918      |

3. Выбор k: Допустим, мы выберем $ k = 2$.

4. Голосование: Определим класс нового текста по ближайшим соседям. У нас два ближайших соседа:

   - "This is terrible" (Отрицательное)
   - "I hate this" (Отрицательное)

   Оба соседа относятся к классу "Отрицательное". Таким образом, новый текст классифицируется как Отрицательное.

### Заключение

Оба примера иллюстрируют, как работает алгоритм k-ближайших соседей для классификации текстов в задачах NLP. Правильный выбор метрики расстояния, определение ближайших соседей и голосование позволяют эффективно классифицировать новые тексты на основе обучающей выборки.










Давайте реализуем оба примера на Python с использованием библиотеки `matplotlib` для визуализации. В данном примере мы будем использовать `numpy` для вычисления расстояний и обработки данных. Также применим библиотеку `sklearn` для работы с векторами и визуализации.

### Пример 1: Классификация текстов по категориям (Спам и Не спам)



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize

# Данные
texts = ["Buy cheap shoes", "Lowest price on books", "Meeting at noon", "Your invoice"]
vectors = np.array([[0.7, 0.2], [0.6, 0.3], [0.1, 0.9], [0.2, 0.8]])  # TF-IDF векторы
labels = ["Spam", "Spam", "Not Spam", "Not Spam"]

# Новый текст
new_text_vector = np.array([0.65, 0.25])

# Функция для вычисления евклидова расстояния
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

# Вычисляем расстояния
distances = np.array([euclidean_distance(new_text_vector, vector) for vector in vectors])

# Сортируем расстояния и соответствующие метки
sorted_indices = np.argsort(distances)
sorted_labels = [labels[i] for i in sorted_indices]

# Выбираем k=2
k = 2
nearest_labels = sorted_labels[:k]

# Голосование
from collections import Counter
vote_counts = Counter(nearest_labels)
predicted_label = vote_counts.most_common(1)[0][0]

# Визуализация
plt.figure(figsize=(8, 6))
plt.scatter(vectors[:, 0], vectors[:, 1], c=['red' if label == 'Spam' else 'blue' for label in labels], label='Training Data')
plt.scatter(new_text_vector[0], new_text_vector[1], c='green', label='New Text', marker='X', s=200)

# Настройки графика
for i, txt in enumerate(texts):
    plt.annotate(txt, (vectors[i, 0], vectors[i, 1]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title(f'k-NN Classification (Predicted: {predicted_label})')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()
```


### Пример 2: Классификация настроений (Положительное и Отрицательное)



```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Данные
texts_sentiment = ["I love this product", "This is amazing", "I hate this", "This is terrible"]
vectors_sentiment = np.array([[0.8, 0.1], [0.9, 0.05], [0.1, 0.7], [0.2, 0.6]])  # TF-IDF векторы
labels_sentiment = ["Positive", "Positive", "Negative", "Negative"]

# Новый текст
new_text_vector_sentiment = np.array([0.1, 0.5])

# Функция для вычисления евклидова расстояния
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

# Вычисляем расстояния
distances_sentiment = np.array([euclidean_distance(new_text_vector_sentiment, vector) for vector in vectors_sentiment])

# Сортируем расстояния и соответствующие метки
sorted_indices_sentiment = np.argsort(distances_sentiment)
sorted_labels_sentiment = [labels_sentiment[i] for i in sorted_indices_sentiment]

# Выбираем k=2
k_sentiment = 2
nearest_labels_sentiment = sorted_labels_sentiment[:k_sentiment]

# Голосование
vote_counts_sentiment = Counter(nearest_labels_sentiment)
predicted_label_sentiment = vote_counts_sentiment.most_common(1)[0][0]

# Визуализация
plt.figure(figsize=(8, 6))
plt.scatter(vectors_sentiment[:, 0], vectors_sentiment[:, 1],
            c=['orange' if label == 'Positive' else 'purple' for label in labels_sentiment],
            label='Training Data')
plt.scatter(new_text_vector_sentiment[0], new_text_vector_sentiment[1], c='green', label='New Text', marker='X', s=200)

# Настройки графика
for i, txt in enumerate(texts_sentiment):
    plt.annotate(txt, (vectors_sentiment[i, 0], vectors_sentiment[i, 1]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title(f'k-NN Sentiment Classification (Predicted: {predicted_label_sentiment})')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()
```
