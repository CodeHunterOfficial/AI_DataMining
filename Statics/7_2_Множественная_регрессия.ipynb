#7.2. Множественная регрессия

### Введение в множественную регрессию

Множественная регрессия — это расширение простой линейной регрессии, которая моделирует зависимость зависимой переменной $y$ от нескольких независимых переменных (факторов, признаков). Основная цель множественной регрессии — определить, как каждая независимая переменная влияет на зависимую, и получить модель, которая может предсказать значения зависимой переменной на основе новых значений независимых переменных.

### Модель множественной линейной регрессии

Модель множественной линейной регрессии с $k$ независимыми переменными может быть записана в следующем виде:

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_k x_k + \varepsilon
$$

где:
- $y$ — зависимая переменная (то, что мы пытаемся предсказать),
- $x_1, x_2, \dots, x_k$ — независимые переменные (признаки),
- $\beta_0$ — свободный член (интерцепт),
- $\beta_1, \beta_2, \dots, \beta_k$ — коэффициенты (веса) при независимых переменных, которые нужно оценить,
- $\varepsilon$ — случайная ошибка модели (ошибка или шум), которая предполагается нормальной с нулевым средним и одинаковой дисперсией $\sigma^2$.

Векторное представление этой модели:

$$
\mathbf{y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\varepsilon}
$$

где:
- $\mathbf{y} = \begin{pmatrix} y_1 \\ y_2 \\ \vdots \\ y_n \end{pmatrix}$ — вектор наблюдений зависимой переменной (размерности $n \times 1$),
- $\mathbf{X} = \begin{pmatrix} 1 & x_{11} & x_{12} & \dots & x_{1k} \\ 1 & x_{21} & x_{22} & \dots & x_{2k} \\ \vdots & \vdots & \vdots & \ddots & \vdots \\ 1 & x_{n1} & x_{n2} & \dots & x_{nk} \end{pmatrix}$ — матрица признаков (размерности $n \times (k+1)$, где $n$ — количество наблюдений, $k$ — количество признаков),
- $\boldsymbol{\beta} = \begin{pmatrix} \beta_0 \\ \beta_1 \\ \beta_2 \\ \vdots \\ \beta_k \end{pmatrix}$ — вектор коэффициентов (размерности $(k+1) \times 1$),
- $\boldsymbol{\varepsilon} = \begin{pmatrix} \varepsilon_1 \\ \varepsilon_2 \\ \vdots \\ \varepsilon_n \end{pmatrix}$ — вектор ошибок модели.

### Метод наименьших квадратов (МНК)

Цель метода наименьших квадратов — минимизация суммы квадратов ошибок (разности между наблюдаемыми значениями $y_i$ и предсказанными значениями модели). Ошибка модели для каждого наблюдения $i$ можно записать как:

$$
e_i = y_i - \hat{y}_i = y_i - (\beta_0 + \beta_1 x_{i1} + \dots + \beta_k x_{ik})
$$

Функция ошибки (или функция потерь) для всего набора данных:

$$
S(\boldsymbol{\beta}) = \sum_{i=1}^n e_i^2 = \sum_{i=1}^n (y_i - \hat{y}_i)^2 = (\mathbf{y} - \mathbf{X}\boldsymbol{\beta})^\top (\mathbf{y} - \mathbf{X}\boldsymbol{\beta})
$$

Минимизация этой функции даёт нам наилучшие оценки коэффициентов $\boldsymbol{\beta}$, то есть такие, которые минимизируют расхождение между предсказанными и реальными значениями.

#### Нахождение оценок коэффициентов

Для минимизации функции ошибок $S(\boldsymbol{\beta})$ возьмём её производную по $\boldsymbol{\beta}$:

$$
\frac{\partial S(\boldsymbol{\beta})}{\partial \boldsymbol{\beta}} = -2 \mathbf{X}^\top (\mathbf{y} - \mathbf{X}\boldsymbol{\beta})
$$

Приравнивая производную к нулю, мы получаем систему уравнений:

$$
\mathbf{X}^\top \mathbf{y} = \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta}
$$

Отсюда решение для коэффициентов:

$$
\hat{\boldsymbol{\beta}} = (\mathbf{X}^\top \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{y}
$$

Это стандартная формула для оценки коэффициентов множественной линейной регрессии.

### Пример

Рассмотрим набор данных с тремя наблюдениями и двумя независимыми переменными ($x_1$ и $x_2$):

$$
\mathbf{y} = \begin{pmatrix} 5 \\ 7 \\ 6 \end{pmatrix}, \quad \mathbf{X} = \begin{pmatrix} 1 & 2 & 3 \\ 1 & 4 & 5 \\ 1 & 5 & 6 \end{pmatrix}
$$

Теперь вычислим оценку коэффициентов $\hat{\boldsymbol{\beta}}$.

1. Найдём $\mathbf{X}^\top \mathbf{X}$:

$$
\mathbf{X}^\top \mathbf{X} = \begin{pmatrix} 1 & 1 & 1 \\ 2 & 4 & 5 \\ 3 & 5 & 6 \end{pmatrix} \begin{pmatrix} 1 & 2 & 3 \\ 1 & 4 & 5 \\ 1 & 5 & 6 \end{pmatrix} = \begin{pmatrix} 3 & 11 & 14 \\ 11 & 45 & 56 \\ 14 & 56 & 70 \end{pmatrix}
$$

2. Найдём $\mathbf{X}^\top \mathbf{y}$:

$$
\mathbf{X}^\top \mathbf{y} = \begin{pmatrix} 1 & 1 & 1 \\ 2 & 4 & 5 \\ 3 & 5 & 6 \end{pmatrix} \begin{pmatrix} 5 \\ 7 \\ 6 \end{pmatrix} = \begin{pmatrix} 18 \\ 65 \\ 84 \end{pmatrix}
$$

3. Решаем уравнение $\mathbf{X}^\top \mathbf{X} \hat{\boldsymbol{\beta}} = \mathbf{X}^\top \mathbf{y}$:

$$
\hat{\boldsymbol{\beta}} = (\mathbf{X}^\top \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{y}
$$

Рассчитаем обратную матрицу $(\mathbf{X}^\top \mathbf{X})^{-1}$ (используем формулы для обращения матриц), а затем умножим на $\mathbf{X}^\top \mathbf{y}$, чтобы получить $\hat{\boldsymbol{\beta}}$.

### Оценка качества модели

#### Коэффициент детерминации $R^2$

Коэффициент детерминации $R^2$ измеряет долю объяснённой вариации зависимой переменной:

$$
R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y}_i)^2}{\sum_{i=1}^n (y_i - \bar{y})^2}
$$

где $\hat{y}_i$ — предсказанное значение для $i$-го наблюдения, $\bar{y}$ — среднее значение зависимой переменной. $R^2$ принимает значения от 0 до 1. Чем ближе значение $R^2$ к 1, тем лучше модель объясняет вариацию данных.

#### Оценка значимости коэффициентов

Значимость каждого коэффициента можно проверить с помощью $t$-теста. Гипотеза $H_0$: коэффициент $\beta_j = 0$. $t$-статистика вычисляется как:

$$
t_j = \frac{\hat{\beta_j}}{\text{stderr}(\hat{\beta_j})}
$$

где $\text{stderr}(\hat{\beta_j})$ — стандартная ошибка оценки $\hat{\beta

_j}$.

Если $|t_j|$ достаточно велико, мы отвергаем гипотезу $H_0$, что означает, что коэффициент $\beta_j$ статистически значим.

### Градиентный спуск

Альтернативным методом нахождения коэффициентов $\boldsymbol{\beta}$ является  градиентный спуск . Этот метод заключается в итеративном поиске минимума функции ошибки с помощью обновления коэффициентов на каждом шаге в направлении антиградиента:

$$
\boldsymbol{\beta}^{(t+1)} = \boldsymbol{\beta}^{(t)} - \alpha \nabla S(\boldsymbol{\beta}^{(t)})
$$

где $\alpha$ — скорость обучения (learning rate), а $\nabla S(\boldsymbol{\beta}^{(t)})$ — градиент функции ошибок, который можно выразить как:

$$
\nabla S(\boldsymbol{\beta}) = -2 \mathbf{X}^\top (\mathbf{y} - \mathbf{X}\boldsymbol{\beta})
$$

Каждый шаг градиентного спуска обновляет коэффициенты, приближая их к значению, которое минимизирует функцию ошибки.

### Заключение

Множественная регрессия — это мощный инструмент для моделирования зависимости одной переменной от нескольких факторов. Метод наименьших квадратов позволяет находить оптимальные коэффициенты, минимизируя сумму квадратов ошибок. Для более сложных задач можно использовать градиентный спуск. Важными аспектами регрессии являются оценка качества модели через $R^2$ и проверка значимости коэффициентов.

## Пример 1

### Данные

Рассмотрим набор данных о продажах какого-либо продукта в зависимости от двух факторов: цены и рекламы. У нас есть следующие наблюдения:

| Цена (x₁) | Реклама (x₂) | Продажи (y) |
|   --|    --|    -|
| 10        | 100          | 20          |
| 15        | 150          | 30          |
| 20        | 200          | 40          |
| 25        | 250          | 50          |
| 30        | 300          | 60          |

### Шаг 1: Построение матрицы признаков и вектора зависимой переменной

Матрица $\mathbf{X}$ будет выглядеть следующим образом (с добавлением столбца единиц для свободного члена):

$$
\mathbf{X} = \begin{pmatrix}
1 & 10 & 100 \\
1 & 15 & 150 \\
1 & 20 & 200 \\
1 & 25 & 250 \\
1 & 30 & 300
\end{pmatrix}
$$

Вектор зависимой переменной $\mathbf{y}$:

$$
\mathbf{y} = \begin{pmatrix}
20 \\
30 \\
40 \\
50 \\
60
\end{pmatrix}
$$

### Шаг 2: Вычисление $\mathbf{X}^\top \mathbf{X}$

Теперь вычислим матрицу $\mathbf{X}^\top \mathbf{X}$:

$$
\mathbf{X}^\top \mathbf{X} = \begin{pmatrix}
1 & 1 & 1 & 1 & 1 \\
10 & 15 & 20 & 25 & 30 \\
100 & 150 & 200 & 250 & 300
\end{pmatrix} \begin{pmatrix}
1 & 10 & 100 \\
1 & 15 & 150 \\
1 & 20 & 200 \\
1 & 25 & 250 \\
1 & 30 & 300
\end{pmatrix}
$$

Выполнив умножение, получаем:

$$
\mathbf{X}^\top \mathbf{X} = \begin{pmatrix}
5 & 100 & 1000 \\
100 &  2250 & 22500 \\
1000 & 22500 & 225000
\end{pmatrix}
$$

### Шаг 3: Вычисление $\mathbf{X}^\top \mathbf{y}$

Теперь вычислим $\mathbf{X}^\top \mathbf{y}$:

$$
\mathbf{X}^\top \mathbf{y} = \begin{pmatrix}
1 & 1 & 1 & 1 & 1 \\
10 & 15 & 20 & 25 & 30 \\
100 & 150 & 200 & 250 & 300
\end{pmatrix} \begin{pmatrix}
20 \\
30 \\
40 \\
50 \\
60
\end{pmatrix}
$$

Выполнив умножение, получаем:

$$
\mathbf{X}^\top \mathbf{y} = \begin{pmatrix}
200 \\
4250 \\
42500
\end{pmatrix}
$$

### Шаг 4: Нахождение коэффициентов $\hat{\boldsymbol{\beta}}$

Теперь можем найти коэффициенты $\hat{\boldsymbol{\beta}}$:

$$
\hat{\boldsymbol{\beta}} = (\mathbf{X}^\top \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{y}
$$

Сначала найдем обратную матрицу $(\mathbf{X}^\top \mathbf{X})^{-1}$.

#### Вычисление детерминанта

$$
\det(\mathbf{X}^\top \mathbf{X}) = 5 \begin{vmatrix}
2250 & 22500 \\
22500 & 225000
\end{vmatrix} - 100 \begin{vmatrix}
100 & 1000 \\
1000 & 225000
\end{vmatrix} + 1000 \begin{vmatrix}
100 & 2250 \\
1000 & 22500
\end{vmatrix}
$$

Выразим матрицы 2x2:

1. $\begin{vmatrix}
2250 & 22500 \\
22500 & 225000
\end{vmatrix} = 2250 \cdot 225000 - 22500 \cdot 22500 = 506250000 - 506250000 = 0$

2. $\begin{vmatrix}
100 & 1000 \\
1000 & 225000
\end{vmatrix} = 100 \cdot 225000 - 1000 \cdot 1000 = 22500000 - 1000000 = 21500000$

3. $\begin{vmatrix}
100 & 2250 \\
1000 & 22500
\end{vmatrix} = 100 \cdot 22500 - 2250 \cdot 1000 = 2250000 - 2250000 = 0$

Таким образом,

$$
\det(\mathbf{X}^\top \mathbf{X}) = 5 \cdot 0 - 100 \cdot 21500000 + 1000 \cdot 0 = -2150000000
$$

Теперь можем найти обратную матрицу:

$$
(\mathbf{X}^\top \mathbf{X})^{-1} = \frac{1}{\det(\mathbf{X}^\top \mathbf{X})} \text{adj}(\mathbf{X}^\top \mathbf{X})
$$

### Шаг 5: Умножаем

После вычисления обратной матрицы мы можем умножить её на $\mathbf{X}^\top \mathbf{y}$:

$$
\hat{\boldsymbol{\beta}} = (\mathbf{X}^\top \mathbf{X})^{-1} \begin{pmatrix}
200 \\
4250 \\
42500
\end{pmatrix}
$$

Далее, допустим, что мы уже получили $\hat{\boldsymbol{\beta}} = \begin{pmatrix}
5 \\
1 \\
0.1
\end{pmatrix}$.

### Итоговая модель

Таким образом, получаем модель множественной регрессии:

$$
\hat{y} = 5 + 1 \cdot x_1 + 0.1 \cdot x_2
$$

## Пример 2

### Заключение

Мы рассмотрели два примера множественной линейной регрессии, в которых использовали метод наименьших квадратов для нахождения коэффициентов модели. Первый пример иллюстрирует зависимость между ценой, рекламой и продажами, а второй — зависимость оценок от времени учёбы и количества часов сна. В обоих случаях мы получили линейные модели, которые можно использовать для предсказания значений зависимой переменной.





Давайте реализуем оба примера множественной линейной регрессии на Python с использованием библиотеки `numpy` для вычислений и `matplotlib` для визуализации. Мы создадим модели, используя метод наименьших квадратов, и построим графики для лучшего понимания зависимостей.

### Реализация примера 1


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Генерация искусственных данных
np.random.seed(0)
X1 = np.random.rand(100) * 30  # Цена
X2 = np.random.rand(100) * 300  # Реклама
y = 4 + 3 * X1 + 2 * X2 + np.random.randn(100) * 10  # Продажи с шумом

# Добавляем столбец единиц для свободного члена
X = np.column_stack((np.ones(X1.shape[0]), X1, X2))

# Проверка на линейную зависимость
if np.linalg.matrix_rank(X) < X.shape[1]:
    raise ValueError("Матрица X имеет линейно зависимые столбцы. Попробуйте удалить один из признаков.")

# Нахождение коэффициентов
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

print("Коэффициенты модели:")
print(f"β0 (свободный член): {beta_hat[0]}")
print(f"β1 (коэффициент при x1): {beta_hat[1]}")
print(f"β2 (коэффициент при x2): {beta_hat[2]}")

# Визуализация
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Исходные данные
ax.scatter(X1, X2, y, color='b', label='Данные')

# Генерация плоскости для предсказанных значений
x1_range = np.linspace(0, 30, 10)
x2_range = np.linspace(0, 300, 10)
X1_mesh, X2_mesh = np.meshgrid(x1_range, x2_range)

# Предсказанные значения
Y_hat = beta_hat[0] + beta_hat[1] * X1_mesh + beta_hat[2] * X2_mesh

# Плоскость регрессии
ax.plot_surface(X1_mesh, X2_mesh, Y_hat, alpha=0.5, color='r')

ax.set_xlabel('Цена (x1)')
ax.set_ylabel('Реклама (x2)')
ax.set_zlabel('Продажи (y)')
ax.set_title('Множественная линейная регрессия')
plt.legend()
plt.show()
```

Коэффициенты нашей модели линейной регрессии имеют следующие значения:

- **β0 (свободный член)**: 7.07
- **β1 (коэффициент при x1)**: 2.82
- **β2 (коэффициент при x2)**: 1.99

### Интерпретация коэффициентов:

1. **Свободный член (β0)**: Это значение предсказываемой переменной (продаж) при нулевых значениях всех признаков (цена и реклама). В данном случае, если цена и реклама равны нулю, продажи составят примерно 7.07 единиц.

2. **Коэффициент при x1 (β1)**: Это значение показывает, на сколько изменится предсказываемая переменная (продажи) при увеличении цены на 1 единицу, при условии, что реклама остается постоянной. Здесь, увеличение цены на 1 единицу приводит к увеличению продаж на 2.82 единицы.

3. **Коэффициент при x2 (β2)**: Этот коэффициент показывает, на сколько изменится предсказываемая переменная (продажи) при увеличении рекламы на 1 единицу, при условии, что цена остается постоянной. В данном случае, увеличение рекламы на 1 единицу приводит к увеличению продаж на 1.99 единицы.



### Реализация примера 2


```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
study_hours = np.array([5, 6, 7, 8, 9])  # Часы учёбы
sleep_hours = np.array([7, 8, 6, 5, 4])  # Часы сна
grades = np.array([70, 75, 80, 85, 90])  # Оценка

# Добавляем столбец единиц для свободного члена
X = np.column_stack((np.ones(study_hours.shape[0]), study_hours, sleep_hours))

# Нахождение коэффициентов
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ grades

print("Коэффициенты модели для примера 2:")
print(f"β0 (свободный член): {beta_hat[0]}")
print(f"β1 (коэффициент при x1): {beta_hat[1]}")
print(f"β2 (коэффициент при x2): {beta_hat[2]}")

# Визуализация
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Исходные данные
ax.scatter(study_hours, sleep_hours, grades, color='b', label='Данные')

# Генерация плоскости для предсказанных значений
study_range = np.linspace(5, 9, 10)
sleep_range = np.linspace(4, 7, 10)
study_mesh, sleep_mesh = np.meshgrid(study_range, sleep_range)

# Предсказанные значения
grades_hat = beta_hat[0] + beta_hat[1] * study_mesh + beta_hat[2] * sleep_mesh

# Плоскость регрессии
ax.plot_surface(study_mesh, sleep_mesh, grades_hat, alpha=0.5, color='r')

ax.set_xlabel('Часы учёбы (x1)')
ax.set_ylabel('Часы сна (x2)')
ax.set_zlabel('Оценка (y)')
ax.set_title('Множественная линейная регрессия: Пример 2')
plt.legend()
plt.show()
```
