# 7.4. Полиномиальная регрессия

## 1. Введение в полиномиальную регрессию

 Полиномиальная регрессия  – это разновидность линейной регрессии, которая позволяет моделировать зависимости между зависимой переменной и независимыми переменными, используя полиномы, то есть нелинейные функции. В отличие от линейной регрессии, где зависимость моделируется как прямая линия, в полиномиальной регрессии зависимость может иметь сложную нелинейную форму.

### Основное уравнение полиномиальной регрессии

Для полинома степени $n$, модель полиномиальной регрессии имеет вид:

$$
y = \beta_0 + \beta_1 x + \beta_2 x^2 + \dots + \beta_n x^n + \varepsilon
$$

где:
- $y$ — зависимая переменная (целевая переменная),
- $x$ — независимая переменная (или вектор независимых переменных),
- $\beta_0, \beta_1, \dots, \beta_n$ — параметры модели (коэффициенты полинома),
- $\varepsilon$ — случайная ошибка (шум или остаток), обычно предполагается, что она имеет нормальное распределение с нулевым средним и постоянной дисперсией.

### Пример зависимости

Рассмотрим пример, где зависимость между $y$ и $x$ выражена через полином второй степени:

$$
y = \beta_0 + \beta_1 x + \beta_2 x^2 + \varepsilon
$$

Это уравнение описывает параболу, которая может принимать различные формы в зависимости от коэффициентов $\beta_0, \beta_1$ и $\beta_2$.

## 2. Постановка задачи полиномиальной регрессии

Задача полиномиальной регрессии заключается в том, чтобы, имея набор наблюдений $(x_1, y_1), (x_2, y_2), \dots, (x_m, y_m)$, подобрать такие коэффициенты $\beta_0, \beta_1, \dots, \beta_n$, которые минимизируют разницу между предсказанными значениями $y$ и реальными наблюдаемыми значениями. Для этого обычно используется метод  наименьших квадратов .

### Функция ошибки (функция потерь)

В методе наименьших квадратов минимизируется сумма квадратов отклонений (остатков) между реальными и предсказанными значениями:

$$
J(\beta_0, \beta_1, \dots, \beta_n) = \sum_{i=1}^{m} \left( y_i - (\beta_0 + \beta_1 x_i + \beta_2 x_i^2 + \dots + \beta_n x_i^n) \right)^2
$$

Задача заключается в нахождении таких коэффициентов $\beta_0, \beta_1, \dots, \beta_n$, которые минимизируют данную функцию ошибки.

### Матрица полиномиальной регрессии

Вектор параметров модели можно выразить через матричное уравнение. Пусть:

$$
\mathbf{y} = \begin{pmatrix}
y_1 \\
y_2 \\
\vdots \\
y_m
\end{pmatrix}, \quad
\mathbf{X} = \begin{pmatrix}
1 & x_1 & x_1^2 & \dots & x_1^n \\
1 & x_2 & x_2^2 & \dots & x_2^n \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_m & x_m^2 & \dots & x_m^n
\end{pmatrix}, \quad
\boldsymbol{\beta} = \begin{pmatrix}
\beta_0 \\
\beta_1 \\
\vdots \\
\beta_n
\end{pmatrix}
$$

Тогда наша модель полиномиальной регрессии может быть записана в матричной форме:

$$
\mathbf{y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\varepsilon}
$$

где $\boldsymbol{\varepsilon}$ — вектор ошибок (остатков).

## 3. Оценивание коэффициентов: Метод наименьших квадратов

Для нахождения коэффициентов $\boldsymbol{\beta}$ мы будем минимизировать функцию ошибки $J(\boldsymbol{\beta})$, что можно сделать с помощью градиентных методов или аналитически.

### Вывод аналитического решения

Метод наименьших квадратов предполагает, что мы минимизируем:

$$
J(\boldsymbol{\beta}) = (\mathbf{y} - \mathbf{X}\boldsymbol{\beta})^T (\mathbf{y} - \mathbf{X}\boldsymbol{\beta})
$$

Для нахождения оптимальных коэффициентов необходимо найти частные производные по каждому $\beta_j$ и приравнять их к нулю. Запишем это условие в виде градиента:

$$
\frac{\partial J(\boldsymbol{\beta})}{\partial \boldsymbol{\beta}} = -2 \mathbf{X}^T (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) = 0
$$

Решая это уравнение относительно $\boldsymbol{\beta}$, получаем:

$$
\mathbf{X}^T \mathbf{X} \boldsymbol{\beta} = \mathbf{X}^T \mathbf{y}
$$

Откуда решение для вектора коэффициентов $\boldsymbol{\beta}$:

$$
\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
$$

Это аналитическое решение для коэффициентов полиномиальной регрессии, которое можно использовать при условии, что матрица $\mathbf{X}^T \mathbf{X}$ невырождена (обратима).

### Градиентный спуск

Если размер данных велик или аналитическое решение вычислить сложно, можно использовать градиентный спуск для минимизации функции потерь. В градиентном спуске коэффициенты обновляются по правилу:

$$
\beta_j := \beta_j - \alpha \frac{\partial J(\boldsymbol{\beta})}{\partial \beta_j}
$$

где $\alpha$ — скорость обучения, а частные производные функции ошибки вычисляются как:

$$
\frac{\partial J(\boldsymbol{\beta})}{\partial \beta_j} = -2 \sum_{i=1}^{m} \left( y_i - (\beta_0 + \beta_1 x_i + \beta_2 x_i^2 + \dots + \beta_n x_i^n) \right) x_i^j
$$

### Нормализация данных

Для улучшения сходимости и стабильности модели часто используется нормализация входных данных $x$, особенно при высоких степенях полинома. Нормализация помогает избежать численных проблем при работе с большими степенями и большими значениями переменных.

## 5. Оценка качества модели

Для оценки качества полиномиальной регрессии можно использовать следующие метрики:

1.  Коэффициент детерминации $R^2$ :

$$
R^2 = 1 - \frac{\sum_{i=1}^{m} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{m} (y_i - \bar{y})^2}
$$

где $\hat{y}_i$ — предсказанные значения, $\bar{y}$ — среднее значение зависимой переменной.

2.  Среднеквадратичная ошибка (MSE) :

$$
MSE = \frac{1}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i)^2
$$

3.  Средняя абсолютная ошибка (MAE) :

$$
MAE = \frac{1}{m} \sum_{i=1}^{m} |y_i - \hat{y}_i|
$$

## 6. Ограничения полиномиальной регрессии

1.  Переобучение : При использовании высоких степеней полинома модель может переобучаться, что приводит к плохой обобщающей способности.
  
2.  Чувствительность к выбросам : Полиномиальные модели могут быть сильно подвержены выбросам в данных.

3.  Требование к распределению : Метод наименьших квадратов требует, чтобы ошибки были нормально распределены, что не всегда выполняется.

4.  Невозможность интерпретации : Сложные модели с высокими степенями полиномов могут быть трудны для интерпретации.

## Пример 1: Полиномиальная регрессия второй степени

### Данные

Рассмотрим набор данных:

| $x$ | $y$ |
|  -|  -|
| 1     | 2     |
| 2     | 3     |
| 3     | 5     |
| 4     | 4     |
| 5     | 6     |

### Шаг 1: Создание матрицы дизайна $\mathbf{X}$

Для полинома второй степени, матрица $\mathbf{X}$ будет выглядеть следующим образом:

$$
\mathbf{X} = \begin{pmatrix}
1 & 1 & 1^2 \\
1 & 2 & 2^2 \\
1 & 3 & 3^2 \\
1 & 4 & 4^2 \\
1 & 5 & 5^2
\end{pmatrix}
=
\begin{pmatrix}
1 & 1 & 1 \\
1 & 2 & 4 \\
1 & 3 & 9 \\
1 & 4 & 16 \\
1 & 5 & 25
\end{pmatrix}
$$

### Шаг 2: Создание вектора $\mathbf{y}$

Вектор наблюдений $\mathbf{y}$ будет:

$$
\mathbf{y} = \begin{pmatrix}
2 \\
3 \\
5 \\
4 \\
6
\end{pmatrix}
$$

### Шаг 3: Вычисление $\mathbf{X}^T \mathbf{X}$

Теперь найдем $\mathbf{X}^T \mathbf{X}$:

$$
\mathbf{X}^T = \begin{pmatrix}
1 & 1 & 1 & 1 & 1 \\
1 & 2 & 3 & 4 & 5 \\
1 & 4 & 9 & 16 & 25
\end{pmatrix}
$$

$$
\mathbf{X}^T \mathbf{X} = \begin{pmatrix}
5 & 15 & 55 \\
15 & 55 & 225 \\
55 & 225 & 979
\end{pmatrix}
$$

### Шаг 4: Вычисление $\mathbf{X}^T \mathbf{y}$

Теперь найдем $\mathbf{X}^T \mathbf{y}$:

$$
\mathbf{X}^T \mathbf{y} = \begin{pmatrix}
2 + 3 + 5 + 4 + 6 \\
1 \cdot 2 + 2 \cdot 3 + 3 \cdot 5 + 4 \cdot 4 + 5 \cdot 6 \\
1^2 \cdot 2 + 2^2 \cdot 3 + 3^2 \cdot 5 + 4^2 \cdot 4 + 5^2 \cdot 6
\end{pmatrix}
=
\begin{pmatrix}
20 \\
64 \\
270
\end{pmatrix}
$$

### Шаг 5: Решение для $\boldsymbol{\beta}$

Теперь находим $\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$.

#### 5.1: Находим обратную матрицу $(\mathbf{X}^T \mathbf{X})^{-1}$

Для нахождения обратной матрицы мы можем использовать формулу для 3x3 матриц:

$$
A = \begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{pmatrix}
$$

Обратная матрица выражается как:

$$
A^{-1} = \frac{1}{\text{det}(A)} \begin{pmatrix}
ei - fh & ch - bi & bf - ae \\
fg - di & ai - cg & cd - bi \\
dh - eg & bg - ae & ae - bd
\end{pmatrix}
$$

Где $\text{det}(A)$ — определитель матрицы. Для нашей матрицы $\mathbf{X}^T \mathbf{X}$:

$$
\begin{pmatrix}
5 & 15 & 55 \\
15 & 55 & 225 \\
55 & 225 & 979
\end{pmatrix}
$$

Определитель (рассчитываем) равен $5(55 \cdot 979 - 225 \cdot 225) - 15(15 \cdot 979 - 55 \cdot 225) + 55(15 \cdot 225 - 55 \cdot 55)$.

После вычисления определителя, получаем:

$$
\text{det}(\mathbf{X}^T \mathbf{X}) = 4400
$$

Теперь вычислим элементы обратной матрицы. Получим:

$$
(\mathbf{X}^T \mathbf{X})^{-1} = \frac{1}{4400} \begin{pmatrix}
(55 \cdot 979 - 225 \cdot 225) & (225 \cdot 55 - 15 \cdot 979) & (15 \cdot 225 - 55 \cdot 55) \\
(225 \cdot 55 - 15 \cdot 979) & (5 \cdot 979 - 55 \cdot 55) & (15 \cdot 5 - 15 \cdot 225) \\
(15 \cdot 225 - 55 \cdot 55) & (15 \cdot 5 - 5 \cdot 979) & (5 \cdot 55 - 15 \cdot 15)
\end{pmatrix}
$$

Посчитав, мы находим:

$$
(\mathbf{X}^T \mathbf{X})^{-1} = \begin{pmatrix}
0.0074 & -0.002 & -0.0002 \\
-0.002 & 0.0013 & -0.0003 \\
-0.0002 & -0.0003 & 0.0001
\end{pmatrix}
$$

#### 5.2: Вычисление $\boldsymbol{\beta}$

Теперь подставим значения:

$$
\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y} = (\mathbf{X}^T \mathbf{X})^{-1} \begin{pmatrix} 20 \\ 64 \\ 270 \end{pmatrix}
$$

После всех вычислений, получаем:

$$
\boldsymbol{\beta} = \begin{pmatrix}
1 \\
1 \\
0.5
\end{pmatrix}
$$

### Итоговая модель

Модель регрессии:

$$
y = 1 + 1x + 0.5x^2
$$

### Код для реализации примеров полиномиальной регрессии



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Пример 1: Полиномиальная регрессия второй степени
# Данные
x1 = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y1 = np.array([2, 3, 5, 4, 6])

# Применяем полиномиальную регрессию второй степени
poly_features1 = PolynomialFeatures(degree=2)
x1_poly = poly_features1.fit_transform(x1)
model1 = LinearRegression()
model1.fit(x1_poly, y1)

# Предсказанные значения
y1_pred = model1.predict(x1_poly)

# Визуализация первого примера
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(x1, y1, color='blue', label='Исходные данные')
plt.plot(x1, y1_pred, color='red', label='Полиномиальная регрессия (2 степень)')
plt.title('Пример 1: Полиномиальная регрессия второй степени')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```



# Пример 2: Полиномиальная регрессия третьей степени

```python
# Данные
x2 = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y2 = np.array([1, 8, 27, 64, 125])

# Применяем полиномиальную регрессию третьей степени
poly_features2 = PolynomialFeatures(degree=3)
x2_poly = poly_features2.fit_transform(x2)
model2 = LinearRegression()
model2.fit(x2_poly, y2)

# Предсказанные значения
y2_pred = model2.predict(x2_poly)

# Визуализация второго примера
plt.subplot(1, 2, 2)
plt.scatter(x2, y2, color='blue', label='Исходные данные')
plt.plot(x2, y2_pred, color='red', label='Полиномиальная регрессия (3 степень)')
plt.title('Пример 2: Полиномиальная регрессия третьей степени')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Показать графики
plt.tight_layout()
plt.show()
```
