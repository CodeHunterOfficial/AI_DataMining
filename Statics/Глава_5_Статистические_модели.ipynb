#1. Генеративные и дискриминативные модели
Генеративные и дискриминативные модели - это две основные категории алгоритмов машинного обучения, которые используются для решения задач классификации и моделирования данных. В этой лекции будут рассмотрены основные концепции, математические формулы и примеры для каждой из этих категорий.

## 1. Введение в генеративные и дискриминативные модели

Машинное обучение можно разделить на две большие категории в зависимости от подхода к решению задач классификации:
- **Дискриминативные модели**: ориентированы на классификацию данных, то есть на разделение различных классов данных.
- **Генеративные модели**: направлены на моделирование вероятностного распределения данных, чтобы иметь возможность генерировать новые данные, которые выглядят как данные из исходного набора.

Чтобы лучше понять разницу между генеративными и дискриминативными моделями, давайте кратко рассмотрим, как каждая из них решает задачу классификации.

- **Дискриминативные модели** оценивают условную вероятность $P(y | x)$, где $y$ — это метка класса, а $x$ — входные данные. Это вероятность того, что объект принадлежит классу $y$, при условии наличия входных данных $x$.
  
- **Генеративные модели** оценивают совместное распределение вероятностей $P(x, y)$ и могут быть использованы для моделирования $P(x | y)$, что представляет собой вероятность наблюдения данных $x$, принадлежащих классу $y$. Затем можно использовать теорему Байеса для вычисления условной вероятности $P(y | x)$.

## 2. Дискриминативные модели

### 2.1. Основные концепции

Дискриминативные модели фокусируются на построении границы принятия решений, которая разделяет различные классы данных. В таких моделях используется информация о различиях между классами, чтобы оптимизировать критерии разделения данных.

Примеры дискриминативных моделей:
- Логистическая регрессия
- Метод опорных векторов (SVM)
- Классификаторы, основанные на нейронных сетях (например, сверточные нейронные сети)

### 2.2. Математическая формулировка дискриминативных моделей

Для дискриминативных моделей цель состоит в том, чтобы найти параметрическое или непараметрическое приближение функции $P(y | x; \theta)$, где $\theta$ — это параметры модели, которые нужно определить в процессе обучения. Например, в случае логистической регрессии функция классификации имеет вид:

$$
P(y = 1 | x) = \frac{1}{1 + \exp(-(\theta_0 + \theta_1 x_1 + \ldots + \theta_n x_n))},
$$

где $\theta_i$ — это веса, которые характеризуют вклад каждой входной переменной $x_i$.

### 2.3. Алгоритм обучения дискриминативной модели

Алгоритм обучения обычно состоит из следующих этапов:
1. **Определение функции потерь**: Это функция, которая измеряет, насколько хорошо модель соответствует данным. Например, для логистической регрессии используется логистическая функция потерь:

   $$
   L(\theta) = -\sum_{i=1}^{N} [y_i \log(P(y_i | x_i; \theta)) + (1 - y_i) \log(1 - P(y_i | x_i; \theta))],
   $$

   где $N$ — количество наблюдений в обучающем наборе.

2. **Минимизация функции потерь**: Оптимизация функции потерь для нахождения таких параметров $\theta$, которые минимизируют ошибку классификации. Обычно используется градиентный спуск.

3. **Оценка качества модели**: Оценка точности классификации на тестовых данных.

## 3. Генеративные модели

### 3.1. Основные концепции

Генеративные модели направлены на изучение совместного распределения данных $P(x, y)$ или $P(x)$. Цель состоит в том, чтобы понять, как формируются данные, и быть в состоянии создавать новые примеры, которые соответствуют распределению реальных данных.

Примеры генеративных моделей:
- Наивный байесовский классификатор
- Скрытые марковские модели (HMM)
- Генеративно-состязательные сети (GAN)
- Вариационные автокодировщики (VAE)

### 3.2. Математическая формулировка генеративных моделей

Генеративная модель строится на основе оценки вероятности $P(x, y)$, которая может быть разложена на произведение $P(y) \times P(x | y)$. Теорема Байеса позволяет выразить условную вероятность класса $P(y | x)$:

$$
P(y | x) = \frac{P(x | y) P(y)}{P(x)},
$$

где:
- $P(x | y)$ — вероятность наблюдения данных при условии, что они принадлежат классу $y$;
- $P(y)$ — априорная вероятность класса $y$;
- $P(x)$ — нормирующий коэффициент, представляющий собой сумму всех возможных $P(x | y) P(y)$.

### 3.3. Пример генеративной модели: Наивный байесовский классификатор

Для иллюстрации генеративных моделей рассмотрим наивный байесовский классификатор, который предполагает условную независимость признаков. Условная вероятность для каждого признака выражается как произведение вероятностей для каждого признака:

$$
P(x | y) = \prod_{i=1}^{n} P(x_i | y),
$$

где $n$ — количество признаков. Затем можно найти $P(y | x)$ и использовать его для классификации.

### 3.4. Генеративно-состязательные сети (GAN)

GAN — это более сложный пример генеративной модели, который включает две нейронные сети:
- **Генератор**, который создает новые данные.
- **Дискриминатор**, который пытается отличить реальные данные от сгенерированных.

Алгоритм обучения GAN нацелен на минимизацию функции стоимости дискриминатора и максимизацию функции стоимости генератора, что формально выражается как:

$$
\min_G \max_D V(D, G) = \mathbb{E}_{x \sim P_{\text{data}}(x)} [\log D(x)] + \mathbb{E}_{z \sim P_z(z)} [\log (1 - D(G(z)))].
$$

Здесь $D(x)$ — вероятность того, что $x$ является реальными данными, а $G(z)$ — сгенерированные данные.

## 4. Сравнение генеративных и дискриминативных моделей

### 4.1. Преимущества и недостатки дискриминативных моделей
- **Преимущества**: Легко обучаются, хорошая точность на задачах классификации, оптимизация направлена непосредственно на уменьшение ошибки классификации.
- **Недостатки**: Не могут использоваться для генерации новых данных, зависят от качества размеченных данных.

### 4.2. Преимущества и недостатки генеративных моделей
- **Преимущества**: Могут использоваться для создания новых данных, позволяют лучше моделировать структуру данных.
- **Недостатки**: Более сложные для обучения, могут страдать от проблем устойчивости и сходиться к неоптимальным решениям.

### 4.3. Комбинированные подходы
Иногда для улучшения результатов комбинируют генеративные и дискриминативные модели. Например, в алгоритмах полусупервизорного обучения генеративные модели могут использоваться для предварительного обучения, а дискриминативные — для последующей точной настройки.

## 5. Заключение

Генеративные и дискриминативные модели являются важными компонентами в арсенале машинного обучения, и каждый из этих подходов имеет свои области применения. Дискриминативные модели особенно полезны в задачах классификации, в то время как генеративные модели применимы там, где требуется моделирование распределения данных и генерация новых примеров.

### Задача 1: Логистическая регрессия для бинарной классификации

#### 1. Постановка задачи

Предположим, что у нас есть набор данных с признаками о пациентах, и мы хотим предсказать, есть ли у пациента диабет (1 — диабет, 0 — нет диабета).

#### 2. Данные

Данные состоят из двух признаков:
- Индекс массы тела (BMI)
- Уровень глюкозы в крови

Пример данных:
| BMI  | Глюкоза | Диабет |
|||--|
| 22.0 | 85      | 0      |
| 28.0 | 90      | 1      |
| 25.0 | 95      | 0      |
| 30.0 | 120     | 1      |
| 35.0 | 150     | 1      |

#### 3. Модель

Используем логистическую регрессию, которая оценивает вероятность диагноза диабета.

#### 4. Шаги решения

1. **Инициализация параметров**:
   $$
   \theta = [\theta_0, \theta_1, \theta_2]
   $$

2. **Определение функции вероятности**:
   $$
   P(y=1 | x) = \frac{1}{1 + e^{-(\theta_0 + \theta_1 \cdot \text{BMI} + \theta_2 \cdot \text{Глюкоза})}}
   $$

3. **Определение функции потерь (логистическая функция потерь)**:
   $$
   L(\theta) = -\frac{1}{N} \sum_{i=1}^{N} [y_i \log(P(y_i | x_i; \theta)) + (1 - y_i) \log(1 - P(y_i | x_i; \theta))]
   $$

4. **Оптимизация параметров с помощью градиентного спуска**:
   - Обновление параметров:
   $$
   \theta_j \leftarrow \theta_j - \alpha \frac{\partial L(\theta)}{\partial \theta_j}
   $$
   где $\alpha$ — скорость обучения.

5. **Обучение модели**: Применяем градиентный спуск до сходимости.

6. **Предсказание**: Для нового пациента с заданными значениями BMI и уровня глюкозы вычисляем вероятность диабета и делаем предсказание.

### Задача 2: Метод опорных векторов (SVM)

#### 1. Постановка задачи

Предположим, что мы хотим классифицировать объекты на два класса на основе двух признаков.

#### 2. Данные

Пример данных:
| Признак 1 | Признак 2 | Класс |
|||-|
| 2.0        | 3.0        | 0     |
| 3.0        | 3.5        | 0     |
| 5.0        | 4.0        | 1     |
| 6.0        | 5.0        | 1     |

#### 3. Модель

Используем метод опорных векторов.

#### 4. Шаги решения

1. **Построение модели**:
   $$
   \text{minimize} \quad \frac{1}{2} \|\theta\|^2
   $$
   при условии, что:
   $$
   y_i (\theta^T x_i + b) \geq 1
   $$

2. **Обучение модели**: Используем алгоритм SMO (Sequential Minimal Optimization) для поиска оптимальных параметров $\theta$ и $b$.

3. **Граница принятия решений**: Найдем уравнение границы:
   $$
   \theta^T x + b = 0
   $$

4. **Классификация новых данных**:
   - Для нового объекта $x'$:
   $$
   y' = \begin{cases}
   1, & \text{если } \theta^T x' + b > 0 \\
   0, & \text{иначе}
   \end{cases}
   $$

### Задача 3: Наивный байесовский классификатор

#### 1. Постановка задачи

Классификация текстовых документов (например, спам/не спам).

#### 2. Данные

Пример документов:
| Документ                       | Спам |
|--||
| "Скидки на товары"            | 1    |
| "Важное сообщение"            | 0    |
| "Промо акции"                 | 1    |
| "Ваша подписка"               | 0    |

#### 3. Модель

Наивный байесовский классификатор с предположением о независимости признаков.

#### 4. Шаги решения

1. **Предобработка данных**:
   - Токенизация
   - Стемминг

2. **Подсчет вероятностей**:
   $$
   P(y=1) = \frac{\text{количество спама}}{\text{общее количество документов}}, \quad P(y=0) = \frac{\text{количество не спама}}{\text{общее количество документов}}
   $$

3. **Подсчет условных вероятностей** для каждого слова:
   $$
   P(w_i | y) = \frac{\text{количество } w_i \text{ в } y}{\text{общее количество слов в } y}
   $$

4. **Классификация нового документа**:
   $$
   P(y=1 | d) \propto P(y=1) \prod_{i} P(w_i | y=1), \quad P(y=0 | d) \propto P(y=0) \prod_{i} P(w_i | y=0)
   $$
   Выбираем класс с максимальной вероятностью.

### Задача 4: Генеративно-состязательные сети (GAN)

#### 1. Постановка задачи

Создание реалистичных изображений (например, лиц) на основе генеративной модели.

#### 2. Модель

Генератор и дискриминатор, которые обучаются в конкурентном режиме.

#### 3. Шаги решения

1. **Определение генератора** $G$:
   - Генерирует изображения из случайного шума $z$.
   - Модель:
   $$
   G(z) = \text{нейронная сеть}
   $$

2. **Определение дискриминатора** $D$:
   - Принимает изображение и выдает вероятность того, что оно реальное.
   - Модель:
   $$
   D(x) = \text{нейронная сеть}
   $$

3. **Определение функции потерь**:
   $$
   V(D, G) = \mathbb{E}_{x \sim P_{\text{data}}(x)} [\log D(x)] + \mathbb{E}_{z \sim P_z(z)} [\log (1 - D(G(z)))]
   $$

4. **Обучение**:
   - Обучение дискриминатора:
   $$
   D \leftarrow D - \alpha \nabla_D V(D, G)
   $$
   - Обучение генератора:
   $$
   G \leftarrow G - \alpha \nabla_G V(D, G)
   $$

5. **Генерация новых изображений**: Используем обученный генератор для создания новых изображений.

### Задача 5: Вариационные автокодировщики (VAE)

#### 1. Постановка задачи

Создание латентного представления данных, например, изображений.

#### 2. Модель

Автокодировщик с вариационным подходом для генерации новых данных.

#### 3. Шаги решения

1. **Определение кодировщика**:
   $$
   q(z | x) = \mathcal{N}(z; \mu(x), \sigma^2(x))
   $$
   где $\mu(x)$ и $\sigma^2(x)$ — функции, которые преобразуют входные данные $x$ в параметры распределения.

2. **Определение декодировщика**:
   $$
   p(x | z) = \mathcal{N}(x; \mu'(z), \sigma'^2(z))
   $$

3. **Определение функции потерь**:
   $$
   L(x, z) = -D_{KL}(q(z | x) || p(z)) + \mathbb{E}_{q(z|x)}[\log p(x | z)]
   $$
   где $D_{KL}$ — дивергенция Кульбака-Лейблера.

4. **Обучение модели**: Использ

уем стохастический градиентный спуск для минимизации функции потерь.

5. **Генерация новых данных**: После обучения модели можно сгенерировать новые данные, взяв случайные точки из латентного пространства и пропустив их через декодировщик.


#2. Модели скрытых Марковских процессов (HMM)

#### Введение

Модели скрытых Марковских процессов (HMM) являются мощным инструментом для моделирования последовательностей данных, где предполагается наличие скрытых состояний, которые наблюдателю неизвестны напрямую, но они определяют видимые данные. Эти модели широко используются в обработке естественного языка, распознавании речи, биоинформатике и других областях.

#### Основные понятия

1. Определение HMM:

   HMM можно определить как стохастическую модель, которая описывает последовательность наблюдаемых событий, где каждое событие зависит от некоторого скрытого состояния, переходы между которыми также подчиняются вероятностным законам.

2. Компоненты HMM:

   - Скрытые состояния (Hidden States): $S = \{ S_1, S_2, \ldots, S_N \}$ - множество скрытых состояний.
   - Видимые состояния (Observations): $O = \{ O_1, O_2, \ldots, O_T \}$ - наблюдаемые данные.
   - Матрица вероятностей переходов (Transition Probability Matrix): $A = [a_{ij}]$, где $a_{ij} = P(q_t = S_j \mid q_{t-1} = S_i)$.
   - Матрица вероятностей эмиссии (Emission Probability Matrix): $B = [b_j(k)]$, где $b_j(k) = P(O_t = O_k \mid q_t = S_j)$.
   - Начальное распределение состояний (Initial State Distribution): $\pi = [\pi_i]$, где $\pi_i = P(q_1 = S_i)$.

3. Задачи HMM:

   - Определение вероятности наблюдаемой последовательности: $P(O \mid \lambda)$, где $\lambda = (A, B, \pi)$ - параметры модели.
   - Декодирование: Определение наиболее вероятной последовательности скрытых состояний $q$ при заданных параметрах $\lambda$ и наблюдаемой последовательности $O$.
   - Оценка параметров модели: Настройка параметров $A, B, \pi$ на основе обучающих данных.

#### Математическая формализация

1. Определение вероятности наблюдаемой последовательности:

   Для последовательности $O = (O_1, O_2, \ldots, O_T)$, вероятность $P(O \mid \lambda)$ вычисляется с использованием алгоритма прямого прохода (forward algorithm):

   $$
   P(O \mid \lambda) = \sum_{i=1}^N \alpha_T(i)
   $$

   где $\alpha_t(i) = P(O_1, O_2, \ldots, O_t, q_t = S_i \mid \lambda)$.

2. Алгоритм Витерби (Viterbi algorithm):

   Для определения наиболее вероятной последовательности скрытых состояний $q = (q_1, q_2, \ldots, q_T)$ используется алгоритм Витерби, который находит последовательность с максимальной вероятностью $P(q \mid O, \lambda)$.

3. Оценка параметров модели:

   Используется метод максимального правдоподобия для оценки параметров $\lambda = (A, B, \pi)$ на основе обучающих данных.

   - Оценка матрицы $A$: $a_{ij} = \frac{\sum_{t=1}^{T-1} \xi_t(i,j)}{\sum_{t=1}^{T-1} \gamma_t(i)}$, где $\xi_t(i,j)$ - вероятность нахождения в состоянии $S_i$ в момент $t$ и переходе в состояние $S_j$ в момент $t+1$.
   - Оценка матрицы $B$: $b_j(k) = \frac{\sum_{t \mid O_t = O_k} \gamma_t(j)}{\sum_{t=1}^T \gamma_t(j)}$, где $\gamma_t(j)$ - апостериорная вероятность нахождения в состоянии $S_j$ в момент $t$.
   - Оценка начального распределения $\pi$: $\pi_i = \gamma_1(i)$, где $\gamma_1(i)$ - апостериорная вероятность начального состояния $S_i$.

#### Примеры

Давайте рассмотрим пример применения HMM для моделирования последовательности слов:

1. Определение HMM для моделирования частей речи:

   - Скрытые состояния: $S = \{ \text{существительное}, \text{глагол}, \text{прилагательное}, \ldots \}$.
   - Видимые состояния: $O = \{ \text{слова в тексте} \}$.
   - Матрицы $A$ и $B$: определяют вероятности переходов между частями речи и вероятности появления слов в каждой части речи.
   - Обучение модели: настройка параметров $A$, $B$ и $\pi$ на размеченных данных.

#### Заключение

Модели скрытых Марковских процессов представляют собой мощный инструмент для моделирования последовательностей данных с неявной структурой. Они находят применение в различных областях, требующих анализа последовательностей, и играют ключевую роль в обработке естественного языка, распознавании речи и биоинформатике.



##Алгоритм Баума-Велча

Алгоритм Баума-Велча — это метод оценки параметров модели скрытых марковских процессов (HMM). Он используется для нахождения наилучших параметров модели (матрицы переходов $A$, матрицы эмиссий $B$ и начального распределения $\pi$), которые максимизируют вероятность наблюдаемой последовательности. Этот алгоритм относится к классу алгоритмов максимизации ожидания (EM, Expectation-Maximization).

### Описание алгоритма Баума-Велча

Алгоритм Баума-Велча состоит из двух основных шагов, которые повторяются итеративно:

1. **Шаг ожидания (E-шаг)**: Рассчитываются вероятности, связанные с состояниями (временные вероятности), используя текущие оценки параметров модели.

2. **Шаг максимизации (M-шаг)**: Параметры модели обновляются на основе рассчитанных вероятностей.

### Основные обозначения

- $N$ — количество скрытых состояний.
- $M$ — количество возможных наблюдаемых символов.
- $T$ — длина последовательности наблюдений.
- $O = (O_1, O_2, \ldots, O_T)$ — последовательность наблюдений.
- $A = [a_{ij}]$ — матрица переходов, где $a_{ij}$ — вероятность перехода из состояния $i$ в состояние $j$.
- $B = [b_j(k)]$ — матрица эмиссий, где $b_j(k)$ — вероятность наблюдения символа $k$ в состоянии $j$.
- $\pi = [\pi_i]$ — начальное распределение состояний, где $\pi_i$ — вероятность того, что начальное состояние равно $i$.

### Шаг 1: Инициализация

Инициализируем начальные параметры модели $A$, $B$ и $\pi$ произвольными значениями или оценками на основе априорных знаний.

### Шаг 2: Алгоритм прямого и обратного прохода

Перед выполнением основного цикла обновления параметров вычислим две вспомогательные матрицы:

1. **Прямой проход (Forward Algorithm)**: Вычисляется матрица $\alpha$, где $\alpha_t(i)$ — это вероятность того, что мы находимся в состоянии $i$ в момент времени $t$, и уже наблюдали первую часть последовательности $O_1, O_2, \ldots, O_t$:
   $$
   \alpha_1(i) = \pi_i \cdot b_i(O_1), \quad i = 1, \ldots, N
  $$
   $$
   \alpha_{t+1}(j) = \left( \sum_{i=1}^N \alpha_t(i) \cdot a_{ij} \right) \cdot b_j(O_{t+1}), \quad j = 1, \ldots, N
  $$

2. **Обратный проход (Backward Algorithm)**: Вычисляется матрица $\beta$, где $\beta_t(i)$ — это вероятность наблюдения последовательности от момента времени $t+1$ до $T$, находясь в состоянии $i$ в момент времени $t$:
   $$
   \beta_T(i) = 1, \quad i = 1, \ldots, N
  $$
   $$
   \beta_t(i) = \sum_{j=1}^N a_{ij} \cdot b_j(O_{t+1}) \cdot \beta_{t+1}(j), \quad t = T-1, T-2, \ldots, 1
  $$

### Шаг 3: Вычисление временных вероятностей $\gamma$ и $\xi$

- **Временная вероятность** $\gamma_t(i)$ — это вероятность того, что в момент времени $t$ система находилась в состоянии $i$, учитывая всю последовательность наблюдений:
  $$
  \gamma_t(i) = \frac{\alpha_t(i) \cdot \beta_t(i)}{\sum_{j=1}^N \alpha_t(j) \cdot \beta_t(j)}
 $$

- **Совместная временная вероятность** $\xi_t(i, j)$ — это вероятность того, что в момент времени $t$ система находилась в состоянии $i$, а в момент времени $t+1$ — в состоянии $j$:
  $$
  \xi_t(i, j) = \frac{\alpha_t(i) \cdot a_{ij} \cdot b_j(O_{t+1}) \cdot \beta_{t+1}(j)}{\sum_{k=1}^N \sum_{l=1}^N \alpha_t(k) \cdot a_{kl} \cdot b_l(O_{t+1}) \cdot \beta_{t+1}(l)}
 $$

### Шаг 4: Обновление параметров (M-шаг)

Используя вычисленные временные вероятности, обновляем параметры $A$, $B$ и $\pi$:

1. **Обновление начального распределения**:
   $$
   \pi_i = \gamma_1(i)
  $$

2. **Обновление матрицы переходов**:
   $$
   a_{ij} = \frac{\sum_{t=1}^{T-1} \xi_t(i, j)}{\sum_{t=1}^{T-1} \gamma_t(i)}
  $$

3. **Обновление матрицы эмиссий**:
   $$
   b_j(k) = \frac{\sum_{t=1}^T \gamma_t(j) \cdot \mathbb{I}(O_t = k)}{\sum_{t=1}^T \gamma_t(j)}
  $$
   где $\mathbb{I}(O_t = k)$ — индикаторная функция, равная 1, если наблюдение $O_t$ равно $k$, и 0 в противном случае.

### Шаг 5: Повторение шагов 2–4

Продолжаем повторять шаги 2–4, пока изменение параметров модели не станет достаточно малым (по заданному критерию сходимости) или пока не будет достигнуто максимальное количество итераций.

### Пример реализации на Python

Вот пример кода, который реализует алгоритм Баума-Велча:



```python
import numpy as np

def forward_algorithm(O, A, B, pi):
    N = A.shape[0]
    T = len(O)
    alpha = np.zeros((T, N))

    # Инициализация
    alpha[0, :] = pi * B[:, O[0]]

    # Рекурсия
    for t in range(1, T):
        for j in range(N):
            alpha[t, j] = np.sum(alpha[t-1, :] * A[:, j]) * B[j, O[t]]

    return alpha

def backward_algorithm(O, A, B):
    N = A.shape[0]
    T = len(O)
    beta = np.zeros((T, N))

    # Инициализация
    beta[T-1, :] = 1

    # Рекурсия
    for t in range(T-2, -1, -1):
        for i in range(N):
            beta[t, i] = np.sum(A[i, :] * B[:, O[t+1]] * beta[t+1, :])

    return beta

def baum_welch(O, N, M, max_iterations=100):
    T = len(O)
    A = np.random.rand(N, N)
    A /= A.sum(axis=1, keepdims=True)
    B = np.random.rand(N, M)
    B /= B.sum(axis=1, keepdims=True)
    pi = np.random.rand(N)
    pi /= pi.sum()

    for iteration in range(max_iterations):
        # Шаг ожидания (E-шаг)
        alpha = forward_algorithm(O, A, B, pi)
        beta = backward_algorithm(O, A, B)

        gamma = (alpha * beta) / np.sum(alpha * beta, axis=1, keepdims=True)
        xi = np.zeros((T-1, N, N))

        for t in range(T-1):
            denominator = np.sum(alpha[t, :, np.newaxis] * A * B[:, O[t+1]] * beta[t+1, :])
            xi[t] = (alpha[t, :, np.newaxis] * A * B[:, O[t+1]] * beta[t+1, :]) / denominator

        # Шаг максимизации (M-шаг)
        pi = gamma[0]
        A = np.sum(xi, axis=0) / np.sum(gamma[:-1], axis=0, keepdims=True)

        for k in range(M):
            mask = (O == k)
            B[:, k] = np.sum(gamma[mask], axis=0) / np.sum(gamma, axis=0)

        # Нормализация матрицы эмиссий
        B /= B.sum(axis=1, keepdims=True)

    return A, B, pi

# Пример использования
O = np.array([0, 1, 0, 2])  # Наблюдаемая последовательность
N = 2  # Количество скрытых состояний
M = 3  # Количество наблюдаемых символов

A, B, pi = baum_welch(O, N, M)
print("Матрица переходов A:")
print(A)
print("\nМатрица эмиссий B:")
print(B)
print("\nНачальное распределение pi:")
print(pi)
```

### Задача 1: Вычисление вероятности наблюдаемой последовательности

Условия:

Рассмотрим модель HMM с двумя скрытыми состояниями (S1 и S2) и наблюдаемыми состояниями (O1, O2). Даны следующие параметры:

- Матрица переходов $A$:

$$
A = \begin{bmatrix}
0.7 & 0.3 \\
0.4 & 0.6
\end{bmatrix}
$$

- Матрица эмиссии $B$:

$$
B = \begin{bmatrix}
0.5 & 0.5 \\
0.1 & 0.9
\end{bmatrix}
$$

- Начальное распределение $\pi$:

$$
\pi = \begin{bmatrix}
0.6 \\
0.4
\end{bmatrix}
$$

- Наблюдаемая последовательность: $O = (O_1, O_2) = (O_1, O_2)$

Задача: Найти вероятность наблюдаемой последовательности $P(O \mid \lambda)$.

#### Решение:

1. Определим значения наблюдений:
   - Пусть $O_1 = 1$ и $O_2 = 2$. Мы обозначим это через индексы: $k_1 = 1$, $k_2 = 2$.

2. Инициализация:
   - Вычислим $\alpha_1(i)$ для $i = 1, 2$:
   
$$
\alpha_1(1) = \pi_1 \cdot B_{1,1} = 0.6 \cdot 0.5 = 0.3
$$
$$
\alpha_1(2) = \pi_2 \cdot B_{2,1} = 0.4 \cdot 0.1 = 0.04
$$

3. Рекурсия для $t = 2$:
   - Вычислим $\alpha_2(i)$:
   
$$
\alpha_2(1) = \left(\alpha_1(1) \cdot A_{11} + \alpha_1(2) \cdot A_{21}\right) \cdot B_{1,2} = (0.3 \cdot 0.7 + 0.04 \cdot 0.4) \cdot 0.5 = (0.21 + 0.016) \cdot 0.5 = 0.118
$$
$$
\alpha_2(2) = \left(\alpha_1(1) \cdot A_{12} + \alpha_1(2) \cdot A_{22}\right) \cdot B_{2,2} = (0.3 \cdot 0.3 + 0.04 \cdot 0.6) \cdot 0.9 = (0.09 + 0.024) \cdot 0.9 = 0.1026
$$

4. Общая вероятность наблюдаемой последовательности:
   - Суммируем значения $\alpha_2(i)$:

$$
P(O \mid \lambda) = \alpha_2(1) + \alpha_2(2) = 0.118 + 0.1026 = 0.2206
$$

Таким образом, вероятность наблюдаемой последовательности $P(O \mid \lambda)$ равна 0.2206.



### Задача 2: Декодирование наиболее вероятной последовательности состояний

Условия:

Используем ту же модель HMM, что и в предыдущей задаче. Дана наблюдаемая последовательность:

- $O = (O_1, O_2) = (1, 2)$

Задача: Найти наиболее вероятную последовательность скрытых состояний $q$.

#### Решение:

1. Инициализация:
   - Вычисляем $\delta_1(i)$:
   
$$
\delta_1(1) = \pi_1 \cdot B_{1,1} = 0.6 \cdot 0.5 = 0.3
$$
$$
\delta_1(2) = \pi_2 \cdot B_{2,1} = 0.4 \cdot 0.1 = 0.04
$$

2. Рекурсия для $t = 2$:
   - Вычисляем $\delta_2(i)$ и следим за предыдущими состояниями:

$$
\delta_2(1) = \max(\delta_1(1) \cdot A_{11}, \delta_1(2) \cdot A_{21}) \cdot B_{1,2} = \max(0.3 \cdot 0.7, 0.04 \cdot 0.4) \cdot 0.5 = \max(0.21, 0.016) \cdot 0.5 = 0.105
$$
$$
\delta_2(2) = \max(\delta_1(1) \cdot A_{12}, \delta_1(2) \cdot A_{22}) \cdot B_{2,2} = \max(0.3 \cdot 0.3, 0.04 \cdot 0.6) \cdot 0.9 = \max(0.09, 0.024) \cdot 0.9 = 0.081
$$

3. Обратный проход для восстановления последовательности состояний:
   - Выбираем состояние с максимальным значением $\delta_2(i)$:
   - Наиболее вероятное состояние для $t=2$ - это состояние 1, так как $\delta_2(1) = 0.105 > \delta_2(2) = 0.081$.

4. Последовательность состояний:
   - Находим предшествующее состояние для $t=2$:
   - Для состояния 1:
   $$
   \psi_2(1) = \text{argmax}(\delta_1(1) \cdot A_{11}, \delta_1(2) \cdot A_{21}) = 1
   $$
   - Таким образом, $q_2 = 1$ и $q_1 = 1$.

Таким образом, наиболее вероятная последовательность скрытых состояний - $(S_1, S_1)$.



### Задача 3: Оценка параметров модели

Условия:

У нас есть данные о наблюдениях:

- Наблюдаемая последовательность: $O = (O_1, O_2, O_1)$

Задача: Оценить параметры HMM $A$ и $B$ на основе наблюдений, используя метод максимального правдоподобия.

#### Решение:

1. Инициализация:
   - Предположим, что у нас есть 2 состояния $S_1$ и $S_2$, и наблюдения $O_1, O_2$ соответствуют $k_1, k_2$.

2. Подсчет переходов:

#### Решение:

1. Шаг 1: Прямой проход (алгоритм Форвард):

   Сначала мы уже рассчитали вероятности $\alpha_t$ для наблюдений $O = (O_1, O_2, O_1)$.

2. Шаг 2: Обратный проход (алгоритм Бэкворд):

\beta_T(i) = 1 \quad \text{для всех } i
$$

$$

   - И так далее для других $t$.

3. Шаг 3: Вычисление \(\xi_t(i,j)\) и \(\gamma_t(i)\):

$$
\xi_t(i,j) = \frac{\alpha_t(i) A_{ij} B_{j,k} \beta_{t+1}(j)}{P(O \mid \lambda)}
$$
$$
\gamma_t(i) = \sum_{j} \xi_t(i,j)
$$

4. Шаг 4: Обновление параметров:

### Задача 5: Применение HMM для классификации текстов

Условия:

Рассмотрим задачу классификации текстов на две категории: положительная и отрицательная. Мы хотим построить модель HMM для этой задачи.

Задача: Построить HMM для классификации текстов и провести обучение на размеченных данных.

#### Решение:

1. Подготовка данных:
   - Разделим тексты на положительные и отрицательные.
   - Извлечем наблюдения (слова) и определим скрытые состояния (категории).

2. Обучение модели:
   - Сформируем набор наблюдаемых состояний $O$ и скрытых состояний $S$.
   - Подсчитаем частоты переходов и эмиссий, чтобы сформировать матрицы $A$ и $B$.

3. Тестирование модели:
   - На основе обученной модели применим алгоритм Витерби на новых текстах, чтобы предсказать их категорию.

4. Оценка качества:
   - Рассчитаем точность, полноту и F1-мера для оценки производительности модели.

Таким образом, мы можем использовать HMM для классификации текстов, позволяя учитывать скрытые зависимости между словами и категориями.



```python
import numpy as np
import matplotlib.pyplot as plt

# Установим фиксированные параметры модели HMM
states = ['S1', 'S2']
observations = ['O1', 'O2']
A = np.array([[0.7, 0.3],  # Матрица переходов
              [0.4, 0.6]])
B = np.array([[0.5, 0.5],  # Матрица эмиссии
              [0.1, 0.9]])
pi = np.array([0.6, 0.4])  # Начальное распределение

# Наблюдаемая последовательность O = (O1, O2) -> индексы [0, 1]
observed_sequence = [0, 1]  # O1, O2

# Задача 1: Вычисление вероятности наблюдаемой последовательности (алгоритм Форвард)
def forward_algorithm(observed_sequence, A, B, pi):
    n_states = A.shape[0]
    T = len(observed_sequence)
    alpha = np.zeros((T, n_states))

    # Инициализация
    for i in range(n_states):
        alpha[0, i] = pi[i] * B[i, observed_sequence[0]]

    # Рекурсия
    for t in range(1, T):
        for j in range(n_states):
            alpha[t, j] = sum(alpha[t-1, i] * A[i, j] for i in range(n_states)) * B[j, observed_sequence[t]]

    return alpha

# Вычисляем вероятность
alpha = forward_algorithm(observed_sequence, A, B, pi)
probability = sum(alpha[-1])
print(f"Вероятность наблюдаемой последовательности: {probability:.4f}")

# Визуализация вероятностей
plt.figure(figsize=(10, 5))
plt.plot(alpha, marker='o')
plt.title('Алгоритм Форвард: Вероятности состояний')
plt.xlabel('Время (t)')
plt.ylabel('Вероятности (alpha)')
plt.xticks(ticks=range(len(observed_sequence)), labels=[f'O{obs+1}' for obs in observed_sequence])
plt.legend(states)
plt.grid()
plt.show()

# Задача 2: Декодирование наиболее вероятной последовательности состояний (алгоритм Витерби)
def viterbi_algorithm(observed_sequence, A, B, pi):
    n_states = A.shape[0]
    T = len(observed_sequence)
    delta = np.zeros((T, n_states))
    psi = np.zeros((T, n_states), dtype=int)

    # Инициализация
    for i in range(n_states):
        delta[0, i] = pi[i] * B[i, observed_sequence[0]]

    # Рекурсия
    for t in range(1, T):
        for j in range(n_states):
            delta[t, j] = max(delta[t-1, i] * A[i, j] for i in range(n_states)) * B[j, observed_sequence[t]]
            psi[t, j] = np.argmax(delta[t-1] * A[:, j])

    # Обратный проход
    q = np.zeros(T, dtype=int)
    q[T-1] = np.argmax(delta[T-1])

    for t in range(T-2, -1, -1):
        q[t] = psi[t+1, q[t+1]]

    return q

# Находим наиболее вероятную последовательность
most_probable_sequence = viterbi_algorithm(observed_sequence, A, B, pi)
print(f"Наиболее вероятная последовательность состояний: {', '.join(states[i] for i in most_probable_sequence)}")

# Задача 4: Применение алгоритма Баума-Велча
def baum_welch_algorithm(observed_sequence, A, B, pi, n_iterations=100):
    n_states = A.shape[0]
    T = len(observed_sequence)

    for _ in range(n_iterations):
        alpha = forward_algorithm(observed_sequence, A, B, pi)
        beta = backward_algorithm(observed_sequence, A, B)

        # Обновление параметров
        xi = np.zeros((T-1, n_states, n_states))
        gamma = np.zeros((T, n_states))

        for t in range(T-1):
            denominator = np.sum(alpha[t, :, np.newaxis] * A * B[:, observed_sequence[t+1]] * beta[t+1, :])
            for i in range(n_states):
                for j in range(n_states):
                    xi[t, i, j] = (alpha[t, i] * A[i, j] * B[j, observed_sequence[t+1]] * beta[t+1, j]) / denominator

        for t in range(T):
            gamma[t] = np.sum(xi[t-1], axis=1) if t > 0 else alpha[t] / np.sum(alpha[t])

        # Обновляем A и B
        for i in range(n_states):
            for j in range(n_states):
                A[i, j] = np.sum(xi[:, i, j]) / np.sum(gamma[:-1, i])
            for k in range(len(observations)):
                B[i, k] = np.sum(gamma[np.array(observed_sequence) == k, i]) / np.sum(gamma[:, i])

    return A, B

# Функция обратного прохода для алгоритма Баума-Велча
def backward_algorithm(observed_sequence, A, B):
    n_states = A.shape[0]
    T = len(observed_sequence)
    beta = np.zeros((T, n_states))

    beta[T-1] = 1  # Базовый случай

    for t in range(T-2, -1, -1):
        for i in range(n_states):
            beta[t, i] = np.sum(A[i, :] * B[:, observed_sequence[t+1]] * beta[t+1, :])

    return beta

# Запускаем алгоритм Баума-Велча
A_final, B_final = baum_welch_algorithm(observed_sequence, A, B, pi)
print("Итоговая матрица переходов:\n", A_final)
print("Итоговая матрица эмиссий:\n", B_final)

# Задача 5: Применение HMM для классификации текстов
texts = ['good movie', 'bad movie', 'fantastic plot', 'terrible acting']
labels = ['positive', 'negative', 'positive', 'negative']

# Словарь для представления слов
word_dict = {word: i for i, word in enumerate(set(' '.join(texts).split()))}
text_sequences = [[word_dict[word] for word in text.split() if word in word_dict] for text in texts]

# Применение HMM для классификации
def classify_texts(text_sequences, A, B, pi):
    classifications = []
    for seq in text_sequences:
        # Проверка, чтобы последовательность содержала только допустимые индексы
        seq = [obs for obs in seq if obs < B.shape[1]]

        if not seq:  # Если последовательность пуста после фильтрации, пропускаем её
            classifications.append('unknown')
            continue

        state_sequence = viterbi_algorithm(seq, A, B, pi)
        classification = 'positive' if np.mean(state_sequence) > 0 else 'negative'
        classifications.append(classification)
    return classifications

# Классифицируем тексты
classifications = classify_texts(text_sequences, A_final, B_final, pi)
print(f"Классификации текстов: {classifications}")
```
