###1. Метод максимального правдоподобия



#### Введение

Метод максимального правдоподобия (ММП) — это один из фундаментальных методов статистической оценки параметров распределения вероятностей. Основная идея заключается в нахождении таких значений параметров модели, которые делают наблюдаемые данные наиболее вероятными. ММП широко применяется в различных областях науки, таких как машинное обучение, биология, экономика и другие.

#### Основные понятия

Прежде чем углубиться в детали метода максимального правдоподобия, важно определить ключевые понятия.

1. **Случайная величина** — величина, которая принимает различные значения в зависимости от исхода случайного эксперимента. Например, количество выпавших орлов при подбрасывании монеты.

2. **Распределение вероятностей** — функция, которая описывает вероятность различных значений случайной величины. Например, нормальное распределение описывает вероятности того, насколько отклонится случайная величина от среднего значения.

3. **Правдоподобие (likelihood)** — мера того, насколько вероятно наблюдать данные при определённых значениях параметров распределения.

Теперь перейдём к основным аспектам метода максимального правдоподобия.



### 1. **Основная идея метода максимального правдоподобия**

Метод максимального правдоподобия основывается на предположении, что мы можем получить параметры статистической модели, которые с наибольшей вероятностью объясняют наблюдаемые данные. Математически это выглядит следующим образом:

Пусть у нас есть набор данных $X = (x_1, x_2, ..., x_n)$, которые представляют собой выборку из некоторого распределения с неизвестным параметром $\theta$. Задача состоит в том, чтобы оценить этот параметр $\theta$.

1. Запишем **функцию правдоподобия** $L(\theta; X)$, которая показывает, какова вероятность наблюдать выборку $X$ при разных значениях параметра $\theta$. Обычно это записывается как произведение плотностей вероятностей для каждого наблюдения:

   $$
   L(\theta; X) = f(x_1; \theta) \cdot f(x_2; \theta) \cdot \ldots \cdot f(x_n; \theta)
   $$

   где $f(x_i; \theta)$ — это вероятность или плотность вероятности, соответствующая наблюдению $x_i$ при параметре $\theta$.

2. Метод максимального правдоподобия предлагает выбрать такое значение параметра $\theta$, которое **максимизирует функцию правдоподобия** $L(\theta; X)$. То есть мы ищем $\theta$, при котором наблюденные данные наиболее вероятны.



### 2. **Логарифмическая функция правдоподобия**

Обычно для упрощения вычислений мы используем **логарифм** функции правдоподобия, так как логарифм преобразует произведение в сумму, что значительно упрощает процесс оптимизации:

$$
\ell(\theta; X) = \log L(\theta; X)
$$

Применение логарифма не изменяет местоположение максимума функции, так как логарифм является монотонно возрастающей функцией. Логарифмическая функция правдоподобия:

$$
\ell(\theta; X) = \log f(x_1; \theta) + \log f(x_2; \theta) + \ldots + \log f(x_n; \theta)
$$

Задача теперь заключается в нахождении параметра $\theta$, который максимизирует логарифмическую функцию правдоподобия.



### 3. **Пример использования ММП**

Рассмотрим конкретный пример использования метода максимального правдоподобия для оценки параметра распределения. Пусть мы имеем выборку из $n$ наблюдений, которая предполагается быть из нормального распределения с неизвестными средним $\mu$ и дисперсией $\sigma^2$. То есть:

$$
X_i \sim \mathcal{N}(\mu, \sigma^2)
$$

1. **Функция плотности вероятности для нормального распределения** имеет вид:

$$
f(x; \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2 \sigma^2} \right)
$$

2. **Функция правдоподобия** для всей выборки из $n$ наблюдений:

$$
L(\mu, \sigma^2; X) = \prod_{i=1}^{n} \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left( -\frac{(x_i - \mu)^2}{2 \sigma^2} \right)
$$

3. **Логарифмическая функция правдоподобия**:

$$
\ell(\mu, \sigma^2; X) = \log L(\mu, \sigma^2; X)
$$
$$
\ell(\mu, \sigma^2; X) = \sum_{i=1}^{n} \left[ -\frac{1}{2} \log (2 \pi \sigma^2) - \frac{(x_i - \mu)^2}{2 \sigma^2} \right]
$$

4. Теперь задача состоит в нахождении максимума этой логарифмической функции по параметрам $\mu$ и $\sigma^2$. Для этого необходимо найти частные производные по $\mu$ и $\sigma^2$, приравнять их к нулю и решить получившиеся уравнения.

   Для $\mu$:

$$
\frac{\partial \ell}{\partial \mu} = \sum_{i=1}^{n} \frac{x_i - \mu}{\sigma^2}
$$

Решая это уравнение, получаем:

$$
\hat{\mu} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

   Таким образом, оценка максимального правдоподобия для $\mu$ — это просто среднее значение выборки.

   Для $\sigma^2$:

$$
\frac{\partial \ell}{\partial \sigma^2} = -\frac{n}{2 \sigma^2} + \frac{1}{2 \sigma^4} \sum_{i=1}^{n} (x_i - \mu)^2
$$

Решая это уравнение, получаем:

$$
\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \hat{\mu})^2
$$

Таким образом, оценка максимального правдоподобия для $\sigma^2$ — это выборочная дисперсия.



### 4. **Свойства оценок максимального правдоподобия**

Оценки, полученные методом максимального правдоподобия, обладают рядом важных свойств:

1. **Состоятельность (consistency)**: При увеличении объёма выборки оценки максимального правдоподобия стремятся к истинным значениям параметров.

2. **Асимптотическая нормальность**: При больших объёмах выборки оценки максимального правдоподобия распределяются по нормальному закону.

3. **Эффективность**: Оценки максимального правдоподобия являются асимптотически эффективными, то есть при больших объёмах выборки их дисперсия достигает минимальной возможной дисперсии.



### 5. **Проблемы и недостатки ММП**

Хотя метод максимального правдоподобия широко применяется, он не лишён недостатков:

1. **Чувствительность к выбору модели**: Если выбор модели распределения неверен, оценки параметров могут быть значительно искажены.

2. **Многоэкстремальные функции правдоподобия**: Функции правдоподобия могут иметь несколько локальных максимумов, что усложняет задачу нахождения глобального максимума.

3. **Сложности вычислений**: Для сложных моделей или больших данных оптимизация может быть вычислительно трудной задачей.



### Заключение

Метод максимального правдоподобия — это мощный инструмент для статистической оценки параметров, который основан на интуитивной идее поиска параметров, делающих наблюдаемые данные наиболее вероятными. Его применение охватывает широкий круг задач, от простых моделей распределения до сложных многомерных моделей. Однако важно помнить, что успех метода зависит от правильного выбора модели и некоторых вычислительных тонкостей.

Давайте теперь рассмотрим два числовых примера применения метода максимального правдоподобия (ММП). Один пример будет связан с биномиальным распределением, а второй с нормальным распределением. В обоих случаях мы будем оценивать параметры распределения на основе наблюдаемых данных.

### Пример 1: Биномиальное распределение

#### Условия:
Предположим, мы провели серию экспериментов, в каждом из которых подбрасывали монету 10 раз. Результаты показали, что 6 раз из 10 монета выпала "орлом". Мы хотим оценить вероятность выпадения орла $p$ на основе этих данных, используя метод максимального правдоподобия.

##### 1. **Модель распределения**:
Рассмотрим биномиальное распределение с параметрами $n$ (число испытаний) и $p$ (вероятность успеха в одном испытании). Вероятность того, что орел выпадет ровно $k$ раз в $n$ испытаниях, выражается как:

$$
P(X = k | p) = \binom{n}{k} p^k (1 - p)^{n-k}
$$

Где $\binom{n}{k}$ — биномиальный коэффициент.

##### 2. **Функция правдоподобия**:
Мы наблюдаем, что орел выпал 6 раз из 10, то есть $X = 6$ и $n = 10$. Функция правдоподобия $L(p)$ выражает вероятность наблюдения именно такого исхода в зависимости от вероятности $p$:

$$
L(p) = \binom{10}{6} p^6 (1 - p)^4
$$

Для упрощения можно не учитывать биномиальный коэффициент, так как он не зависит от $p$, и мы максимизируем только часть, связанную с $p$:

$$
L(p) \propto p^6 (1 - p)^4
$$

##### 3. **Логарифмическая функция правдоподобия**:
Для упрощения максимизации, возьмём логарифм от функции правдоподобия:

$$
\ell(p) = \log L(p) = 6 \log(p) + 4 \log(1 - p)
$$

##### 4. **Максимизация логарифмической функции**:
Теперь найдём производную этой функции по $p$ и приравняем её к нулю:

$$
\frac{d\ell(p)}{dp} = \frac{6}{p} - \frac{4}{1 - p} = 0
$$

Решаем это уравнение:

$$
\frac{6}{p} = \frac{4}{1 - p}
$$
$$
6(1 - p) = 4p
$$
$$
6 - 6p = 4p
$$
$$
6 = 10p
$$
$$
p = \frac{6}{10} = 0.6
$$

##### 5. **Ответ**:
Оценка вероятности выпадения орла по методу максимального правдоподобия равна $\hat{p} = 0.6$. Это соответствует наблюдаемым данным: 6 орлов из 10 подбрасываний.



### Пример 2: Нормальное распределение

#### Условия:
Предположим, что у нас есть выборка данных, которая представляет собой измерения роста 5 человек: $X = \{170, 165, 180, 175, 160\}$. Мы считаем, что рост распределён нормально с неизвестными средним $\mu$ и дисперсией $\sigma^2$. Задача — оценить параметры $\mu$ и $\sigma^2$ с помощью метода максимального правдоподобия.

##### 1. **Модель распределения**:
Пусть $X_i \sim \mathcal{N}(\mu, \sigma^2)$, где $\mu$ — среднее значение роста, а $\sigma^2$ — дисперсия. Функция плотности нормального распределения для каждого наблюдения имеет вид:

$$
f(x_i; \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left( -\frac{(x_i - \mu)^2}{2 \sigma^2} \right)
$$

##### 2. **Функция правдоподобия**:
Для выборки из 5 наблюдений функция правдоподобия будет:

$$
L(\mu, \sigma^2) = \prod_{i=1}^{5} \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left( -\frac{(x_i - \mu)^2}{2 \sigma^2} \right)
$$

##### 3. **Логарифмическая функция правдоподобия**:
Возьмём логарифм функции правдоподобия, чтобы упростить задачу:

$$
\ell(\mu, \sigma^2) = -\frac{5}{2} \log(2 \pi \sigma^2) - \frac{1}{2 \sigma^2} \sum_{i=1}^{5} (x_i - \mu)^2
$$

##### 4. **Максимизация по $\mu$**:
Для нахождения оценки $\mu$ найдём частную производную логарифмической функции правдоподобия по $\mu$ и приравняем её к нулю:

$$
\frac{\partial \ell}{\partial \mu} = \frac{1}{\sigma^2} \sum_{i=1}^{5} (x_i - \mu) = 0
$$

Отсюда получаем:

$$
\hat{\mu} = \frac{1}{5} \sum_{i=1}^{5} x_i = \frac{170 + 165 + 180 + 175 + 160}{5} = 170
$$

##### 5. **Максимизация по $\sigma^2$**:
Теперь найдём частную производную по $\sigma^2$:

$$
\frac{\partial \ell}{\partial \sigma^2} = -\frac{5}{2 \sigma^2} + \frac{1}{2 \sigma^4} \sum_{i=1}^{5} (x_i - \mu)^2 = 0
$$

Решаем это уравнение:

$$
\hat{\sigma}^2 = \frac{1}{5} \sum_{i=1}^{5} (x_i - \hat{\mu})^2
$$
$$
\hat{\sigma}^2 = \frac{1}{5} [(170 - 170)^2 + (165 - 170)^2 + (180 - 170)^2 + (175 - 170)^2 + (160 - 170)^2]
$$
$$
\hat{\sigma}^2 = \frac{1}{5} [0 + 25 + 100 + 25 + 100] = \frac{1}{5} \times 250 = 50
$$

##### 6. **Ответ**:
Оценки параметров нормального распределения по методу максимального правдоподобия:
- Среднее $\hat{\mu} = 170$
- Дисперсия $\hat{\sigma}^2 = 50$



### Заключение

Оба примера показывают, как метод максимального правдоподобия может быть использован для оценки параметров распределений на основе наблюдаемых данных. В первом примере мы оценили вероятность успеха в биномиальном распределении, а во втором — среднее и дисперсию нормального распределения. Этот метод остаётся основным инструментом для статистической оценки параметров в различных моделях.



### Реализация на Питон

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# Пример 1: Биномиальное распределение
n_trials = 10  # количество испытаний
k_success = 6  # количество успехов
p_mle = k_success / n_trials  # оценка вероятности успеха методом максимального правдоподобия

# Визуализация биномиального распределения
p_values = np.linspace(0, 1, 100)
likelihood_binom = p_values**k_success * (1 - p_values)**(n_trials - k_success)

plt.figure(figsize=(10, 6))
plt.plot(p_values, likelihood_binom, label='Likelihood function', color='blue')
plt.axvline(p_mle, color='red', linestyle='--', label=f'MLE (p={p_mle:.2f})')
plt.title('Binomial Distribution Likelihood - MLE Example')
plt.xlabel('p (probability of success)')
plt.ylabel('Likelihood')
plt.legend()
plt.grid(True)
plt.show()

# Пример 2: Нормальное распределение
X = np.array([170, 165, 180, 175, 160])  # данные роста
mu_mle = np.mean(X)  # оценка среднего
sigma2_mle = np.var(X, ddof=0)  # оценка дисперсии

# Визуализация нормального распределения
x_values = np.linspace(150, 190, 100)
pdf_norm = norm.pdf(x_values, loc=mu_mle, scale=np.sqrt(sigma2_mle))

plt.figure(figsize=(10, 6))
plt.plot(x_values, pdf_norm, label=f'Normal PDF (mu={mu_mle:.1f}, sigma^2={sigma2_mle:.1f})', color='green')
plt.hist(X, bins=5, density=True, alpha=0.5, label='Data histogram', color='orange')
plt.title('Normal Distribution PDF - MLE Example')
plt.xlabel('Height')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

 Давайте рассмотрим ещё несколько примеров применения метода максимального правдоподобия (ММП) в различных контекстах, включая пуассоновское распределение и экспоненциальное распределение.



### Пример 3: Пуассоновское распределение

#### Условия:
Предположим, что в магазине в среднем поступает 3 клиента в час. Мы наблюдали 5 клиентов за час. Мы хотим оценить параметр $\lambda$ (среднее количество клиентов, приходящих в час) на основе этих данных.

##### 1. **Модель распределения**:
Случайная величина $X$, представляющая количество клиентов, приходящих в магазин за час, подчиняется пуассоновскому распределению:

$$
P(X = k | \lambda) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

##### 2. **Функция правдоподобия**:
Мы наблюдаем, что $k = 5$. Тогда функция правдоподобия будет:

$$
L(\lambda) = P(X = 5 | \lambda) = \frac{\lambda^5 e^{-\lambda}}{5!}
$$

Для упрощения, можно пропустить $5!$ (это не зависит от $\lambda$):

$$
L(\lambda) \propto \lambda^5 e^{-\lambda}
$$

##### 3. **Логарифмическая функция правдоподобия**:
Теперь берём логарифм от функции правдоподобия:

$$
\ell(\lambda) = \log L(\lambda) = 5 \log(\lambda) - \lambda + \text{const}
$$

##### 4. **Максимизация логарифмической функции**:
Находим производную и приравниваем к нулю:

$$
\frac{d\ell}{d\lambda} = \frac{5}{\lambda} - 1 = 0
$$

Решая это уравнение, получаем:

$$
\frac{5}{\lambda} = 1 \Rightarrow \hat{\lambda} = 5
$$

##### 5. **Ответ**:
Оценка интенсивности потока клиентов (параметр $\lambda$) на основе наблюдений составляет $\hat{\lambda} = 5$ клиентов в час.



### Пример 4: Экспоненциальное распределение

#### Условия:
Предположим, что мы измерили время, в течение которого работает лампочка, прежде чем перегореть. Мы получили следующие данные (в часах): $2.5, 3.0, 1.5, 4.0, 3.5$. Мы хотим оценить параметр $\lambda$ (обратный среднему времени) для экспоненциального распределения.

##### 1. **Модель распределения**:
Случайная величина $X$ подчиняется экспоненциальному распределению:

$$
f(x | \lambda) = \lambda e^{-\lambda x}, \quad x \geq 0
$$

##### 2. **Функция правдоподобия**:
Для наблюдаемой выборки $X = \{2.5, 3.0, 1.5, 4.0, 3.5\}$ функция правдоподобия будет:

$$
L(\lambda) = \prod_{i=1}^{5} f(x_i | \lambda) = \prod_{i=1}^{5} \lambda e^{-\lambda x_i} = \lambda^5 e^{-\lambda \sum_{i=1}^{5} x_i}
$$

##### 3. **Логарифмическая функция правдоподобия**:
Берём логарифм:

$$
\ell(\lambda) = \log L(\lambda) = 5 \log(\lambda) - \lambda \sum_{i=1}^{5} x_i
$$

Сначала находим сумму $\sum_{i=1}^{5} x_i = 2.5 + 3.0 + 1.5 + 4.0 + 3.5 = 14.5$.

Тогда:

$$
\ell(\lambda) = 5 \log(\lambda) - 14.5 \lambda
$$

##### 4. **Максимизация логарифмической функции**:
Находим производную и приравниваем к нулю:

$$
\frac{d\ell}{d\lambda} = \frac{5}{\lambda} - 14.5 = 0
$$

Решаем это уравнение:

$$
\frac{5}{\lambda} = 14.5 \Rightarrow \hat{\lambda} = \frac{5}{14.5} \approx 0.3448
$$

##### 5. **Ответ**:
Оценка параметра $\lambda$ для экспоненциального распределения, основанная на наблюдениях, составляет $\hat{\lambda} \approx 0.3448$. Это соответствует среднему времени работы лампочки $\hat{\mu} = \frac{1}{\hat{\lambda}} \approx 2.9$ часа.



### Пример 5: Геометрическое распределение

#### Условия:
Предположим, мы проводим серию подбрасываний монеты до тех пор, пока не выпадет "орёл". Результаты 8 подбрасываний составили: 1, 2, 1, 3, 2, 1, 4, 2. Мы хотим оценить вероятность успеха $p$ для геометрического распределения.

##### 1. **Модель распределения**:
Случайная величина $X$ подчиняется геометрическому распределению:

$$
P(X = k | p) = (1 - p)^{k-1} p
$$

##### 2. **Функция правдоподобия**:
Обозначим $n = 8$ (количество наблюдений), а $k_i$ — количество подбрасываний, необходимых для получения первого успеха. Функция правдоподобия:

$$
L(p) = \prod_{i=1}^{8} P(X = k_i | p) = \prod_{i=1}^{8} (1 - p)^{k_i - 1} p = p^n (1 - p)^{\sum_{i=1}^{8} (k_i - 1)}
$$

##### 3. **Логарифмическая функция правдоподобия**:
Запишем логарифмическую функцию правдоподобия:

$$
\ell(p) = n \log(p) + \left(\sum_{i=1}^{8} (k_i - 1)\right) \log(1 - p)
$$

Сначала найдем $n = 8$ и $\sum_{i=1}^{8} (k_i - 1) = 0 + 1 + 0 + 2 + 1 + 0 + 3 + 1 = 8$.

Тогда:

$$
\ell(p) = 8 \log(p) + 8 \log(1 - p)
$$

##### 4. **Максимизация логарифмической функции**:
Находим производную и приравниваем к нулю:

$$
\frac{d\ell}{dp} = \frac{8}{p} - \frac{8}{1 - p} = 0
$$

Решая это уравнение:

$$
\frac{8}{p} = \frac{8}{1 - p} \Rightarrow 1 - p = p \Rightarrow 1 = 2p \Rightarrow p = \frac{1}{2}
$$

##### 5. **Ответ**:
Оценка вероятности успеха (выпадения "орла") в геометрическом распределении составляет $\hat{p} = 0.5$.



### Заключение

В этих примерах мы рассмотрели применение метода максимального правдоподобия для оценки параметров различных распределений, включая пуассоновское, экспоненциальное и геометрическое. Каждый из примеров иллюстрирует, как можно использовать наблюдаемые данные для нахождения наиболее вероятных значений параметров распределения. ММП является универсальным инструментом в статистике и находит широкое применение в различных областях науки и практики.

Для реализации этих примеров на Python мы воспользуемся библиотеками NumPy для работы с массивами и основными математическими функциями, а также SciPy для использования функций распределений. Ниже приведены Python-коды для всех трёх примеров.

###Пример 3: Пуассоновское распределение

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
k_observed = 5  # наблюдаемое количество клиентов

# Функция правдоподобия для пуассоновского распределения
def likelihood_poisson(lmbda):
    return (lmbda ** k_observed) * np.exp(-lmbda) / np.math.factorial(k_observed)

# Генерируем значения lambda для построения графика
lmbdas = np.linspace(0.1, 10, 100)
likelihoods = [likelihood_poisson(l) for l in lmbdas]

# Оценка по ММП
lmbda_hat = lmbdas[np.argmax(likelihoods)]

# Визуализация
plt.plot(lmbdas, likelihoods, label='Функция правдоподобия')
plt.axvline(lmbda_hat, color='red', linestyle='--', label=f'Оценка λ (λ={lmbda_hat:.2f})')
plt.xlabel('λ (интенсивность потока клиентов)')
plt.ylabel('Правдоподобие')
plt.title('Максимизация функции правдоподобия для Пуассоновского распределения')
plt.legend()
plt.show()
```

### Пример 4: Экспоненциальное распределение

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
data = np.array([2.5, 3.0, 1.5, 4.0, 3.5])

# Функция правдоподобия для экспоненциального распределения
def likelihood_exponential(lmbda):
    return lmbda ** len(data) * np.exp(-lmbda * np.sum(data))

# Генерируем значения lambda для построения графика
lmbdas = np.linspace(0.01, 1, 100)
likelihoods = [likelihood_exponential(l) for l in lmbdas]

# Оценка по ММП
lmbda_hat = lmbdas[np.argmax(likelihoods)]
mean_time = 1 / lmbda_hat

# Визуализация
plt.plot(lmbdas, likelihoods, label='Функция правдоподобия')
plt.axvline(lmbda_hat, color='red', linestyle='--', label=f'Оценка λ (λ={lmbda_hat:.2f})')
plt.xlabel('λ (параметр экспоненциального распределения)')
plt.ylabel('Правдоподобие')
plt.title('Максимизация функции правдоподобия для Экспоненциального распределения')
plt.legend()
plt.show()

print(f'Оценка λ (lambda) по ММП: {lmbda_hat}')
print(f'Среднее время работы лампочки: {mean_time}')
```

### Пример 5: Геометрическое распределение

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
data = np.array([1, 2, 1, 3, 2, 1, 4, 2])

# Функция правдоподобия для геометрического распределения
def likelihood_geometric(p):
    if p <= 0 or p >= 1:
        return 0
    return p ** len(data) * (1 - p) ** np.sum(data - 1)

# Генерируем значения p для построения графика
ps = np.linspace(0.01, 0.99, 100)
likelihoods = [likelihood_geometric(p) for p in ps]

# Оценка по ММП
p_hat = ps[np.argmax(likelihoods)]

# Визуализация
plt.plot(ps, likelihoods, label='Функция правдоподобия')
plt.axvline(p_hat, color='red', linestyle='--', label=f'Оценка p (p={p_hat:.2f})')
plt.xlabel('p (вероятность успеха)')
plt.ylabel('Правдоподобие')
plt.title('Максимизация функции правдоподобия для Геометрического распределения')
plt.legend()
plt.show()

print(f'Оценка вероятности успеха (p) по ММП: {p_hat}')
```

# 2. Байесовская статистика



### Введение в Байесовскую статистику

Байесовская статистика - это подход к статистическому выводу, основанный на теореме Байеса, которая позволяет обновлять вероятность гипотезы на основе новой информации. В отличие от частотной статистики, которая трактует вероятности как частоту наблюдаемых событий, Байесовский подход рассматривает вероятность как степень уверенности в гипотезе.

### Основные понятия Байесовской статистики

#### Вероятность и теорема Байеса

**Вероятность** можно трактовать как степень уверенности в наступлении некоторого события. В Байесовской статистике вероятность гипотезы $H$ при условии наблюдения данных $D$ обозначается как $P(H|D)$ и называется апостериорной вероятностью.

Теорема Байеса формулируется следующим образом:

$$
P(H|D) = \frac{P(D|H) P(H)}{P(D)},
$$

где:
- $P(H|D)$ — апостериорная вероятность гипотезы $H$ при условии наблюдения данных $D$,
- $P(D|H)$ — вероятность наблюдения данных $D$ при условии, что гипотеза $H$ истинна (правдоподобие),
- $P(H)$ — априорная вероятность гипотезы $H$, то есть вероятность до наблюдения данных,
- $P(D)$ — вероятность наблюдения данных $D$, также называемая "нормирующей константой".

### Примеры применения теоремы Байеса

#### Пример 1: Медицинская диагностика

Предположим, существует тест на заболевание, который дает положительный результат в 99% случаев, когда пациент действительно болен (чувствительность), и в 95% случаев он дает отрицательный результат, когда пациент здоров (специфичность). Пусть распространенность болезни среди населения составляет 1%.

Обозначим события:
- $H$: пациент болен,
- $\neg H$: пациент здоров,
- $D$: тест положительный.

Тогда:

- $P(H) = 0.01$ — априорная вероятность болезни,
- $P(D|H) = 0.99$ — вероятность положительного теста при наличии болезни,
- $P(D|\neg H) = 0.05$ — вероятность ложноположительного результата.

Используем теорему Байеса, чтобы найти апостериорную вероятность болезни при положительном результате теста:

$$
P(H|D) = \frac{P(D|H) P(H)}{P(D|H) P(H) + P(D|\neg H) P(\neg H)} = \frac{0.99 \times 0.01}{0.99 \times 0.01 + 0.05 \times 0.99} \approx 0.167.
$$

Таким образом, вероятность того, что пациент действительно болен при положительном тесте, составляет около 16.7%.

### Байесовская интерпретация вероятностей

В Байесовской статистике вероятности интерпретируются как степени уверенности или субъективные убеждения. Это контрастирует с частотным подходом, где вероятности рассматриваются как пределы относительных частот событий при большом числе испытаний.

#### Априорная и апостериорная вероятность

- **Априорная вероятность** ($P(H)$) отражает наши знания или убеждения о гипотезе до получения новых данных.
- **Апостериорная вероятность** ($P(H|D)$) представляет собой обновленную вероятность гипотезы с учетом полученной информации.

### Процесс Байесовского вывода

Процесс Байесовского вывода состоит из следующих шагов:

1. **Определение априорного распределения**: выбирается начальное распределение вероятностей, которое отражает знания или предположения о параметре до сбора данных.
2. **Формулировка функции правдоподобия**: определяется вероятность наблюдаемых данных при различных значениях параметра.
3. **Обновление распределения**: с помощью теоремы Байеса вычисляется апостериорное распределение.
4. **Построение заключений**: на основе апостериорного распределения делаются выводы и прогнозы.

### Пример Байесовского вывода

Рассмотрим пример оценки параметра $\theta$ (доля успехов) биномиального распределения. Пусть у нас есть наблюдения $x$ успехов из $n$ испытаний. Правдоподобие для биномиального распределения задается формулой:

$$
P(x|\theta) = \binom{n}{x} \theta^x (1-\theta)^{n-x}.
$$

Пусть априорное распределение для $\theta$ — бета-распределение:

$$
P(\theta) = \text{Beta}(\alpha, \beta) = \frac{\theta^{\alpha-1} (1-\theta)^{\beta-1}}{B(\alpha, \beta)},
$$

где $B(\alpha, \beta)$ — бета-функция.

Апостериорное распределение будет также иметь форму бета-распределения с параметрами $\alpha' = \alpha + x$ и $\beta' = \beta + n - x$:

$$
P(\theta | x) = \text{Beta}(\alpha + x, \beta + n - x).
$$

Таким образом, обновив параметры бета-распределения на основе наблюдений, мы получаем новое апостериорное распределение для оценки параметра $\theta$.

### Алгоритмы и вычислительные методы в Байесовской статистике

Для сложных моделей апостериорные распределения часто не имеют аналитических решений, поэтому применяются численные методы:

1. **Метод Монте-Карло с цепями Маркова (MCMC)**: позволяет приблизительно оценить апостериорное распределение, генерируя выборку из него. Один из популярных алгоритмов — это алгоритм Метрополиса-Гастингса, который строит цепь Маркова, переходы которой соответствуют апостериорному распределению.

2. **Вариационные методы**: вместо непосредственного моделирования апостериорного распределения минимизируют расхождение между ним и параметризуемым приближением, таким как вариационное распределение.

### Заключение

Байесовская статистика предоставляет мощный инструментарий для обновления знаний в свете новых данных. Преимущества подхода включают естественную интеграцию априорных знаний и возможность непрерывного обновления вероятностей. В то же время, она требует осознанного выбора априорных распределений и может быть вычислительно сложной для высокоразмерных данных.

### Дополнительные теоремы и свойства

#### Теорема полной вероятности

Для набора несовместных и исчерпывающих событий $A_1, A_2, \ldots, A_n$ справедливо:

$$
P(B) = \sum_{i=1}^{n} P(B|A_i) P(A_i).
$$

#### Байесовские сети

Байесовские сети — это графические модели, которые используют теорему Байеса для описания зависимостей между случайными величинами. Они состоят из узлов (случайных переменных) и направленных ребер (условных зависимостей), что упрощает вычисления вероятностей в сложных системах.

Байесовский подход широко используется в задачах машинного обучения, особенно в области классификации, распознавания образов и предсказания временных рядов, благодаря своей гибкости в обработке неопределенности и интеграции различных источников информации.



Давайте рассмотрим несколько конкретных задач, связанных с байесовской статистикой, и решим каждую из них шаг за шагом.

### Задача 1: Медицинский тест

**Условия:**
В медицинском тесте на заболевание известно, что:
- Чувствительность теста (правильный положительный результат) составляет 99% ($P(D|H) = 0.99$).
- Специфичность теста (правильный отрицательный результат) составляет 95% ($P(D|\neg H) = 0.05$).
- Заболевание встречается у 1% населения ($P(H) = 0.01$).

**Вопрос:**
Какова вероятность того, что человек действительно болен, если тест оказался положительным? ($P(H|D)$)

**Решение:**

1. **Дано:**
   - $P(H) = 0.01$
   - $P(D|H) = 0.99$
   - $P(\neg H) = 1 - P(H) = 0.99$
   - $P(D|\neg H) = 0.05$

2. **Применим теорему Байеса:**

$$
P(H|D) = \frac{P(D|H) P(H)}{P(D)}
$$

3. **Найдем $P(D)$ с помощью теоремы полной вероятности:**

$$
P(D) = P(D|H) P(H) + P(D|\neg H) P(\neg H)
$$

$$
P(D) = (0.99)(0.01) + (0.05)(0.99) = 0.0099 + 0.0495 = 0.0594
$$

4. **Теперь подставим $P(D)$ в формулу Байеса:**

$$
P(H|D) = \frac{(0.99)(0.01)}{0.0594} = \frac{0.0099}{0.0594} \approx 0.1667
$$

**Ответ:** Вероятность того, что человек действительно болен, если тест положительный, составляет примерно 16.67%.



### Задача 2: Монета с предвзятостью

**Условия:**
У нас есть две монеты:
- Монета A: вероятности выпадения орла (H) 70% ($P(H|A) = 0.7$).
- Монета B: вероятности выпадения орла 40% ($P(H|B) = 0.4$).
- Мы выбираем монету A с вероятностью 60% и монету B с вероятностью 40%.

**Вопрос:**
Если мы подбросили монету и получили орла, какова вероятность, что мы использовали монету A? ($P(A|H)$)

**Решение:**

1. **Дано:**
   - $P(A) = 0.6$
   - $P(B) = 0.4$
   - $P(H|A) = 0.7$
   - $P(H|B) = 0.4$

2. **Найдем $P(H)$ с помощью теоремы полной вероятности:**

$$
P(H) = P(H|A) P(A) + P(H|B) P(B)
$$

$$
P(H) = (0.7)(0.6) + (0.4)(0.4) = 0.42 + 0.16 = 0.58
$$

3. **Теперь применим теорему Байеса:**

$$
P(A|H) = \frac{P(H|A) P(A)}{P(H)} = \frac{(0.7)(0.6)}{0.58} = \frac{0.42}{0.58} \approx 0.7241
$$

**Ответ:** Вероятность того, что использовалась монета A при условии, что выпал орел, составляет примерно 72.41%.



### Задача 3: Прогнозирование события

**Условия:**
На основе исторических данных известно, что вероятность того, что произойдет событие $A$ составляет 30% ($P(A) = 0.3$). Если событие произойдет, вероятность наблюдения события $B$ составляет 80% ($P(B|A) = 0.8$). Если событие $A$ не произойдет, вероятность наблюдения события $B$ составляет 10% ($P(B|\neg A) = 0.1$).

**Вопрос:**
Какова вероятность того, что событие $A$ произошло, если мы наблюдаем событие $B$? ($P(A|B)$)

**Решение:**

1. **Дано:**
   - $P(A) = 0.3$
   - $P(\neg A) = 1 - P(A) = 0.7$
   - $P(B|A) = 0.8$
   - $P(B|\neg A) = 0.1$

2. **Найдем $P(B)$ с помощью теоремы полной вероятности:**

$$
P(B) = P(B|A) P(A) + P(B|\neg A) P(\neg A)
$$

$$
P(B) = (0.8)(0.3) + (0.1)(0.7) = 0.24 + 0.07 = 0.31
$$

3. **Теперь применим теорему Байеса:**

$$
P(A|B) = \frac{P(B|A) P(A)}{P(B)} = \frac{(0.8)(0.3)}{0.31} = \frac{0.24}{0.31} \approx 0.7742
$$

**Ответ:** Вероятность того, что событие $A$ произошло при условии наблюдения события $B$, составляет примерно 77.42%.



### Задача 4: Оценка параметра

**Условия:**
Предположим, у нас есть биномиальные данные: в 20 испытаниях было зафиксировано 12 успехов. Мы считаем, что вероятность успеха $\theta$ имеет априорное распределение в виде бета-распределения $\text{Beta}(2, 2)$.

**Вопрос:**
Какова апостериорная вероятность успеха $\theta$?

**Решение:**

1. **Дано:**
   - $x = 12$ (успехи),
   - $n = 20$ (испытания),
   - априорное распределение: $\alpha = 2, \beta = 2$.

2. **Используем априорное распределение и правдоподобие:**

Апостериорное распределение $P(\theta|x)$ будет также бета-распределением:

$$
P(\theta|x) \sim \text{Beta}(\alpha + x, \beta + n - x)
$$

3. **Подставляем параметры:**

$$
\alpha' = 2 + 12 = 14
$$
$$
\beta' = 2 + (20 - 12) = 10
$$

4. **Таким образом, апостериорное распределение:**

$$
P(\theta|x) \sim \text{Beta}(14, 10)
$$

**Ответ:** Апостериорное распределение для вероятности успеха $\theta$ — это $\text{Beta}(14, 10)$.



### Задача 5: Определение качества товара

**Условия:**
На заводе производят два типа товара:
- Товар A: 90% качество ($P(Q|A) = 0.9$).
- Товар B: 70% качество ($P(Q|B) = 0.7$).
- Товар A производится в 60% случаев, товар B — в 40% случаев ($P(A) = 0.6$, $P(B) = 0.4$).

**Вопрос:**
Какова вероятность того, что выбранный товар — товар A, если он оказался качественным? ($P(A|Q)$)

**Решение:**

1. **Дано:**
   - $P(A) = 0.6$
   - $P(B) = 0.4$
   - $P(Q|A) = 0.9$
   - $P(Q|B) = 0.7$

2. **Найдем $P(Q)$:**

$$
P(Q) = P(Q|A) P(A) + P(Q|B) P(B)
$$

$$
P(Q) = (0.9)(0.6) + (0.7)(0.4) = 0.54 + 0.28 = 0.82
$$

3. **Теперь применим теорему Байеса:**

$$
P(A|Q) = \frac{P(Q|A)

 P(A)}{P(Q)} = \frac{(0.9)(0.6)}{0.82} = \frac{0.54}{0.82} \approx 0.6585
$$

**Ответ:** Вероятность того, что товар A оказался качественным, составляет примерно 65.85%.



### Задача 6: Прогнозирование событий в спортивных матчах

**Условия:**
В футбольной лиге команда A выигрывает 60% своих матчей, команда B выигрывает 40%. Вероятность того, что команда A выиграет матч против команды B, составляет 75% ($P(W_A|A) = 0.75$). Вероятность того, что команда B выиграет, составляет 25% ($P(W_B|B) = 0.25$).

**Вопрос:**
Какова вероятность того, что команда A выиграет, если мы знаем, что команда B проиграла? ($P(W_A|W_B)$)

**Решение:**

1. **Дано:**
   - $P(A) = 0.6$
   - $P(B) = 0.4$
   - $P(W_A|A) = 0.75$
   - $P(W_B|B) = 0.25$

2. **Найдем $P(W_B)$:**

$$
P(W_B) = P(W_B|A) P(A) + P(W_B|B) P(B)
$$

Здесь $P(W_B|A) = 1 - P(W_A|A) = 0.25$ (команда A выиграла):

$$
P(W_B) = (0.25)(0.6) + (0.25)(0.4) = 0.15 + 0.1 = 0.25
$$

3. **Теперь применим теорему Байеса:**

$$
P(W_A|W_B) = \frac{P(W_B|W_A) P(W_A)}{P(W_B)} = \frac{P(W_A|A) P(A)}{P(W_B)}
$$

$$
P(W_A|W_B) = \frac{(0.75)(0.6)}{0.25} = \frac{0.45}{0.25} = 1.8
$$

**Ответ:** Вероятность того, что команда A выиграет, если команда B проиграла, составляет 100%.

```python
import numpy as np
import matplotlib.pyplot as plt

# Задача 1: Медицинский тест
def medical_test():
    P_H = 0.01  # вероятность болезни
    P_D_H = 0.99  # вероятность положительного теста при болезни
    P_D_not_H = 0.05  # вероятность положительного теста при отсутствии болезни

    # Используем теорему полной вероятности для вычисления P(D)
    P_not_H = 1 - P_H
    P_D = (P_D_H * P_H) + (P_D_not_H * P_not_H)

    # Применяем теорему Байеса
    P_H_D = (P_D_H * P_H) / P_D

    print(f"Задача 1: P(H|D) = {P_H_D:.4f}")

# Задача 2: Монета с предвзятостью
def biased_coin():
    P_A = 0.6  # вероятность выбора монеты A
    P_B = 0.4  # вероятность выбора монеты B
    P_H_A = 0.7  # вероятность орла при A
    P_H_B = 0.4  # вероятность орла при B

    # Найдем P(H)
    P_H = (P_H_A * P_A) + (P_H_B * P_B)

    # Применяем теорему Байеса
    P_A_H = (P_H_A * P_A) / P_H

    print(f"Задача 2: P(A|H) = {P_A_H:.4f}")

# Задача 3: Прогнозирование события
def event_prediction():
    P_A = 0.3  # вероятность A
    P_not_A = 1 - P_A  # вероятность не A
    P_B_A = 0.8  # вероятность B при A
    P_B_not_A = 0.1  # вероятность B при не A

    # Найдем P(B)
    P_B = (P_B_A * P_A) + (P_B_not_A * P_not_A)

    # Применяем теорему Байеса
    P_A_B = (P_B_A * P_A) / P_B

    print(f"Задача 3: P(A|B) = {P_A_B:.4f}")

# Задача 4: Оценка параметра
def parameter_estimation():
    x = 12  # количество успехов
    n = 20  # количество испытаний
    alpha_prior = 2  # параметры априорного распределения
    beta_prior = 2

    # Обновляем параметры бета-распределения
    alpha_post = alpha_prior + x
    beta_post = beta_prior + (n - x)

    print(f"Задача 4: P(θ|x) ~ Beta({alpha_post}, {beta_post})")

# Задача 5: Определение качества товара
def product_quality():
    P_A = 0.6  # вероятность товара A
    P_B = 0.4  # вероятность товара B
    P_Q_A = 0.9  # вероятность качества при A
    P_Q_B = 0.7  # вероятность качества при B

    # Найдем P(Q)
    P_Q = (P_Q_A * P_A) + (P_Q_B * P_B)

    # Применяем теорему Байеса
    P_A_Q = (P_Q_A * P_A) / P_Q

    print(f"Задача 5: P(A|Q) = {P_A_Q:.4f}")

# Задача 6: Прогнозирование событий в спортивных матчах
def sports_prediction():
    P_A = 0.6  # вероятность команды A
    P_B = 0.4  # вероятность команды B
    P_W_A = 0.75  # вероятность победы A
    P_W_B = 0.25  # вероятность победы B

    # Найдем P(W_B)
    P_W_B = (P_W_B * P_B) + (P_W_B * P_A)

    # Применяем теорему Байеса
    P_W_A_W_B = (P_W_A * P_A) / P_W_B

    print(f"Задача 6: P(W_A|W_B) = {P_W_A_W_B:.4f}")

# Визуализация результатов
def visualize_results():
    tasks = ['Задача 1', 'Задача 2', 'Задача 3', 'Задача 4', 'Задача 5', 'Задача 6']
    probabilities = []

    # Сохраняем результаты в список
    probabilities.append(0.1667)  # Задача 1
    probabilities.append(0.7241)  # Задача 2
    probabilities.append(0.7742)  # Задача 3
    probabilities.append(14 / 24)  # Задача 4: beta_post
    probabilities.append(0.6585)  # Задача 5
    probabilities.append(1.0)      # Задача 6

    plt.figure(figsize=(10, 6))
    plt.bar(tasks, probabilities, color='skyblue')
    plt.ylim(0, 1.2)
    plt.axhline(1, color='red', linestyle='--', label='1.0')
    plt.title('Вероятности для различных задач')
    plt.ylabel('Вероятность')
    plt.xlabel('Задачи')
    plt.legend()
    plt.grid(axis='y')
    plt.show()

# Выполнение задач
if __name__ == "__main__":
    medical_test()
    biased_coin()
    event_prediction()
    parameter_estimation()
    product_quality()
    sports_prediction()

    visualize_results()
```

#3. Точечные и интервальные оценки параметров

#### 1. Введение

Статистическое оценивание — это область математической статистики, занимающаяся построением оценок параметров распределения. Оценивание параметров случайных величин — одна из ключевых задач, так как параметры распределения позволяют делать выводы о генеральной совокупности на основании выборки.

Оценивание может быть:
- **Точечным**: нахождение одного значения, которое служит оценкой неизвестного параметра.
- **Интервальным**: нахождение интервала, в котором с заданной вероятностью (доверительная вероятность) лежит истинное значение параметра.

#### 2. Точечные оценки параметров

Точечная оценка — это одно значение, которое используется для приближенного определения неизвестного параметра генеральной совокупности. Пусть $\theta$ — неизвестный параметр, тогда точечная оценка $\hat{\theta}$ параметра $\theta$ — это функция от выборочных данных.

##### 2.1. Основные свойства точечных оценок

1. **Несмещенность (unbiasedness)**: Оценка называется несмещённой, если математическое ожидание этой оценки равно истинному значению оцениваемого параметра:
   $$
   \mathbb{E}(\hat{\theta}) = \theta.
   $$
   Если это не так, то оценка называется смещённой, и смещение определяется как разность:
   $$
   \text{Смещение} = \mathbb{E}(\hat{\theta}) - \theta.
   $$

2. **Состоятельность (consistency)**: Оценка называется состоятельной, если с увеличением объема выборки она стремится к истинному значению оцениваемого параметра:
   $$
   \hat{\theta}_n \xrightarrow{P} \theta, \quad \text{при}\; n \to \infty.
   $$

3. **Эффективность (efficiency)**: Эффективная оценка обладает минимальной дисперсией среди всех несмещённых оценок. Если для двух несмещённых оценок $\hat{\theta}_1$ и $\hat{\theta}_2$, дисперсия первой меньше:
   $$
   \text{Var}(\hat{\theta}_1) < \text{Var}(\hat{\theta}_2),
   $$
   то $\hat{\theta}_1$ считается более эффективной.

4. **Справедливость (sufficiency)**: Оценка является справедливой, если она использует всю информацию, содержащуюся в выборке, о параметре $\theta$.

##### 2.2. Методы нахождения точечных оценок

Существуют различные методы нахождения точечных оценок, основные из них включают:

- **Метод моментов**: Основывается на приравнивании теоретических моментов к выборочным моментам.
  
- **Метод максимального правдоподобия**: Строится на максимизации функции правдоподобия.

##### 2.2.1. Метод моментов

Метод моментов предполагает приравнивание первых $k$ выборочных моментов соответствующим теоретическим моментам.

Пусть $X_1, X_2, \ldots, X_n$ — выборка объема $n$, и $m_k$ — $k$-й выборочный момент:
$$
m_k = \frac{1}{n} \sum_{i=1}^{n} X_i^k.
$$

Теоретический момент порядка $k$ определяется как:
$$
\mu_k = \mathbb{E}(X^k).
$$

Метод моментов состоит в решении системы уравнений:
$$
m_1 = \mu_1, \quad m_2 = \mu_2, \ldots
$$

##### 2.2.2. Метод максимального правдоподобия

Метод максимального правдоподобия заключается в нахождении такого значения параметра, при котором вероятность получить данную выборку будет максимальной.

Функция правдоподобия для выборки $X_1, X_2, \ldots, X_n$, состоящей из независимых одинаково распределённых случайных величин с функцией плотности распределения $f(X; \theta)$, определяется как:
$$
L(\theta) = \prod_{i=1}^{n} f(X_i; \theta).
$$

Для нахождения оценки максимального правдоподобия нужно найти максимум функции правдоподобия или её логарифма:
$$
\ln L(\theta) = \sum_{i=1}^{n} \ln f(X_i; \theta),
$$
и затем решить уравнение:
$$
\frac{d}{d\theta} \ln L(\theta) = 0.
$$

##### Пример 1: Оценка параметра нормального распределения методом максимального правдоподобия

Пусть $X_1, X_2, \ldots, X_n$ — выборка из нормального распределения $N(\mu, \sigma^2)$. Функция плотности:
$$
f(X; \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(X - \mu)^2}{2\sigma^2} \right).
$$

Функция правдоподобия:
$$
L(\mu, \sigma^2) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(X_i - \mu)^2}{2\sigma^2} \right).
$$

Логарифмическая функция правдоподобия:
$$
\ln L(\mu, \sigma^2) = -\frac{n}{2} \ln(2\pi) - \frac{n}{2} \ln \sigma^2 - \frac{1}{2\sigma^2} \sum_{i=1}^{n} (X_i - \mu)^2.
$$

Решая уравнение для нахождения максимумов, получим:
$$
\hat{\mu} = \frac{1}{n} \sum_{i=1}^{n} X_i, \quad \hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (X_i - \hat{\mu})^2.
$$

#### 3. Интервальные оценки

Интервальная оценка параметра задаётся в виде интервала, в котором с определённой доверительной вероятностью находится истинное значение параметра.

##### 3.1. Понятие доверительного интервала

Доверительный интервал для параметра $\theta$ представляет собой пару значений $(L, U)$, таких что вероятность того, что параметр попадёт в этот интервал, равна $1 - \alpha$:
$$
P(L < \theta < U) = 1 - \alpha,
$$
где $\alpha$ — уровень значимости.

##### 3.2. Построение доверительного интервала

Для построения доверительного интервала необходимо знать распределение соответствующей статистики, например, оценку параметра. Далее представим несколько типичных случаев построения доверительных интервалов.

##### 3.2.1. Доверительный интервал для математического ожидания нормального распределения при известной дисперсии

Пусть $X_1, X_2, \ldots, X_n$ — независимые одинаково распределённые случайные величины из нормального распределения $N(\mu, \sigma^2)$, где $\sigma^2$ известно. Тогда доверительный интервал для математического ожидания с доверительной вероятностью $1 - \alpha$ имеет вид:
$$
\left( \overline{X} - z_{\alpha/2} \frac{\sigma}{\sqrt{n}}, \; \overline{X} + z_{\alpha/2} \frac{\sigma}{\sqrt{n}} \right),
$$
где $\overline{X} = \frac{1}{n} \sum_{i=1}^{n} X_i$ — выборочное среднее, а $z_{\alpha/2}$ — квантиль стандартного нормального распределения уровня $1 - \alpha/2$.

##### 3.2.2. Доверительный интервал для дисперсии нормального распределения

Если $X_1, X_2, \ldots, X_n$ — выборка из нормального распределения, то для дисперсии с доверительной вероятностью $1 - \alpha$ доверительный интервал определяется следующим образом:
$$
\left( \frac{(n-1)S^2}{\chi^2_{1-\alpha/2, n-1}}, \; \frac{(n-1)S^2}{\chi^2_{\alpha/2, n-1}} \right),
$$
где $S^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \overline{X})^2$ — выборочная дисперсия, а $\chi^2$ — распределение хи-квадрат.

##### Пример 2: Доверительный интервал для математического ожидания

Предположим, что у нас есть выборка из 25 наблюдений $X_1, X_2, \ldots, X_{25}$, где $\overline{X} = 50$, а дисперсия известна и равна $\sigma^2 = 16$. Нам необходимо построить 95%-ный доверительный интервал для математического ожидания.

Вычислим квантиль:
$$
z_{\alpha/2} = z_{0.025} \approx 1.96.
$$

Теперь построим доверительный интервал:
$$
\left( 50 - 1.96 \frac{4}{\sqrt{25}}, \; 50 + 1.96 \frac{4}{\sqrt{25}} \right) = (48.44, 51.56).
$$

Таким образом, с вероятностью 95% истинное значение математического ожидания находится в интервале $(48.44, 51.56)$.

### Задача 1: Оценка математического ожидания

**Условие:** Известно, что выборка $X_1, X_2, X_3, X_4, X_5$ из нормально распределенной случайной величины с известной дисперсией $\sigma^2 = 9$ имеет следующие значения: $X_1 = 5$, $X_2 = 7$, $X_3 = 8$, $X_4 = 10$, $X_5 = 6$. Найдите точечную оценку математического ожидания $\mu$ и 95% доверительный интервал для $\mu$.

#### Решение:

1. **Найдем выборочное среднее:**
   $$
   \overline{X} = \frac{X_1 + X_2 + X_3 + X_4 + X_5}{5} = \frac{5 + 7 + 8 + 10 + 6}{5} = \frac{36}{5} = 7.2.
   $$

2. **Найдём стандартное отклонение:**
   Поскольку дисперсия известна ($\sigma^2 = 9$), стандартное отклонение:
   $$
   \sigma = \sqrt{9} = 3.
   $$

3. **Определим значение квантиля для 95% доверительного интервала:**
   Для нормального распределения $z_{0.025} \approx 1.96$.

4. **Построим доверительный интервал:**
   $$
   \left( \overline{X} - z_{0.025} \frac{\sigma}{\sqrt{n}}, \; \overline{X} + z_{0.025} \frac{\sigma}{\sqrt{n}} \right).
   $$
   $$
   \left( 7.2 - 1.96 \cdot \frac{3}{\sqrt{5}}, \; 7.2 + 1.96 \cdot \frac{3}{\sqrt{5}} \right).
   $$
   $$
   \left( 7.2 - 2.63, \; 7.2 + 2.63 \right) \approx (4.57, 9.83).
   $$

**Ответ:** Точечная оценка математического ожидания $\mu = 7.2$. Доверительный интервал: $(4.57, 9.83)$.

### Задача 2: Оценка дисперсии

**Условие:** Известно, что выборка из нормального распределения содержит $n = 10$ наблюдений: $X = [12, 14, 15, 10, 13, 16, 14, 12, 18, 20]$. Найдите 95% доверительный интервал для дисперсии $\sigma^2$.

#### Решение:

1. **Найдем выборочную среднюю:**
   $$
   \overline{X} = \frac{12 + 14 + 15 + 10 + 13 + 16 + 14 + 12 + 18 + 20}{10} = \frac{144}{10} = 14.4.
   $$

2. **Найдем выборочную дисперсию $S^2$:**
   $$
   S^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \overline{X})^2.
   $$
   $$
   S^2 = \frac{1}{9} \left((12-14.4)^2 + (14-14.4)^2 + (15-14.4)^2 + (10-14.4)^2 + (13-14.4)^2 + (16-14.4)^2 + (14-14.4)^2 + (12-14.4)^2 + (18-14.4)^2 + (20-14.4)^2\right).
   $$
   $$
   = \frac{1}{9} \left(5.76 + 0.16 + 0.36 + 19.36 + 1.96 + 2.56 + 0.16 + 5.76 + 12.96 + 30.76\right).
   $$
   $$
   = \frac{1}{9} \cdot 79.68 \approx 8.85.
   $$

3. **Построим доверительный интервал для $\sigma^2$:**
   Используем распределение хи-квадрат:
   $$
   \left( \frac{(n-1)S^2}{\chi^2_{1-\alpha/2, n-1}}, \frac{(n-1)S^2}{\chi^2_{\alpha/2, n-1}} \right).
   $$
   Для $n = 10$ и $\alpha = 0.05$:
   $\chi^2_{0.975, 9} \approx 2.7$, $\chi^2_{0.025, 9} \approx 19.02$.
   $$
   \left( \frac{9 \cdot 8.85}{19.02}, \frac{9 \cdot 8.85}{2.7} \right) \approx (4.18, 29.55).
   $$

**Ответ:** Доверительный интервал для дисперсии $\sigma^2$ составляет $(4.18, 29.55)$.

### Задача 3: Оценка пропорции

**Условие:** В результате опроса 200 респондентов 60 человек выразили одобрение новой политики. Найдите точечную оценку и 95% доверительный интервал для истинной пропорции $\pi$.

#### Решение:

1. **Точечная оценка пропорции:**
   $$
   \hat{\pi} = \frac{x}{n} = \frac{60}{200} = 0.3.
   $$

2. **Определим стандартную ошибку:**
   $$
   SE = \sqrt{\frac{\hat{\pi}(1 - \hat{\pi})}{n}} = \sqrt{\frac{0.3(1 - 0.3)}{200}} = \sqrt{\frac{0.3 \cdot 0.7}{200}} \approx 0.034.
   $$

3. **Определим значение квантиля для 95% доверительного интервала:**
   Для нормального распределения $z_{0.025} \approx 1.96$.

4. **Построим доверительный интервал:**
   $$
   \left( \hat{\pi} - z_{0.025} \cdot SE, \hat{\pi} + z_{0.025} \cdot SE \right).
   $$
   $$
   \left( 0.3 - 1.96 \cdot 0.034, 0.3 + 1.96 \cdot 0.034 \right) \approx (0.3 - 0.067, 0.3 + 0.067) \approx (0.233, 0.367).
   $$

**Ответ:** Точечная оценка $\pi = 0.3$. Доверительный интервал: $(0.233, 0.367)$.

### Задача 4: Интервальный оценка разности средних

**Условие:** Две группы респондентов оценили новый продукт. В первой группе (n1=30) средняя оценка составила 8, с дисперсией 2. Во второй группе (n2=35) средняя оценка составила 6, с дисперсией 3. Постройте 95% доверительный интервал для разности средних.

#### Решение:

1. **Найдем стандартные отклонения для обеих групп:**
   $$
   S_1 = \sqrt{2} \approx 1.414, \quad S_2 = \sqrt{3} \approx 1.732.
   $$

2. **Определим стандартную ошибку для разности средних:**
   $$
   SE = \sqrt{\frac{S_1^2}{n_1} + \frac{S_2^2}{n_2}} = \sqrt{\frac{2}{30} + \frac{3}{35}} = \sqrt{0.0667 + 0.0857} \approx \sqrt{0.1524} \approx 0.39.
   $$

3. **Построим доверительный интервал для разности средних:**
   $$
   \text{Разность средних} = \overline{X_1} - \overline{X_2} = 8 - 6 = 2.
   $$

4. **Квантиль для 95% доверительного интервала (при большом объеме

 выборки, можно использовать нормальное распределение):**
   $z_{0.025} \approx 1.96$.

5. **Построим доверительный интервал:**
   $$
   (2 - 1.96 \cdot 0.39, 2 + 1.96 \cdot 0.39) \approx (2 - 0.765, 2 + 0.765) \approx (1.235, 2.765).
   $$

**Ответ:** Доверительный интервал для разности средних составляет $(1.235, 2.765)$.

### Задача 5: Оценка медианы

**Условие:** Имеется выборка из 11 наблюдений: $X = [5, 8, 7, 10, 6, 9, 4, 12, 11, 15, 14]$. Найдите точечную оценку медианы и 95% доверительный интервал для медианы.

#### Решение:

1. **Отсортируем выборку:**
   $X_{sorted} = [4, 5, 6, 7, 8, 9, 10, 11, 12, 14, 15]$.

2. **Найдем медиану:**
   Для нечетного количества наблюдений медиана — это средний элемент:
   $$
   \text{Медиана} = X_{(n+1)/2} = X_{6} = 9.
   $$

3. **Для построения доверительного интервала медианы можно использовать метод бутстрэпинга или центральной предельной теоремы, но для простоты воспользуемся нормальным приближением.**
   Для малого количества наблюдений (например, $n < 30$), будем использовать непараметрический подход, но так как $n = 11$, предполагаем нормальное распределение:

   **Необходимый шаг:** Найти стандартное отклонение.

   $$

4. **Определим доверительный интервал для медианы:**

   \left( \hat{m} - z_{0.025} \cdot \frac{S}{\sqrt{n}}, \hat{m} + z_{0.025} \cdot \frac{S}{\sqrt{n}} \right) \approx \left( 9 - 1.96 \cdot \frac{2.63}{\sqrt{11}}, 9 + 1.96 \cdot \frac{2.63}{\sqrt{11}} \right).
   $$
   $$
   \approx \left( 9 - 1.56, 9 + 1.56 \right) \approx (7.44, 10.56).
   $$

**Ответ:** Точечная оценка медианы = 9. Доверительный интервал: $(7.44, 10.56)$.

### Задача 6: Оценка параметров распределения

**Условие:** В выборке из 20 наблюдений (неизвестное распределение) среднее значение составляет 50, а дисперсия 36. Найдите 90% доверительный интервал для математического ожидания.

#### Решение:

1. **Выборочные параметры:**
   $\overline{X} = 50$, $S^2 = 36$, следовательно $S = 6$.

2. **Определим уровень значимости:**
   $\alpha = 0.10$ и $n = 20$.

3. **Квантиль для t-распределения:**
   Для $df = n-1 = 19$ (при 90% доверительной вероятности):
   $t_{0.05, 19} \approx 1.645$.

4. **Построим доверительный интервал:**
   $$
   \left( \overline{X} - t_{0.05, 19} \cdot \frac{S}{\sqrt{n}}, \overline{X} + t_{0.05, 19} \cdot \frac{S}{\sqrt{n}} \right) = \left( 50 - 1.645 \cdot \frac{6}{\sqrt{20}}, 50 + 1.645 \cdot \frac{6}{\sqrt{20}} \right).
   $$
   $$
   = \left( 50 - 1.645 \cdot 1.341, 50 + 1.645 \cdot 1.341 \right) \approx \left( 50 - 2.20, 50 + 2.20 \right) \approx (47.8, 52.2).
   $$

**Ответ:** Доверительный интервал для математического ожидания $\mu$ составляет $(47.8, 52.2)$.



Эти шесть задач охватывают различные аспекты оценивания параметров и построения доверительных интервалов, включая как точечные, так и интервальные оценки для различных статистических параметров.



Давайте реализуем каждую из шести задач на Python с использованием библиотек NumPy и Matplotlib для визуализации. В каждом примере мы будем выводить результаты расчетов и графически представлять доверительные интервалы.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Задача 1: Оценка математического ожидания
data1 = np.array([5, 7, 8, 10, 6])
sigma1 = 3  # известное стандартное отклонение
n1 = len(data1)

mean1 = np.mean(data1)
std_error1 = sigma1 / np.sqrt(n1)
z_critical1 = stats.norm.ppf(0.975)  # 95% доверительный интервал

conf_interval1 = (mean1 - z_critical1 * std_error1, mean1 + z_critical1 * std_error1)

# Визуализация
plt.figure(figsize=(10, 6))
plt.axhline(y=mean1, color='r', linestyle='--', label='Точечная оценка (среднее)')
plt.fill_betweenx([0, 1], conf_interval1[0], conf_interval1[1], alpha=0.2, label='95% доверительный интервал')
plt.title('Задача 1: Доверительный интервал для математического ожидания')
plt.xticks([conf_interval1[0], mean1, conf_interval1[1]], ['Нижняя граница', 'Среднее', 'Верхняя граница'])
plt.yticks([])
plt.legend()
plt.show()

# Задача 2: Оценка дисперсии
data2 = np.array([12, 14, 15, 10, 13, 16, 14, 12, 18, 20])
n2 = len(data2)
s2 = np.var(data2, ddof=1)

chi2_lower = stats.chi2.ppf(0.975, df=n2-1)
chi2_upper = stats.chi2.ppf(0.025, df=n2-1)

conf_interval2 = ((n2-1) * s2 / chi2_upper, (n2-1) * s2 / chi2_lower)

# Визуализация
plt.figure(figsize=(10, 6))
plt.axhline(y=s2, color='r', linestyle='--', label='Точечная оценка (дисперсия)')
plt.fill_betweenx([0, 1], conf_interval2[0], conf_interval2[1], alpha=0.2, label='95% доверительный интервал')
plt.title('Задача 2: Доверительный интервал для дисперсии')
plt.xticks([conf_interval2[0], s2, conf_interval2[1]], ['Нижняя граница', 'Дисперсия', 'Верхняя граница'])
plt.yticks([])
plt.legend()
plt.show()

# Задача 3: Оценка пропорции
n3 = 200
x3 = 60
pi_hat = x3 / n3
std_error3 = np.sqrt(pi_hat * (1 - pi_hat) / n3)
z_critical3 = stats.norm.ppf(0.975)

conf_interval3 = (pi_hat - z_critical3 * std_error3, pi_hat + z_critical3 * std_error3)

# Визуализация
plt.figure(figsize=(10, 6))
plt.axhline(y=pi_hat, color='r', linestyle='--', label='Точечная оценка (пропорция)')
plt.fill_betweenx([0, 1], conf_interval3[0], conf_interval3[1], alpha=0.2, label='95% доверительный интервал')
plt.title('Задача 3: Доверительный интервал для пропорции')
plt.xticks([conf_interval3[0], pi_hat, conf_interval3[1]], ['Нижняя граница', 'Пропорция', 'Верхняя граница'])
plt.yticks([])
plt.legend()
plt.show()

# Задача 4: Интервальный оценка разности средних
data4_group1 = np.array([8, 8, 9, 7, 8, 9, 8, 7, 8, 9, 10, 8, 7, 9, 8, 10, 8, 9, 7, 9, 8, 10, 9, 7, 9, 10, 9, 8, 8, 8])
data4_group2 = np.array([5, 6, 7, 5, 5, 6, 7, 5, 6, 7, 8, 5, 5, 6, 6, 5, 5, 6, 5, 6, 7, 6, 6, 5, 6, 5, 6, 5, 6, 5])

n4_group1 = len(data4_group1)
n4_group2 = len(data4_group2)

mean4_group1 = np.mean(data4_group1)
mean4_group2 = np.mean(data4_group2)

std_error4 = np.sqrt(np.var(data4_group1, ddof=1) / n4_group1 + np.var(data4_group2, ddof=1) / n4_group2)
diff_means = mean4_group1 - mean4_group2
z_critical4 = stats.norm.ppf(0.975)

conf_interval4 = (diff_means - z_critical4 * std_error4, diff_means + z_critical4 * std_error4)

# Визуализация
plt.figure(figsize=(10, 6))
plt.axhline(y=diff_means, color='r', linestyle='--', label='Разность средних')
plt.fill_betweenx([-5, 5], conf_interval4[0], conf_interval4[1], alpha=0.2, label='95% доверительный интервал')
plt.title('Задача 4: Доверительный интервал для разности средних')
plt.xticks([conf_interval4[0], diff_means, conf_interval4[1]], ['Нижняя граница', 'Разность средних', 'Верхняя граница'])
plt.yticks([])
plt.legend()
plt.show()

# Задача 5: Оценка медианы
data5 = np.array([5, 8, 7, 10, 6, 9, 4, 12, 11, 15, 14])
n5 = len(data5)

median5 = np.median(data5)

# Используем бутстрэп для оценки доверительного интервала медианы
bootstraps = np.random.choice(data5, (1000, n5), replace=True)
medians_bootstrap = np.median(bootstraps, axis=1)
conf_interval5 = np.percentile(medians_bootstrap, [2.5, 97.5])

# Визуализация
plt.figure(figsize=(10, 6))
plt.axhline(y=median5, color='r', linestyle='--', label='Точечная оценка (медиана)')
plt.fill_betweenx([0, 1], conf_interval5[0], conf_interval5[1], alpha=0.2, label='95% доверительный интервал')
plt.title('Задача 5: Доверительный интервал для медианы')
plt.xticks([conf_interval5[0], median5, conf_interval5[1]], ['Нижняя граница', 'Медиана', 'Верхняя граница'])
plt.yticks([])
plt.legend()
plt.show()

# Задача 6: Оценка параметров распределения
n6 = 20
mean6 = 50
var6 = 36
sigma6 = np.sqrt(var6)

t_critical6 = stats.t.ppf(0.95, df=n6-1)

conf_interval6 = (mean6 - t_critical6 * (sigma6 / np.sqrt(n6)), mean6 + t_critical6 * (sigma6 / np.sqrt(n6)))

# Визуализация
plt.figure(figsize=(10, 6))
plt.axhline(y=mean6, color='r', linestyle='--', label='Точечная оценка (мат. ожидание)')
plt.fill_betweenx([0, 1], conf_interval6[0], conf_interval6[1], alpha=0.2, label='90% доверительный интервал')
plt.title('Задача 6: Доверительный интервал для математического ожидания')
plt.xticks([conf_interval6[0], mean6, conf_interval6[1]], ['Нижняя граница', 'Мат. ожидание', 'Верхняя граница'])
plt.yticks([])
plt.legend()
plt.show()
```

#3. Методы оценивания в статистике – метод моментов

#### Введение в методы оценивания

В математической статистике под **оцениванием параметров** понимается процесс определения значений неизвестных параметров распределения вероятностей случайной величины на основе имеющихся выборочных данных. Пусть у нас есть выборка $X_1, X_2, \ldots, X_n$, состоящая из независимых и одинаково распределённых случайных величин, которые имеют неизвестное распределение, описываемое параметром $\theta$. Целью оценивания является построение статистики $\hat{\theta} = \hat{\theta}(X_1, X_2, \ldots, X_n)$, которая служит оценкой для истинного значения параметра $\theta$.

Существует несколько методов построения оценок параметров, среди которых наиболее известны метод максимального правдоподобия и метод моментов. В данной лекции мы подробно рассмотрим метод моментов, который используется для построения оценок параметров распределения.

#### Основы метода моментов

**Метод моментов** – это способ оценивания параметров распределения случайной величины, основанный на приравнивании теоретических моментов к выборочным моментам.

Рассмотрим основные понятия:

1. **Математическое ожидание** (первый момент):
   \[
   \mu = \mathbb{E}[X],
   \]
   где $\mathbb{E}[X]$ – математическое ожидание случайной величины $X$.

2. **Моменты случайной величины**. Теоретический $k$-й момент случайной величины $X$ относительно начала (нулевой момент) определяется формулой:
   \[
   \mu_k' = \mathbb{E}[X^k].
   \]

   Центральный момент порядка $k$ определяется как:
   \[
   \mu_k = \mathbb{E}[(X - \mu)^k],
   \]
   где $\mu$ – математическое ожидание.

3. **Выборочные моменты**. Для выборки $X_1, X_2, \ldots, X_n$ выборочный момент порядка $k$ относительно начала вычисляется по формуле:
   \[
   m_k' = \frac{1}{n} \sum_{i=1}^{n} X_i^k.
   \]

#### Принцип метода моментов

Идея метода моментов заключается в следующем: мы приравниваем выборочные моменты к соответствующим теоретическим моментам и решаем полученные уравнения относительно неизвестных параметров.

Пусть распределение случайной величины $X$ зависит от параметров $\theta_1, \theta_2, \ldots, \theta_r$. В этом случае теоретические моменты $\mu_1', \mu_2', \ldots, \mu_r'$ будут функциями этих параметров:
\[
\mu_k' = \mu_k'(\theta_1, \theta_2, \ldots, \theta_r), \quad k = 1, 2, \ldots, r.
\]

Метод моментов предполагает приравнивание первых $r$ выборочных моментов к соответствующим теоретическим моментам:
\[
m_1' = \mu_1'(\theta_1, \theta_2, \ldots, \theta_r),
\]
\[
m_2' = \mu_2'(\theta_1, \theta_2, \ldots, \theta_r),
\]
\[
\ldots
\]
\[
m_r' = \mu_r'(\theta_1, \theta_2, \ldots, \theta_r).
\]

Решив эту систему уравнений относительно $\theta_1, \theta_2, \ldots, \theta_r$, получим оценки параметров $\hat{\theta}_1, \hat{\theta}_2, \ldots, \hat{\theta}_r$.

#### Пример использования метода моментов

Рассмотрим пример, когда случайная величина $X$ имеет распределение Пуассона с параметром $\lambda$, то есть
\[
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad k = 0, 1, 2, \ldots
\]

1. **Теоретический момент**. Известно, что математическое ожидание распределения Пуассона равно его параметру:
   \[
   \mathbb{E}[X] = \lambda.
   \]

2. **Выборочный момент**. Для выборки $X_1, X_2, \ldots, X_n$ выборочное среднее (первый выборочный момент) вычисляется как:
   \[
   m_1' = \frac{1}{n} \sum_{i=1}^{n} X_i.
   \]

3. **Применение метода моментов**. Приравниваем выборочный момент к теоретическому:
   \[
   m_1' = \lambda.
   \]

Таким образом, оценка параметра $\lambda$ методом моментов будет равна:
\[
\hat{\lambda} = \frac{1}{n} \sum_{i=1}^{n} X_i.
\]

#### Пример для нормального распределения

Рассмотрим случай нормального распределения $N(\mu, \sigma^2)$, где параметры $\mu$ и $\sigma^2$ неизвестны.

1. **Теоретические моменты**:
   \[
   \mathbb{E}[X] = \mu, \quad \mathbb{E}[(X - \mu)^2] = \sigma^2.
   \]

2. **Выборочные моменты**:
   \[
   m_1' = \frac{1}{n} \sum_{i=1}^{n} X_i, \quad m_2 = \frac{1}{n} \sum_{i=1}^{n} (X_i - m_1')^2.
   \]

3. **Система уравнений**:
   \[
   m_1' = \mu,
   \]
   \[
   m_2 = \sigma^2.
   \]

Таким образом, оценки методом моментов для параметров нормального распределения будут:
\[
\hat{\mu} = \frac{1}{n} \sum_{i=1}^{n} X_i,
\]
\[
\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (X_i - \hat{\mu})^2.
\]

#### Свойства метода моментов

Метод моментов обладает рядом свойств:

1. **Простота применения**. Метод моментов часто используется, так как его применение не требует сложных вычислений и он дает простые аналитические формулы.

2. **Состоятельность**. Оценки, полученные методом моментов, могут быть состоятельными, то есть они сходятся по вероятности к истинному значению параметра при увеличении объема выборки.

3. **Не всегда эффективен**. Оценки методом моментов могут быть неэффективными, то есть их дисперсия может быть больше, чем у оценок, полученных методом максимального правдоподобия.

4. **Зависимость от выборки**. Точность оценок, полученных методом моментов, зависит от объема выборки и её качества.

#### Заключение

Метод моментов является классическим и широко используемым методом оценивания параметров распределений в статистике. Он прост в применении и не требует сложных вычислений. Однако в некоторых случаях его оценки могут быть менее точными по сравнению с другими методами, например, методом максимального правдоподобия. Тем не менее, он остаётся полезным инструментом для получения начальных приближений к оценкам параметров.



Вот шесть конкретных задач, которые иллюстрируют применение метода моментов для оценки параметров различных распределений. Каждая задача включает подробное пошаговое решение.

### Задача 1: Оценка параметра распределения Пуассона

**Условие:** Пусть случайная величина $X$ имеет распределение Пуассона с параметром $\lambda$. На основании выборки $X_1 = 2, X_2 = 3, X_3 = 4, X_4 = 5, X_5 = 3$ необходимо найти оценку параметра $\lambda$ методом моментов.

#### Решение:
1. **Запишем выборочные данные:**
   \[
   X_1 = 2, \quad X_2 = 3, \quad X_3 = 4, \quad X_4 = 5, \quad X_5 = 3
   \]

2. **Вычислим выборочное среднее (первый выборочный момент):**
   \[
   m_1' = \frac{1}{5} (X_1 + X_2 + X_3 + X_4 + X_5) = \frac{1}{5} (2 + 3 + 4 + 5 + 3) = \frac{17}{5} = 3.4
   \]

3. **Приравняем выборочный момент к теоретическому:**
   \[
   \lambda = m_1' = 3.4
   \]

4. **Ответ:**
   Оценка параметра $\lambda$ методом моментов равна $\hat{\lambda} = 3.4$.



### Задача 2: Оценка параметров нормального распределения

**Условие:** Пусть случайная величина $X$ распределена нормально $N(\mu, \sigma^2)$. На основании выборки $X_1 = 5, X_2 = 7, X_3 = 6$ необходимо оценить параметры $\mu$ и $\sigma^2$ методом моментов.

#### Решение:
1. **Запишем выборочные данные:**
   \[
   X_1 = 5, \quad X_2 = 7, \quad X_3 = 6
   \]

2. **Вычислим выборочное среднее (первый выборочный момент):**
   \[
   m_1' = \frac{1}{3} (X_1 + X_2 + X_3) = \frac{1}{3} (5 + 7 + 6) = \frac{18}{3} = 6
   \]

3. **Вычислим выборочный дисперсию (второй момент):**
   \[
   m_2' = \frac{1}{3} \sum_{i=1}^{3} X_i^2 = \frac{1}{3} (5^2 + 7^2 + 6^2) = \frac{1}{3} (25 + 49 + 36) = \frac{110}{3} \approx 36.67
   \]

4. **Приравняем выборочные моменты к теоретическим:**
   - Первый момент:
   \[
   \hat{\mu} = m_1' = 6
   \]
   - Второй момент:
   \[
   \hat{\sigma}^2 = m_2' - \hat{\mu}^2 = \frac{110}{3} - 6^2 = \frac{110}{3} - 36 = \frac{110 - 108}{3} = \frac{2}{3} \approx 0.67
   \]

5. **Ответ:**
   Оценка параметров: $\hat{\mu} = 6$ и $\hat{\sigma}^2 \approx 0.67$.



### Задача 3: Оценка параметра равномерного распределения

**Условие:** Пусть случайная величина $X$ имеет равномерное распределение на отрезке $[a, b]$. На основании выборки $X_1 = 1, X_2 = 2, X_3 = 4$ необходимо найти оценки параметров $a$ и $b$ методом моментов.

#### Решение:
1. **Запишем выборочные данные:**
   \[
   X_1 = 1, \quad X_2 = 2, \quad X_3 = 4
   \]

2. **Вычислим выборочное среднее (первый выборочный момент):**
   \[
   m_1' = \frac{1}{3} (X_1 + X_2 + X_3) = \frac{1}{3} (1 + 2 + 4) = \frac{7}{3} \approx 2.33
   \]

3. **Вычислим выборочный момент второго порядка:**
   \[
   m_2' = \frac{1}{3} \sum_{i=1}^{3} X_i^2 = \frac{1}{3} (1^2 + 2^2 + 4^2) = \frac{1}{3} (1 + 4 + 16) = \frac{21}{3} = 7
   \]

4. **Теоретические моменты для равномерного распределения:**
   \[
   \mu = \frac{a + b}{2}, \quad \mathbb{E}[X^2] = \frac{a^2 + ab + b^2}{3}
   \]

5. **Составим систему уравнений:**
   - Из первого уравнения:
   \[
   \frac{a + b}{2} = \frac{7}{3} \Rightarrow a + b = \frac{14}{3}
   \]
   - Из второго уравнения:
   \[
   \frac{a^2 + ab + b^2}{3} = 7 \Rightarrow a^2 + ab + b^2 = 21
   \]

6. **Подставим $b = \frac{14}{3} - a$ в уравнение для второго момента:**
   \[
   a^2 + a\left(\frac{14}{3} - a\right) + \left(\frac{14}{3} - a\right)^2 = 21
   \]
   Преобразуем:
   \[
   a^2 + \frac{14a}{3} - a^2 + \frac{196}{9} - \frac{28a}{3} + a^2 = 21
   \]
   Упрощаем:
   \[
   \frac{14a}{3} - \frac{28a}{3} + \frac{196}{9} = 21
   \]
   Умножаем на 9:
   \[
   -14a + 196 = 189 \Rightarrow -14a = -7 \Rightarrow a = \frac{1}{2}
   \]

7. **Найдем $b$:**
   \[
   b = \frac{14}{3} - \frac{1}{2} = \frac{14}{3} - \frac{3}{6} = \frac{28}{6} - \frac{3}{6} = \frac{25}{6}
   \]

8. **Ответ:**
   Оценки параметров: $\hat{a} = 0.5$ и $\hat{b} \approx 4.17$.



### Задача 4: Оценка параметров экспоненциального распределения

**Условие:** Пусть случайная величина $X$ имеет экспоненциальное распределение с параметром $\lambda$. На основании выборки $X_1 = 1, X_2 = 2, X_3 = 3$ необходимо оценить параметр $\lambda$ методом моментов.

#### Решение:
1. **Запишем выборочные данные:**
   \[
   X_1 = 1, \quad X_2 = 2, \quad X_3 = 3
   \]

2. **Вычислим выборочное среднее (первый выборочный момент):**
   \[
   m_1' = \frac{1}{3} (X_1 + X_2 + X_3) = \frac{1}{3} (1 + 2 + 3) = \frac{6}{3} = 2
   \]

3. **Приравняем выборочный момент к теоретическому:**
   Для экспоненциального распределения:
   \[
   \mathbb{E}[X] = \frac{1}{\lambda} \Rightarrow \lambda = \frac{1}{m_1'}
   \]
   Следовательно:
   \[
   \hat{\lambda} = \frac{1}{m_1'} = \frac{1}{2} = 0.5
   \]

4. **Ответ:**
   Оценка параметра $\lambda$ методом моментов рав

на $\hat{\lambda} = 0.5$.



### Задача 5: Оценка параметров бета-распределения

**Условие:** Пусть случайная величина $X$ имеет бета-распределение $B(\alpha, \beta)$. На основании выборки $X_1 = 0.2, X_2 = 0.5, X_3 = 0.7$ необходимо оценить параметры $\alpha$ и $\beta$ методом моментов.

#### Решение:
1. **Запишем выборочные данные:**
   \[
   X_1 = 0.2, \quad X_2 = 0.5, \quad X_3 = 0.7
   \]

2. **Вычислим выборочное среднее (первый выборочный момент):**
   \[
   m_1' = \frac{1}{3} (X_1 + X_2 + X_3) = \frac{1}{3} (0.2 + 0.5 + 0.7) = \frac{1.4}{3} \approx 0.467
   \]

3. **Вычислим выборочный момент второго порядка:**
   \[
   m_2' = \frac{1}{3} \sum_{i=1}^{3} X_i^2 = \frac{1}{3} (0.2^2 + 0.5^2 + 0.7^2) = \frac{1}{3} (0.04 + 0.25 + 0.49) = \frac{0.78}{3} \approx 0.26
   \]

4. **Теоретические моменты для бета-распределения:**
   \[
   \mu = \frac{\alpha}{\alpha + \beta}, \quad \mathbb{E}[X^2] = \frac{\alpha(\alpha + 1)}{(\alpha + \beta)(\alpha + \beta + 1)}
   \]

5. **Составим систему уравнений:**
   - Из первого уравнения:
   \[
   \frac{\alpha}{\alpha + \beta} = m_1' \Rightarrow \alpha = m_1'(\alpha + \beta) \Rightarrow \alpha - m_1' \alpha = m_1' \beta \Rightarrow \alpha(1 - m_1') = m_1' \beta
   \]
   Отсюда:
   \[
   \beta = \frac{\alpha(1 - m_1')}{m_1'}
   \]

6. **Подставим $\beta$ во второе уравнение:**
   \[
   m_2' = \frac{\alpha(\alpha + 1)}{(\alpha + \beta)(\alpha + \beta + 1)} \Rightarrow m_2' = \frac{\alpha(\alpha + 1)}{(\alpha + \frac{\alpha(1 - m_1')}{m_1'})(\alpha + \frac{\alpha(1 - m_1')}{m_1'} + 1)}
   \]

7. **Найдем численные значения:**
   После подстановки и решения уравнения (по шагам), получим значения $\hat{\alpha}$ и $\hat{\beta}$.

8. **Ответ:**
   Оценки параметров $\hat{\alpha}$ и $\hat{\beta}$ (дополнительно проведем вычисления для получения числовых значений).



### Задача 6: Оценка параметров гамма-распределения

**Условие:** Пусть случайная величина $X$ имеет гамма-распределение с параметрами $\alpha$ и $\beta$. На основании выборки $X_1 = 3, X_2 = 4, X_3 = 5$ необходимо оценить параметры $\alpha$ и $\beta$ методом моментов.

#### Решение:
1. **Запишем выборочные данные:**
   \[
   X_1 = 3, \quad X_2 = 4, \quad X_3 = 5
   \]

2. **Вычислим выборочное среднее (первый выборочный момент):**
   \[
   m_1' = \frac{1}{3} (X_1 + X_2 + X_3) = \frac{1}{3} (3 + 4 + 5) = \frac{12}{3} = 4
   \]

3. **Вычислим выборочный момент второго порядка:**
   \[
   m_2' = \frac{1}{3} \sum_{i=1}^{3} X_i^2 = \frac{1}{3} (3^2 + 4^2 + 5^2) = \frac{1}{3} (9 + 16 + 25) = \frac{50}{3} \approx 16.67
   \]

4. **Теоретические моменты для гамма-распределения:**
   \[
   \mu = \frac{\alpha}{\beta}, \quad \mathbb{E}[X^2] = \frac{\alpha(\alpha + 1)}{\beta^2}
   \]

5. **Составим систему уравнений:**
   - Из первого уравнения:
   \[
   \frac{\alpha}{\beta} = m_1' \Rightarrow \alpha = m_1' \beta
   \]
   - Из второго уравнения:
   \[
   \frac{\alpha(\alpha + 1)}{\beta^2} = m_2'
   \]

6. **Подставим $\alpha$ во второе уравнение:**
   \[
   \frac{m_1' \beta (m_1' \beta + 1)}{\beta^2} = m_2'
   \]

7. **Решим уравнение для нахождения $\alpha$ и $\beta$:**
   После упрощения получим значения $\hat{\alpha}$ и $\hat{\beta}$.

8. **Ответ:**
   Оценки параметров $\hat{\alpha}$ и $\hat{\beta}$ (также потребуются дальнейшие вычисления).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Настройки для графиков
sns.set(style='whitegrid')

# Задача 1: Оценка параметра распределения Пуассона
data_poisson = np.array([2, 3, 4, 5, 3])
lambda_hat = np.mean(data_poisson)
print(f"Оценка параметра λ (Пуассон): {lambda_hat}")

plt.figure(figsize=(10, 6))
sns.histplot(data_poisson, bins=5, kde=True, stat="density", label='Выборка')
x = np.arange(0, 10)
plt.plot(x, stats.poisson.pmf(x, lambda_hat), 'ro', label=f'Пуассон PMF (λ={lambda_hat:.2f})')
plt.title('Распределение Пуассона')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()

# Задача 2: Оценка параметров нормального распределения
data_normal = np.array([5, 7, 6])
mu_hat = np.mean(data_normal)
sigma_hat = np.var(data_normal, ddof=0)
print(f"Оценка параметра μ (Нормальное): {mu_hat}, σ²: {sigma_hat}")

plt.figure(figsize=(10, 6))
sns.histplot(data_normal, bins=5, kde=True

, stat="density", label='Выборка')
x = np.linspace(0, 10, 100)
plt.plot(x, stats.norm.pdf(x, mu_hat, np.sqrt(sigma_hat)), 'ro', label=f'Нормальное PDF (μ={mu_hat:.2f}, σ²={sigma_hat:.2f})')
plt.title('Нормальное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()

# Задача 3: Оценка параметров равномерного распределения
data_uniform = np.array([1, 2, 4])
m1_hat = np.mean(data_uniform)
m2_hat = np.mean(data_uniform**2)
a_hat = m1_hat - np.sqrt(3 * (m2_hat - m1_hat**2))
b_hat = m1_hat + np.sqrt(3 * (m2_hat - m1_hat**2))
print(f"Оценка параметров равномерного распределения: a = {a_hat:.2f}, b = {b_hat:.2f}")

plt.figure(figsize=(10, 6))
sns.histplot(data_uniform, bins=5, kde=True, stat="density", label='Выборка')
x = np.linspace(0, 5, 100)
plt.plot(x, stats.uniform.pdf(x, a_hat, b_hat - a_hat), 'ro', label=f'Равномерное PDF (a={a_hat:.2f}, b={b_hat:.2f})')
plt.title('Равномерное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()

# Задача 4: Оценка параметров экспоненциального распределения
data_exponential = np.array([1, 2, 3])
lambda_hat = 1 / np.mean(data_exponential)
print(f"Оценка параметра λ (Экспоненциальное): {lambda_hat:.2f}")

plt.figure(figsize=(10, 6))
sns.histplot(data_exponential, bins=5, kde=True, stat="density", label='Выборка')
x = np.linspace(0, 5, 100)
plt.plot(x, stats.expon.pdf(x, scale=1/lambda_hat), 'ro', label=f'Экспоненциальное PDF (λ={lambda_hat:.2f})')
plt.title('Экспоненциальное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()

# Задача 5: Оценка параметров бета-распределения
data_beta = np.array([0.2, 0.5, 0.7])
m1_hat = np.mean(data_beta)
m2_hat = np.mean(data_beta**2)
alpha_hat = m1_hat * (m1_hat * (1 - m1_hat) / (m2_hat - m1_hat**2))
beta_hat = (1 - m1_hat) * alpha_hat / m1_hat
print(f"Оценка параметров бета-распределения: α = {alpha_hat:.2f}, β = {beta_hat:.2f}")

plt.figure(figsize=(10, 6))
sns.histplot(data_beta, bins=10, kde=True, stat="density", label='Выборка')
x = np.linspace(0, 1, 100)
plt.plot(x, stats.beta.pdf(x, alpha_hat, beta_hat), 'ro', label=f'Бета PDF (α={alpha_hat:.2f}, β={beta_hat:.2f})')
plt.title('Бета распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()

# Задача 6: Оценка параметров гамма-распределения
data_gamma = np.array([3, 4, 5])
m1_hat = np.mean(data_gamma)
m2_hat = np.mean(data_gamma**2)
beta_hat = m1_hat / (m2_hat - m1_hat**2)
alpha_hat = m1_hat * beta_hat
print(f"Оценка параметров гамма-распределения: α = {alpha_hat:.2f}, β = {beta_hat:.2f}")

plt.figure(figsize=(10, 6))
sns.histplot(data_gamma, bins=10, kde=True, stat="density", label='Выборка')
x = np.linspace(0, 8, 100)
plt.plot(x, stats.gamma.pdf(x, alpha_hat, scale=1/beta_hat), 'ro', label=f'Гамма PDF (α={alpha_hat:.2f}, β={beta_hat:.2f})')
plt.title('Гамма распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

###5.   Доверительные интервалы и их интерпретация

### 1. Введение в доверительные интервалы

**Доверительный интервал** (Confidence Interval, CI) — это диапазон значений, который используется для оценки истинного значения параметра генеральной совокупности. В статистике он позволяет нам оценить неизвестный параметр с некоторым уровнем уверенности. Доверительный интервал определяется двумя границами — нижней и верхней, которые покрывают истинное значение параметра с заданной вероятностью.

### 2. Основные понятия и определения

- **Генеральная совокупность (Population)** — это полный набор объектов или событий, для которых мы хотим сделать выводы.
- **Выборка (Sample)** — подмножество генеральной совокупности, на основании которого мы делаем выводы о всей совокупности.
- **Параметр** — это числовая характеристика генеральной совокупности (например, среднее, дисперсия, пропорция).
- **Статистика** — это числовая характеристика выборки, которая используется для оценки параметра генеральной совокупности.

Доверительный интервал строится на основе выборочной статистики (например, среднего выборочного значения) и дает диапазон, в котором, с определенной степенью уверенности, находится истинное значение параметра генеральной совокупности.

### 3. Основная идея доверительных интервалов

Допустим, мы хотим оценить среднее значение $\mu$ в генеральной совокупности. Для этого мы выбираем выборку объема $n$, вычисляем среднее выборки $\bar{x}$, и затем строим интервал вокруг этого среднего. Интервал строится таким образом, чтобы с заданной вероятностью (например, 95%) истинное среднее $\mu$ попадало в этот интервал.

#### Пример
Если мы построили 95% доверительный интервал для среднего значения $\mu$, это означает, что при многократном повторении эксперимента 95% из построенных интервалов будут содержать истинное значение параметра $\mu$.

### 4. Построение доверительных интервалов

#### 4.1 Доверительный интервал для среднего при известной дисперсии

Рассмотрим случай, когда генеральная дисперсия $\sigma^2$ известна.

- Пусть $X_1, X_2, \ldots, X_n$ — случайная выборка объема $n$ из нормального распределения с математическим ожиданием $\mu$ и дисперсией $\sigma^2$.

- Выборочное среднее:
$$
\bar{X} = \frac{1}{n} \sum_{i=1}^n X_i
\]

- Известно, что при нормальности данных выборочное среднее $\bar{X}$ распределено по нормальному закону с параметрами $\mu$ и $\frac{\sigma^2}{n}$:
$$
\bar{X} \sim N\left(\mu, \frac{\sigma^2}{n}\right)
\]

Теперь построим доверительный интервал для среднего. Нам нужно найти такой интервал, который будет содержать истинное значение параметра $\mu$ с вероятностью $1-\alpha$, где $\alpha$ — уровень значимости.

Для этого воспользуемся стандартным нормальным распределением. Определим значение $z_{\alpha/2}$, которое соответствует квантилю стандартного нормального распределения. Тогда доверительный интервал будет иметь вид:
$$
\left( \bar{X} - z_{\alpha/2} \frac{\sigma}{\sqrt{n}}, \; \bar{X} + z_{\alpha/2} \frac{\sigma}{\sqrt{n}} \right)
\]
где $z_{\alpha/2}$ — квантиль стандартного нормального распределения, удовлетворяющий равенству:
$$
P\left(-z_{\alpha/2} \leq \frac{\bar{X} - \mu}{\sigma/\sqrt{n}} \leq z_{\alpha/2}\right) = 1 - \alpha
\]

#### Пример

Пусть $n = 25$, выборочное среднее равно $\bar{X} = 100$, а известное значение стандартного отклонения $\sigma = 15$. Мы хотим построить 95% доверительный интервал.

Значение квантиля $z_{0.025} \approx 1.96$.

Тогда доверительный интервал:
$$
\left(100 - 1.96 \cdot \frac{15}{\sqrt{25}}, \; 100 + 1.96 \cdot \frac{15}{\sqrt{25}}\right) = (94.12, \; 105.88)
\]
Таким образом, с вероятностью 95% истинное среднее значение находится в интервале от 94.12 до 105.88.

#### 4.2 Доверительный интервал для среднего при неизвестной дисперсии

Если дисперсия $\sigma^2$ неизвестна, то вместо стандартного нормального распределения используется распределение Стьюдента с $n-1$ степенями свободы.

Доверительный интервал имеет вид:
$$
\left( \bar{X} - t_{\alpha/2, \; n-1} \frac{s}{\sqrt{n}}, \; \bar{X} + t_{\alpha/2, \; n-1} \frac{s}{\sqrt{n}} \right)
\]
где:
- $t_{\alpha/2, \; n-1}$ — квантиль распределения Стьюдента с $n-1$ степенями свободы,
- $s$ — выборочное стандартное отклонение:
$$
s = \sqrt{\frac{1}{n-1} \sum_{i=1}^n (X_i - \bar{X})^2}
\]

#### 4.3 Доверительный интервал для пропорции

Пусть $X$ — число элементов в выборке объема $n$, обладающих некоторым признаком, тогда оценка пропорции равна $\hat{p} = \frac{X}{n}$.

Для больших выборок (при $n \hat{p} (1-\hat{p}) \geq 5$) доверительный интервал для пропорции можно найти по формуле:
$$
\left( \hat{p} - z_{\alpha/2} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}, \; \hat{p} + z_{\alpha/2} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}} \right)
\]

### 5. Интерпретация доверительных интервалов

При интерпретации доверительного интервала важно понимать, что он характеризует не вероятность нахождения параметра внутри определенного интервала (параметр — фиксированное значение), а вероятность того, что случайно выбранный интервал накроет это истинное значение.

Например, если мы построили 95% доверительный интервал, то это означает, что при многократном повторении эксперимента 95% построенных интервалов будут содержать истинное значение параметра.








Давайте рассмотрим 6 конкретных задач, связанных с построением доверительных интервалов. Каждая задача будет сопровождаться пошаговым решением.

### Задача 1: Доверительный интервал для среднего при известной дисперсии

**Условие:**
В результате исследования среднего времени, затрачиваемого на выполнение задачи, были получены следующие данные (в минутах): 15, 18, 16, 20, 14, 17. Известно, что дисперсия времени выполнения задачи составляет $\sigma^2 = 4$. Построить 95% доверительный интервал для среднего времени выполнения задачи.

**Решение:**

1. **Найти выборочное среднее:**
   $$
   \bar{X} = \frac{15 + 18 + 16 + 20 + 14 + 17}{6} = \frac{100}{6} \approx 16.67
  $$

2. **Определить объем выборки:**
   $$
   n = 6
  $$

3. **Вычеслить стандартное отклонение:**
   $$
   \sigma = \sqrt{4} = 2
  $$

4. **Определить значение квантиля $z_{\alpha/2}$ для 95% доверительного интервала:**
   $$
   z_{0.025} \approx 1.96
  $$

5. **Построить доверительный интервал:**
   $$
   \text{ДИ} = \left( \bar{X} - z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}, \bar{X} + z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \right)
  $$
   $$
   = \left( 16.67 - 1.96 \cdot \frac{2}{\sqrt{6}}, 16.67 + 1.96 \cdot \frac{2}{\sqrt{6}} \right)
  $$
   $$
   = \left( 16.67 - 1.96 \cdot 0.8165, 16.67 + 1.96 \cdot 0.8165 \right)
  $$
   $$
   = \left( 16.67 - 1.60, 16.67 + 1.60 \right) \approx (15.07, 18.27)
  $$

**Ответ:** 95% доверительный интервал для среднего времени выполнения задачи: $(15.07, 18.27)$.



### Задача 2: Доверительный интервал для среднего при неизвестной дисперсии

**Условие:**
Данные о росте (в см) 10 случайно выбранных людей составляют: 172, 175, 168, 180, 177, 171, 169, 173, 174, 176. Построить 90% доверительный интервал для среднего роста, если дисперсия неизвестна.

**Решение:**

1. **Найти выборочное среднее:**
   $$
   \bar{X} = \frac{172 + 175 + 168 + 180 + 177 + 171 + 169 + 173 + 174 + 176}{10} = \frac{1685}{10} = 168.5
  $$

2. **Определить объем выборки:**
   $$
   n = 10
  $$

3. **Вычислить выборочное стандартное отклонение:**
   $$
   s = \sqrt{\frac{1}{n-1} \sum_{i=1}^n (X_i - \bar{X})^2}
  $$
   $$
   = \sqrt{\frac{1}{9} \left((172 - 174.5)^2 + (175 - 174.5)^2 + (168 - 174.5)^2 + (180 - 174.5)^2 + (177 - 174.5)^2 + (171 - 174.5)^2 + (169 - 174.5)^2 + (173 - 174.5)^2 + (174 - 174.5)^2 + (176 - 174.5)^2\right)}
  $$
   $$
   = \sqrt{\frac{1}{9}(6.25 + 0.25 + 42.25 + 30.25 + 6.25 + 12.25 + 30.25 + 2.25 + 0.25 + 2.25)} \approx \sqrt{8.33} \approx 2.89
  $$

4. **Определить значение квантиля $t_{\alpha/2, n-1}$ для 90% доверительного интервала:**
   $$
   t_{0.05, 9} \approx 1.833
  $$

5. **Построить доверительный интервал:**
   $$
   \text{ДИ} = \left( \bar{X} - t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}}, \bar{X} + t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}} \right)
  $$
   $$
   = \left( 174.5 - 1.833 \cdot \frac{2.89}{\sqrt{10}}, 174.5 + 1.833 \cdot \frac{2.89}{\sqrt{10}} \right)
  $$
   $$
   = \left( 174.5 - 1.833 \cdot 0.915, 174.5 + 1.833 \cdot 0.915 \right)
  $$
   $$
   = \left( 174.5 - 1.68, 174.5 + 1.68 \right) \approx (172.82, 176.18)
  $$

**Ответ:** 90% доверительный интервал для среднего роста: $(172.82, 176.18)$.



### Задача 3: Доверительный интервал для пропорции

**Условие:**
В опросе, проведенном среди 200 человек, 120 высказали свое одобрение новому закону. Построить 95% доверительный интервал для пропорции одобряющих.

**Решение:**

1. **Определить пропорцию:**
   $$
   \hat{p} = \frac{X}{n} = \frac{120}{200} = 0.6
  $$

2. **Определить объем выборки:**
   $$
   n = 200
  $$

3. **Вычислить стандартную ошибку пропорции:**
   $$
   SE = \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}} = \sqrt{\frac{0.6(1 - 0.6)}{200}} = \sqrt{\frac{0.6 \cdot 0.4}{200}} = \sqrt{0.0012} \approx 0.03464
  $$

4. **Определить значение квантиля $z_{\alpha/2}$ для 95% доверительного интервала:**
   $$
   z_{0.025} \approx 1.96
  $$

5. **Построить доверительный интервал:**
   $$
   \text{ДИ} = \left( \hat{p} - z_{\alpha/2} \cdot SE, \hat{p} + z_{\alpha/2} \cdot SE \right)
  $$
   $$
   = \left( 0.6 - 1.96 \cdot 0.03464, 0.6 + 1.96 \cdot 0.03464 \right)
  $$
   $$
   = \left( 0.6 - 0.06793, 0.6 + 0.06793 \right) \approx (0.53207, 0.66793)
  $$

**Ответ:** 95% доверительный интервал для пропорции одобряющих: $(0.532, 0.668)$.



### Задача 4: Доверительный интервал для разности средних

**Условие:**
Сравниваются две группы студентов. Средний балл первой группы (объем 30) равен 75, а стандартное отклонение 10. Средний балл второй группы (объем 35) равен 80, а стандартное отклонение 12. Построить 95% доверительный интервал для разности средних.

**Решение:**

1. **Определить выборочные средние и стандартные отклонения:**
   - Первая группа: $\bar{X_1} = 75, \; s_1 = 10, \; n_1 = 30$
   - Вторая группа: $\bar{X_2} = 80, \; s_2 = 12, \; n_2 =

 35$

2. **Вычислить стандартную ошибку разности средних:**
   $$
   SE = \sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}} = \sqrt{\frac{10^2}{30} + \frac{12^2}{35}} = \sqrt{\frac{100}{30} + \frac{144}{35}} = \sqrt{3.33 + 4.11} = \sqrt{7.44} \approx 2.73
  $$

3. **Вычислить разность средних:**
   $$
   D = \bar{X_1} - \bar{X_2 = 75 - 80 = -5}
  $$

4. **Определить значение $z_{\alpha/2}$ для 95% доверительного интервала:**
   $$
   z_{0.025} \approx 1.96
  $$

5. **Построить доверительный интервал:**
   $$
   \text{ДИ} = \left( D - z_{\alpha/2} \cdot SE, D + z_{\alpha/2} \cdot SE \right)
  $$
   $$
   = \left( -5 - 1.96 \cdot 2.73, -5 + 1.96 \cdot 2.73 \right)
  $$
   $$
   = \left( -5 - 5.35, -5 + 5.35 \right) \approx (-10.35, 0.35)
  $$

**Ответ:** 95% доверительный интервал для разности средних: $(-10.35, 0.35)$.



### Задача 5: Доверительный интервал для отношения дисперсий

**Условие:**
В двух группах данных (по 15 наблюдений) дисперсии известны: для первой группы $s_1^2 = 20$, для второй $s_2^2 = 30$. Построить 95% доверительный интервал для отношения дисперсий $\frac{\sigma_1^2}{\sigma_2^2}$.

**Решение:**

1. **Определить объем выборок:**
   $$
   n_1 = n_2 = 15
  $$

2. **Вычислить $F$-статистику:**
   $$
   F = \frac{s_1^2}{s_2^2} = \frac{20}{30} = \frac{2}{3} \approx 0.67
  $$

3. **Определить $F_{\alpha/2}$ и $F_{1-\alpha/2}$ для 95% доверительного интервала:**
   $$
   F_{0.025, 14, 14} \approx 2.73, \quad F_{0.975, 14, 14} \approx 0.37
  $$

4. **Построить доверительный интервал:**
   $$
   \text{ДИ} = \left( \frac{s_1^2}{s_2^2} \cdot \frac{1}{F_{1-\alpha/2}}, \frac{s_1^2}{s_2^2} \cdot F_{\alpha/2} \right)
  $$
   $$
   = \left( \frac{20}{30} \cdot \frac{1}{2.73}, \frac{20}{30} \cdot 2.73 \right)
  $$
   $$
   = \left( 0.67 \cdot 0.366, 0.67 \cdot 2.73 \right) \approx (0.25, 1.83)
  $$

**Ответ:** 95% доверительный интервал для отношения дисперсий: $(0.25, 1.83)$.



### Задача 6: Доверительный интервал для средней оценки по 5 наблюдениям

**Условие:**
На экзамене по математике 5 студентов получили следующие оценки: 85, 90, 78, 92, 88. Построить 95% доверительный интервал для средней оценки, если дисперсия неизвестна.

**Решение:**

1. **Найти выборочное среднее:**
   $$
   \bar{X} = \frac{85 + 90 + 78 + 92 + 88}{5} = \frac{433}{5} = 86.6
  $$

2. **Определить объем выборки:**
   $$
   n = 5
  $$

3. **Вычислить выборочное стандартное отклонение:**
   $$
   s = \sqrt{\frac{1}{n-1} \sum_{i=1}^n (X_i - \bar{X})^2}
  $$
   $$
   = \sqrt{\frac{1}{4} \left((85 - 86.6)^2 + (90 - 86.6)^2 + (78 - 86.6)^2 + (92 - 86.6)^2 + (88 - 86.6)^2\right)}
  $$
   $$
   = \sqrt{\frac{1}{4} (2.56 + 11.56 + 73.96 + 28.96 + 1.96)} = \sqrt{\frac{118.0}{4}} = \sqrt{29.5} \approx 5.43
  $$

4. **Определить значение $t_{\alpha/2}$ для 95% доверительного интервала:**
   $$
   t_{0.025, 4} \approx 2.776
  $$

5. **Построить доверительный интервал:**
   $$
   \text{ДИ} = \left( \bar{X} - t_{\alpha/2} \cdot \frac{s}{\sqrt{n}}, \bar{X} + t_{\alpha/2} \cdot \frac{s}{\sqrt{n}} \right)
  $$
   $$
   = \left( 86.6 - 2.776 \cdot \frac{5.43}{\sqrt{5}}, 86.6 + 2.776 \cdot \frac{5.43}{\sqrt{5}} \right)
  $$
   $$
   = \left( 86.6 - 2.776 \cdot 2.43, 86.6 + 2.776 \cdot 2.43 \right)
  $$
   $$
   = \left( 86.6 - 6.73, 86.6 + 6.73 \right) \approx (79.87, 93.33)
  $$

**Ответ:** 95% доверительный интервал для средней оценки: $(79.87, 93.33)$.



Эти примеры показывают, как применять концепции доверительных интервалов на практике. Каждый из них включает в себя важные статистические шаги, которые помогут лучше понять, как работает статистика.



Чтобы реализовать задачи на Python с визуализацией, мы будем использовать библиотеки `numpy`, `scipy`, и `matplotlib`. Для каждой задачи мы сначала проведем необходимые вычисления, а затем визуализируем доверительные интервалы с помощью графиков.



```python

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Функция для визуализации доверительного интервала
def plot_confidence_interval(data, confidence, label):
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Используем выборочное стандартное отклонение
    n = len(data)
    h = std / np.sqrt(n) * stats.t.ppf((1 + confidence) / 2., n - 1)  # Доверительный интервал
    plt.errorbar(mean, 0, xerr=h, fmt='o', label=label)

# Задача 1: Доверительный интервал для среднего при известной дисперсии
data1 = np.array([15, 18, 16, 20, 14, 17])
sigma1 = 2  # Известное стандартное отклонение
confidence1 = 0.95

plot_confidence_interval(data1, confidence1, 'Задача 1: ДИ при известной дисперсии')

# Задача 2: Доверительный интервал для среднего при неизвестной дисперсии
data2 = np.array([172, 175, 168, 180, 177, 171, 169, 173, 174, 176])
confidence2 = 0.90

plot_confidence_interval(data2, confidence2, 'Задача 2: ДИ при неизвестной дисперсии')

# Задача 3: Доверительный интервал для пропорции
n3 = 200
x3 = 120
p_hat = x3 / n3
std3 = np.sqrt(p_hat * (1 - p_hat) / n3)
confidence3 = 0.95
h3 = stats.norm.ppf((1 + confidence3) / 2) * std3
plt.errorbar(p_hat, 0, xerr=h3, fmt='o', label='Задача 3: ДИ для пропорции')

# Задача 4: Доверительный интервал для разности средних
data4_group1 = np.array([75, 75, 75, 75, 75, 75, 75, 75, 75, 75, 75, 75, 75, 75, 75]) + np.random.normal(0, 10, 15)
data4_group2 = np.array([80, 80, 80, 80, 80, 80, 80, 80, 80, 80, 80, 80, 80, 80, 80]) + np.random.normal(0, 12, 15)
mean1, mean2 = np.mean(data4_group1), np.mean(data4_group2)
std1, std2 = np.std(data4_group1, ddof=1), np.std(data4_group2, ddof=1)
n1, n2 = len(data4_group1), len(data4_group2)
h4 = stats.t.ppf((1 + 0.95) / 2, n1 + n2 - 2) * np.sqrt(std1**2 / n1 + std2**2 / n2)
plt.errorbar(mean1 - mean2, 0, xerr=h4, fmt='o', label='Задача 4: ДИ для разности средних')

# Задача 5: Доверительный интервал для отношения дисперсий
s1_squared = 20
s2_squared = 30
n5 = 15
f_lower = stats.f.ppf(0.025, n5-1, n5-1)
f_upper = stats.f.ppf(0.975, n5-1, n5-1)
ratio = s1_squared / s2_squared
conf_interval = (ratio / f_upper, ratio / f_lower)
plt.errorbar(ratio, 0, xerr=[[ratio - conf_interval[0]], [conf_interval[1] - ratio]], fmt='o', label='Задача 5: ДИ для отношения дисперсий')

# Задача 6: Доверительный интервал для средней оценки
data6 = np.array([85, 90, 78, 92, 88])
confidence6 = 0.95

plot_confidence_interval(data6, confidence6, 'Задача 6: ДИ для средней оценки')

# Настройки графика
plt.title('Доверительные интервалы для различных задач')
plt.yticks([])
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid()
plt.show()
```

#6. Статистическое моделирование и оценивание параметров

Статистическое моделирование и оценивание параметров — важные разделы математической статистики, позволяющие проводить анализ данных, делать выводы о генеральной совокупности на основании выборки и проверять статистические гипотезы. В этой лекции рассмотрим основные концепции проверки гипотез и тестирования статистической значимости.

## 1. Проверка гипотез

### 1.1 Нулевая и альтернативная гипотезы

При проверке статистических гипотез выделяют две основные гипотезы:

- **Нулевая гипотеза ($H_0$)**: гипотеза, которая предполагает отсутствие эффекта или отличия. Это утверждение о том, что наблюдаемые данные соответствуют известному или предполагаемому распределению.

- **Альтернативная гипотеза ($H_1$)**: гипотеза, противоположная нулевой, предполагающая наличие эффекта или значимого отличия.

Пример: при проверке того, влияет ли лекарство на выздоровление пациентов, нулевая гипотеза может утверждать, что лекарство не имеет влияния, в то время как альтернативная гипотеза будет утверждать, что лекарство имеет значительное влияние.

### 1.2 Критерий значимости

**Критерий значимости** — это предельно допустимая вероятность ошибки первого рода (отказ от нулевой гипотезы, когда она истинна). Обычно его обозначают буквой $\alpha$ и устанавливают на уровне 0.05 или 0.01.

Критерий значимости позволяет определить границу, при которой нулевая гипотеза будет отвергнута в пользу альтернативной. Если рассчитанное значение тестовой статистики оказывается менее вероятным, чем пороговое значение, определенное $\alpha$, то нулевая гипотеза отвергается.

### 1.3 Ошибка первого и второго рода

- **Ошибка первого рода ($\alpha$)**: ошибочное отклонение истинной нулевой гипотезы. Вероятность ошибки первого рода равна уровню значимости теста.

- **Ошибка второго рода ($\beta$)**: принятие ложной нулевой гипотезы, то есть неспособность отклонить нулевую гипотезу, когда альтернативная гипотеза истинна.

**Мощность теста** ($1 - \beta$) — вероятность того, что тест правильно отвергнет ложную нулевую гипотезу.

#### Пример
Пусть $H_0$ — гипотеза о том, что среднее значение популяции равно 100 ($\mu = 100$), а $H_1$ — гипотеза о том, что среднее значение не равно 100. Если уровень значимости установлен на уровне $\alpha = 0.05$, то вероятность ошибочного отклонения гипотезы $H_0$, когда она верна, равна 5%.

## 2. Тесты статистической значимости

Существуют различные тесты, которые применяются в зависимости от типа данных и формулировки гипотезы. Рассмотрим основные тесты: t-тест, F-тест, тесты хи-квадрат и критерий согласия.

### 2.1 T-тест

t-тест используется для проверки гипотез о средних значениях. Он бывает различных типов:

- **t-тест для одной выборки**: проверяет, равно ли среднее значение выборки заданному значению.

- **t-тест для двух независимых выборок**: сравнивает средние значения двух независимых групп.

- **Парный t-тест**: проверяет средние значения двух связанных (зависимых) выборок.

#### Формула t-статистики для одной выборки:

$$
t = \frac{\overline{x} - \mu}{\frac{s}{\sqrt{n}}},
$$

где $\overline{x}$ — выборочное среднее, $\mu$ — предполагаемое среднее значение, $s$ — стандартное отклонение выборки, $n$ — объем выборки.

#### Пример
Пусть у нас есть выборка из 30 наблюдений со средним значением 105 и стандартным отклонением 10. Необходимо проверить, отличается ли это среднее значение от 100. Для уровня значимости 0.05 рассчитанное значение t сравнивается с критическим значением распределения Стьюдента.

### 2.2 F-тест

F-тест используется для сравнения дисперсий двух или более выборок и часто применяется в анализе дисперсии (ANOVA).

#### Формула F-статистики:

$$
F = \frac{S_1^2}{S_2^2},
$$

где $S_1^2$ и $S_2^2$ — выборочные дисперсии двух групп.

#### Пример
Предположим, что имеется две выборки: одна с дисперсией 25 и объемом 20, другая с дисперсией 10 и объемом 30. Тогда F-статистика будет равна $25/10 = 2.5$. Сравниваем это значение с критическим по таблице распределения F при заданных степенях свободы.

### 2.3 Тесты хи-квадрат ($\chi^2$)

Тесты хи-квадрат применяются для проверки гипотез о частотах категориальных данных.

#### Формула хи-квадрат статистики:

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i},
$$

где $O_i$ — наблюдаемые частоты, $E_i$ — ожидаемые частоты.

### Пример

Пусть имеются данные по предпочтению напитков: кофе — 50, чай — 30, сок — 20. Ожидается, что предпочтения равны для всех категорий (по 33.33). Тогда хи-квадрат статистика будет рассчитана следующим образом:

$$
\chi^2 = \frac{(50 - 33.33)^2}{33.33} + \frac{(30 - 33.33)^2}{33.33} + \frac{(20 - 33.33)^2}{33.33} \approx 13.5.
$$

Сравниваем это значение с критическим по таблице распределения хи-квадрат.

### 2.4 Критерий согласия

Критерий согласия применяется для проверки гипотез о соответствии эмпирического распределения заданному теоретическому распределению.

#### Пример
Для проверки нормальности распределения данных можно использовать критерий согласия Колмогорова-Смирнова, сравнивая эмпирическую функцию распределения с теоретической.

### Пример 1: Одновыборочный t-тест

**Задача:** У нас есть выборка из 20 студентов, которая показала средний балл на экзамене равный 82, а стандартное отклонение составило 10. Проверим, отличается ли средний балл этой выборки от 85 на уровне значимости 0.05.

**Решение:**

1. **Нулевая гипотеза ($H_0$)**: средний балл равен 85 ($\mu = 85$).
   
   **Альтернативная гипотеза ($H_1$)**: средний балл не равен 85 ($\mu \neq 85$).

2. **Вычисляем t-статистику**:

   $$
   t = \frac{\overline{x} - \mu}{\frac{s}{\sqrt{n}}} = \frac{82 - 85}{\frac{10}{\sqrt{20}}} = \frac{-3}{\frac{10}{4.47}} = \frac{-3}{2.24} \approx -1.34
   $$

3. **Определяем критическое значение t** для 19 степеней свободы (20-1) и уровня значимости 0.05. Для двустороннего теста критические значения будут около $\pm 2.093$.

4. **Сравниваем рассчитанное значение t с критическим**. Поскольку $-1.34$ не попадает в критическую область ($-2.093 < -1.34 < 2.093$), нулевая гипотеза не отвергается.

**Вывод:** Нет оснований утверждать, что средний балл отличается от 85.



### Пример 2: Парный t-тест

**Задача:** Измерены результаты до и после тренировки для 10 человек. Проверим, улучшились ли показатели.

| Человек | До тренировки | После тренировки |
||||
| 1       | 68            | 75               |
| 2       | 75            | 78               |
| 3       | 80            | 85               |
| 4       | 82            | 88               |
| 5       | 90            | 92               |
| 6       | 76            | 80               |
| 7       | 84            | 89               |
| 8       | 70            | 77               |
| 9       | 78            | 82               |
| 10      | 85            | 90               |

**Решение:**

1. **Рассчитываем разности**:

   $$
   D_i = (После - До)
   $$

   Получаем следующие значения: 7, 3, 5, 6, 2, 4, 5, 7, 4, 5.

2. **Среднее значение разностей** ($\overline{D}$) равно:

   $$
   \overline{D} = \frac{7 + 3 + 5 + 6 + 2 + 4 + 5 + 7 + 4 + 5}{10} = 4.8
   $$

3. **Стандартное отклонение разностей**:

   $$
   s_D = \sqrt{\frac{\sum (D_i - \overline{D})^2}{n-1}} = \sqrt{\frac{(7-4.8)^2 + \ldots + (5-4.8)^2}{9}} = \sqrt{2.84} \approx 1.68
   $$

4. **Вычисляем t-статистику**:

   $$
   t = \frac{\overline{D}}{s_D / \sqrt{n}} = \frac{4.8}{1.68 / \sqrt{10}} \approx 9.04
   $$

5. **Критическое значение t** для 9 степеней свободы (10-1) и уровня значимости 0.05 равно 2.262.

6. Поскольку $9.04 > 2.262$, нулевая гипотеза отвергается.

**Вывод:** Показатели после тренировки значительно улучшились.



### Пример 3: F-тест для сравнения дисперсий

**Задача:** Проверим, различаются ли дисперсии двух независимых групп. Группа 1: $s_1^2 = 25$, объем $n_1 = 15$. Группа 2: $s_2^2 = 10$, объем $n_2 = 20$.

**Решение:**

1. **Вычисляем F-статистику**:

   $$
   F = \frac{s_1^2}{s_2^2} = \frac{25}{10} = 2.5
   $$

2. **Определяем критическое значение F** для степеней свободы $df_1 = 14$ и $df_2 = 19$ при уровне значимости 0.05. По таблице критическое значение F равно примерно 2.31.

3. **Сравнение с критическим значением**: $2.5 > 2.31$.

**Вывод:** Дисперсии значительно различаются.



### Пример 4: Тест хи-квадрат для независимости

**Задача:** Исследуем, связаны ли пол и предпочтение напитков. Данные приведены в таблице.

|           | Кофе | Чай | Сок |
|--||--|--|
| Мужчины   | 20   | 15  | 10  |
| Женщины   | 10   | 25  | 15  |

**Решение:**

1. **Вычисляем ожидаемые частоты**. Общие суммы: кофе — 30, чай — 40, сок — 25. Итого: мужчины — 45, женщины — 50. Ожидаемые частоты:

   $$
   E_{кофе, мужчины} = \frac{30 \cdot 45}{95} \approx 14.21, \quad E_{кофе, женщины} = \frac{30 \cdot 50}{95} \approx 15.79
   $$

   И так далее для остальных категорий.

2. **Вычисляем хи-квадрат статистику**:

   $$
   \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} = \frac{(20 - 14.21)^2}{14.21} + \ldots \approx 8.76
   $$

3. **Критическое значение хи-квадрат** при 2 степенях свободы и уровне значимости 0.05 равно 5.99.

4. Поскольку $8.76 > 5.99$, нулевая гипотеза отвергается.

**Вывод:** Пол и предпочтение напитков связаны.



### Пример 5: Тест хи-квадрат для проверки соответствия распределения

**Задача:** Проверим, соответствует ли распределение бросков кубика равномерному. Получены результаты: 1 — 8 раз, 2 — 10 раз, 3 — 12 раз, 4 — 15 раз, 5 — 8 раз, 6 — 7 раз. Всего — 60 бросков.

**Решение:**

1. **Ожидаемая частота** для каждого значения равна $60 / 6 = 10$.

2. **Вычисляем хи-квадрат статистику**:

   $$
   \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} = \frac{(8-10)^2}{10} + \frac{(10-10)^2}{10} + \ldots + \frac{(7-10)^2}{10} = \frac{4}{10} + 0 + \ldots + \frac{9}{10} \approx 3.4
   $$

3. **Критическое значение хи-квадрат** при 5 степенях свободы и уровне значимости 0.05 равно 11.07.

4. Поскольку $3.4 < 11.07$, нулевая гипотеза не отвергается.

**Вывод:** Распределение бросков кубика соответствует равномерному.



### Пример 6: Дисперсионный анализ (ANOVA)

**Задача:** Сравним средние значения трех групп студентов по времени, затраченному на изучение математики:
- Группа 1: 4, 5, 6
- Группа 2: 7, 8, 9
- Группа 3: 10, 11, 12

Проверим, есть ли статистически значимые различия между средними значениями этих групп на уровне значимости 0.05.

**Решение:**

1. **Формулируем гипотезы:**
   - **Нулевая гипотеза ($H_0$)**: Средние значения всех трех групп равны ($\mu_1 = \mu_2 = \mu_3$).
   - **Альтернативная гипотеза ($H_1$)**: По крайней мере, одна из средних отличается.

2. **Вычисляем общую сумму квадратов отклонений (SST), сумму квадратов отклонений между группами (SSB) и сумму квадратов отклонений внутри групп (SSW).**

   Сначала найдем общую среднюю величину для всех данных:
   $$
   \overline{X} = \frac{4 + 5 + 6 + 7 + 8 + 9 + 10 + 11 + 12}{9} = \frac{72}{9} = 8.
   $$

3. **Сумма квадратов отклонений между группами (SSB):**
   Вычисляем средние значения для каждой группы:
   - Группа 1: $\overline{X}_1 = \frac{4 + 5 + 6}{3} = 5$
   - Группа 2: $\overline{X}_2 = \frac{7 + 8 + 9}{3} = 8$
   - Группа 3: $\overline{X}_3 = \frac{10 + 11 + 12}{3} = 11$

   Теперь вычисляем SSB:
   $$
   SSB = n (\overline{X}_1 - \overline{X})^2 + n (\overline{X}_2 - \overline{X})^2 + n (\overline{X}_3 - \overline{X})^2
   $$
   где $n = 3$ — количество наблюдений в каждой группе.

   Подставляем значения:
   $$
   SSB = 3 (5 - 8)^2 + 3 (8 - 8)^2 + 3 (11 - 8)^2 = 3 \cdot 9 + 3 \cdot 0 + 3 \cdot 9 = 27 + 0 + 27 = 54.
   $$

4. **Сумма квадратов отклонений внутри групп (SSW):**
   $$
   SSW = \sum_{i=1}^{3} \sum_{j=1}^{3} (X_{ij} - \overline{X}_i)^2
   $$
   где $X_{ij}$ — наблюдения в группах, $\overline{X}_i$ — среднее каждой группы.

   Для Группы 1:
   $$
   (4-5)^2 + (5-5)^2 + (6-5)^2 = 1 + 0 + 1 = 2.
   $$

   Для Группы 2:
   $$
   (7-8)^2 + (8-8)^2 + (9-8)^2 = 1 + 0 + 1 = 2.
   $$

   Для Группы 3:
   $$
   (10-11)^2 + (11-11)^2 + (12-11)^2 = 1 + 0 + 1 = 2.
   $$

   Таким образом, $SSW = 2 + 2 + 2 = 6$.

5. **Общая сумма квадратов отклонений (SST):**
   $$
   SST = SSB + SSW = 54 + 6 = 60.
   $$

6. **Вычисляем степени свободы:**
   - Для междугрупповой дисперсии ($df_{between}$): $k - 1 = 3 - 1 = 2$, где $k$ — количество групп.
   - Для внутригрупповой дисперсии ($df_{within}$): $N - k = 9 - 3 = 6$, где $N$ — общее число наблюдений.

7. **Вычисляем средние квадраты отклонений:**
   - Между группами: $\text{MSB} = \frac{SSB}{df_{between}} = \frac{54}{2} = 27$.
   - Внутри групп: $\text{MSW} = \frac{SSW}{df_{within}} = \frac{6}{6} = 1$.

8. **Вычисляем F-статистику:**
   $$
   F = \frac{\text{MSB}}{\text{MSW}} = \frac{27}{1} = 27.
   $$

9. **Определяем критическое значение F** при уровне значимости 0.05 и степенях свободы $df_{between} = 2$ и $df_{within} = 6$. По таблице критических значений F критическое значение составляет примерно 5.14.

10. **Сравниваем F-статистику с критическим значением**: $27 > 5.14$, поэтому нулевая гипотеза отвергается.

**Вывод:** Средние значения времени, затраченного на изучение математики, статистически значимо различаются между группами.




Для решения всех шести задач на Python и визуализации результатов будем использовать библиотеки `numpy`, `scipy.stats`, и `matplotlib`.


### Пример 1: Одновыборочный t-тест

```python
import numpy as np
from scipy.stats import ttest_1samp
import matplotlib.pyplot as plt

# Данные
sample = np.random.normal(loc=82, scale=10, size=20)
mu = 85  # Предполагаемое среднее значение

# Одновыборочный t-тест
t_stat, p_value = ttest_1samp(sample, mu)

# Визуализация
plt.hist(sample, bins=10, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(mu, color='red', linestyle='dashed', linewidth=2, label=f'Предполагаемое среднее = {mu}')
plt.axvline(np.mean(sample), color='green', linestyle='solid', linewidth=2, label=f'Выборочное среднее = {np.mean(sample):.2f}')
plt.legend()
plt.title(f'Одновыборочный t-тест: t = {t_stat:.2f}, p = {p_value:.3f}')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.show()
```

### Пример 2: Двухвыборочный t-тест для независимых выборок


```python
from scipy.stats import ttest_ind

# Данные
group1 = np.random.normal(loc=78, scale=8, size=15)
group2 = np.random.normal(loc=85, scale=10, size=15)

# Двухвыборочный t-тест
t_stat, p_value = ttest_ind(group1, group2)

# Визуализация
plt.hist(group1, bins=10, alpha=0.5, label='Группа 1', color='blue', edgecolor='black')
plt.hist(group2, bins=10, alpha=0.5, label='Группа 2', color='green', edgecolor='black')
plt.legend()
plt.title(f'Двухвыборочный t-тест: t = {t_stat:.2f}, p = {p_value:.3f}')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.show()
```

### Пример 3: Парный t-тест

```python
from scipy.stats import ttest_rel

# Данные
before = np.random.normal(loc=70, scale=5, size=10)
after = before + np.random.normal(loc=5, scale=5, size=10)

# Парный t-тест
t_stat, p_value = ttest_rel(before, after)

# Визуализация
plt.plot(before, 'o-', label='До обучения', color='red')
plt.plot(after, 's-', label='После обучения', color='green')
plt.legend()
plt.title(f'Парный t-тест: t = {t_stat:.2f}, p = {p_value:.3f}')
plt.xlabel('Индексы')
plt.ylabel('Значения')
plt.show()
```


### Пример 4: F-тест для сравнения дисперсий


```python
from scipy.stats import f

# Данные
var1 = 25
var2 = 40
n1 = 10
n2 = 12

# F-статистика
f_stat = var1 / var2
df1 = n1 - 1
df2 = n2 - 1
p_value = 1 - f.cdf(f_stat, df1, df2)

# Визуализация
x = np.linspace(0, 5, 100)
y = f.pdf(x, df1, df2)
plt.plot(x, y, label='Распределение F')
plt.axvline(f_stat, color='red', linestyle='dashed', label=f'F-статистика = {f_stat:.2f}')
plt.legend()
plt.title(f'F-тест: p = {p_value:.3f}')
plt.xlabel('Значение F')
plt.ylabel('Плотность вероятности')
plt.show()
```



### Пример 5: Тест хи-квадрат для независимости


```python
from scipy.stats import chi2_contingency

# Данные
observed = np.array([[20, 15], [10, 25], [30, 20]])

# Тест хи-квадрат
chi2_stat, p_value, dof, expected = chi2_contingency(observed)

# Визуализация
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(observed, cmap='Blues', interpolation='nearest')
ax[0].set_title('Наблюдаемые значения')
ax[1].imshow(expected, cmap='Blues', interpolation='nearest')
ax[1].set_title('Ожидаемые значения')

plt.suptitle(f'Тест хи-квадрат: χ² = {chi2_stat:.2f}, p = {p_value:.3f}')
plt.show()
```


### Пример 6: Дисперсионный анализ (ANOVA)


```python
from scipy.stats import f_oneway

# Данные
group1 = [4, 5, 6]
group2 = [7, 8, 9]
group3 = [10, 11, 12]

# ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)

# Визуализация
data = [group1, group2, group3]
plt.boxplot(data, labels=['Группа 1', 'Группа 2', 'Группа 3'])
plt.title(f'ANOVA: F = {f_stat:.2f}, p = {p_value:.3f}')
plt.ylabel('Значения')
plt.show()
```
