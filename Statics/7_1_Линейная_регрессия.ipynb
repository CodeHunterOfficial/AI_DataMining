#7.1. Линейная регрессия

#### Введение

Линейная регрессия — это один из базовых и наиболее используемых методов машинного обучения и статистики, предназначенный для прогнозирования зависимости одной переменной от другой (или нескольких других). Этот метод особенно популярен благодаря своей простоте и интерпретируемости. В данной лекции мы рассмотрим  одномерную линейную регрессию , также известную как  простая линейная регрессия .

### 1. Постановка задачи

Предположим, что у нас есть набор данных, состоящий из $n$ наблюдений $\{ (x_i, y_i) \}_{i=1}^{n}$, где $x_i$ — это независимая переменная (признак), а $y_i$ — зависимая переменная (целевой признак). Задача линейной регрессии заключается в том, чтобы найти такую линейную зависимость между $x_i$ и $y_i$, которая наилучшим образом описывает данные.

Математически модель линейной регрессии записывается в виде:

$$
y_i = \beta_0 + \beta_1 x_i + \epsilon_i
$$

где:
- $\beta_0$ — свободный член (иногда называемый  перехват ),
- $\beta_1$ — коэффициент наклона (или  коэффициент регрессии ),
- $\epsilon_i$ — случайная ошибка, которая считается нормально распределенной с математическим ожиданием $\mathbb{E}[\epsilon_i] = 0$ и дисперсией $\sigma^2$.

Цель состоит в том, чтобы оценить параметры $\beta_0$ и $\beta_1$, которые минимизируют ошибку между предсказанными значениями $\hat{y}_i = \beta_0 + \beta_1 x_i$ и фактическими значениями $y_i$.

### 2. Метод наименьших квадратов (МНК)

Самым популярным методом для оценки коэффициентов линейной регрессии является  метод наименьших квадратов  (МНК). Он минимизирует сумму квадратов отклонений предсказанных значений $\hat{y}_i$ от реальных значений $y_i$:

$$
S(\beta_0, \beta_1) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \sum_{i=1}^{n} \left( y_i - (\beta_0 + \beta_1 x_i) \right)^2
$$

Задача состоит в минимизации этой функции по $\beta_0$ и $\beta_1$.

### 3. Вывод формул для коэффициентов

Чтобы найти значения $\beta_0$ и $\beta_1$, при которых сумма квадратов отклонений минимальна, нам нужно взять производные этой суммы по $\beta_0$ и $\beta_1$ и приравнять их к нулю.

#### Производная по $\beta_1$:

$$
\frac{\partial S}{\partial \beta_1} = -2 \sum_{i=1}^{n} x_i (y_i - \beta_0 - \beta_1 x_i)
$$

#### Производная по $\beta_0$:

$$
\frac{\partial S}{\partial \beta_0} = -2 \sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i)
$$

Приравняем обе производные к нулю:

$$
\sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i) = 0
$$

$$
\sum_{i=1}^{n} x_i (y_i - \beta_0 - \beta_1 x_i) = 0
$$

Из первого уравнения можно вывести $\beta_0$:

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

где $\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$ — среднее значение $x$, а $\bar{y} = \frac{1}{n} \sum_{i=1}^{n} y_i$ — среднее значение $y$.

Теперь подставим выражение для $\beta_0$ во второе уравнение и решим его относительно $\beta_1$:

$$
\sum_{i=1}^{n} x_i (y_i - (\bar{y} - \beta_1 \bar{x}) - \beta_1 x_i) = 0
$$

Упростим это уравнение:

$$
\sum_{i=1}^{n} x_i (y_i - \bar{y}) = \beta_1 \sum_{i=1}^{n} x_i (x_i - \bar{x})
$$

Отсюда:

$$
\beta_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

Коэффициент $\beta_0$ можно найти по формуле:

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

Таким образом, мы получили следующие формулы для оценки коэффициентов линейной регрессии:

$$
\beta_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

### 4. Пример вычисления

Рассмотрим простой пример, где у нас есть набор данных:

| $x_i$ | $y_i$ |
|  --|  --|
|   1    |   1    |
|   2    |   2    |
|   3    |   2    |
|   4    |   4    |
|   5    |   4    |

1. Рассчитаем средние значения $\bar{x}$ и $\bar{y}$:

$$
\bar{x} = \frac{1 + 2 + 3 + 4 + 5}{5} = 3
$$

$$
\bar{y} = \frac{1 + 2 + 2 + 4 + 4}{5} = 2.6
$$

2. Рассчитаем числитель и знаменатель для коэффициента $\beta_1$:

$$
\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y}) = (1 - 3)(1 - 2.6) + (2 - 3)(2 - 2.6) + (3 - 3)(2 - 2.6) + (4 - 3)(4 - 2.6) + (5 - 3)(4 - 2.6)
$$

$$
= (-2)(-1.6) + (-1)(-0.6) + (0)(-0.6) + (1)(1.4) + (2)(1.4) = 3.2 + 0.6 + 0 + 1.4 + 2.8 = 8
$$

$$
\sum_{i=1}^{n} (x_i - \bar{x})^2 = (1 - 3)^2 + (2 - 3)^2 + (3 - 3)^2 + (4 - 3)^2 + (5 - 3)^2
$$

$$
= (-2)^2 + (-1)^2 + (0)^2 + (1)^2 + (2)^2 = 4 + 1 + 0 + 1 + 4 = 10
$$

3. Найдем $\beta_1$:

$$
\beta_1 = \frac{8}{10} = 0.8
$$

4. Найдем $\beta_0$:

$$
\beta_0 = \bar{y} - \beta_1 \bar{x} = 2.6 - 0.8 \times 3 = 2.6 - 2.4 = 0.2
$$

Итак, уравнение регрессии имеет вид:

$$
\hat{y} = 0.2 + 0.8x
$$

### 5. Оценка качества модели

После того, как модель построена, важно оценить её качество. Для этого существует несколько методов.

#### 5.1. Коэффициент детерминации $R^2$

Один из наиболее часто используемых показателей качества линейной регрессии — это  коэффициент детерминации  $R^2$. Он показывает,

 какая доля вариации зависимой переменной объясняется моделью. Формула для $R^2$ выглядит следующим образом:

$$
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
$$

Где:
- $\sum_{i=1}^{n} (y_i - \hat{y}_i)^2$ — сумма квадратов остатков ( остаточная сумма квадратов , RSS),
- $\sum_{i=1}^{n} (y_i - \bar{y})^2$ — общая сумма квадратов ( сумма квадратов отклонений , TSS).

Чем ближе значение $R^2$ к 1, тем лучше модель объясняет данные.

#### 5.2. Среднеквадратическая ошибка (MSE)

Среднеквадратическая ошибка показывает среднюю величину отклонений между предсказанными значениями и реальными значениями. Её можно рассчитать по следующей формуле:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

### 6. Свойства коэффициентов линейной регрессии

Коэффициенты $\beta_0$ и $\beta_1$ обладают следующими свойствами:

1.  Несмещённость : Оценки $\hat{\beta_0}$ и $\hat{\beta_1}$ являются несмещёнными, то есть их математическое ожидание равно истинным значениям параметров $\beta_0$ и $\beta_1$.

2.  Эффективность : Оценки параметров являются эффективными, то есть они имеют минимальную дисперсию среди всех несмещённых оценок.

3.  Нормальность : При нормальности ошибок $\epsilon_i$ коэффициенты $\hat{\beta_0}$ и $\hat{\beta_1}$ имеют нормальное распределение.

### 7. Заключение

Линейная регрессия — это мощный инструмент для анализа данных, который может использоваться в самых различных областях: от эконометрики до биостатистики.







Давайте рассмотрим два конкретных примера применения простой линейной регрессии, решив их шаг за шагом.

### Пример 1

Предположим, у нас есть следующие данные о количестве часов, проведенных за учебой, и оценках студентов на экзамене:

| Часы учебы ($x_i$) | Оценка ($y_i$) |
|       |     --|
| 1                   | 2               |
| 2                   | 3               |
| 3                   | 5               |
| 4                   | 4               |
| 5                   | 5               |

#### Шаг 1: Найдем средние значения

1.  Найдем среднее значение $\bar{x}$ (часы учебы):

$$
\bar{x} = \frac{1 + 2 + 3 + 4 + 5}{5} = \frac{15}{5} = 3
$$

2.  Найдем среднее значение $\bar{y}$ (оценки):

$$
\bar{y} = \frac{2 + 3 + 5 + 4 + 5}{5} = \frac{19}{5} = 3.8
$$

#### Шаг 2: Найдем коэффициент наклона $\beta_1$

Используем формулу:

$$
\beta_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

1.  Найдем числитель:

$$
\sum_{i=1}^{5} (x_i - \bar{x})(y_i - \bar{y}) = (1-3)(2-3.8) + (2-3)(3-3.8) + (3-3)(5-3.8) + (4-3)(4-3.8) + (5-3)(5-3.8)
$$

Рассчитаем по каждому элементу:

- Для $i = 1$: $(1-3)(2-3.8) = (-2)(-1.8) = 3.6$
- Для $i = 2$: $(2-3)(3-3.8) = (-1)(-0.8) = 0.8$
- Для $i = 3$: $(3-3)(5-3.8) = (0)(1.2) = 0$
- Для $i = 4$: $(4-3)(4-3.8) = (1)(0.2) = 0.2$
- Для $i = 5$: $(5-3)(5-3.8) = (2)(1.2) = 2.4$

Теперь сложим все результаты:

$$
\sum = 3.6 + 0.8 + 0 + 0.2 + 2.4 = 7
$$

2.  Найдем знаменатель:

$$
\sum_{i=1}^{n} (x_i - \bar{x})^2 = (1-3)^2 + (2-3)^2 + (3-3)^2 + (4-3)^2 + (5-3)^2
$$

Рассчитаем по каждому элементу:

- Для $i = 1$: $(1-3)^2 = 4$
- Для $i = 2$: $(2-3)^2 = 1$
- Для $i = 3$: $(3-3)^2 = 0$
- Для $i = 4$: $(4-3)^2 = 1$
- Для $i = 5$: $(5-3)^2 = 4$

Теперь сложим все результаты:

$$
\sum = 4 + 1 + 0 + 1 + 4 = 10
$$

Теперь можем найти $\beta_1$:

$$
\beta_1 = \frac{7}{10} = 0.7
$$

#### Шаг 3: Найдем свободный член $\beta_0$

Используем формулу:

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

Подставим значения:

$$
\beta_0 = 3.8 - 0.7 \cdot 3 = 3.8 - 2.1 = 1.7
$$

#### Шаг 4: Запишем уравнение регрессии

Таким образом, уравнение регрессии имеет вид:

$$
\hat{y} = 1.7 + 0.7x
$$

### Пример 2

Теперь рассмотрим другой набор данных о температуре (в градусах Цельсия) и количестве проданных мороженых:

| Температура ($x_i$) | Продажи мороженого ($y_i$) |
|       --|          |
| 10                    | 20                           |
| 12                    | 25                           |
| 15                    | 35                           |
| 18                    | 45                           |
| 20                    | 50                           |

#### Шаг 1: Найдем средние значения

1.  Найдем среднее значение $\bar{x}$ (температура):

$$
\bar{x} = \frac{10 + 12 + 15 + 18 + 20}{5} = \frac{75}{5} = 15
$$

2.  Найдем среднее значение $\bar{y}$ (продажи):

$$
\bar{y} = \frac{20 + 25 + 35 + 45 + 50}{5} = \frac{175}{5} = 35
$$

#### Шаг 2: Найдем коэффициент наклона $\beta_1$

Используем формулу:

$$
\beta_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

1.  Найдем числитель:

$$
\sum_{i=1}^{5} (x_i - \bar{x})(y_i - \bar{y}) = (10-15)(20-35) + (12-15)(25-35) + (15-15)(35-35) + (18-15)(45-35) + (20-15)(50-35)
$$

Рассчитаем по каждому элементу:

- Для $i = 1$: $(10-15)(20-35) = (-5)(-15) = 75$
- Для $i = 2$: $(12-15)(25-35) = (-3)(-10) = 30$
- Для $i = 3$: $(15-15)(35-35) = (0)(0) = 0$
- Для $i = 4$: $(18-15)(45-35) = (3)(10) = 30$
- Для $i = 5$: $(20-15)(50-35) = (5)(15) = 75$

Теперь сложим все результаты:

$$
\sum = 75 + 30 + 0 + 30 + 75 = 210
$$

2.  Найдем знаменатель:

$$
\sum_{i=1}^{n} (x_i - \bar{x})^2 = (10-15)^2 + (12-15)^2 + (15-15)^2 + (18-15)^2 + (20-15)^2
$$

Рассчитаем по каждому элементу:

- Для $i = 1$: $(10-15)^2 = 25$
- Для $i = 2$: $(12-15)^2 = 9$
- Для $i = 3$: $(15-15)^2 = 0$
- Для $i = 4$: $(18-15)^2 = 9$
- Для $i = 5$: $(20-15)^2 = 25$

Теперь сложим все результаты:

$$
\sum = 25 + 9 + 0 + 9 + 25 = 68
$$

Теперь можем найти $\beta_1$:

$$
\beta_1 = \frac{210}{68} \approx 3.09
$$

#### Шаг 3: Найдем свободный член $\beta_0$

Используем формулу:

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

Подставим значения:

$$
\beta_0 = 35 - 3.09 \cdot 15 \approx 35 - 46.35 \approx -11.35
$$

#### Шаг 4: Запишем уравнение регрессии

Таким образом, уравнение рег

рессии имеет вид:

$$
\hat{y} = -11.35 + 3.09x
$$

### Заключение

В обоих примерах мы успешно вычислили параметры линейной регрессии, что позволяет нам использовать полученные модели для прогнозирования значений зависимой переменной на основе независимой переменной.


Для реализации примеров линейной регрессии с визуализацией на Python мы будем использовать библиотеки `numpy`, `matplotlib` и `scikit-learn`.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Пример 1: Часы учебы и оценки
x1 = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Часы учебы
y1 = np.array([2, 3, 5, 4, 5])                 # Оценки

# Создание и обучение модели линейной регрессии
model1 = LinearRegression()
model1.fit(x1, y1)

# Получаем коэффициенты
beta_0_1 = model1.intercept_
beta_1_1 = model1.coef_[0]

# Предсказания
y_pred_1 = model1.predict(x1)

# Визуализация
plt.figure(figsize=(10, 6))
plt.scatter(x1, y1, color='blue', label='Данные')
plt.plot(x1, y_pred_1, color='red', label='Линейная регрессия')
plt.title('Линейная регрессия: Часы учебы и оценки')
plt.xlabel('Часы учебы')
plt.ylabel('Оценки')
plt.legend()
plt.grid()
plt.show()

# Вывод коэффициентов
print(f"Коэффициенты для примера 1: β0 = {beta_0_1:.2f}, β1 = {beta_1_1:.2f}")


# Пример 2: Температура и продажи мороженого
x2 = np.array([10, 12, 15, 18, 20]).reshape(-1, 1)  # Температура
y2 = np.array([20, 25, 35, 45, 50])                  # Продажи мороженого

# Создание и обучение модели линейной регрессии
model2 = LinearRegression()
model2.fit(x2, y2)

# Получаем коэффициенты
beta_0_2 = model2.intercept_
beta_1_2 = model2.coef_[0]

# Предсказания
y_pred_2 = model2.predict(x2)

# Визуализация
plt.figure(figsize=(10, 6))
plt.scatter(x2, y2, color='green', label='Данные')
plt.plot(x2, y_pred_2, color='orange', label='Линейная регрессия')
plt.title('Линейная регрессия: Температура и продажи мороженого')
plt.xlabel('Температура (°C)')
plt.ylabel('Продажи мороженого')
plt.legend()
plt.grid()
plt.show()

# Вывод коэффициентов
print(f"Коэффициенты для примера 2: β0 = {beta_0_2:.2f}, β1 = {beta_1_2:.2f}")
```
