#Глава 7. Матричные разложения

#### Введение

Матричные разложения — это фундаментальные инструменты в линейной алгебре и численных методах, которые позволяют представлять матрицы в различных формах, облегчая их обработку и анализ. Два из наиболее важных разложений, широко используемых в теории и практике, — это сингулярное разложение (SVD) и QR-разложение. Оба метода находят применение в решении систем линейных уравнений, вычислении собственных значений, сжатии данных и других численных задачах.



### 1. Сингулярное разложение (SVD)

#### Определение

Сингулярное разложение произвольной матрицы $A \in \mathbb{R}^{m \times n}$ — это представление матрицы в виде произведения трех матриц:

$$
A = U \Sigma V^T
$$

где:
- $U \in \mathbb{R}^{m \times m}$ — ортогональная матрица (столбцы которой являются ортонормированными векторами),
- $\Sigma \in \mathbb{R}^{m \times n}$ — диагональная матрица сингулярных значений,
- $V \in \mathbb{R}^{n \times n}$ — ортогональная матрица.

#### Теорема SVD

Каждая прямоугольная матрица $A \in \mathbb{R}^{m \times n}$ может быть разложена в виде $A = U \Sigma V^T$, где:
- $U$ и $V$ — ортогональные матрицы,
- $\Sigma$ — матрица с сингулярными значениями на диагонали.

##### Доказательство теоремы

1. Для начала найдем собственные значения матрицы $A^T A$, которая является симметричной матрицей размера $n \times n$. Так как симметричные матрицы диагонализуемы, существует такая ортогональная матрица $V$, что:
$$
   A^T A = V \Lambda V^T
$$
   где $\Lambda$ — диагональная матрица собственных значений (все собственные значения $\lambda_i \geq 0$).

2. Собственные значения $\lambda_i$ матрицы $A^T A$ — это квадраты сингулярных значений $\sigma_i$ матрицы $A$, т.е. $\sigma_i = \sqrt{\lambda_i}$.

3. Введем матрицу сингулярных значений $\Sigma$, которая содержит сингулярные значения $\sigma_i$ на диагонали. Столбцы матрицы $U$ определяются как нормализованные векторы $A v_i / \sigma_i$, где $v_i$ — столбцы из матрицы $V$, соответствующие ненулевым сингулярным значениям $\sigma_i$.

4. Таким образом, мы получаем разложение $A = U \Sigma V^T$.

#### Свойства сингулярного разложения

1. Ортогональность матриц $U$ и $V$:
$$
   U^T U = I_m, \quad V^T V = I_n
$$
   Это означает, что столбцы $U$ и $V$ являются ортонормированными.

2. Диагональная матрица $\Sigma$:
   Сингулярные значения $\sigma_1 \geq \sigma_2 \geq \ldots \geq \sigma_{\min(m,n)} \geq 0$. Эти значения равны квадратным корням собственных значений матрицы $A^T A$.

3. Ранг матрицы:
   Количество ненулевых сингулярных значений совпадает с рангом матрицы $A$.

4. Норма Фробениуса и спектральная норма:
   Норма Фробениуса матрицы $A$ может быть выражена через сингулярные значения:
$$
   \|A\|_F = \sqrt{\sum_{i=1}^{\min(m,n)} \sigma_i^2}
$$
   Спектральная норма:
$$
   \|A\|_2 = \sigma_1
$$
   где $\sigma_1$ — наибольшее сингулярное значение.

5. Псевдообратная матрица:
   Псевдообратная матрица $A^+$ (по Муру-Пенроузу) вычисляется как:
$$
   A^+ = V \Sigma^+ U^T
$$
   где $\Sigma^+$ — матрица, полученная обратным преобразованием ненулевых сингулярных значений матрицы $\Sigma$.

#### Применения SVD

1. Сжатие изображений: Сингулярное разложение позволяет приблизить исходную матрицу изображений матрицей более низкого ранга, сохраняя основные особенности изображения.
2. Решение плохо обусловленных систем линейных уравнений: SVD позволяет стабилизировать решение, используя только значимые сингулярные значения.
3. Анализ главных компонент (PCA): SVD используется для поиска главных компонент, уменьшающих размерность данных при минимальных потерях информации.



### 2. QR-разложение

#### Определение

QR-разложение матрицы $A \in \mathbb{R}^{m \times n}$ — это представление матрицы в виде произведения ортогональной матрицы $Q \in \mathbb{R}^{m \times m}$ и верхнетреугольной матрицы $R \in \mathbb{R}^{m \times n}$:

$$
A = QR
$$

где:
- $Q \in \mathbb{R}^{m \times m}$ — ортогональная матрица,
- $R \in \mathbb{R}^{m \times n}$ — верхнетреугольная матрица.

#### Теорема QR-разложения

Для любой матрицы $A \in \mathbb{R}^{m \times n}$ (где $m \geq n$) существует разложение $A = QR$, где $Q$ — ортогональная матрица, а $R$ — верхнетреугольная матрица.

##### Доказательство теоремы

QR-разложение можно получить с помощью процедуры Грамма-Шмидта, которая позволяет ортогонализовать столбцы матрицы $A$.

1. Алгоритм Грамма-Шмидта:
   Рассмотрим столбцы матрицы $A = [a_1, a_2, \dots, a_n]$. Процедура Грамма-Шмидта заключается в последовательном ортогонализировании этих столбцов:
   - Пусть $q_1 = \frac{a_1}{\|a_1\|}$,
   - Следующие векторы вычисляются как:
 $$
     r_{i,j} = a_j^T q_i, \quad a_j := a_j - \sum_{i=1}^{j-1} r_{i,j} q_i
 $$
   - Каждый новый вектор нормализуется: $q_j = \frac{a_j}{\|a_j\|}$.

2. В результате мы получаем ортонормированный набор векторов $Q = [q_1, q_2, \dots, q_n]$, и верхнетреугольную матрицу $R$, которая содержит коэффициенты ортогонализации $r_{i,j}$.

#### Свойства QR-разложения

1. Ортогональность матрицы $Q$:
$$
   Q^T Q = I
$$
   Это свойство гарантирует, что столбцы матрицы $Q$ являются ортонормированными.

2. Верхнетреугольная структура матрицы $R$:
   Все элементы ниже главной диагонали матрицы $R$ равны нулю.

3. Уникальность:
   Если матрица $A$ имеет полный ранг, то разложение $A = QR$ уникально при условии, что диагональные элементы матрицы $R$ положительны.

#### Применения QR-разложения

1. Решение систем линейных уравнений: Разложение QR используется для решения переопределенных систем уравнений $Ax = b$ через метод наименьших квадратов

2. Вычисление собственных значений: QR-алгоритм используется для вычисления собственных значений и собственных векторов матрицы.
3. Ортогонализация и ортогональные преобразования: QR-разложение используется для ортогонализации базисов и вычисления QR-преобразований.





### Свойства QR-разложения (продолжение)

4. Численная устойчивость:
   QR-разложение является численно устойчивым методом. Ортогональные матрицы $Q$ хорошо сохраняют числовую точность при вычислениях, так как такие матрицы не изменяют длину векторов при умножении на них, что помогает избегать накопления ошибок при численных расчетах.

5. Методы вычисления QR-разложения:
   QR-разложение может быть вычислено различными способами, в том числе:

   - Метод Грамма-Шмидта, как было описано выше.
   - Метод отражений Хаусхолдера (Householder reflections), который обеспечивает более стабильные численные результаты, чем метод Грамма-Шмидта.
   - Метод вращений Гивенса (Givens rotations), который позволяет эффективно вычислять QR-разложение для разреженных матриц и используется в оптимизационных задачах.



### Методы вычисления QR-разложения

#### 1. Метод отражений Хаусхолдера

Отражения Хаусхолдера — это линейные преобразования, которые позволяют обнулить элементы под диагональю матрицы, приводя её к верхнетреугольному виду. Такой метод считается более устойчивым, чем метод Грамма-Шмидта, особенно в задачах с плохой обусловленностью.

Основная идея метода:

Для матрицы $A \in \mathbb{R}^{m \times n}$ мы хотим найти последовательность ортогональных матриц Хаусхолдера $H_1, H_2, \dots, H_{n-1}$, таких что:

$$
H_{n-1} \dots H_2 H_1 A = R
$$

где $R$ — верхнетреугольная матрица.

Отражение Хаусхолдера $H$ определяется как:

$$
H = I - 2 \frac{vv^T}{v^T v}
$$

где $v$ — специально подобранный вектор, который обнуляет нижние элементы столбца матрицы $A$.

Алгоритм:
1. На первом шаге мы выбираем вектор $v$, который обнуляет все элементы первого столбца под первым элементом.
2. Применяем отражение $H_1$ к матрице $A$, чтобы обнулить элементы под первой строкой.
3. Повторяем процесс для каждой последующей подматрицы, пока не получим верхнетреугольную матрицу.

Этот метод позволяет вычислить QR-разложение за $O(mn^2)$ операций, что делает его эффективным для численных вычислений.

#### 2. Метод вращений Гивенса

Вращения Гивенса — это линейные преобразования, которые позволяют обнулять отдельные элементы матрицы путём вращения двухмерного подпространства. В отличие от отражений Хаусхолдера, которые действуют сразу на весь столбец, вращения Гивенса работают локально, обнуляя отдельные элементы под диагональю.

Матрица вращения Гивенса $G(i, j, \theta)$ имеет вид:

$$
G(i, j, \theta) =
\begin{pmatrix}
1 &  &  &  & \\
 & \cos(\theta) &  & \sin(\theta) & \\
 &  & 1 &  & \\
 & -\sin(\theta) &  & \cos(\theta) & \\
 &  &  &  & 1
\end{pmatrix}
$$

где $\theta$ — угол вращения, а $i$-ая и $j$-ая строки матрицы изменяются.

Алгоритм:
1. Для каждого элемента ниже диагонали в матрице $A$ выбираем угол $\theta$ таким образом, чтобы обнулить данный элемент.
2. Вращение применяется к строкам матрицы $A$, чтобы обнулить выбранный элемент.
3. Продолжаем процесс, пока все элементы под диагональю не будут обнулены.

Метод Гивенса особенно эффективен для разреженных матриц, так как он действует на локальные части матрицы.



### Пример QR-разложения

Рассмотрим простую матрицу $A \in \mathbb{R}^{3 \times 2}$:

$$
A = \begin{pmatrix}
1 & 1 \\
1 & 0 \\
0 & 1
\end{pmatrix}
$$

Мы хотим найти её QR-разложение.

#### Шаг 1: Метод Грамма-Шмидта

1. Первый столбец матрицы $A$ — это $a_1 = (1, 1, 0)^T$. Нормализуем его:

$$
q_1 = \frac{a_1}{\|a_1\|} = \frac{(1, 1, 0)^T}{\sqrt{1^2 + 1^2 + 0^2}} = \frac{(1, 1, 0)^T}{\sqrt{2}} = \left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0 \right)^T
$$

2. Второй столбец $a_2 = (1, 0, 1)^T$. Ортогонализируем его относительно $q_1$:

$$
a_2' = a_2 - (q_1^T a_2) q_1 = (1, 0, 1)^T - \left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0 \right)^T \cdot (1, 0, 1)^T \cdot \frac{1}{\sqrt{2}} = (0, -\frac{1}{2}, 1)^T
$$

Теперь нормализуем вектор $a_2'$, чтобы получить $q_2$:

$$
q_2 = \frac{(0, -\frac{1}{2}, 1)^T}{\sqrt{0^2 + (-\frac{1}{2})^2 + 1^2}} = (0, -\frac{1}{\sqrt{5}}, \frac{2}{\sqrt{5}})^T
$$

3. Теперь можем собрать матрицу $Q$:

$$
Q = \begin{pmatrix}
\frac{1}{\sqrt{2}} & 0 \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{5}} \\
0 & \frac{2}{\sqrt{5}}
\end{pmatrix}
$$

4. Для получения $R$, мы перемножаем $Q^T A$:

$$
R = Q^T A = \begin{pmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} & 0 \\
0 & -\frac{1}{\sqrt{5}} & \frac{2}{\sqrt{5}}
\end{pmatrix}
$$



### Заключение

Матричные разложения, такие как SVD и QR-разложения, являются основой для многих методов численного анализа и линейной алгебры. Они помогают разложить сложные матрицы на более простые компоненты, обеспечивая эффективные вычисления и решение широкого круга прикладных задач, таких как сжатие данных, анализ сигналов и решение систем линейных уравнений.


Давайте рассмотрим конкретные примеры сингулярного разложения (SVD) и QR-разложения. Для каждого из разложений я предоставлю пошаговые решения.

### Пример 1: Сингулярное разложение (SVD)

Рассмотрим матрицу $A$:

$$
A = \begin{pmatrix}
3 & 1 \\
1 & 3 \\
0 & 0
\end{pmatrix}
$$

#### Шаг 1: Вычисление $A^T A$

Сначала мы найдем матрицу $A^T A$:

$$
A^T = \begin{pmatrix}
3 & 1 & 0 \\
1 & 3 & 0
\end{pmatrix}
$$

Теперь перемножим $A^T$ и $A$:

$$
A^T A = \begin{pmatrix}
3 & 1 & 0 \\
1 & 3 & 0
\end{pmatrix} \begin{pmatrix}
3 & 1 \\
1 & 3 \\
0 & 0
\end{pmatrix} = \begin{pmatrix}
3 \cdot 3 + 1 \cdot 1 & 3 \cdot 1 + 1 \cdot 3 \\
1 \cdot 3 + 3 \cdot 1 & 1 \cdot 1 + 3 \cdot 3
\end{pmatrix}
$$

$$
= \begin{pmatrix}
10 & 6 \\
6 & 10
\end{pmatrix}
$$

#### Шаг 2: Нахождение собственных значений и собственных векторов

Теперь найдем собственные значения матрицы $A^T A$ решая характеристическое уравнение:

$$
\text{det}(A^T A - \lambda I) = 0
$$

где $I$ — единичная матрица. Мы можем записать это как:

$$
\text{det}\begin{pmatrix}
10 - \lambda & 6 \\
6 & 10 - \lambda
\end{pmatrix} = (10 - \lambda)^2 - 36 = 0
$$

Решаем это уравнение:

$$
(10 - \lambda)^2 = 36 \implies 10 - \lambda = 6 \quad \text{или} \quad 10 - \lambda = -6
$$

$$
\lambda_1 = 16, \quad \lambda_2 = 4
$$

Теперь найдем собственные векторы.

Для $\lambda_1 = 16$:

$$
(A^T A - 16I) \begin{pmatrix}
x_1 \\
x_2
\end{pmatrix} = 0 \implies \begin{pmatrix}
-6 & 6 \\
6 & -6
\end{pmatrix} \begin{pmatrix}
x_1 \\
x_2
\end{pmatrix} = 0
$$

Это дает нам уравнение $-6x_1 + 6x_2 = 0 \implies x_1 = x_2$.

Выберем $x_1 = 1$, тогда $v_1 = \begin{pmatrix}
1 \\
1
\end{pmatrix}$.

Для $\lambda_2 = 4$:

$$
(A^T A - 4I) \begin{pmatrix}
x_1 \\
x_2
\end{pmatrix} = 0 \implies \begin{pmatrix}
6 & 6 \\
6 & 6
\end{pmatrix} \begin{pmatrix}
x_1 \\
x_2
\end{pmatrix} = 0
$$

Это дает нам уравнение $6x_1 + 6x_2 = 0 \implies x_1 = -x_2$.

Выберем $x_1 = 1$, тогда $v_2 = \begin{pmatrix}
1 \\
-1
\end{pmatrix}$.

#### Шаг 3: Нормализация собственных векторов

Ортонормируем собственные векторы:

$$
v_1 = \frac{1}{\sqrt{2}} \begin{pmatrix}
1 \\
1
\end{pmatrix}, \quad v_2 = \frac{1}{\sqrt{2}} \begin{pmatrix}
1 \\
-1
\end{pmatrix}
$$

#### Шаг 4: Нахождение сингулярных значений

Сингулярные значения матрицы $A$ равны квадратным корням собственных значений матрицы $A^T A$:

$$
\sigma_1 = \sqrt{16} = 4, \quad \sigma_2 = \sqrt{4} = 2
$$

#### Шаг 5: Нахождение матрицы $U$

Теперь найдем $U$ с использованием сингулярных значений:

1. Для $v_1 = \begin{pmatrix}
1 \\
1
\end{pmatrix}$:
$$
   u_1 = \frac{1}{\sigma_1} A v_1 = \frac{1}{4} A \begin{pmatrix}
   1 \\
   1
   \end{pmatrix} = \frac{1}{4} \begin{pmatrix}
   4 \\
   4 \\
   0
   \end{pmatrix} = \begin{pmatrix}
   1 \\
   1 \\
   0
   \end{pmatrix}
$$

2. Для $v_2 = \begin{pmatrix}
1 \\
-1
\end{pmatrix}$:
$$
   u_2 = \frac{1}{\sigma_2} A v_2 = \frac{1}{2} A \begin{pmatrix}
   1 \\
   -1
   \end{pmatrix} = \frac{1}{2} \begin{pmatrix}
   2 \\
   0 \\
   -1
   \end{pmatrix} = \begin{pmatrix}
   1 \\
   0 \\
   -\frac{1}{2}
   \end{pmatrix}
$$

Нормализуем $u_2$:

$$
u_2 = \frac{1}{\sqrt{1 + \frac{1}{4}}} \begin{pmatrix}
1 \\
0 \\
-\frac{1}{2}
\end{pmatrix} = \begin{pmatrix}
1 \\
0 \\
-\frac{1}{2}
\end{pmatrix}
$$

#### Шаг 6: Формирование матриц $U$, $\Sigma$ и $V$

1. Теперь соберем матрицы:
$$
   U = \begin{pmatrix}
   1 & 1 \\
   1 & 0 \\
   0 & -\frac{1}{2}
   \end{pmatrix}, \quad \Sigma = \begin{pmatrix}
   4 & 0 \\
   0 & 2 \\
   0 & 0
   \end{pmatrix}, \quad V = \begin{pmatrix}
   \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
   \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
   \end{pmatrix}
$$

Итак, окончательное разложение SVD:

$$
A = U \Sigma V^T = \begin{pmatrix}
1 & 1 \\
1 & 0 \\
0 & -\frac{1}{2}
\end{pmatrix} \begin{pmatrix}
4 & 0 \\
0 & 2
\end{pmatrix} \begin{pmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{pmatrix}
$$



### Пример 2: QR-разложение

Рассмотрим ту же матрицу $A$:

$$
A = \begin{pmatrix}
3 & 1 \\
1 & 3 \\
0 & 0
\end{pmatrix}
$$

#### Шаг 1: Применение метода Грамма-Шмидта

1. Нормализация первого столбца $a_1 = \begin{pmatrix} 3 \\ 1 \\ 0 \end{pmatrix}$:
$$
   \|a_1\| = \sqrt{3^2 + 1^2 + 0^2} = \sqrt{10}
$$
$$
   q_1 = \frac{a_1}{\|a_1\|} = \frac{1}{\sqrt{10}} \begin{pmatrix} 3 \\ 1

 \\ 0 \end{pmatrix} = \begin{pmatrix} \frac{3}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} \\ 0 \end{pmatrix}
$$

2. Ортогонализация второго столбца $a_2 = \begin{pmatrix} 1 \\ 3 \\ 0 \end{pmatrix}$:
$$
   \text{proj}_{q_1}(a_2) = \left( \frac{a_2^T q_1}{q_1^T q_1} \right) q_1 = \left( \frac{\begin{pmatrix} 1 \\ 3 \\ 0 \end{pmatrix}^T \begin{pmatrix} \frac{3}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} \\ 0 \end{pmatrix}}{1} \right) \begin{pmatrix} \frac{3}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} \\ 0 \end{pmatrix}
$$

   Вычислим:

$$
   a_2^T q_1 = \frac{3}{\sqrt{10}} + \frac{3}{\sqrt{10}} = \frac{6}{\sqrt{10}} \quad \text{и} \quad q_1^T q_1 = 1
$$

   Значит,
$$
   \text{proj}_{q_1}(a_2) = \frac{6}{\sqrt{10}} \begin{pmatrix} \frac{3}{\sqrt{10}} \\ \frac{1}{\sqrt{10}} \\ 0 \end{pmatrix} = \begin{pmatrix} \frac{18}{10} \\ \frac{6}{10} \\ 0 \end{pmatrix} = \begin{pmatrix} 1.8 \\ 0.6 \\ 0 \end{pmatrix}
$$

   Теперь найдем ортогональный компонент:
$$
   a_2' = a_2 - \text{proj}_{q_1}(a_2) = \begin{pmatrix} 1 \\ 3 \\ 0 \end{pmatrix} - \begin{pmatrix} 1.8 \\ 0.6 \\ 0 \end{pmatrix} = \begin{pmatrix} -0.8 \\ 2.4 \\ 0 \end{pmatrix}
$$

3. Нормализация $a_2'$:
$$
   \|a_2'\| = \sqrt{(-0.8)^2 + (2.4)^2} = \sqrt{0.64 + 5.76} = \sqrt{6.4}
$$
$$
   q_2 = \frac{a_2'}{\|a_2'\|} = \frac{1}{\sqrt{6.4}} \begin{pmatrix} -0.8 \\ 2.4 \\ 0 \end{pmatrix} = \begin{pmatrix} -\frac{0.8}{\sqrt{6.4}} \\ \frac{2.4}{\sqrt{6.4}} \\ 0 \end{pmatrix}
$$

#### Шаг 2: Формирование матрицы $Q$ и $R$

1. Соберем матрицу $Q$:
$$
   Q = \begin{pmatrix}
   \frac{3}{\sqrt{10}} & -\frac{0.8}{\sqrt{6.4}} \\
   \frac{1}{\sqrt{10}} & \frac{2.4}{\sqrt{6.4}} \\
   0 & 0
   \end{pmatrix}
$$

2. Найдем матрицу $R$:
$$
   R = Q^T A = \begin{pmatrix}
   q_1^T a_1 & q_1^T a_2
   \end{pmatrix}
$$

В данном случае матрица $R$ будет верхнетреугольной, и мы получим окончательное QR-разложение.

### Заключение

Таким образом, мы выполнили сингулярное разложение и QR-разложение для матрицы $A$ с пошаговыми расчетами. Оба метода имеют свои применения в численных методах, анализе данных и решении систем уравнений. Если у вас есть дополнительные вопросы или вы хотите рассмотреть другие примеры, дайте знать!




Давайте реализуем примеры SVD и QR-разложения на Python и добавим визуализацию с помощью библиотеки `matplotlib`. Для этого мы будем использовать библиотеки `numpy` для численных расчетов и `matplotlib` для графиков.


```python
import numpy as np
import matplotlib.pyplot as plt

# Определение матрицы A
A = np.array([[3, 1],
              [1, 3],
              [0, 0]])

# Функция для визуализации векторов
def plot_vectors(vectors, colors, title):
    origin = np.zeros((2, len(vectors[0])))
    plt.quiver(*origin, vectors[0], vectors[1], color=colors, scale=1, scale_units='xy', angles='xy')
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.axhline(0, color='grey', lw=0.5)
    plt.axvline(0, color='grey', lw=0.5)
    plt.grid()
    plt.title(title)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# 1. Сингулярное разложение (SVD)
U, Sigma, VT = np.linalg.svd(A)

# Вывод сингулярных значений и матриц U и V
print("Сингулярные значения (Sigma):", Sigma)
print("Матрица U:\n", U)
print("Матрица V^T:\n", VT)

# Визуализация векторов U и V
plot_vectors(U[:, :2].T, ['r', 'g'], 'Векторы U (SVD)')
plot_vectors(VT[:2, :].T, ['b', 'm'], 'Векторы V^T (SVD)')

# 2. QR-разложение
Q, R = np.linalg.qr(A)

# Вывод матриц Q и R
print("Матрица Q:\n", Q)
print("Матрица R:\n", R)

# Визуализация векторов Q
plot_vectors(Q[:, :2].T, ['c', 'y'], 'Векторы Q (QR)')

# Визуализация R как прямоугольника
plt.figure()
plt.imshow(R, cmap='viridis', aspect='auto')
plt.title('Матрица R (QR-разложение)')
plt.colorbar()
plt.show()
```
