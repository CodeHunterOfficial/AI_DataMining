#Глава 6. Скалярные и векторные пространства


### Введение

Скалярные и векторные пространства являются основополагающими понятиями линейной алгебры и математики в целом. Векторные пространства служат основой для анализа и решения многих задач в математике, физике и инженерии. В этой лекции мы рассмотрим базисы и размерности векторных пространств, а также нормы и расстояния.



### 1. Векторные пространства

Определение векторного пространства:
Векторное пространство $V$ над полем $F$ (обычно это поле действительных чисел $\mathbb{R}$ или комплексных чисел $\mathbb{C}$) — это множество объектов, называемых векторами, вместе с двумя операциями:

1. Сложение векторов: Если $\mathbf{u}, \mathbf{v} \in V$, то $\mathbf{u} + \mathbf{v} \in V$.
2. Умножение вектора на скаляр: Если $\alpha \in F$ и $\mathbf{u} \in V$, то $\alpha \mathbf{u} \in V$.

Векторное пространство $V$ должно удовлетворять следующим аксиомам:

1. Коммутативность сложения: $\mathbf{u} + \mathbf{v} = \mathbf{v} + \mathbf{u}$ для всех $\mathbf{u}, \mathbf{v} \in V$.
2. Ассоциативность сложения: $\mathbf{u} + (\mathbf{v} + \mathbf{w}) = (\mathbf{u} + \mathbf{v}) + \mathbf{w}$ для всех $\mathbf{u}, \mathbf{v}, \mathbf{w} \in V$.
3. Существование нулевого вектора: Существует такой вектор $\mathbf{0} \in V$, что $\mathbf{u} + \mathbf{0} = \mathbf{u}$ для любого $\mathbf{u} \in V$.
4. Существование обратного элемента: Для любого $\mathbf{u} \in V$ существует такой вектор $-\mathbf{u} \in V$, что $\mathbf{u} + (-\mathbf{u}) = \mathbf{0}$.
5. Дистрибутивность относительно сложения скалярей: $\alpha(\mathbf{u} + \mathbf{v}) = \alpha\mathbf{u} + \alpha\mathbf{v}$.
6. Дистрибутивность относительно сложения векторов: $(\alpha + \beta) \mathbf{u} = \alpha \mathbf{u} + \beta \mathbf{u}$.
7. Ассоциативность умножения на скаляр: $\alpha(\beta \mathbf{u}) = (\alpha \beta) \mathbf{u}$.
8. Существование единичного элемента: $1 \cdot \mathbf{u} = \mathbf{u}$ для любого $\mathbf{u} \in V$.


### 2. Базисы и размерность

Определение линейной комбинации: Векторы $\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_n \in V$ называются линейно зависимыми, если существуют скаляры $\alpha_1, \alpha_2, \ldots, \alpha_n \in F$, не все равные нулю, такие что:

$$
\alpha_1 \mathbf{v}_1 + \alpha_2 \mathbf{v}_2 + \ldots + \alpha_n \mathbf{v}_n = \mathbf{0}.
$$

В противном случае они называются линейно независимыми.

Определение базиса: Набор векторов $\{\mathbf{b}_1, \mathbf{b}_2, \ldots, \mathbf{b}_k\}$ называется базисом векторного пространства $V$, если:

1. Векторы линейно независимы.
2. Любой вектор $\mathbf{v} \in V$ может быть представлен как линейная комбинация векторов базиса:

$$
\mathbf{v} = \alpha_1 \mathbf{b}_1 + \alpha_2 \mathbf{b}_2 + \ldots + \alpha_k \mathbf{b}_k.
$$

Размерность векторного пространства: Размерность $V$ обозначается как $\dim(V)$ и равна числу векторов в базисе $V$. Например, если базис состоит из $n$ векторов, то $\dim(V) = n$.

Теорема о базисе: Если векторное пространство $V$ имеет конечное число векторов в базисе, то все его базисы имеют одинаковое число векторов.


### 3. Нормы и расстояния

Определение нормы: Норма вектора $\mathbf{v} \in V$ — это функция $||\cdot|| : V \to \mathbb{R}$, которая удовлетворяет следующим свойствам:

1. Неотрицательность: $||\mathbf{v}|| \geq 0$ и $||\mathbf{v}|| = 0$ только тогда, когда $\mathbf{v} = \mathbf{0}$.
2. Однородность: $||\alpha \mathbf{v}|| = |\alpha| \cdot ||\mathbf{v}||$ для любого скаляра $\alpha$ и вектора $\mathbf{v}$.
3. Неравенство треугольника: $||\mathbf{u} + \mathbf{v}|| \leq ||\mathbf{u}|| + ||\mathbf{v}||$ для любых $\mathbf{u}, \mathbf{v} \in V$.

Примеры норм:

1. Евклидова норма (L2-норма):

$$
||\mathbf{v}||_2 = \sqrt{v_1^2 + v_2^2 + \ldots + v_n^2}.
$$

2. Манхэттенская норма (L1-норма):

$$
||\mathbf{v}||_1 = |v_1| + |v_2| + \ldots + |v_n|.
$$

3. Максимальная норма (L∞-норма):

$$
||\mathbf{v}||_\infty = \max\{|v_1|, |v_2|, \ldots, |v_n|\}.
$$

Определение расстояния: Расстояние между двумя векторами $\mathbf{u}$ и $\mathbf{v}$ определяется как норма их разности:

$$
d(\mathbf{u}, \mathbf{v}) = ||\mathbf{u} - \mathbf{v}||.
$$

### 4. Свойства норм и расстояний

1. Норма является расстоянием: Норма вектора $\mathbf{v}$ может использоваться для измерения расстояния от нуля до $\mathbf{v}$.
2. Неотрицательность расстояния: $d(\mathbf{u}, \mathbf{v}) \geq 0$ и $d(\mathbf{u}, \mathbf{v}) = 0$ только при $\mathbf{u} = \mathbf{v}$.
3. Симметричность: $d(\mathbf{u}, \mathbf{v}) = d(\mathbf{v}, \mathbf{u})$.
4. Неравенство треугольника: $d(\mathbf{u}, \mathbf{v}) \leq d(\mathbf{u}, \mathbf{w}) + d(\mathbf{w}, \mathbf{v})$ для любого вектора $\mathbf{w}$.

Они находят широкое применение в различных областях науки и техники, таких как физика, экономика, информатика и многие другие. ### Пример 1: Базис и размерность векторного пространства

Дано: Векторы $\mathbf{v_1} = \begin{pmatrix} 1 \\ 2 \end{pmatrix}$ и $\mathbf{v_2} = \begin{pmatrix} 3 \\ 4 \end{pmatrix}$.

Шаг 1: Проверка линейной независимости векторов

Для того чтобы векторы $\mathbf{v_1}$ и $\mathbf{v_2}$ образовывали базис, нужно проверить их линейную независимость. Это значит, что уравнение

$$
\alpha_1 \begin{pmatrix} 1 \\ 2 \end{pmatrix} + \alpha_2 \begin{pmatrix} 3 \\ 4 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}
$$

должно иметь только тривиальное решение $\alpha_1 = 0$ и $\alpha_2 = 0$.

Шаг 2: Запишем систему уравнений

Разложим уравнение:

$$
\begin{pmatrix} \alpha_1 + 3\alpha_2 \\ 2\alpha_1 + 4\alpha_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}
$$

Это дает две линейные системы:

1. $\alpha_1 + 3\alpha_2 = 0$
2. $2\alpha_1 + 4\alpha_2 = 0$

Шаг 3: Решим первую систему

Из первого уравнения выразим $\alpha_1$:

$$
\alpha_1 = -3\alpha_2
$$

Подставим это значение во второе уравнение:

$$
2(-3\alpha_2) + 4\alpha_2 = 0 \implies -6\alpha_2 + 4\alpha_2 = 0 \implies -2\alpha_2 = 0 \implies \alpha_2 = 0
$$

Теперь подставим $\alpha_2 = 0$ в $\alpha_1 = -3\alpha_2$:

$$
\alpha_1 = 0
$$

Шаг 4: Заключение

Поскольку единственное решение $\alpha_1 = 0$ и $\alpha_2 = 0$, векторы $\mathbf{v_1}$ и $\mathbf{v_2}$ линейно независимы и образуют базис в $\mathbb{R}^2$. Размерность этого векторного пространства равна 2:

$$
\dim(V) = 2.
$$

### Пример 2: Норма вектора и расстояние

Дано: Вектор $\mathbf{u} = \begin{pmatrix} 3 \\ 4 \end{pmatrix}$ и $\mathbf{v} = \begin{pmatrix} 1 \\ 2 \end{pmatrix}$.

Шаг 1: Найдем норму вектора $\mathbf{u}$

Используем евклидову норму:

$$
||\mathbf{u}|| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5.
$$

Шаг 2: Найдем норму вектора $\mathbf{v}$

$$
||\mathbf{v}|| = \sqrt{1^2 + 2^2} = \sqrt{1 + 4} = \sqrt{5} \approx 2.24.
$$

Шаг 3: Найдем расстояние между векторами $\mathbf{u}$ и $\mathbf{v}$

Расстояние вычисляется как норма их разности:

$$
d(\mathbf{u}, \mathbf{v}) = ||\mathbf{u} - \mathbf{v}||.
$$

Шаг 4: Вычислим разность

$$
\mathbf{u} - \mathbf{v} = \begin{pmatrix} 3 \\ 4 \end{pmatrix} - \begin{pmatrix} 1 \\ 2 \end{pmatrix} = \begin{pmatrix} 3 - 1 \\ 4 - 2 \end{pmatrix} = \begin{pmatrix} 2 \\ 2 \end{pmatrix}.
$$

Шаг 5: Найдем норму разности

Теперь вычислим норму разности:

$$
d(\mathbf{u}, \mathbf{v}) = ||\begin{pmatrix} 2 \\ 2 \end{pmatrix}|| = \sqrt{2^2 + 2^2} = \sqrt{4 + 4} = \sqrt{8} = 2\sqrt{2} \approx 2.83.
$$


### Пример 3: Проверка линейной зависимости

Дано: Векторы $\mathbf{a_1} = \begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}, \mathbf{a_2} = \begin{pmatrix} 2 \\ 2 \\ 2 \end{pmatrix}, \mathbf{a_3} = \begin{pmatrix} 3 \\ 3 \\ 3 \end{pmatrix}$.

Шаг 1: Проверим линейную зависимость

Составим линейное уравнение:

$$
\alpha_1 \begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix} + \alpha_2 \begin{pmatrix} 2 \\ 2 \\ 2 \end{pmatrix} + \alpha_3 \begin{pmatrix} 3 \\ 3 \\ 3 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \\ 0 \end{pmatrix}.
$$

Шаг 2: Запишем систему уравнений

Получим три уравнения:

1. $\alpha_1 + 2\alpha_2 + 3\alpha_3 = 0$
2. $\alpha_1 + 2\alpha_2 + 3\alpha_3 = 0$
3. $\alpha_1 + 2\alpha_2 + 3\alpha_3 = 0$

Все три уравнения идентичны. Таким образом, у нас есть бесконечное множество решений, например:

- Если $\alpha_3 = 1$, то $\alpha_1 + 2\alpha_2 + 3 \cdot 1 = 0$ даёт $\alpha_1 + 2\alpha_2 = -3$.

Это означает, что векторы линейно зависимы.

## Пример 4: Определение базиса и размерности в $\mathbb{R}^3$

Дано: Векторы $\mathbf{v_1} = \begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix}$, $\mathbf{v_2} = \begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix}$, $\mathbf{v_3} = \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix}$.

### Шаг 1: Проверка линейной независимости

Чтобы проверить, являются ли векторы $\mathbf{v_1}, \mathbf{v_2}, \mathbf{v_3}$ линейно независимыми, мы составим уравнение:

$$
\alpha_1 \begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix} + \alpha_2 \begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix} + \alpha_3 \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \\ 0 \end{pmatrix}.
$$

### Шаг 2: Записываем систему уравнений

Разложим уравнение:

$$
\begin{pmatrix} \alpha_1 \\ \alpha_2 \\ \alpha_3 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \\ 0 \end{pmatrix}.
$$

Это дает нам три уравнения:

1. $\alpha_1 = 0$
2. $\alpha_2 = 0$
3. $\alpha_3 = 0$

### Шаг 3: Заключение

Поскольку единственное решение $\alpha_1 = \alpha_2 = \alpha_3 = 0$, векторы $\mathbf{v_1}, \mathbf{v_2}, \mathbf{v_3}$ линейно независимы. Следовательно, они образуют базис в $\mathbb{R}^3$, и размерность этого пространства равна 3:

$$
\dim(V) = 3.
$$



## Пример 5: Нормы и расстояния в $\mathbb{R}^3$

Дано: Векторы $\mathbf{u} = \begin{pmatrix} 1 \\ 2 \\ 2 \end{pmatrix}$ и $\mathbf{v} = \begin{pmatrix} 2 \\ 3 \\ 4 \end{pmatrix}$.

### Шаг 1: Найдем норму векторов

Норма вектора $\mathbf{u}$:

$$
||\mathbf{u}|| = \sqrt{1^2 + 2^2 + 2^2} = \sqrt{1 + 4 + 4} = \sqrt{9} = 3.
$$

Норма вектора $\mathbf{v}$:

$$
||\mathbf{v}|| = \sqrt{2^2 + 3^2 + 4^2} = \sqrt{4 + 9 + 16} = \sqrt{29} \approx 5.39.
$$

### Шаг 2: Найдем расстояние между векторами

Расстояние $d(\mathbf{u}, \mathbf{v})$ вычисляется как норма разности векторов:

$$
d(\mathbf{u}, \mathbf{v}) = ||\mathbf{u} - \mathbf{v}||.
$$

### Шаг 3: Вычислим разность

$$
\mathbf{u} - \mathbf{v} = \begin{pmatrix} 1 \\ 2 \\ 2 \end{pmatrix} - \begin{pmatrix} 2 \\ 3 \\ 4 \end{pmatrix} = \begin{pmatrix} 1 - 2 \\ 2 - 3 \\ 2 - 4 \end{pmatrix} = \begin{pmatrix} -1 \\ -1 \\ -2 \end{pmatrix}.
$$

### Шаг 4: Найдем норму разности

Теперь вычислим норму разности:

$$
d(\mathbf{u}, \mathbf{v}) = ||\begin{pmatrix} -1 \\ -1 \\ -2 \end{pmatrix}|| = \sqrt{(-1)^2 + (-1)^2 + (-2)^2} = \sqrt{1 + 1 + 4} = \sqrt{6} \approx 2.45.
$$



## Пример 6: Линейная комбинация векторов

Дано: Векторы $\mathbf{a_1} = \begin{pmatrix} 1 \\ 2 \\ 3 \end{pmatrix}$, $\mathbf{a_2} = \begin{pmatrix} 4 \\ 5 \\ 6 \end{pmatrix}$.

### Задача: Найти $\alpha_1$ и $\alpha_2$, чтобы:

$$
\alpha_1 \mathbf{a_1} + \alpha_2 \mathbf{a_2} = \begin{pmatrix} 5 \\ 7 \\ 9 \end{pmatrix}.
$$

### Шаг 1: Запишем уравнения

Составим уравнения:

$$
\alpha_1 \begin{pmatrix} 1 \\ 2 \\ 3 \end{pmatrix} + \alpha_2 \begin{pmatrix} 4 \\ 5 \\ 6 \end{pmatrix} = \begin{pmatrix} 5 \\ 7 \\ 9 \end{pmatrix}.
$$

Это дает систему уравнений:

1. $\alpha_1 + 4\alpha_2 = 5$
2. $2\alpha_1 + 5\alpha_2 = 7$
3. $3\alpha_1 + 6\alpha_2 = 9$

### Шаг 2: Решим первую систему уравнений

Из первого уравнения выразим $\alpha_1$:

$$
\alpha_1 = 5 - 4\alpha_2.
$$

Подставим $\alpha_1$ во второе уравнение:

$$
2(5 - 4\alpha_2) + 5\alpha_2 = 7 \implies 10 - 8\alpha_2 + 5\alpha_2 = 7 \implies 10 - 3\alpha_2 = 7.
$$

Решим для $\alpha_2$:

$$
-3\alpha_2 = 7 - 10 \implies -3\alpha_2 = -3 \implies \alpha_2 = 1.
$$

### Шаг 3: Найдем $\alpha_1$

Теперь подставим $\alpha_2$ обратно в уравнение для $\alpha_1$:

$$
\alpha_1 = 5 - 4 \cdot 1 = 5 - 4 = 1.
$$

### Шаг 4: Заключение

Мы нашли, что $\alpha_1 = 1$ и $\alpha_2 = 1$, что означает, что:

$$
\begin{pmatrix} 5 \\ 7 \\ 9 \end{pmatrix} = 1 \cdot \begin{pmatrix} 1 \\ 2 \\ 3 \end{pmatrix} + 1 \cdot \begin{pmatrix} 4 \\ 5 \\ 6 \end{pmatrix}.
$$



## Пример 7: Применение нормы в различных пространствах

Дано: Векторы $\mathbf{x} = \begin{pmatrix} 3 \\ -1 \\ 2 \end{pmatrix}$ и $\mathbf{y} = \begin{pmatrix} -1 \\ 4 \\ 0 \end{pmatrix}$.

### Шаг 1: Вычисление L1-нормы для $\mathbf{x}$ и $\mathbf{y}$

L1-норма:

$$
||\mathbf{x}||_1 = |3| + |-1| + |2| = 3 + 1 + 2 = 6,
$$
$$
||\mathbf{y}||_1 = |-1| + |4| + |0| = 1 + 4 + 0 = 5.
$$

### Шаг 2: Вычисление L∞-нормы для $\mathbf{x}$ и $\mathbf{y}$

$L∞$-норма:

$$
||\mathbf{x}||_\infty = \max\{|3|, |-1|, |2|\} = 3,
$$
$$
||\mathbf{y}||_\infty = \max\{|-1|, |4|, |0|\} = 4.
$$



Визуализация векторов для различных примеров выполнена:

1. Пример 1: Векторы $v_1 = \begin{pmatrix} 1 \\ 2 \end{pmatrix}$ и $v_2 = \begin{pmatrix} 3 \\ 4 \end{pmatrix}$ в $\mathbb{R}^2$, иллюстрирующие базисное пространство.
   
2. Пример 2: Векторы $u = \begin{pmatrix} 3 \\ 4 \end{pmatrix}$ и $v = \begin{pmatrix} 1 \\ 2 \end{pmatrix}$, которые использовались для вычисления нормы и расстояния между ними.

3. Пример 4: Векторы $v_1$, $v_2$ и $v_3$ в $\mathbb{R}^3$, образующие стандартный базис (ортонормальный) в трехмерном пространстве.

4. Пример 5: Векторы $u = \begin{pmatrix} 1 \\ 2 \\ 2 \end{pmatrix}$ и $v = \begin{pmatrix} 2 \\ 3 \\ 4 \end{pmatrix}$, используемые для вычисления нормы в трехмерном пространстве $\mathbb{R}^3$.

Эти графики наглядно демонстрируют геометрическое положение векторов и связь между ними в различных пространствах.

### Задачи для самостоятельной работы

#### Задачи на базисы и размерности

1. Найдите базис для векторного пространства, заданного векторами $\begin{pmatrix} 2 \\ 4 \\ 6 \end{pmatrix}$ и $\begin{pmatrix} 1 \\ 3 \\ 5 \end{pmatrix}$.
2. Определите размерность пространства, натянутого на векторы $\begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix}$, $\begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix}$ и $\begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix}$.
3. Являются ли векторы $\begin{pmatrix} 1 \\ 2 \\ 3 \end{pmatrix}$, $\begin{pmatrix} 2 \\ 4 \\ 6 \end{pmatrix}$ и $\begin{pmatrix} 3 \\ 6 \\ 9 \end{pmatrix}$ линейно независимыми? Обоснуйте ответ.
4. Найдите базис для векторного пространства, натянутого на векторы $\begin{pmatrix} 1 \\ 2 \end{pmatrix}$ и $\begin{pmatrix} 3 \\ 6 \end{pmatrix}$.
5. Вычислите размерность векторного пространства, заданного векторами $\begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}$, $\begin{pmatrix} 0 \\ 1 \\ 1 \end{pmatrix}$ и $\begin{pmatrix} 1 \\ 0 \\ 1 \end{pmatrix}$.

#### Задачи на нормы и расстояния

6. Вычислите норму вектора $\mathbf{v} = \begin{pmatrix} 3 \\ 4 \\ 5 \end{pmatrix}$.
7. Найдите расстояние между векторами $\mathbf{u} = \begin{pmatrix} 1 \\ 2 \\ 3 \end{pmatrix}$ и $\mathbf{v} = \begin{pmatrix} 4 \\ 5 \\ 6 \end{pmatrix}$.
8. Вычислите $L^1$-норму и $L^\infty$-норму для вектора $\mathbf{a} = \begin{pmatrix} -2 \\ 3 \\ 1 \end{pmatrix}$.
9. Найдите расстояние между векторами $\mathbf{u} = \begin{pmatrix} 2 \\ -1 \end{pmatrix}$ и $\mathbf{v} = \begin{pmatrix} 1 \\ 2 \end{pmatrix}$.
10. Вычислите норму вектора $\mathbf{b} = \begin{pmatrix} 1 \\ -1 \\ 2 \\ -2 \end{pmatrix}$ в пространстве $\mathbb{R}^4$.

#### Задачи на линейные комбинации

11. Найдите коэффициенты $\alpha_1$ и $\alpha_2$ такие, чтобы $\alpha_1 \begin{pmatrix} 1 \\ 1 \end{pmatrix} + \alpha_2 \begin{pmatrix} 2 \\ 2 \end{pmatrix} = \begin{pmatrix} 3 \\ 3 \end{pmatrix}$.
12. Являются ли векторы $\begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix}$ и $\begin{pmatrix} 0 \\ 1 \\ 1 \end{pmatrix}$ линейно независимыми?
13. Найдите вектор $\mathbf{w}$, который можно представить как линейную комбинацию векторов $\mathbf{u} = \begin{pmatrix} 1 \\ 0 \\ 1 \end{pmatrix}$ и $\mathbf{v} = \begin{pmatrix} 0 \\ 1 \\ 1 \end{pmatrix}$, где $\mathbf{w} = \begin{pmatrix} 2 \\ 2 \\ 2 \end{pmatrix}$.
14. Найдите коэффициенты для линейной комбинации $\mathbf{a_1}$ и $\mathbf{a_2}$, если $\mathbf{b} = 2\mathbf{a_1} - 3\mathbf{a_2}$ и $\mathbf{b} = \begin{pmatrix} 4 \\ 5 \end{pmatrix}$.
15. Найдите, является ли вектор $\begin{pmatrix} 0 \\ 2 \\ 4 \end{pmatrix}$ линейной комбинацией векторов $\begin{pmatrix} 1 \\ 0 \\ 1 \end{pmatrix}$ и $\begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix}$.

#### Задачи на применение норм и расстояний

16. Найдите расстояние между точками $A(1, 2)$ и $B(4, 6)$ в $\mathbb{R}^2$.
17. Вычислите расстояние между точками $P(2, 3, 5)$ и $Q(1, 1, 1)$ в $\mathbb{R}^3$.
18. Найдите норму вектора $\mathbf{v} = \begin{pmatrix} -3 \\ 4 \\ -2 \end{pmatrix}$.
19. Вычислите $L^2$-норму для вектора $\mathbf{x} = \begin{pmatrix} 1 \\ 2 \\ 3 \\ 4 \end{pmatrix}$.
20. Найдите $L^1$-норму для вектора $\mathbf{y} = \begin{pmatrix} -2 \\ 0 \\ 3 \\ -1 \end{pmatrix}$.

#### Задачи на анализ свойств векторов

#### Проблемы на применение теорем и аксиом

26. Используя аксиомы векторного пространства, докажите, что сумма двух линейно независимых векторов остается линейно независимой.
27. Определите, является ли множество векторов $\left\{ \begin{pmatrix} 1 \\ 2 \\ 0 \end{pmatrix}, \begin{pmatrix} 2 \\ 4 \\ 0 \end{pmatrix}, \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix} \right\}$ базисом в $\mathbb{R}^3$.
28. Используя свойства норм, докажите, что для любого вектора $\mathbf{a}$ выполняется $||\mathbf{a}|| \geq 0$.
29. Используя аксиомы векторных пространств, докажите, что для любых векторов $\mathbf{u}, \mathbf{v}, \mathbf{w}$ выполняется $\mathbf{u} + (\mathbf{v} + \mathbf{w}) = (\mathbf{u} + \mathbf{v}) + \mathbf{w}$.
30. Примените теорему о существовании базиса, чтобы показать, что любой конечномерный вектор может быть представлен как линейная комбинация базисных векторов.
